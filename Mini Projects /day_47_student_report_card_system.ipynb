```python
# ================================================================
# ðŸŽ“ MINI PROJECT: STUDENT REPORT CARD SYSTEM
# ================================================================

# This program will:
# 1. Ask student name
# 2. Ask marks for 5 subjects
# 3. Store them in a list
# 4. Calculate total
# 5. Calculate average
# 6. Decide grade
# 7. Print full report card

# ================================================================
# STEP 1: FUNCTION TO CALCULATE GRADE
# ================================================================

# A function is like a MACHINE.
# You give something to it (input).
# It processes.
# It gives something back (output).

# Example:
# Blender machine:
# Put fruits -> Blend -> Juice comes out

# Same idea here.

def calculate_grade(average):
    # "def" means: define a function
    # calculate_grade is function name
    # (average) is input parameter (like fruit going inside blender)

    # Now we check average and decide grade

    if average >= 90:
        return "A+"
        # return means: send result back and stop function

    elif average >= 75:
        return "A"

    elif average >= 60:
        return "B"

    elif average >= 50:
        return "C"

    else:
        return "Fail"


# ================================================================
# STEP 2: ASK STUDENT NAME
# ================================================================

# input() always gives STRING (text)

student_name = input("Enter student name: ")

# Example:
# If user types: John
# student_name = "John"

# ================================================================
# STEP 3: CREATE EMPTY LIST TO STORE MARKS
# ================================================================

# A LIST is like a basket that can store multiple values.
# Example:
# basket = [10, 20, 30]

marks = []
# [] means empty list
# Right now basket is empty

# ================================================================
# STEP 4: LOOP TO TAKE 5 SUBJECT MARKS
# ================================================================

# range(5) means:
# Repeat 5 times.
# It produces numbers: 0,1,2,3,4

for i in range(5):
    # for loop repeats code inside it

    # i is just a counter number
    # First time: i = 0
    # Second time: i = 1
    # ...

    mark_input = input(f"Enter marks for subject {i+1}: ")

    # f"" is formatted string
    # {i+1} means:
    # If i=0 â†’ show subject 1
    # If i=1 â†’ show subject 2

    mark = int(mark_input)
    # Convert text into integer

    marks.append(mark)
    # append means: add item into list
    # Example:
    # If marks was []
    # After append(80)
    # marks = [80]

# After loop finishes,
# marks might look like:
# [80, 70, 90, 60, 85]


# ================================================================
# STEP 5: CALCULATE TOTAL
# ================================================================

total = 0
# Start total from 0

for m in marks:
    # This loop goes through each value inside marks list
    # Example:
    # If marks = [80, 70, 90]
    # First m = 80
    # Then m = 70
    # Then m = 90

    total = total + m
    # Add each mark into total

# If marks = [80, 70, 90]
# total calculation:
# 0 + 80 = 80
# 80 + 70 = 150
# 150 + 90 = 240


# ================================================================
# STEP 6: CALCULATE AVERAGE
# ================================================================

average = total / len(marks)

# len(marks) means:
# How many items inside list
# If 5 subjects â†’ len = 5

# Example:
# total = 400
# len = 5
# average = 400 / 5 = 80


# ================================================================
# STEP 7: GET GRADE USING FUNCTION
# ================================================================

grade = calculate_grade(average)

# We send average into function
# Function checks and returns grade
# Example:
# average = 82
# grade = "A"


# ================================================================
# STEP 8: CREATE DICTIONARY FOR REPORT
# ================================================================

# Dictionary is like labeled data
# Example:
# {"name": "John", "age": 10}

report = {
    "Name": student_name,
    "Marks": marks,
    "Total": total,
    "Average": average,
    "Grade": grade
}

# This stores all results in organized way


# ================================================================
# STEP 9: PRINT REPORT CARD
# ================================================================

print("\n================ REPORT CARD ================")

for key, value in report.items():
    # report.items() gives:
    # key and value pair

    print(f"{key} : {value}")

print("=============================================")


# ================================================================
# ðŸŽ‰ END OF PROGRAM
# ================================================================
```
