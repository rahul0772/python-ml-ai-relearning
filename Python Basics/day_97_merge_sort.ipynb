```python
# ================================================================
# MERGE SORT (Divide & Conquer Algorithm)
# ================================================================
# Imagine you have a messy pile of numbers like cards 🃏:
#   [38, 27, 43, 3, 9, 82, 10]
#
# Instead of sorting everything at once (hard 🤯),
# we use a SUPER SMART STRATEGY:
#
#   1. -234567890-BREAK the list into smaller parts again and again
#   2. SORT those tiny parts
#   3. MERGE them back together in sorted order
#
# This is called: MERGE SORT
#
# 🧠 CORE IDEA:
#   "Break → Solve → Combine"
#
# ⏱️ Time Complexity:
#   O(n log n) → VERY FAST for large data
#
# ================================================================


# -------------------------------
# 🔧 STEP 1: DEFINE MERGE FUNCTION
# -------------------------------
def merge(left, right):
    # left → first sorted half
    # right → second sorted half

    merged = []  # This will store final sorted result

    i = 0  # pointer for left list
    j = 0  # pointer for right list

    # 🧠 COMPARE BOTH LISTS ELEMENT BY ELEMENT
    while i < len(left) and j < len(right):

        # Compare current elements
        if left[i] < right[j]:
            # If left element is smaller → add it first
            merged.append(left[i])
            i += 1  # move left pointer forward
        else:
            # If right element is smaller → add it first
            merged.append(right[j])
            j += 1  # move right pointer forward

    # 🧠 ADD REMAINING ELEMENTS (if any)
    # Because one list may still have leftover items

    # Add remaining left elements
    while i < len(left):
        merged.append(left[i])
        i += 1

    # Add remaining right elements
    while j < len(right):
        merged.append(right[j])
        j += 1

    return merged


# -------------------------------
# 🔧 STEP 2: DEFINE MERGE SORT FUNCTION
# -------------------------------
def merge_sort(arr):
    # arr → list we want to sort

    # 🧠 BASE CASE:
    # If list has 1 or 0 elements → already sorted
    if len(arr) <= 1:
        return arr

    # 🧠 DIVIDE STEP:
    # Split list into two halves
    mid = len(arr) // 2

    left_half = arr[:mid]    # first half
    right_half = arr[mid:]   # second half

    # 🧠 RECURSION:
    # Sort both halves separately
    left_sorted = merge_sort(left_half)
    right_sorted = merge_sort(right_half)

    # 🧠 MERGE STEP:
    # Combine sorted halves into one sorted list
    return merge(left_sorted, right_sorted)


# -------------------------------
# 🧪 STEP 3: TEST MERGE SORT
# -------------------------------

# Unsorted list
arr = [38, 27, 43, 3, 9, 82, 10]

print("Original array:", arr)

# Apply merge sort
sorted_arr = merge_sort(arr)

print("Sorted array:", sorted_arr)


# ================================================================
# 📊 OUTPUT EXPLANATION (VERY IMPORTANT 🧠)
# ================================================================
# Input:
#   [38, 27, 43, 3, 9, 82, 10]
#
# ----------------------------------------
# 🧠 STEP-BY-STEP BREAKDOWN:
# ----------------------------------------
#
# STEP 1 (Divide):
#   [38, 27, 43, 3, 9, 82, 10]
# → split →
#   [38, 27, 43]   and   [3, 9, 82, 10]
#
# STEP 2 (Divide more):
#   Keep splitting until single elements:
#
#   [38] [27] [43]   and   [3] [9] [82] [10]
#
# STEP 3 (Merge back sorted):
#
#   [38] + [27] → [27, 38]
#   [27, 38] + [43] → [27, 38, 43]
#
#   [3] + [9] → [3, 9]
#   [82] + [10] → [10, 82]
#   [3, 9] + [10, 82] → [3, 9, 10, 82]
#
# STEP 4 (Final merge):
#   [27, 38, 43] + [3, 9, 10, 82]
# → [3, 9, 10, 27, 38, 43, 82]
#
# ----------------------------------------
# FINAL OUTPUT:
#   Sorted array: [3, 9, 10, 27, 38, 43, 82]
#
# ----------------------------------------
# 🧠 DEEP MENTAL MODEL:
# ----------------------------------------
# Think like this:
#
# ❌ Don't try to solve BIG problem directly
# ✅ Break BIG → into SMALL pieces
# ✅ Solve SMALL → easily
# ✅ Combine → into final answer
#
# This pattern is called:
#   👉 Divide and Conquer
#
# ----------------------------------------
# ⚡ WHY MERGE SORT IS POWERFUL:
# ----------------------------------------
# ✔ Always O(n log n) → very reliable
# ✔ Works great for large datasets
# ✔ Stable sorting (keeps order of equal elements)
#
# ----------------------------------------
# 🧠 MEMORY IMPRINT:
# ----------------------------------------
# MERGE SORT =
#   1. Split
#   2. Sort recursively
#   3. Merge carefully
```
