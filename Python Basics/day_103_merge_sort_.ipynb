```python
# ==============================================
# MERGE SORT LESSON
# ==============================================

# Merge Sort is a classic example of Divide and Conquer.
# Divide and Conquer is a strategy where we:
# 1. Divide the problem into smaller sub-problems
# 2. Solve the smaller sub-problems recursively
# 3. Combine the solved sub-problems into the final solution

# Merge Sort works as follows:
# - Split the list into two halves
# - Recursively sort each half
# - Merge the two sorted halves into one sorted list

# Let's write and explain the code in extreme detail:

def merge_sort(arr):
    """
    Function to perform Merge Sort on a list 'arr'
    """
    # Step 1: Base case of recursion
    # If the array has 0 or 1 element, it is already sorted
    if len(arr) <= 1:
        # This is our stopping condition: a list with 1 or 0 elements cannot be split further
        return arr  # Return the list itself

    # Step 2: Divide
    # Find the middle index of the array
    mid = len(arr) // 2  # Using integer division to find exact middle
    # Explanation: If array has 7 elements, mid = 3, so left = arr[0:3], right = arr[3:7]

    # Recursively call merge_sort on the left half
    left_half = merge_sort(arr[:mid])
    # Explanation: arr[:mid] slices the list from index 0 to mid-1
    # This recursion keeps splitting the list until we reach single elements

    # Recursively call merge_sort on the right half
    right_half = merge_sort(arr[mid:])
    # Explanation: arr[mid:] slices the list from index mid to the end
    # This also keeps splitting until single elements

    # Step 3: Merge the sorted halves
    # We now have two sorted halves and we need to merge them in sorted order
    sorted_array = []  # Initialize empty list to store the merged result
    i = 0  # Pointer for left_half
    j = 0  # Pointer for right_half

    # Loop until we reach the end of one of the halves
    while i < len(left_half) and j < len(right_half):
        # Compare elements of left and right halves
        if left_half[i] < right_half[j]:
            sorted_array.append(left_half[i])
            # Explanation: left element is smaller, append it to the sorted array
            i += 1  # Move pointer in left_half
        else:
            sorted_array.append(right_half[j])
            # Explanation: right element is smaller or equal, append it to sorted array
            j += 1  # Move pointer in right_half

    # If any elements remain in left_half, append them
    while i < len(left_half):
        sorted_array.append(left_half[i])
        i += 1

    # If any elements remain in right_half, append them
    while j < len(right_half):
        sorted_array.append(right_half[j])
        j += 1

    # Return the fully merged and sorted array
    return sorted_array

# =============================
# TESTING MERGE SORT
# =============================

# Let's define an example array
example_array = [38, 27, 43, 3, 9, 82, 10]

# Call merge_sort and store the result
sorted_example = merge_sort(example_array)

# Print the original and sorted array
print("Original array:", example_array)
print("Sorted array:", sorted_example)

# ===============================================
# OUTPUT EXPLANATION
# ===============================================
# Original array: [38, 27, 43, 3, 9, 82, 10]
# Explanation: This is the input list that we want to sort.
#
# Sorted array: [3, 9, 10, 27, 38, 43, 82]
# Explanation:
# - Merge Sort split the array recursively into single elements
# - Then it merged the elements back in sorted order step by step
# - 3 came first because it is the smallest
# - 82 came last because it is the largest
# - Every step preserved sorted order, demonstrating the Divide and Conquer logic
#
# This output shows that merge_sort correctly sorts any unsorted list.
```
