```python
#Testing Multiple Conditions

fruits = ['orange', 'lemon', 'apple']

if "orange" in fruits:
  print("orange")
if "cherry" in fruits:
  print("cherry")
if "apple" in fruits: #three independent tests are executed every time this program is run, every condition in this example is evaluated unlike elif
  print("apple")
```

```python
#elif
fruits = ['orange', 'lemon', 'apple']

if "orange" in fruits:
  print("orange")
elif "cherry" in fruits:
  print("cherry")
elif "apple" in fruits: # doesnâ€™t run any tests beyond the first test that passes in an if-elif-else chain
  print("apple")
```

```python
#Using if Statements with Lists

fruits_basket = ["orange", "apple", "banana"]

for fruit in fruits_basket:
  print(fruit + " is added")
print ("fruit basket filled")
```

```python
#if item runs out
for fruit in fruits_basket:
  if fruit == "apple":
    print(fruit +" ran out")
  else:
    print(fruit + " is added")
print ("fruit basket filled")
```

```python
#List not empty check
fruit_basket = [] #returns True if the list contains at least one item; an empty list evaluates to False

if fruit_basket:
  for fruit in fruit_basket:
    print(fruit + " is added")
  print("basket filled")
else:
    print("basket empty")
```

```python
#Multiple list
fruit_basket = ["orange", "apple", "banana", "cherry"] #available fruits
fruits = ['orange', 'banana', "pear"] #needed fruits

for fruit in fruits:
  if fruit in fruit_basket:
    print(fruit + " is added")
else:
  print(fruit + " ran out")
```
