```python
# A hash table is a data structure that allows you to store and retrieve values based on a key.
# The key is hashed to produce an index, and the value is stored at that index.
# Python's built-in dictionary (`dict`) uses a hash table under the hood.

# First, we'll define a simple hash table using a Python dictionary.

# Creating an empty hash table (dictionary).
hash_table = {}

# Let's add some data to the hash table.
# We use a key-value pair for insertion, where 'key' is the index and 'value' is the data.
# In a hash table, the key is hashed (converted into a unique index), and the value is stored at that index.

# Insert 'apple' with key 'fruit' into the hash table.
hash_table['fruit'] = 'apple'  # key = 'fruit', value = 'apple'

# Insert 'carrot' with key 'vegetable' into the hash table.
hash_table['vegetable'] = 'carrot'  # key = 'vegetable', value = 'carrot'

# Insert 'rose' with key 'flower' into the hash table.
hash_table['flower'] = 'rose'  # key = 'flower', value = 'rose'

# Now, let's access the values using their keys.
# We can retrieve the value by providing the key to the hash table.

# Access the value associated with the key 'fruit'
print("Value for 'fruit' key:", hash_table['fruit'])  # Should print 'apple' because we stored 'apple' with 'fruit' as key

# Access the value associated with the key 'vegetable'
print("Value for 'vegetable' key:", hash_table['vegetable'])  # Should print 'carrot' because we stored 'carrot' with 'vegetable' as key

# Access the value associated with the key 'flower'
print("Value for 'flower' key:", hash_table['flower'])  # Should print 'rose' because we stored 'rose' with 'flower' as key

# What happens if we try to access a key that doesn't exist?
# Python will raise a KeyError if we try to access a key that isn't present in the hash table.
# We can handle this using `get()` method, which returns 'None' instead of raising an error if the key doesn't exist.

# Try accessing a key that does not exist (e.g., 'tree')
# The `get()` method returns 'None' when the key is not found, instead of raising an error.
value = hash_table.get('tree', 'Key not found')  # Since 'tree' doesn't exist, it will return 'Key not found'
print("Value for 'tree' key:", value)  # Should print 'Key not found'

# Let's also demonstrate the deletion of a key-value pair from the hash table using `del`.
# Deleting 'vegetable' from the hash table.
del hash_table['vegetable']  # This removes the entry with key 'vegetable'

# Now let's check the hash table after deletion.
# If we try to access 'vegetable' now, we should get an error since it was deleted.
# But we'll use the `get()` method safely to check.
value_after_deletion = hash_table.get('vegetable', 'Key not found')  # This will return 'Key not found'
print("Value for 'vegetable' key after deletion:", value_after_deletion)  # Should print 'Key not found'

# We can also check if a key exists in the hash table using the 'in' keyword.
# This will return True if the key exists, otherwise False.
print("'fruit' key exists:", 'fruit' in hash_table)  # True, because 'fruit' exists in the hash table
print("'vegetable' key exists:", 'vegetable' in hash_table)  # False, because 'vegetable' was deleted

# Finally, let's print the current state of our hash table.
# The hash table should now contain only 'fruit' and 'flower' because 'vegetable' was deleted.
print("Current state of the hash table:", hash_table)  # Should print: {'fruit': 'apple', 'flower': 'rose'}
```

```python
# We are going to build a custom hash table with separate chaining to handle collisions.
# We will also add logic to resize the hash table when it becomes too full.

class CustomHashTable:
    def __init__(self, initial_capacity=4):
        # We start by creating an empty hash table with a fixed initial size.
        # The initial size is usually a small prime number to minimize collisions.
        self.capacity = initial_capacity
        self.size = 0  # Number of elements in the hash table.

        # We create an array of lists to handle collisions (separate chaining).
        # Each index in this array will hold a list of key-value pairs (a chain).
        self.table = [[] for _ in range(self.capacity)]  # Initialize table with empty lists.

    def _hash(self, key):
        # This is the hash function. We'll use Python's built-in hash() function, but you can write your own.
        # The idea is to take a key and compute a hash value, then map it to an index in the table.
        # We use the modulo operator to ensure the index stays within the bounds of the table size.
        return hash(key) % self.capacity  # Hash the key and return the index in the table.

    def _resize(self):
        # This method resizes the hash table when the load factor exceeds 0.75.
        # Load factor = size of table / capacity of table.
        # If the load factor is too high, we double the table's capacity to avoid too many collisions.

        # When resizing, we create a new table with double the size.
        new_capacity = self.capacity * 2
        new_table = [[] for _ in range(new_capacity)]  # New table with double the capacity.

        # We now rehash all the existing elements and insert them into the new table.
        for chain in self.table:
            for key, value in chain:
                # Rehash each key-value pair and insert into the new table.
                new_index = hash(key) % new_capacity
                new_table[new_index].append((key, value))

        # Update the table and capacity.
        self.table = new_table
        self.capacity = new_capacity

    def insert(self, key, value):
        # This method inserts a key-value pair into the hash table.
        # First, we compute the index using the hash function.
        index = self._hash(key)

        # Check if the key already exists in the chain (list at that index).
        for i, (existing_key, existing_value) in enumerate(self.table[index]):
            if existing_key == key:
                # If the key already exists, we update the value.
                self.table[index][i] = (key, value)
                return

        # If the key doesn't exist, we append the new key-value pair to the list at the index.
        self.table[index].append((key, value))

        # Increase the size of the hash table.
        self.size += 1

        # If the load factor exceeds 0.75, we resize the table.
        if self.size / self.capacity > 0.75:
            self._resize()

    def get(self, key):
        # This method retrieves the value associated with a given key.
        # We compute the index of the key using the hash function.
        index = self._hash(key)

        # Search the chain (list) at the computed index for the key.
        for existing_key, existing_value in self.table[index]:
            if existing_key == key:
                return existing_value

        # If the key doesn't exist, return None.
        return None

    def remove(self, key):
        # This method removes a key-value pair from the hash table.
        index = self._hash(key)  # Find the index for the key.

        # Search through the list at that index to find the key.
        for i, (existing_key, existing_value) in enumerate(self.table[index]):
            if existing_key == key:
                # If the key is found, we remove it from the list (chain).
                del self.table[index][i]
                self.size -= 1
                return

        # If the key is not found, we do nothing.

    def __repr__(self):
        # This method provides a string representation of the hash table.
        return str(self.table)  # Return the table as a string for easy visualization.


# Let's test the CustomHashTable with some insertions and show how it handles collisions.

# Create a new hash table.
hash_table = CustomHashTable()

# Insert some key-value pairs.
hash_table.insert('apple', 10)
hash_table.insert('banana', 20)
hash_table.insert('grape', 30)
hash_table.insert('orange', 40)

# We are intentionally adding more items to see if the table resizes as the load factor increases.

# Insert another key-value pair to trigger resizing.
hash_table.insert('pear', 50)

# Let's try retrieving the values for the keys.
print("Value for 'apple':", hash_table.get('apple'))  # Expected: 10
print("Value for 'banana':", hash_table.get('banana'))  # Expected: 20
print("Value for 'grape':", hash_table.get('grape'))  # Expected: 30
print("Value for 'orange':", hash_table.get('orange'))  # Expected: 40
print("Value for 'pear':", hash_table.get('pear'))  # Expected: 50

# Let's try removing a key-value pair.
hash_table.remove('banana')  # Remove 'banana'

# Try accessing 'banana' after removal.
print("Value for 'banana' after removal:", hash_table.get('banana'))  # Expected: None (since it was removed)

# Print the current state of the hash table.
print("Current hash table:", hash_table)  # Should print the table after the removal and resizing.
```
