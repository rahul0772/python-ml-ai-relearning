```python
# =============================================
# Linked List
# This lesson will teach you how to create and manipulate a singly linked list in Python.
# Each line is explained in extreme detail for cognitive upgrade.
# =============================================

# Step 1: Define a Node class to store individual elements of the linked list
class Node:
    # The Node class is the building block of a linked list.
    # Each Node stores a value and a reference (pointer) to the next Node.

    def __init__(self, data):
        # __init__ is a special method called a constructor. It runs when we create a new Node.
        # 'data' is the value we want to store in the node, e.g., a number or string.
        self.data = data  # Store the value in this Node
        self.next = None  # Initially, the node does not point to any other node (None)

# Step 2: Define a LinkedList class to manage the linked list
class LinkedList:
    # LinkedList class helps manage multiple nodes connected in sequence.

    def __init__(self):
        # Initialize an empty linked list
        self.head = None  # 'head' points to the first node in the list. Starts as None because list is empty

    def append(self, data):
        # Append function adds a new node to the end of the linked list

        new_node = Node(data)
        # Create a new node with the given data
        # Think of this like making a new brick to add to a chain of bricks

        if self.head is None:
            # If the list is empty, there is no head node yet
            self.head = new_node
            # So we make the new node the head of the list
            return  # Exit the function as the node is already added

        # If the list is not empty, we need to traverse to the last node
        last = self.head  # Start at the head node
        while last.next:
            # While there is a node after the current one
            last = last.next  # Move to the next node

        last.next = new_node
        # Once we reach the last node (its next is None), we point it to the new node
        # This links the new node at the end of the list

    def print_list(self):
        # Function to print all elements in the linked list
        current = self.head  # Start from the head of the list
        while current:
            # Keep moving as long as the current node exists (not None)
            print(current.data, end=" -> ")  # Print node's data and an arrow to show connection
            current = current.next  # Move to the next node
        print("None")  # End of the list, showing there are no more nodes

# Step 3: Demonstration - using the LinkedList class
my_list = LinkedList()
# Create a new linked list object called my_list

# Append some elements
my_list.append(10)
# Add a node with value 10
my_list.append(20)
# Add a node with value 20
my_list.append(30)
# Add a node with value 30

# Print the list
my_list.print_list()
# Expected Output Explanation:
# The list now has nodes 10 -> 20 -> 30 -> None
# Each number is connected to the next via the 'next' pointer
# 'None' at the end indicates the end of the linked list

# =============================================
# OUTPUT (inside cell):
# 10 -> 20 -> 30 -> None
# Explanation of Output:
# - 10: head node of the linked list
# - 20: second node, connected from 10
# - 30: third node, connected from 20
# - None: marks the end of the linked list
# The arrows show how each node points to the next, which is the core structure of a linked list
# =============================================
```
