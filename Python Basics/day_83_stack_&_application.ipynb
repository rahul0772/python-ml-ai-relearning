```python
# ============================================
# STACK (Using List) + APPLICATION (VALID PARENTHESES)
# ============================================
# A STACK is a data structure that follows:
# LIFO → Last In, First Out
#
# Real life example:
# Stack of plates 🍽️
# You put a plate on top → push
# You remove the top plate → pop
#
# We will:
# 1. Build a stack using Python list
# 2. Use it to solve a real problem:
#    👉 Check if parentheses are valid: "({[]})"

# ============================================
# STEP 1: CREATE STACK CLASS
# ============================================

class Stack:

    def __init__(self):
        # We use a Python list to store stack elements
        self.items = []

    def push(self, item):
        # push means add element to TOP of stack
        self.items.append(item)
        print(f"Pushed {item} → Stack now: {self.items}")

    def pop(self):
        # pop means remove TOP element
        if not self.is_empty():
            removed = self.items.pop()
            print(f"Popped {removed} → Stack now: {self.items}")
            return removed
        return None

    def peek(self):
        # peek means see top element without removing it
        if not self.is_empty():
            return self.items[-1]
        return None

    def is_empty(self):
        # stack is empty if size is 0
        return len(self.items) == 0


# ============================================
# STEP 2: VALID PARENTHESES CHECK FUNCTION
# ============================================

def is_valid_parentheses(expression):
    # Create a stack
    stack = Stack()

    # Dictionary to match closing → opening
    pairs = {
        ')': '(',
        '}': '{',
        ']': '['
    }

    # Loop through each character
    for char in expression:
        print(f"\nProcessing: {char}")

        # If opening bracket → push to stack
        if char in "({[":
            stack.push(char)

        # If closing bracket → check match
        elif char in ")}]":

            # If stack is empty → invalid
            if stack.is_empty():
                print("Stack empty when expecting match → INVALID")
                return False

            # Pop top element
            top = stack.pop()

            # Check if it matches
            if pairs[char] != top:
                print(f"Mismatch: expected {pairs[char]}, got {top} → INVALID")
                return False

    # After processing everything:
    # If stack is empty → valid
    if stack.is_empty():
        return True
    else:
        print("Stack not empty at end → INVALID")
        return False


# ============================================
# STEP 3: TEST INPUT
# ============================================

expression = "({[]})"

print("Expression:", expression)

# ============================================
# STEP 4: RUN THE FUNCTION
# ============================================

result = is_valid_parentheses(expression)

# ============================================
# STEP 5: PRINT RESULT
# ============================================

if result:
    print("\n✅ Parentheses are VALID")
else:
    print("\n❌ Parentheses are INVALID")


# ============================================
# 🖥️ OUTPUT (EXAMPLE RUN)
# ============================================
# Expression: ({[]})
#
# Processing: (
# Pushed ( → Stack now: ['(']
#
# Processing: {
# Pushed { → Stack now: ['(', '{']
#
# Processing: [
# Pushed [ → Stack now: ['(', '{', '[']
#
# Processing: ]
# Popped [ → Stack now: ['(', '{']
#
# Processing: }
# Popped { → Stack now: ['(']
#
# Processing: )
# Popped ( → Stack now: []
#
# ✅ Parentheses are VALID


# ============================================
# 🧾 OUTPUT EXPLANATION (VERY SIMPLE)
# ============================================
# Step-by-step:
#
# 1. Push all opening brackets:
#    ( → { → [
#
# 2. When closing comes:
#    ] matches [
#    } matches {
#    ) matches (
#
# 3. Every closing bracket matched correctly
#
# 4. Stack becomes EMPTY at end → VALID
#
# If something was wrong:
# Example: "({[})"
# Then mismatch would happen → INVALID
#
# ============================================
# ⏱️ TIME COMPLEXITY
# ============================================
# O(n) → we scan each character once
#
# ============================================
# 🎯 KEY IDEA
# ============================================
# Use STACK to remember opening brackets
# Match them with closing brackets in reverse order
#
# LIFO makes it perfect for this problem 🚀
# ============================================
```
