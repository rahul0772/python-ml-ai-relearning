```python
# ============================================
# STACK (Using List) + APPLICATION (VALID PARENTHESES)
# ============================================
# A STACK is a data structure that follows:
# LIFO → Last In, First Out
#
# Real life example:
# Stack of plates 🍽️
# You put a plate on top → push
# You remove the top plate → pop
#
# We will:
# 1. Build a stack using Python list
# 2. Use it to solve a real problem:
#    👉 Check if parentheses are valid: "({[]})"

# ============================================
# STEP 1: CREATE STACK CLASS
# ============================================

class Stack:

    def __init__(self):
        # We use a Python list to store stack elements
        self.items = []

    def push(self, item):
        # push means add element to TOP of stack
        self.items.append(item)
        print(f"Pushed {item} → Stack now: {self.items}")

    def pop(self):
        # pop means remove TOP element
        if not self.is_empty():
            removed = self.items.pop()
            print(f"Popped {removed} → Stack now: {self.items}")
            return removed
        return None

    def peek(self):
        # peek means see top element without removing it
        if not self.is_empty():
            return self.items[-1]
        return None

    def is_empty(self):
        # stack is empty if size is 0
        return len(self.items) == 0


# ============================================
# STEP 2: VALID PARENTHESES CHECK FUNCTION
# ============================================

def is_valid_parentheses(expression):
    # Create a stack
    stack = Stack()

    # Dictionary to match closing → opening
    pairs = {
        ')': '(',
        '}': '{',
        ']': '['
    }

    # Loop through each character
    for char in expression:
        print(f"\nProcessing: {char}")

        # If opening bracket → push to stack
        if char in "({[":
            stack.push(char)

        # If closing bracket → check match
        elif char in ")}]":

            # If stack is empty → invalid
            if stack.is_empty():
                print("Stack empty when expecting match → INVALID")
                return False

            # Pop top element
            top = stack.pop()

            # Check if it matches
            if pairs[char] != top:
                print(f"Mismatch: expected {pairs[char]}, got {top} → INVALID")
                return False

    # After processing everything:
    # If stack is empty → valid
    if stack.is_empty():
        return True
    else:
        print("Stack not empty at end → INVALID")
        return False


# ============================================
# STEP 3: TEST INPUT
# ============================================

expression = "({[]})"

print("Expression:", expression)

# ============================================
# STEP 4: RUN THE FUNCTION
# ============================================

result = is_valid_parentheses(expression)

# ============================================
# STEP 5: PRINT RESULT
# ============================================

if result:
    print("\n✅ Parentheses are VALID")
else:
    print("\n❌ Parentheses are INVALID")


# ============================================
# 🖥️ OUTPUT (EXAMPLE RUN)
# ============================================
# Expression: ({[]})
#
# Processing: (
# Pushed ( → Stack now: ['(']
#
# Processing: {
# Pushed { → Stack now: ['(', '{']
#
# Processing: [
# Pushed [ → Stack now: ['(', '{', '[']
#
# Processing: ]
# Popped [ → Stack now: ['(', '{']
#
# Processing: }
# Popped { → Stack now: ['(']
#
# Processing: )
# Popped ( → Stack now: []
#
# ✅ Parentheses are VALID


# ============================================
# 🧾 OUTPUT EXPLANATION (VERY SIMPLE)
# ============================================
# Step-by-step:
#
# 1. Push all opening brackets:
#    ( → { → [
#
# 2. When closing comes:
#    ] matches [
#    } matches {
#    ) matches (
#
# 3. Every closing bracket matched correctly
#
# 4. Stack becomes EMPTY at end → VALID
#
# If something was wrong:
# Example: "({[})"
# Then mismatch would happen → INVALID
#
# ============================================
# ⏱️ TIME COMPLEXITY
# ============================================
# O(n) → we scan each character once
#
# ============================================
# 🎯 KEY IDEA
# ============================================
# Use STACK to remember opening brackets
# Match them with closing brackets in reverse order
#
# LIFO makes it perfect for this problem 🚀
# ============================================
```

```python
# ============================================
# STACK (SUPER SIMPLE)
# ============================================
# Think of a STACK like a stack of books 📚
#
# If you put books:
# Book A → Book B → Book C
#
# The LAST book you put (C) will be the FIRST one you remove.
#
# This is called:
# 👉 LIFO = Last In, First Out
#
# --------------------------------------------
# We will do ONLY 3 simple operations:
# 1. push → add item
# 2. pop  → remove item
# 3. peek → see top item
# --------------------------------------------


# ============================================
# STEP 1: CREATE EMPTY STACK
# ============================================

stack = []  # empty list = empty stack

print("Start:", stack)


# ============================================
# STEP 2: PUSH (ADD ITEMS)
# ============================================

# push means: add to top (end of list)

stack.append(10)  # add 10
print("After push 10:", stack)

stack.append(20)  # add 20
print("After push 20:", stack)

stack.append(30)  # add 30
print("After push 30:", stack)

# Now stack looks like:
# [10, 20, 30]
# Top = 30 (last item)


# ============================================
# STEP 3: PEEK (SEE TOP)
# ============================================

# peek = check last item without removing it
top_item = stack[-1]  # -1 means last element

print("Top item (peek):", top_item)

# stack is still same:
print("Stack after peek:", stack)


# ============================================
# STEP 4: POP (REMOVE TOP)
# ============================================

# pop removes LAST item

removed = stack.pop()  # removes 30
print("Popped item:", removed)
print("Stack now:", stack)

removed = stack.pop()  # removes 20
print("Popped item:", removed)
print("Stack now:", stack)

# Now stack:
# [10]


# ============================================
# 🖥️ OUTPUT (EXAMPLE RUN)
# ============================================
# Start: []
# After push 10: [10]
# After push 20: [10, 20]
# After push 30: [10, 20, 30]
# Top item (peek): 30
# Stack after peek: [10, 20, 30]
# Popped item: 30
# Stack now: [10, 20]
# Popped item: 20
# Stack now: [10]


# ============================================
# 🧾 VERY EASY EXPLANATION
# ============================================
# We added:
# 10 → 20 → 30
#
# So stack became:
# [10, 20, 30]
#
# TOP = 30 (last added)
#
# When we remove:
# First removed = 30
# Then removed = 20
#
# So removal order is:
# 30 → 20 → 10
#
# This is why:
# 👉 LAST IN → FIRST OUT (LIFO)
#
# ============================================
# 🎯 FINAL IDEA (REMEMBER THIS!)
# ============================================
# push = add to end
# pop  = remove from end
# peek = see last item
#
# STACK always works from ONE SIDE (the TOP)
# ============================================
```
