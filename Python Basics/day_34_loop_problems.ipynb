```python
# ============================================================
# SUPER ADVANCED LOOP PROBLEMS
# ONLY COMMENTS EXPLANATION — NO PARAGRAPHS
# EVERY LINE = EXPLAINED WITH EXAMPLES
# ============================================================


# ------------------------------------------------------------
# PROBLEM 1: CHECK IF A NUMBER IS PRIME
# ------------------------------------------------------------

number = 17                  # number we want to test (example: 17)
is_prime = True              # assume prime first (innocent until proven guilty)

# range(2, number) → 2 to 16
# because 1 and 17 are already known divisors
for i in range(2, number):   # i = 2,3,4,...16

    # % means remainder
    # example: 17 % 2 = 1 (not divisible)
    # example: 17 % 17 = 0 (but we don't check itself)
    if number % i == 0:      # if remainder is 0 → divides perfectly

        is_prime = False     # proof found → not prime anymore
        print(number, "is divisible by", i)  # shows which number broke it
        break                # stop loop early (no need to check more)

# after loop finishes
if is_prime:                 # if never divisible by anything
    print(number, "is PRIME")
else:                        # if divisible by at least one number
    print(number, "is NOT prime")


print("--------------------------------------------------")


# ------------------------------------------------------------
# PROBLEM 2: FACTORIAL USING LOOP
# ------------------------------------------------------------

num = 5                      # factorial target (5!)
factorial = 1                # start from 1 (not 0 or everything dies)

# range(1, num+1) → 1,2,3,4,5
for i in range(1, num + 1):  # i increases step-by-step

    # factorial changes like:
    # 1 * 1 = 1
    # 1 * 2 = 2
    # 2 * 3 = 6
    # 6 * 4 = 24
    # 24 * 5 = 120
    factorial = factorial * i

    print("Multiply by", i, "→ factorial now:", factorial)

print("Final factorial:", factorial)


print("--------------------------------------------------")


# ------------------------------------------------------------
# PROBLEM 3: COUNT DIGITS USING WHILE LOOP
# ------------------------------------------------------------

number = 98765               # starting number
count = 0                    # digit counter starts at 0

# loop runs until number becomes 0
while number > 0:            # as long as digits exist

    # // removes last digit
    # 98765 // 10 = 9876
    # 9876 // 10 = 987
    number = number // 10

    count = count + 1        # one digit removed → count +1

    print("Removed a digit, count:", count)

print("Total digits:", count)


print("--------------------------------------------------")


# ------------------------------------------------------------
# PROBLEM 4: REVERSE A NUMBER
# ------------------------------------------------------------

number = 1234                # original number
reversed_number = 0          # reversed starts empty

while number > 0:            # loop until no digits left

    # % 10 extracts last digit
    # 1234 % 10 = 4
    # 123 % 10 = 3
    digit = number % 10

    # build reversed number
    # 0*10+4 = 4
    # 4*10+3 = 43
    # 43*10+2 = 432
    reversed_number = reversed_number * 10 + digit

    # remove last digit
    # 1234 // 10 = 123
    number = number // 10

    print("Reversed so far:", reversed_number)

print("Final reversed number:", reversed_number)


print("--------------------------------------------------")


# ------------------------------------------------------------
# PROBLEM 5: MULTIPLICATION TABLE (NESTED LOOP)
# ------------------------------------------------------------

# outer loop = rows
for i in range(1, 6):        # i = 1,2,3,4,5

    # inner loop = columns
    for j in range(1, 6):    # j = 1,2,3,4,5

        # multiply row * column
        # example: i=2, j=3 → 6
        print(i * j, end="\t")

    print()                  # new line after each row


print("--------------------------------------------------")


# ------------------------------------------------------------
# PROBLEM 6: FIND DUPLICATES IN LIST
# ------------------------------------------------------------

numbers = [1, 2, 3, 2, 4, 5, 1]
# index:    0  1  2  3  4  5  6
# value:    1  2  3  2  4  5  1
# duplicates here are: 1 and 2

duplicates = []
# empty list → will store duplicate numbers ONLY ONCE


for i in range(len(numbers)):
    # len(numbers) = 7
    # i will be: 0,1,2,3,4,5,6
    # i points to the "current" element

    # example when i = 0:
    # numbers[i] = numbers[0] = 1

    for j in range(i + 1, len(numbers)):
        # j starts AFTER i
        # prevents comparing same element with itself
        # also prevents reverse comparison

        # example when i = 0:
        # j = 1 → compare numbers[0] with numbers[1]
        # j = 2 → compare numbers[0] with numbers[2]
        # ...
        # j = 6 → compare numbers[0] with numbers[6]

        # comparison example:
        # numbers[0] == numbers[6]
        # 1 == 1 → TRUE
        if numbers[i] == numbers[j]:

            # check if already saved
            # avoids duplicate entries like [1,1]
            if numbers[i] not in duplicates:

                duplicates.append(numbers[i])
                # append duplicate value

                print("Found duplicate:", numbers[i])
                # prints when duplicate is detected

# final duplicates list after all comparisons
print("Duplicates list:", duplicates)


print("--------------------------------------------------")


# ------------------------------------------------------------
# PROBLEM 7: NUMBER PYRAMID
# ------------------------------------------------------------

# expected output:
# 1
# 22
# 333
# 4444
# 55555

for num in range(1, 6):
    # num controls WHICH number to print
    # num = 1 → print 1
    # num = 2 → print 22
    # num = 3 → print 333
    # num = 4 → print 4444
    # num = 5 → print 55555

    for count in range(num):
        # count controls HOW MANY TIMES
        # range(num) → repeats num times
        # example:
        # num = 3 → range(3) → 0,1,2 → print 3 three times

        print(num, end="")
        # end="" keeps printing on SAME LINE

    print()
    # moves cursor to next line after inner loop finishes


print("--------------------------------------------------")


# ------------------------------------------------------------
# PROBLEM 8: FIZZBUZZ
# ------------------------------------------------------------

for num in range(1, 21):
    # num = 1,2,3,...20

    # divisible by BOTH 3 and 5
    # example:
    # 15 % 3 = 0
    # 15 % 5 = 0
    # → FizzBuzz
    if num % 3 == 0 and num % 5 == 0:
        print("FizzBuzz")

    # divisible ONLY by 3
    # example:
    # 6 % 3 = 0
    # 6 % 5 ≠ 0
    elif num % 3 == 0:
        print("Fizz")

    # divisible ONLY by 5
    # example:
    # 10 % 5 = 0
    # 10 % 3 ≠ 0
    elif num % 5 == 0:
        print("Buzz")

    # not divisible by 3 or 5
    # example:
    # 7 % 3 ≠ 0
    # 7 % 5 ≠ 0
    else:
        print(num)


print("--------------------------------------------------")


# ------------------------------------------------------------
# PROBLEM 9: SECOND LARGEST NUMBER
# ------------------------------------------------------------

numbers = [10, 5, 20, 8, 15]
# step-by-step comparison:
# start → largest = -1, second = -1

largest = -1
second_largest = -1
# start small so any number replaces them

for num in numbers:
    # loop order:
    # num = 10 → 5 → 20 → 8 → 15

    # first iteration: num = 10
    # 10 > -1 → TRUE
    if num > largest:

        second_largest = largest
        # second_largest = -1

        largest = num
        # largest = 10

    # later when num = 20
    # 20 > 10 → TRUE
    # second_largest = 10
    # largest = 20

    elif num > second_largest and num != largest:
        # example:
        # num = 15
        # 15 > 10 AND 15 ≠ 20 → TRUE
        second_largest = num

    print("Largest:", largest, "Second:", second_largest)

# final answer after loop ends
print("Second largest number:", second_largest)


print("--------------------------------------------------")
```
