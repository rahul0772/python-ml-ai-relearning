```python
# Function to find the length of the longest substring without repeating characters
def longest_unique_substring(s):
    # char_map will store the index of each character we encounter
    char_map = {}  # This will keep track of where each character last appeared in the string
    left = 0  # This is the left boundary of our "window" (sublist) we are checking
    max_length = 0  # This keeps the record of the longest length of unique characters we find

    # Loop through each character in the string with a "right" pointer
    for right in range(len(s)):  # `right` will move from 0 to the end of the string
        # If the current character is already in the dictionary (char_map) and its last index is >= left
        if s[right] in char_map and char_map[s[right]] >= left:
            # This means we found a repeat character within the current window/substring
            # So, we need to move the left pointer to the right of the last position of the repeated character
            # This effectively "shrinks" the window so we can remove the repeated character
            left = char_map[s[right]] + 1  # Move the left pointer just past the repeated character

        # Now we update the dictionary to mark the current character's position
        char_map[s[right]] = right  # Save the index of the current character

        # After every step, we check the size of the current window (right - left + 1)
        # If it's larger than the max length we have found so far, update max_length
        max_length = max(max_length, right - left + 1)  # Update the longest length found

    # After going through the string, return the maximum length of substring without repeating characters
    return max_length

# Example string: "abcabcbb"
result = longest_unique_substring("abcabcbb")
print("Longest unique substring length:", result)  # Expected Output: 3 (substring: "abc")
```
