```python
# ================================================================
# UBREADTH-FIRST SEARCH (BFS) IN PYTHON
# ================================================================

# BFS is a fundamental graph traversal algorithm.
# Key idea: "Explore the nearest nodes first before going deeper."
# Imagine dropping a stone in a pond — waves spread evenly outwards.
# BFS explores nodes layer by layer.

# Step 1: Represent the graph
# We will use an adjacency list because it is memory-efficient and easy to traverse.
graph = {
    'A': ['B', 'C'],  # Node A connects to B and C
    'B': ['A', 'D', 'E'],  # Node B connects to A, D, E
    'C': ['A', 'F'],  # Node C connects to A, F
    'D': ['B'],  # Node D connects to B
    'E': ['B', 'F'],  # Node E connects to B, F
    'F': ['C', 'E']   # Node F connects to C, E
}

# Step 2: Define BFS function
def bfs(graph, start):
    """
    graph : dictionary representing adjacency list of the graph
    start : node where BFS traversal begins
    returns: list of nodes in BFS traversal order
    """

    visited = set()  # Set to keep track of visited nodes (avoid revisiting)
    queue = []       # Queue to maintain nodes to explore next (FIFO structure)
    order = []       # List to store traversal order

    # Start BFS from the starting node
    queue.append(start)  # Add starting node to the queue
    visited.add(start)   # Mark starting node as visited

    # Explanation:
    # BFS works by visiting nodes level by level.
    # Queue ensures we explore nodes in the order they were discovered.

    while queue:  # Continue until there are no nodes left in queue
        current = queue.pop(0)  # Dequeue the first element
        # Explanation:
        # pop(0) removes the first inserted element → ensures FIFO order
        order.append(current)  # Add current node to traversal order

        # DEBUG: Print current node and queue status
        print(f"Visiting node {current}, Queue: {queue}, Traversal so far: {order}")

        # Explore all neighbors of current node
        for neighbor in graph[current]:
            if neighbor not in visited:
                queue.append(neighbor)  # Add neighbor to queue
                visited.add(neighbor)   # Mark neighbor as visited
                # Cognitive insight:
                # - We add unvisited neighbors to the queue for future exploration.
                # - We mark them visited immediately to avoid adding duplicates.

    return order  # Return BFS traversal order

# ================================================================
# Step 3: Run BFS on our graph
# ================================================================

start_node = 'A'
bfs_result = bfs(graph, start_node)

# Print final BFS traversal order
print("\nFinal BFS Traversal Order:", bfs_result)

# ================================================================
# Step 4: Explanation of Output (inside code)
# ================================================================

# Output Explanation:
# 1. Starting at 'A', neighbors are 'B' and 'C'. They are added to the queue → order so far: ['A']
# 2. Next, 'B' is dequeued. Its neighbors 'D' and 'E' are added to queue → order: ['A', 'B']
# 3. Then 'C' is dequeued. Its neighbor 'F' is added → order: ['A', 'B', 'C']
# 4. 'D' is dequeued, no new neighbors → order: ['A', 'B', 'C', 'D']
# 5. 'E' is dequeued, neighbor 'F' already visited → order: ['A', 'B', 'C', 'D', 'E']
# 6. 'F' is dequeued, all neighbors visited → order: ['A', 'B', 'C', 'D', 'E', 'F']
# 7. Queue is empty → BFS traversal complete
# 8. Cognitive logic:
#    - BFS ensures nearest nodes are visited first.
#    - Using a queue (FIFO) guarantees layer-by-layer exploration.
#    - Using a set to track visited nodes prevents cycles or infinite loops.
```
