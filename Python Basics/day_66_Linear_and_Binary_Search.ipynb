```python
# ===============================================================
# PYTHON SEARCHING ALGORITHMS
# Linear Search and Binary Search
# Goal:
# Understand how computers search for numbers inside a list.
#
# Think about this real life example:
# You have a row of boxes with numbers.
# You want to find a specific number.
#
# Two ways:
# 1. Check every box one by one  -> Linear Search
# 2. Jump to the middle and divide -> Binary Search
#
# Binary search is MUCH faster but it only works if the list is sorted.
#
# ===============================================================


# ---------------------------------------------------------------
# PART 1 : WHAT IS A LIST IN PYTHON?
# ---------------------------------------------------------------

# A LIST is like a row of boxes.
# Each box stores a value.

numbers = [2, 5, 8, 12, 16, 23, 38, 56, 72]

# Example of the list visually:
#
# index:   0   1   2   3   4   5   6   7   8
# value:   2   5   8  12  16  23  38  56  72
#
# Index means POSITION.
#
# IMPORTANT RULE:
# Python counting starts from 0
#
# numbers[0] -> 2
# numbers[1] -> 5
# numbers[2] -> 8

print("Our list:", numbers)


# ---------------------------------------------------------------
# PART 2 : LINEAR SEARCH
# ---------------------------------------------------------------

# DEFINITION:
# Linear Search checks each element one by one
# until it finds the target value.

# Example:
# If searching 23 in:
# [2,5,8,12,16,23]
#
# Computer checks:
# 2 -> no
# 5 -> no
# 8 -> no
# 12 -> no
# 16 -> no
# 23 -> YES FOUND

# That means it checks EVERY element until found.


def linear_search(data, target):
    """
    data = the list we want to search inside
    target = the value we want to find
    """

    # len(data) means LENGTH of list
    # Example: len([2,5,8]) -> 3
    length = len(data)

    print("\nStarting LINEAR SEARCH")
    print("Target number:", target)

    # range(length) creates numbers from 0 up to length-1
    # If length = 5
    # range(5) -> 0,1,2,3,4
    for index in range(length):

        # index is the position we are checking

        print("Checking index:", index)

        # data[index] means value stored at that position
        current_value = data[index]

        print("Value at this index:", current_value)

        # Now we compare current value with target
        if current_value == target:

            # == means "is equal to"

            print("FOUND the number!")
            print("Index where found:", index)

            # return means give the answer back
            return index

    # If loop finished and nothing found
    print("Number NOT found in the list.")

    return -1


# Example run
linear_search(numbers, 23)



# ---------------------------------------------------------------
# PART 3 : WHY LINEAR SEARCH IS SLOW
# ---------------------------------------------------------------

# Imagine a list with 1 million numbers.
#
# If the number is last:
# Computer checks ALL numbers.
#
# That is slow.
#
# Time Complexity = O(n)
#
# That means if list doubles,
# time also doubles.


# ---------------------------------------------------------------
# PART 4 : BINARY SEARCH
# ---------------------------------------------------------------

# Binary Search is smarter.
#
# It works like guessing a number game.
#
# Example:
# Think of number between 1 and 100.
#
# You guess 50 first.
#
# If answer is bigger -> search right side
# If smaller -> search left side
#
# Each time we CUT the search space in HALF.
#
# That is why it is called BINARY search.


def binary_search(data, target):

    print("\nStarting BINARY SEARCH")

    # left pointer (start of list)
    left = 0

    # right pointer (end of list)
    right = len(data) - 1

    # Why minus 1?
    # Because last index = length - 1

    # Example
    # list size = 9
    # last index = 8

    while left <= right:

        # Calculate middle position
        mid = (left + right) // 2

        # // means integer division
        # Example:
        # 5//2 = 2

        print("\nLeft index:", left)
        print("Right index:", right)
        print("Middle index:", mid)

        # value at middle
        middle_value = data[mid]

        print("Middle value:", middle_value)

        # If middle value equals target
        if middle_value == target:

            print("FOUND the number!")
            return mid

        # If target is bigger
        elif middle_value < target:

            print("Target bigger than middle")
            print("Searching RIGHT side")

            # move left pointer
            left = mid + 1

        else:

            print("Target smaller than middle")
            print("Searching LEFT side")

            # move right pointer
            right = mid - 1

    print("Number NOT found")

    return -1


# Example run
binary_search(numbers, 23)



# ---------------------------------------------------------------
# PART 5 : STEP BY STEP EXAMPLE OF BINARY SEARCH
# ---------------------------------------------------------------

# Searching 23 in:
# [2,5,8,12,16,23,38,56,72]
#
# Step 1
# middle = 16
# target bigger -> search right
#
# Step 2
# middle = 38
# target smaller -> search left
#
# Step 3
# middle = 23
# FOUND


# ---------------------------------------------------------------
# PART 6 : SPEED COMPARISON
# ---------------------------------------------------------------

# Linear Search
# Worst case: check every element
#
# Time Complexity:
# O(n)

# Binary Search
# Each step divides list in half
#
# Time Complexity:
# O(log n)

# Example
#
# 1 million numbers
#
# Linear search worst case = 1,000,000 checks
#
# Binary search ≈ 20 checks


# ---------------------------------------------------------------
# PART 7 : IMPORTANT RULE
# ---------------------------------------------------------------

# Binary Search ONLY works if list is SORTED.

unsorted_example = [9, 2, 50, 1, 7]

print("\nUnsorted list example:", unsorted_example)

# To use binary search we must SORT first

sorted_example = sorted(unsorted_example)

print("Sorted version:", sorted_example)



# ---------------------------------------------------------------
# PART 8 : PRACTICE
# ---------------------------------------------------------------

# Try searching different numbers
linear_search(numbers, 72)
binary_search(numbers, 72)

linear_search(numbers, 100)
binary_search(numbers, 100)


# Linear Search
# - checks every element
# - simple
# - slow for big lists

# Binary Search
# - divides list in half
# - very fast
# - requires sorted list


print("\nYou finished learning Linear Search and Binary Search!")
```
