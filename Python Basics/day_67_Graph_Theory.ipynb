```python
# ============================================================
# GRAPH THEORY IN PYTHON
# ============================================================

# Think of this like a map:
# - Nodes = things like cities or friends
# - Edges = connections between them like roads or friendships

# -------------------------------
# 1️⃣ CREATE A GRAPH
# -------------------------------
# We use a Python dictionary to store nodes and their connections.
# Dictionary: a way to store "key" -> "value" pairs
# Example:
#   Key = "A"  (a node)
#   Value = ["B", "C"]  (nodes connected to A)

# Let's make a simple graph:
#   A is connected to B and C
#   B is connected to A and D
#   C is connected to A and D
#   D is connected to B and C

graph = {                 # Start dictionary
    "A": ["B", "C"],      # Node A connects to B and C
    "B": ["A", "D"],      # Node B connects to A and D
    "C": ["A", "D"],      # Node C connects to A and D
    "D": ["B", "C"]       # Node D connects to B and C
}                         # End dictionary

# Example: graph["A"] = ["B", "C"]
# This means: If you are at A, you can go to B or C

# -------------------------------
# 2️⃣ PRINT GRAPH
# -------------------------------
print("Graph connections:")
for node in graph:                   # Go through each node (A, B, C, D)
    print(node, "is connected to", graph[node])
    # graph[node] gives the list of neighbors
    # Example output: "A is connected to ['B', 'C']"

# -------------------------------
# 3️⃣ ADD A NODE
# -------------------------------
# Example: Add a new node "E" with no connections yet
graph["E"] = []                      # Node E exists but has no edges
print("\nAdded node E:", graph)

# -------------------------------
# 4️⃣ ADD AN EDGE (CONNECTION)
# -------------------------------
# Connect E to A
graph["E"].append("A")               # Add A to E's connections
graph["A"].append("E")               # Add E to A's connections
print("\nAdded edge between E and A:", graph)

# -------------------------------
# 5️⃣ REMOVE AN EDGE
# -------------------------------
# Remove connection between A and E
graph["A"].remove("E")               # Remove E from A's neighbors
graph["E"].remove("A")               # Remove A from E's neighbors
print("\nRemoved edge between E and A:", graph)

# -------------------------------
# 6️⃣ REMOVE A NODE
# -------------------------------
# Remove node E completely
graph.pop("E")                       # Delete E from graph
print("\nRemoved node E:", graph)

# -------------------------------
# 7️⃣ BREADTH-FIRST SEARCH (BFS)
# -------------------------------
# BFS = Visit all nodes level by level
# Example: Start at A
# 1. Visit A
# 2. Visit neighbors of A: B, C
# 3. Visit neighbors of B and C: D

from collections import deque          # deque is a queue (FIFO)

def bfs(graph, start):
    visited = set()                    # Keep track of nodes we visited
    queue = deque([start])             # Start queue with starting node
    print("\nBFS traversal starting from", start)

    while queue:                       # While there are nodes in queue
        node = queue.popleft()         # Take the first node from queue
        if node not in visited:        # If we haven't visited this node
            print(node, end=" ")       # Print node (we are visiting it)
            visited.add(node)          # Mark it as visited
            for neighbor in graph[node]:  # Look at neighbors of this node
                if neighbor not in visited:  # If neighbor not visited
                    queue.append(neighbor)  # Add neighbor to queue

# Run BFS starting from A
bfs(graph, "A")

# -------------------------------
# 8️⃣ DEPTH-FIRST SEARCH (DFS)
# -------------------------------
# DFS = Go deep first before backtracking
# Example: Start at A
# 1. Visit A
# 2. Visit first neighbor B
# 3. Visit B's first neighbor D
# 4. Go back and visit next neighbor C

def dfs(graph, start, visited=None):
    if visited is None:                # If first call, create visited set
        visited = set()
        print("\nDFS traversal starting from", start)

    print(start, end=" ")              # Visit current node
    visited.add(start)                 # Mark node as visited

    for neighbor in graph[start]:      # Go through neighbors
        if neighbor not in visited:    # Only visit if not visited
            dfs(graph, neighbor, visited)  # Recursive call for neighbor

# Run DFS starting from A
dfs(graph, "A")

# -------------------------------
# 9️⃣ SHORTEST PATH (BFS)
# -------------------------------
# BFS can also find shortest path from one node to another
# Example: Shortest path from A to D

def shortest_path(graph, start, end):
    visited = set()                    # Keep track of visited nodes
    queue = deque([[start]])           # Queue of paths

    while queue:
        path = queue.popleft()         # Take first path from queue
        node = path[-1]                # Last node in path
        if node == end:                # If we reached destination
            return path                # Return this path
        if node not in visited:        # If we didn't visit node yet
            visited.add(node)          # Mark node as visited
            for neighbor in graph[node]:  # Look at neighbors
                new_path = list(path)     # Copy current path
                new_path.append(neighbor) # Add neighbor to path
                queue.append(new_path)    # Add new path to queue

# Find shortest path from A to D
path = shortest_path(graph, "A", "D")
print("\nShortest path from A to D:", path)

# -------------------------------
# 🔟 CHECK IF GRAPH IS CONNECTED
# -------------------------------
# Connected graph = You can reach any node from any node

def is_connected(graph):
    visited = set()
    nodes = list(graph.keys())          # List of all nodes

    def dfs_check(node):
        visited.add(node)              # Mark node visited
        for neighbor in graph[node]:
            if neighbor not in visited:
                dfs_check(neighbor)    # Visit neighbor

    dfs_check(nodes[0])                # Start DFS from first node
    return len(visited) == len(graph)  # True if all nodes visited

print("\nIs the graph connected?", is_connected(graph))

# 1. Graph = Nodes + Edges
# 2. Use dictionary (adjacency list) in Python
# 3. BFS = visit nodes level by level
# 4. DFS = go deep first
# 5. BFS can find shortest path
# 6. Can check if graph is connected
# 7. Can add/remove nodes and edges
# Each line has been explained step by step with examples
```
