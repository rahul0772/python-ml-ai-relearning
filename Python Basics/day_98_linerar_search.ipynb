```python
# 👉 TOPIC: LINEAR SEARCH
#
# ================================================================
# 🧠 WHAT IS LINEAR SEARCH? (VERY EASY)
# ================================================================
# Imagine you are looking for your friend in a line of people:
#
# You check one by one:
# Person 1 ❌
# Person 2 ❌
# Person 3 ✅ FOUND!
#
# 👉 That is LINEAR SEARCH
#
# You go STEP-BY-STEP, checking EVERY element one by one
#
# ================================================================
# 🧠 WHY LEARN THIS?
# ================================================================
# This builds your CORE LOGIC:
# - Looping (very important in programming)
# - Comparing values
# - Decision making
#
# ================================================================
# 🧠 REAL SIMPLE RULE:
# ================================================================
# "Start from beginning → check each element → stop when found"
#
# ================================================================


# 🔢 STEP 1: Create a list (collection of values)
numbers = [10, 25, 7, 89, 32, 56]

# 🧠 Think of this like:
# Index:   0   1   2   3   4   5
# Values: [10, 25,  7, 89, 32, 56]


# 🎯 STEP 2: Define what we are searching for
target = 89

# 🧠 We want to find: where is 89?


# ================================================================
# 🧠 STEP 3: LINEAR SEARCH FUNCTION
# ================================================================
def linear_search(arr, target):

    # ------------------------------------------------------------
    # We use a LOOP to check each element one by one
    # ------------------------------------------------------------
    for index in range(len(arr)):

        # --------------------------------------------------------
        # 🧠 VERY IMPORTANT LINE:
        # arr[index] means:
        # "Get the value at this position"
        #
        # Example:
        # index = 0 → arr[0] = 10
        # index = 1 → arr[1] = 25
        # --------------------------------------------------------
        current_value = arr[index]

        # --------------------------------------------------------
        # Now we COMPARE:
        # "Is this value equal to target?"
        # --------------------------------------------------------
        if current_value == target:

            # 🎯 FOUND!
            return index

        # --------------------------------------------------------
        # If not equal → loop continues automatically
        # and moves to next index
        # --------------------------------------------------------

    # ------------------------------------------------------------
    # If loop finishes → nothing matched
    # ------------------------------------------------------------
    return -1


# ================================================================
# 🧠 STEP 4: CALL FUNCTION
# ================================================================
result = linear_search(numbers, target)


# ================================================================
# 🧠 STEP 5: PRINT RESULT
# ================================================================
if result != -1:
    print("✅ Found at index:", result)
else:
    print("❌ Not found")


# ================================================================
# 🧠 OUTPUT + DEEP EXPLANATION
# ================================================================
# numbers = [10, 25, 7, 89, 32, 56]
# target = 89
#
# Step-by-step execution:
#
# index = 0 → value = 10 → not equal ❌
# index = 1 → value = 25 → not equal ❌
# index = 2 → value = 7  → not equal ❌
# index = 3 → value = 89 → MATCH ✅
#
# So function returns index = 3
#
# Output:
# ✅ Found at index: 3
#

# Linear Search = "Check EVERYTHING one by one"
#
# Pros:
# ✔ Works on ANY list (sorted or not)
#
# Cons:
# ❌ Slow for large data
```
