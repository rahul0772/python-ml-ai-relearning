```python
# ============================================
# 🌳 TREES (Binary Tree Traversal - Depth First Search)
# ============================================

# We will learn about a very important DATA STRUCTURE called a TREE.
# Specifically, we will learn:
# 1. What is a Binary Tree
# 2. How to create nodes
# 3. How to connect nodes to form a tree
# 4. How to TRAVERSE (visit) nodes using DFS (Depth First Search)
#    - Inorder Traversal
#    - Preorder Traversal
#    - Postorder Traversal

# 🧠 MENTAL MODEL (SUPER IMPORTANT):
# Think of a tree like a FAMILY TREE 👨‍👩‍👧‍👦
# - One parent (root)
# - Each parent can have children (left and right)
# - You explore family by visiting members one by one

# ============================================
# STEP 1: CREATE A NODE CLASS
# ============================================

class Node:
    def __init__(self, value):
        # This is the VALUE stored inside the node (like a person’s name in family tree)
        self.value = value

        # LEFT CHILD → like first child
        self.left = None

        # RIGHT CHILD → like second child
        self.right = None

# 🧠 WHY THIS STRUCTURE?
# Each node knows:
# - its own value
# - who is on the left
# - who is on the right
# This creates CONNECTIONS → which forms a TREE

# ============================================
# STEP 2: BUILD A SIMPLE TREE
# ============================================

# Let's manually create nodes
root = Node(1)   # Root node (top of tree)

# Creating children
root.left = Node(2)
root.right = Node(3)

# Adding more children
root.left.left = Node(4)
root.left.right = Node(5)

# 🧠 TREE STRUCTURE LOOKS LIKE THIS:
#         1
#       /   \
#      2     3
#     / \
#    4   5

# ============================================
# STEP 3: TREE TRAVERSAL (DFS METHODS)
# ============================================

# 🌟 1. INORDER TRAVERSAL (LEFT → ROOT → RIGHT)
def inorder(node):
    # BASE CASE: if node is None → stop (like reaching end of branch)
    if node is None:
        return

    # Step 1: Go LEFT (deep dive left side first)
    inorder(node.left)

    # Step 2: Visit ROOT (print value)
    print(node.value, end=" ")

    # Step 3: Go RIGHT
    inorder(node.right)

# 🧠 MENTAL MODEL:
# "Go as LEFT as possible → then come back → then go RIGHT"

# --------------------------------------------

# 🌟 2. PREORDER TRAVERSAL (ROOT → LEFT → RIGHT)
def preorder(node):
    if node is None:
        return

    # Step 1: Visit ROOT first
    print(node.value, end=" ")

    # Step 2: Go LEFT
    preorder(node.left)

    # Step 3: Go RIGHT
    preorder(node.right)

# 🧠 MENTAL MODEL:
# "Visit node FIRST → then explore children"

# --------------------------------------------

# 🌟 3. POSTORDER TRAVERSAL (LEFT → RIGHT → ROOT)
def postorder(node):
    if node is None:
        return

    # Step 1: Go LEFT
    postorder(node.left)

    # Step 2: Go RIGHT
    postorder(node.right)

    # Step 3: Visit ROOT LAST
    print(node.value, end=" ")

# 🧠 MENTAL MODEL:
# "Visit children FIRST → parent LAST"

# ============================================
# STEP 4: RUN ALL TRAVERSALS
# ============================================

print("Inorder Traversal:")
inorder(root)

print("\nPreorder Traversal:")
preorder(root)

print("\nPostorder Traversal:")
postorder(root)

# ============================================
# OUTPUT EXPLANATION (VERY IMPORTANT 🧠🔥)
# ============================================

# Expected Output:
# Inorder Traversal:
# 4 2 5 1 3
#
# Preorder Traversal:
# 1 2 4 5 3
#
# Postorder Traversal:
# 4 5 2 3 1

# 🧠 WHY THIS OUTPUT?

# 👉 INORDER (LEFT → ROOT → RIGHT):
# Go leftmost first:
# 4 → then 2 → then 5 → then 1 → then 3

# 👉 PREORDER (ROOT → LEFT → RIGHT):
# Start at root:
# 1 → then explore left subtree → 2 → 4 → 5 → then right → 3

# 👉 POSTORDER (LEFT → RIGHT → ROOT):
# Visit children first:
# 4 → 5 → then parent 2 → then 3 → then root 1

# ============================================
# 🔥 FINAL SUPER IMPORTANT BRAIN IMPRINT:
# ============================================

# Inorder   = sorted-like visiting (used in BST)
# Preorder  = copying tree / structure
# Postorder = deleting tree (children first)

# 🌳 TREE = CONNECTION + RECURSION + STRUCTURE
# 🔁 DFS = GO DEEP before switching branches
```
