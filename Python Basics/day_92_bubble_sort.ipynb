```python
# We are going to demonstrate the working of a **Basic Sorting Algorithm** called **Bubble Sort**.
# Bubble Sort is a simple algorithm that repeatedly steps through the list, compares adjacent elements, and swaps them if they are in the wrong order.
# The pass through the list is repeated until the list is sorted. This is called "bubbling" the largest unsorted element to the end of the list.

# Let's implement the algorithm in Python and explain each line in extreme detail.

def bubble_sort(arr):
    """
    This is the function that will perform the bubble sort on the input list (arr).
    """

    # We will start with the outer loop to go over the entire list multiple times.
    # In each iteration, we are ensuring that the largest element is placed at the end of the array.

    n = len(arr)  # Get the length of the list to know how many passes we need.
    # Why: We need to know how many elements are there in the list so that we can make the necessary number of passes.

    for i in range(n):
        # Outer loop goes over the list 'n' times, where 'n' is the number of elements in the list.
        # In the first pass, the largest element gets bubbled to the end.
        # In the second pass, the second largest element gets bubbled to the second-last position, and so on.

        swapped = False  # This is a flag to check if we made any swap in this iteration.
        # Why: If no elements were swapped in an entire pass, it means the list is already sorted.
        # We can stop early to improve efficiency and avoid unnecessary checks.

        # Now, let's go through the remaining elements of the list, not including the last i elements that have already been sorted.
        for j in range(0, n-i-1):
            # This inner loop goes from the 0th index to the (n-i-1)th index.
            # The (n-i-1)th index is the last unsorted element because after every pass, the largest element gets placed at the end.

            if arr[j] > arr[j+1]:
                # If the element at index 'j' is greater than the element at index 'j+1',
                # it means they are out of order, so we need to swap them.

                arr[j], arr[j+1] = arr[j+1], arr[j]  # Swap the elements at position j and j+1.
                # Why: Swapping is necessary to move the larger element towards the end of the list.

                swapped = True  # Since we made a swap, we update the flag to True.

        if not swapped:
            # If no swaps were made during this pass, it means the list is already sorted.
            # We can break the loop early to save time.
            break

    return arr  # Finally, return the sorted array.

# Now let's test the bubble_sort function.

arr = [64, 34, 25, 12, 22, 11, 90]  # Unsorted list of numbers
print("Original List:", arr)  # Print the original unsorted list

# Call the bubble_sort function on the unsorted list
sorted_arr = bubble_sort(arr)  # Bubble sort the list

print("Sorted List:", sorted_arr)  # Print the sorted list

# Explanation of Output:

# Original List: [64, 34, 25, 12, 22, 11, 90]
# Sorted List: [11, 12, 22, 25, 34, 64, 90]

# The bubble sort algorithm goes through the list multiple times. Each time, it compares adjacent elements
# and swaps them if they are out of order. On each pass, the largest unsorted element gets moved to the end.
# After enough passes, the list is sorted in ascending order.

# In the first pass, the largest number (90) gets moved to the end.
# In the second pass, the second largest number (64) is placed in its correct position, and so on.
# This continues until the list is fully sorted.

# The time complexity of Bubble Sort is O(n^2) in the worst and average cases, making it inefficient for large lists.
# However, for small lists or nearly sorted lists, it can work relatively efficiently, especially with the optimization where
# we break early if no swaps were made during a pass.
```
