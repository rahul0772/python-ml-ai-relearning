```python
# ======================================================
# PYTHON LOOPS
# ======================================================

# --------------------------------
# PROBLEM 1: Print "Hello" 5 times
# --------------------------------
# range(5) means numbers from 0 to 4 (5 times total)
for i in range(5):
    print("Hello")
    print()


# --------------------------------
# PROBLEM 2: Print numbers from 1 to 5
# --------------------------------
# range(1, 6) starts at 1 and stops before 6
for i in range(1, 6):
    print(i)
    print()


# --------------------------------
# PROBLEM 3: Print numbers from 0 to 4
# --------------------------------
for i in range(5):
    print(i)
    print()


# --------------------------------
# PROBLEM 4: Print your name 3 times
# --------------------------------
for i in range(3):
    print("Your Name")
    print()

# --------------------------------
# PROBLEM 5: Print even numbers from 2 to 10
# --------------------------------
# step = 2 means jump by 2
for i in range(2, 11, 2):
    print(i)
    print()

# --------------------------------
# PROBLEM 6: Print odd numbers from 1 to 9
# --------------------------------
for i in range(1, 10, 2):
    print(i)
    print()

# --------------------------------
# PROBLEM 7: Print numbers from 10 to 1 (reverse)
# --------------------------------
# start at 10, stop at 0, go backward by -1
for i in range(10, 0, -1):
    print(i)
    print()

# --------------------------------
# PROBLEM 8: Print square of numbers 1 to 5
# --------------------------------
for i in range(1, 6):
    print(i * i)   # square means number Ã— itself
    print()

# --------------------------------
# PROBLEM 9: Add numbers from 1 to 5
# --------------------------------
total = 0          # start with 0
for i in range(1, 6):
    total = total + i   # add each number
print("Total is:", total)
print()

# --------------------------------
# PROBLEM 10: Print stars (*) 5 times
# --------------------------------
for i in range(5):
    print("*")
    print()

# --------------------------------
# PROBLEM 11: Print one line of 5 stars
# --------------------------------
print("*****")
print()

# --------------------------------
# PROBLEM 12: Print stars using loop
# --------------------------------
# end="" means stay on same line
for i in range(5):
    print("*", end="")
    print()  # move to next line

# --------------------------------
# PROBLEM 13: Print numbers and their double
# --------------------------------
for i in range(1, 6):
    print(i, "double is", i * 2)
    print()

# --------------------------------
# PROBLEM 14: Count from 1 to 10 using while loop
# --------------------------------
i = 1
while i <= 10:
    print(i)
    i = i + 1   # VERY IMPORTANT or loop never stops!
    print()

# --------------------------------
# PROBLEM 15: Print "Python is fun" 4 times
# --------------------------------
i = 1
while i <= 4:
    print("Python is fun")
    i = i + 1
    print()

# --------------------------------
# PROBLEM 16: Print numbers from 5 to 1 using while
# --------------------------------
i = 5
while i >= 1:
    print(i)
    i = i - 1
    print()

# --------------------------------
# PROBLEM 17: Print first 5 multiples of 3
# --------------------------------
for i in range(1, 6):
    print(3 * i)
    print()

# --------------------------------
# PROBLEM 18: Print numbers from 1 to 5 with smile ðŸ™‚
# --------------------------------
for i in range(1, 6):
    print(i, "ðŸ™‚")
    print()

# --------------------------------
# PROBLEM 19: Print sum of even numbers from 2 to 10
# --------------------------------
total = 0
for i in range(2, 11, 2):
    total = total + i
print("Even sum:", total)
print()

# --------------------------------
# PROBLEM 20: Print table of 2
# --------------------------------
for i in range(1, 11):
    print("2 x", i, "=", 2 * i)
    print()
```

```python
# ======================================================
# PART 2: PYTHON LOOPS (EXTRA DETAILED EXPLANATION)
# ======================================================


# ------------------------------------------------------
# PROBLEM 21: Print numbers from 1 to 5
# ------------------------------------------------------

# We use for loop to repeat code
# range(1, 6) means: start at 1, stop before 6
for i in range(1, 6):
    # i stores the current number
    print(i)

print()  # space after this problem


# ------------------------------------------------------
# PROBLEM 22: Print "I love Python" 3 times
# ------------------------------------------------------

# range(3) means loop will run 3 times
for i in range(3):
    # This line prints the message
    print("I love Python")

print()  # space after this problem


# ------------------------------------------------------
# PROBLEM 23: Print squares of numbers 1 to 4
# ------------------------------------------------------

# Loop starts from 1 and goes to 4
for i in range(1, 5):
    # i * i means square of i
    print(i * i)

print()


# ------------------------------------------------------
# PROBLEM 24: Print numbers from 5 to 1 (reverse)
# ------------------------------------------------------

# start = 5, stop = 0 (not included), step = -1
for i in range(5, 0, -1):
    print(i)

print()


# ------------------------------------------------------
# PROBLEM 25: Add numbers from 1 to 5
# ------------------------------------------------------

# total will store the sum
total = 0

# loop runs from 1 to 5
for i in range(1, 6):
    # add current number to total
    total = total + i

# print final answer
print("Sum is:", total)

print()


# ------------------------------------------------------
# PROBLEM 26: Print first 5 multiples of 2
# ------------------------------------------------------

# loop from 1 to 5
for i in range(1, 6):
    # multiply i by 2
    print(2 * i)

print()


# ------------------------------------------------------
# PROBLEM 27: Print stars in a line
# ------------------------------------------------------

# loop runs 5 times
for i in range(5):
    # end="" means stay on same line
    print("*", end="")

# move to next line
print()

print()


# ------------------------------------------------------
# PROBLEM 28: Print 3 lines of stars
# ------------------------------------------------------

# outer loop controls lines
for i in range(3):
    # inner loop prints stars in one line
    for j in range(5):
        print("*", end="")
    # move to next line after stars
    print()

print()


# ------------------------------------------------------
# PROBLEM 29: Count from 1 to 5 using while loop
# ------------------------------------------------------

# start value
i = 1

# loop runs while condition is true
while i <= 5:
    print(i)
    # increase i by 1
    i = i + 1

print()


# ------------------------------------------------------
# PROBLEM 30: Print table of 3
# ------------------------------------------------------

# loop from 1 to 10
for i in range(1, 11):
    # print multiplication
    print("3 x", i, "=", 3 * i)

print()
```

```python
# ======================================================
# PART 3: PYTHON LOOPS (PROBLEMS 31 TO 50)
# ======================================================


# ------------------------------------------------------
# PROBLEM 31: Print numbers from 1 to 10, skip 5
# ------------------------------------------------------

# Loop from 1 to 10
for i in range(1, 11):
    # if i is equal to 5, skip it
    if i == 5:
        continue  # continue means "skip this round"
    print(i)

print()  # space after problem


# ------------------------------------------------------
# PROBLEM 32: Print numbers from 1 to 10, stop at 7
# ------------------------------------------------------

for i in range(1, 11):
    if i == 7:
        break   # break means "stop the loop"
    print(i)

print()


# ------------------------------------------------------
# PROBLEM 33: Print sum of numbers from 1 to 10
# ------------------------------------------------------

total = 0  # store sum

for i in range(1, 11):
    total = total + i

print("Sum:", total)

print()


# ------------------------------------------------------
# PROBLEM 34: Print all numbers divisible by 3 (1 to 20)
# ------------------------------------------------------

for i in range(1, 21):
    # % means remainder
    if i % 3 == 0:
        print(i)

print()


# ------------------------------------------------------
# PROBLEM 35: Count how many numbers from 1 to 10
# ------------------------------------------------------

count = 0

for i in range(1, 11):
    count = count + 1

print("Count is:", count)

print()


# ------------------------------------------------------
# PROBLEM 36: Print reverse table of 5
# ------------------------------------------------------

for i in range(10, 0, -1):
    print("5 x", i, "=", 5 * i)

print()


# ------------------------------------------------------
# PROBLEM 37: Print first 5 natural numbers using while
# ------------------------------------------------------

i = 1

while i <= 5:
    print(i)
    i = i + 1

print()


# ------------------------------------------------------
# PROBLEM 38: Print sum of even numbers from 1 to 10
# ------------------------------------------------------

total = 0

for i in range(1, 11):
    if i % 2 == 0:
        total = total + i

print("Even sum:", total)

print()


# ------------------------------------------------------
# PROBLEM 39: Print square and cube of numbers 1 to 5
# ------------------------------------------------------

for i in range(1, 6):
    print("Number:", i)
    print("Square:", i * i)
    print("Cube:", i * i * i)

print()


# ------------------------------------------------------
# PROBLEM 40: Print stars in increasing order
# *
# **
# ***
# ****
# ------------------------------------------------------

for i in range(1, 5):
    print("*" * i)

print()


# ------------------------------------------------------
# PROBLEM 41: Print stars in decreasing order
# ****
# ***
# **
# *
# ------------------------------------------------------

for i in range(4, 0, -1):
    print("*" * i)

print()


# ------------------------------------------------------
# PROBLEM 42: Print numbers with their type (even/odd)
# ------------------------------------------------------

for i in range(1, 11):
    if i % 2 == 0:
        print(i, "is Even")
    else:
        print(i, "is Odd")

print()


# ------------------------------------------------------
# PROBLEM 43: Find factorial of 5
# ------------------------------------------------------

fact = 1

for i in range(1, 6):
    fact = fact * i

print("Factorial of 5:", fact)

print()


# ------------------------------------------------------
# PROBLEM 44: Print all characters in a word
# ------------------------------------------------------

word = "Python"

for letter in word:
    print(letter)

print()


# ------------------------------------------------------
# PROBLEM 45: Count letters in a word
# ------------------------------------------------------

word = "Python"
count = 0

for letter in word:
    count = count + 1

print("Total letters:", count)

print()


# ------------------------------------------------------
# PROBLEM 46: Print numbers from 1 to 10 except even
# ------------------------------------------------------

for i in range(1, 11):
    if i % 2 == 0:
        continue
    print(i)

print()


# ------------------------------------------------------
# PROBLEM 47: Print multiplication tables from 1 to 3
# ------------------------------------------------------

for num in range(1, 4):
    print("Table of", num)
    for i in range(1, 6):
        print(num, "x", i, "=", num * i)

print()


# ------------------------------------------------------
# PROBLEM 48: Print sum of digits of number 123
# ------------------------------------------------------

number = 123
total = 0

while number > 0:
    digit = number % 10
    total = total + digit
    number = number // 10

print("Sum of digits:", total)

print()


# ------------------------------------------------------
# PROBLEM 49: Print first 5 multiples of 4 using while
# ------------------------------------------------------

i = 1

while i <= 5:
    print(4 * i)
    i = i + 1

print()


# ------------------------------------------------------
# PROBLEM 50: Print pattern using numbers
# 1
# 12
# 123
# 1234
# ------------------------------------------------------

for i in range(1, 5):
    for j in range(1, i + 1):
        print(j, end="")
    print()

print()
```
