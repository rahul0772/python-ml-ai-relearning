```python

# Insertion Sort is a simple comparison-based sorting algorithm that builds the final sorted list one element at a time.
# It is much like sorting playing cards in your hands. You pick one card at a time and place it in the correct position in
# your hand by comparing it with the already sorted cards.

# In this explanation, we will go into ultra detail, breaking down each and every step of the algorithm in an easy-to-understand way.

def insertion_sort(arr):
    # This function sorts the array 'arr' using the insertion sort algorithm.

    # Outer loop: We start with the second element (index 1) and move through the entire array.
    # The idea is that the portion of the array before the current element is already sorted.
    # We will insert the current element in its correct position in the sorted part of the array.

    for i in range(1, len(arr)):  # Loop through each element from index 1 to the end
        # Key element is the element we are currently trying to insert into the sorted part of the list.
        key = arr[i]

        # 'j' is the index where we will start comparing the key with the previous elements.
        j = i - 1  # Start from the element just before 'key' (i.e., arr[i-1])

        # Inner loop: This is where we shift elements of the sorted part to make space for the key.
        # We keep moving backward through the sorted part (from arr[i-1] down to arr[0])
        # until we find the correct position for the key. The condition arr[j] > key checks if
        # the current element is larger than the key. If so, we move the current element one step to the right.

        while j >= 0 and arr[j] > key:  # Keep moving the larger elements to the right
            arr[j + 1] = arr[j]  # Shift arr[j] to the right, making space for the key
            j -= 1  # Move backward to compare with the next element in the sorted part

        # At this point, we have either:
        # 1. Found a spot for the key (where arr[j] <= key) or
        # 2. Reached the beginning of the array (j < 0).

        # Insert the key into the correct position.
        arr[j + 1] = key  # Insert the key into the position found by shifting elements.

    # At this point, the array 'arr' is now sorted.
    return arr

# Let's test the insertion_sort function with an example:

arr = [12, 11, 13, 5, 6]
# This is an unsorted list of integers. We will pass it to the insertion_sort function to sort it.

sorted_arr = insertion_sort(arr)

# Let's print the sorted list to check the result:
print(sorted_arr)

# Detailed Explanation of how Insertion Sort works:

# Initial list: [12, 11, 13, 5, 6]
#
# Pass 1 (i = 1, key = 11):
# - Compare key (11) with arr[0] (12).
# - Since 12 > 11, move 12 one step to the right: arr[1] = 12.
# - Insert 11 in the correct position: arr[0] = 11.
# Result after pass 1: [11, 12, 13, 5, 6]
#
# Pass 2 (i = 2, key = 13):
# - Compare key (13) with arr[1] (12).
# - Since 13 > 12, no change is made.
# Result after pass 2: [11, 12, 13, 5, 6]
#
# Pass 3 (i = 3, key = 5):
# - Compare key (5) with arr[2] (13).
# - Since 13 > 5, move 13 one step to the right: arr[3] = 13.
# - Compare key (5) with arr[1] (12).
# - Since 12 > 5, move 12 one step to the right: arr[2] = 12.
# - Compare key (5) with arr[0] (11).
# - Since 11 > 5, move 11 one step to the right: arr[1] = 11.
# - Insert 5 in the correct position: arr[0] = 5.
# Result after pass 3: [5, 11, 12, 13, 6]
#
# Pass 4 (i = 4, key = 6):
# - Compare key (6) with arr[3] (13).
# - Since 13 > 6, move 13 one step to the right: arr[4] = 13.
# - Compare key (6) with arr[2] (12).
# - Since 12 > 6, move 12 one step to the right: arr[3] = 12.
# - Compare key (6) with arr[1] (11).
# - Since 11 > 6, move 11 one step to the right: arr[2] = 11.
# - Insert 6 in the correct position: arr[1] = 6.
# Result after pass 4: [5, 6, 11, 12, 13]
#
# Final sorted list: [5, 6, 11, 12, 13]

# Output:
# The sorted list will be: [5, 6, 11, 12, 13]
```
