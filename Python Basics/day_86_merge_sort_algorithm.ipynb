```python
# ============================================================
# MERGE SORT (Divide & Conquer Sorting Algorithm)
# ============================================================
# We will learn:
# 1. What Merge Sort is
# 2. Why it is powerful (very efficient sorting)
# 3. Step-by-step logic in SUPER EASY WAY
# 4. Full working code with explanation on EVERY LINE
# 5. Output + explanation inside this SAME cell
# ============================================================


# ------------------------------------------------------------
# 🔍 WHAT IS MERGE SORT?
# ------------------------------------------------------------
# Merge Sort is a sorting algorithm that:
# 1. Divides the list into smaller parts
# 2. Sorts each part
# 3. Merges them back together in sorted order
#
# Think of it like:
# You break a big problem into smaller easy problems,
# solve them, and combine results.


# ------------------------------------------------------------
# 🧠 SIMPLE IDEA:
# ------------------------------------------------------------
# Example:
# [8, 3, 5, 2]
#
# Step 1: Divide
# [8, 3]   [5, 2]
#
# Step 2: Divide again
# [8] [3]   [5] [2]
#
# Step 3: Merge sorted
# [3, 8]   [2, 5]
#
# Step 4: Final merge
# [2, 3, 5, 8]


# ------------------------------------------------------------
# 🧪 INPUT LIST (UNSORTED)
# ------------------------------------------------------------

numbers = [38, 27, 43, 3, 9, 82, 10]
# This is NOT sorted → we will sort it using Merge Sort


# ------------------------------------------------------------
# 🔧 MERGE FUNCTION
# ------------------------------------------------------------
# This function merges TWO sorted lists into ONE sorted list

def merge(left, right):
    result = []
    # This will store the final merged sorted list

    i = 0
    # Pointer for left list

    j = 0
    # Pointer for right list

    # --------------------------------------------------------
    # Compare elements from both lists
    # --------------------------------------------------------
    while i < len(left) and j < len(right):
        # While both lists still have elements

        if left[i] < right[j]:
            result.append(left[i])
            # Add smaller element to result
            i += 1
            # Move left pointer forward

        else:
            result.append(right[j])
            # Add smaller element to result
            j += 1
            # Move right pointer forward

    # --------------------------------------------------------
    # Add remaining elements (if any)
    # --------------------------------------------------------
    result.extend(left[i:])
    # Add leftover elements from left list

    result.extend(right[j:])
    # Add leftover elements from right list

    return result
    # Return merged sorted list


# ------------------------------------------------------------
# 🔧 MERGE SORT FUNCTION
# ------------------------------------------------------------

def merge_sort(arr):
    # Base case:
    if len(arr) <= 1:
        return arr
        # A list with 1 or 0 elements is already sorted

    # --------------------------------------------------------
    # DIVIDE STEP
    # --------------------------------------------------------
    mid = len(arr) // 2
    # Find middle index

    left_half = arr[:mid]
    # First half of list

    right_half = arr[mid:]
    # Second half of list

    # --------------------------------------------------------
    # RECURSIVE SORTING
    # --------------------------------------------------------
    left_sorted = merge_sort(left_half)
    # Sort left half

    right_sorted = merge_sort(right_half)
    # Sort right half

    # --------------------------------------------------------
    # MERGE STEP
    # --------------------------------------------------------
    return merge(left_sorted, right_sorted)
    # Merge sorted halves into one sorted list


# ------------------------------------------------------------
# ▶️ RUN MERGE SORT
# ------------------------------------------------------------

sorted_numbers = merge_sort(numbers)


# ------------------------------------------------------------
# 📢 PRINT OUTPUT
# ------------------------------------------------------------

print("Original list:", numbers)
print("Sorted list:", sorted_numbers)


# ------------------------------------------------------------
# 📊 OUTPUT EXPLANATION
# ------------------------------------------------------------
# Original list:
# [38, 27, 43, 3, 9, 82, 10]
#
# Merge Sort divides the list again and again:
#
# Step 1:
# Split → [38,27,43] and [3,9,82,10]
#
# Step 2:
# Keep splitting until single elements:
# [38] [27] [43] [3] [9] [82] [10]
#
# Step 3:
# Merge sorted pairs:
# [27,38], [43], [3,9], [10,82]
#
# Step 4:
# Merge again:
# [27,38,43], [3,9,10,82]
#
# Step 5:
# Final merge:
# [3, 9, 10, 27, 38, 43, 82]
#
# ------------------------------------------------------------
# 🧠 WHY MERGE SORT IS POWERFUL?
# ------------------------------------------------------------
# It always divides problem into halves
# Time complexity = O(n log n) → VERY FAST
#
# Works well for large data
# Stable sorting algorithm
#
# ------------------------------------------------------------
# 🏁 FINAL UNDERSTANDING:
# ------------------------------------------------------------
# Merge Sort =
# 1. Divide list
# 2. Sort small parts
# 3. Merge them
# ============================================================
```

```python
# ========================= MERGE SORT (DIVIDE AND CONQUER) =========================
# Merge Sort works using a powerful idea called:
# 1. DIVIDE → break the problem into smaller parts
# 2. CONQUER → solve smaller parts
# 3. COMBINE → merge results together

# Imagine you have cards: [38, 27, 43, 3, 9, 82, 10]
# Instead of sorting all at once, we:
# - Break into halves again and again
# - Then merge sorted halves step by step


def merge_sort(arr):
    # This function sorts the array using Merge Sort

    # Step 1: BASE CONDITION (VERY IMPORTANT)
    # If array has 0 or 1 element, it is already sorted
    # Example: [5] is already sorted
    if len(arr) <= 1:
        return arr  # no work needed

    # Step 2: DIVIDE
    # Find middle index to split array into two halves
    mid = len(arr) // 2  # integer division
    # Example: len = 7 → mid = 3

    # Split array into LEFT and RIGHT halves
    left_half = arr[:mid]   # from start to mid-1
    right_half = arr[mid:]  # from mid to end

    # Step 3: RECURSION (CONQUER)
    # We recursively sort both halves
    left_sorted = merge_sort(left_half)
    right_sorted = merge_sort(right_half)

    # Step 4: MERGE (COMBINE)
    # Now we merge the two sorted halves
    return merge(left_sorted, right_sorted)


def merge(left, right):
    # This function merges TWO SORTED arrays into ONE SORTED array

    result = []  # final merged sorted array

    i = 0  # pointer for left array
    j = 0  # pointer for right array

    # Step 1: Compare elements from both arrays
    # and pick the smaller one
    while i < len(left) and j < len(right):
        # Compare current elements
        if left[i] < right[j]:
            result.append(left[i])  # take from left
            i += 1  # move left pointer
        else:
            result.append(right[j])  # take from right
            j += 1  # move right pointer

    # Step 2: Add remaining elements (IMPORTANT)
    # If left array still has elements
    while i < len(left):
        result.append(left[i])
        i += 1

    # If right array still has elements
    while j < len(right):
        result.append(right[j])
        j += 1

    # Now result is fully sorted
    return result


# ========================= DRIVER CODE =========================

# Original unsorted array
arr = [38, 27, 43, 3, 9, 82, 10]

# Print original array
print("Original Array:", arr)

# Call merge_sort function
sorted_arr = merge_sort(arr)

# Print sorted array
print("Sorted Array:", sorted_arr)


# ========================= OUTPUT EXPLANATION =========================
# Expected Output:
# Original Array: [38, 27, 43, 3, 9, 82, 10]
# Sorted Array: [3, 9, 10, 27, 38, 43, 82]

# WHY THIS OUTPUT?
# Step-by-step mental visualization:

# Step 1: Divide
# [38, 27, 43, 3, 9, 82, 10]
# → [38, 27, 43] and [3, 9, 82, 10]

# Step 2: Keep dividing
# [38, 27, 43] → [38] [27, 43] → [27] [43]
# [3, 9, 82, 10] → [3, 9] [82, 10] → [3] [9] [82] [10]

# Step 3: Merge small sorted arrays
# [27] + [43] → [27, 43]
# [38] + [27, 43] → [27, 38, 43]

# [3] + [9] → [3, 9]
# [82] + [10] → [10, 82]
# [3, 9] + [10, 82] → [3, 9, 10, 82]

# Step 4: Final merge
# [27, 38, 43] + [3, 9, 10, 82]
# → Compare one by one → final sorted list:
# [3, 9, 10, 27, 38, 43, 82]

# ========================= KEY LEARNING =========================
# - Merge Sort is VERY EFFICIENT: O(n log n)
# - Works by breaking problems → solving → merging
# - Uses RECURSION (function calling itself)
# - Stable sorting algorithm (keeps order of equal elements)
# Think like this:
# Big problem → break → solve small → combine smartly
# This pattern is used in MANY advanced algorithms 🚀
```
