```python
# ============================
# MERGE SORT (Divide & Conquer Sorting Algorithm)
# ============================

# Merge Sort is a FAST and efficient sorting algorithm.
# It follows the "Divide and Conquer" approach:
# 1. Divide the list into smaller parts
# 2. Sort each part
# 3. Merge them back together in sorted order

# ------------------------------------------------------------
# STEP 1: DEFINE MERGE FUNCTION (Combines two sorted lists)
# ------------------------------------------------------------

def merge(left, right):
    # left → first sorted half
    # right → second sorted half

    merged = []  # This will store the final merged sorted list

    i = 0  # pointer for left list
    j = 0  # pointer for right list

    # Compare elements from both lists and add smaller one
    while i < len(left) and j < len(right):
        # Check which element is smaller
        if left[i] < right[j]:
            merged.append(left[i])  # add smaller element
            i += 1  # move left pointer forward
        else:
            merged.append(right[j])  # add smaller element
            j += 1  # move right pointer forward

        # Why this works:
        # We always pick the smallest available element first

    # If elements remain in left list
    while i < len(left):
        merged.append(left[i])
        i += 1

    # If elements remain in right list
    while j < len(right):
        merged.append(right[j])
        j += 1

    return merged


# ------------------------------------------------------------
# STEP 2: DEFINE MERGE SORT FUNCTION
# ------------------------------------------------------------

def merge_sort(arr):
    # Base case: If list has 1 or 0 elements → already sorted
    if len(arr) <= 1:
        return arr

    # Find middle index
    mid = len(arr) // 2

    # Divide the list into two halves
    left_half = arr[:mid]
    right_half = arr[mid:]

    print(f"Dividing: {arr} → {left_half} and {right_half}")

    # Recursively sort both halves
    sorted_left = merge_sort(left_half)
    sorted_right = merge_sort(right_half)

    # Merge the sorted halves
    merged = merge(sorted_left, sorted_right)

    print(f"Merging: {sorted_left} + {sorted_right} → {merged}")

    return merged


# ------------------------------------------------------------
# STEP 3: INPUT DATA
# ------------------------------------------------------------

numbers = [38, 27, 43, 3, 9, 82, 10]

print("Original List:", numbers)
print("\n--- Merge Sort Process ---")

# ------------------------------------------------------------
# STEP 4: CALL MERGE SORT
# ------------------------------------------------------------

sorted_numbers = merge_sort(numbers)

# ------------------------------------------------------------
# STEP 5: PRINT FINAL RESULT
# ------------------------------------------------------------

print("\n--- Final Sorted List ---")
print(sorted_numbers)


# ------------------------------------------------------------
# 🧾 OUTPUT EXPLANATION
# ------------------------------------------------------------

# The program first DIVIDES the list again and again:
# [38, 27, 43, 3, 9, 82, 10]
# → [38, 27, 43] and [3, 9, 82, 10]
# → continues until each list has only 1 element

# Then it MERGES them step by step:
# [38] + [27] → [27, 38]
# [27, 38] + [43] → [27, 38, 43]
# [3] + [9] → [3, 9]
# [82] + [10] → [10, 82]
# [3, 9] + [10, 82] → [3, 9, 10, 82]

# Final merge:
# [27, 38, 43] + [3, 9, 10, 82]
# → [3, 9, 10, 27, 38, 43, 82]

# WHY THIS OUTPUT?
# Because Merge Sort ensures that at every step,
# smaller sorted lists are combined into bigger sorted lists.

# TIME COMPLEXITY:
# O(n log n) → very efficient for large data 🚀
```
