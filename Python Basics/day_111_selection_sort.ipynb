```python
# Selection Sort Algorithm
# Selection Sort is another simple sorting algorithm that repeatedly selects the smallest
# element from the unsorted portion of the list and swaps it with the first unsorted element.

# Let's break down the Selection Sort Algorithm step by step in a very detailed manner!

# Initialize a list of numbers
numbers = [64, 25, 12, 22, 11]

# Print the original list to see the unsorted state
print("Original List:", numbers)

# Now let's define the Selection Sort function
def selection_sort(arr):
    # Step 1: Iterate over the entire list
    # The outer loop keeps track of the position where the smallest element should go.
    for i in range(len(arr)):
        # Step 2: Assume the current index 'i' is the minimum
        min_idx = i

        # Step 3: Look for the smallest element in the remaining unsorted portion of the list
        for j in range(i + 1, len(arr)):
            # Compare current minimum with the rest of the unsorted list (arr[j])
            # If arr[j] is smaller, then update min_idx to point to j
            if arr[j] < arr[min_idx]:
                min_idx = j

        # Step 4: Swap the found minimum element with the element at index i
        arr[i], arr[min_idx] = arr[min_idx], arr[i]

        # We print the array after each swap to show the progression of sorting
        print(f"After swapping index {i} and {min_idx}: {arr}")

    # Step 5: Return the sorted array
    return arr

# Calling the selection_sort function and passing the 'numbers' list to it
sorted_numbers = selection_sort(numbers)

# Print the sorted list
print("Sorted List:", sorted_numbers)

# Explanation of the output:
# 1. Selection Sort finds the smallest element in the unsorted portion and places it in the correct position.
# 2. This process is repeated for each element, shrinking the unsorted portion until the list is fully sorted.
# 3. Selection Sort works by selecting the smallest element from the unsorted section and swapping it with the
#    element at the current position, ensuring that the list is sorted progressively from left to right.

# Let's now understand the result of sorting and what happens to the list after each pass:
#
# - Initially, the unsorted list was [64, 25, 12, 22, 11]
# - After each selection, the smallest unsorted element is placed in its correct position.
# - Eventually, after all passes, the list becomes sorted: [11, 12, 22, 25, 64]
#
# The algorithm guarantees that in each pass, the smallest element is found and moved to its correct position.
```
