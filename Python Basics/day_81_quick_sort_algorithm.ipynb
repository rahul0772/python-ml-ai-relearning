```python
# ===========================
# QUICK SORT (Divide and Conquer Algorithm)
# ===========================
# Quick Sort is a very fast sorting algorithm based on the idea:
# 1. Pick a "pivot" element.
# 2. Rearrange the list so:
#    - elements smaller than pivot go to the left
#    - elements larger than pivot go to the right
# 3. Recursively apply the same process to left and right parts.


def quick_sort(arr):
    # This function takes a list 'arr' and returns a sorted version of it.

    # STEP 1: BASE CASE (Stopping condition)
    # If the list has 0 or 1 element, it is already sorted.
    # Example: [] or [5] → no sorting needed
    if len(arr) <= 1:
        return arr  # Just return it as it is

    # STEP 2: CHOOSE A PIVOT
    # We choose the middle element as pivot (you can choose first/last too)
    pivot = arr[len(arr) // 2]

    # Why middle?
    # It often gives better balance compared to always picking first element.

    # STEP 3: CREATE 3 LISTS
    # These will store elements based on comparison with pivot
    left = []    # Elements smaller than pivot
    middle = []  # Elements equal to pivot
    right = []   # Elements greater than pivot

    # STEP 4: PARTITIONING (Rearranging elements)
    # We go through each element and compare it with pivot
    for x in arr:
        if x < pivot:
            left.append(x)   # Smaller → goes to left
        elif x == pivot:
            middle.append(x) # Equal → stays in middle
        else:
            right.append(x)  # Greater → goes to right

    # Example:
    # arr = [5, 3, 8, 4, 2]
    # pivot = 8 (middle)
    # left = [5, 3, 4, 2]
    # middle = [8]
    # right = []

    # STEP 5: RECURSION
    # Now we sort left and right parts again using same function
    sorted_left = quick_sort(left)
    sorted_right = quick_sort(right)

    # STEP 6: COMBINE RESULTS
    # Final sorted list = sorted_left + middle + sorted_right
    return sorted_left + middle + sorted_right


# ===========================
# DRIVER CODE (Testing Quick Sort)
# ===========================

# Let's define an unsorted list
numbers = [10, 7, 8, 9, 1, 5]

# Print original list
print("Original List:", numbers)

# Call quick_sort function
sorted_numbers = quick_sort(numbers)

# Print sorted list
print("Sorted List:", sorted_numbers)


# ===========================
# EXPECTED OUTPUT:
# ===========================
# Original List: [10, 7, 8, 9, 1, 5]
# Sorted List: [1, 5, 7, 8, 9, 10]


# ===========================
# OUTPUT EXPLANATION:
# ===========================
# Let's understand WHY we got this output:

# 1. The algorithm picked a pivot (middle element).
# 2. It split the list into smaller parts repeatedly:
#    Example breakdown:
#    [10, 7, 8, 9, 1, 5]
#       pivot = 9
#       left = [7, 8, 1, 5]
#       middle = [9]
#       right = [10]

# 3. Then it recursively sorted left side:
#    [7, 8, 1, 5] → becomes [1, 5, 7, 8]

# 4. Right side [10] is already sorted.

# 5. Finally combined everything:
#    [1, 5, 7, 8] + [9] + [10]

# 6. Result:
#    [1, 5, 7, 8, 9, 10]

# So the output is sorted in ascending order.

# ===========================
# KEY IDEA:
# ===========================
# Quick Sort works by:
# "Divide → Sort → Combine"
```

```python
# ===========================
# QUICK SORT WITH RECURSION TRACE (PRINT EVERY STEP)
# ===========================
# In this version, we will PRINT each recursive call step-by-step
# so you can SEE how recursion actually works internally.

def quick_sort(arr, depth=0):
    # 'depth' tells us how deep we are in recursion
    # We use it to format output (indentation) for better understanding

    indent = "  " * depth
    # This creates spaces based on recursion level
    # depth=0 → ""
    # depth=1 → "  "
    # depth=2 → "    "
    # This makes the output look like a tree

    print(f"{indent}quick_sort({arr}) called")
    # This prints current function call
    # Example: quick_sort([10, 7, 8])

    # BASE CASE
    if len(arr) <= 1:
        print(f"{indent}Returning (base case): {arr}")
        return arr
        # If array has 1 or 0 elements → already sorted

    # CHOOSE PIVOT
    pivot = arr[len(arr) // 2]
    print(f"{indent}Pivot chosen: {pivot}")

    # PARTITION
    left = []
    middle = []
    right = []

    for x in arr:
        if x < pivot:
            left.append(x)
        elif x == pivot:
            middle.append(x)
        else:
            right.append(x)

    print(f"{indent}Left: {left}")
    print(f"{indent}Middle: {middle}")
    print(f"{indent}Right: {right}")

    # RECURSION on left side
    print(f"{indent}Recursively sorting LEFT part")
    sorted_left = quick_sort(left, depth + 1)

    # RECURSION on right side
    print(f"{indent}Recursively sorting RIGHT part")
    sorted_right = quick_sort(right, depth + 1)

    # COMBINE
    result = sorted_left + middle + sorted_right

    print(f"{indent}Combined result: {result}")
    print(f"{indent}Returning: {result}")

    return result


# ===========================
# DRIVER CODE
# ===========================

numbers = [10, 7, 8, 9, 1, 5]

print("Original List:", numbers)
print("\n--- RECURSION TRACE START ---\n")

sorted_numbers = quick_sort(numbers)

print("\n--- FINAL RESULT ---")
print("Sorted List:", sorted_numbers)


# ===========================
# EXPECTED OUTPUT (SIMPLIFIED VIEW)
# ===========================
# You will see a TREE-LIKE STRUCTURE such as:
#
# quick_sort([10, 7, 8, 9, 1, 5]) called
# Pivot chosen: 9
# Left: [7, 8, 1, 5]
# ...
#   quick_sort([7, 8, 1, 5]) called
#   Pivot chosen: 1
#   ...
#     quick_sort([]) called
#     Returning (base case): []
#
# This indentation shows how recursion goes deeper and comes back.


# ===========================
# OUTPUT EXPLANATION:
# ===========================
# 1. Each "quick_sort(...)" line shows a function call.
# 2. Indentation (spaces) shows recursion depth:
#    More spaces = deeper recursion.
# 3. You can SEE:
#    - How list is divided
#    - When base case is reached
#    - When results are combined
# 4. Finally, everything merges into:
#    [1, 5, 7, 8, 9, 10]
```
