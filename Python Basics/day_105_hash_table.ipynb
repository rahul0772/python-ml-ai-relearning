```python
# ============================================
# HASH TABLES (DICTIONARIES)
#
# Think of this as a SUPER SMART STORAGE SYSTEM 🧠
# Instead of searching slowly through a list,
# it can find things ALMOST INSTANTLY ⚡
#
# REAL LIFE ANALOGY:
# Imagine a dictionary (book 📖):
# - You don't read every page to find a word ❌
# - You jump directly using the first letter ✅
#
# Hash tables work similarly:
# They convert a "key" into a special number (called a HASH)
# which tells EXACTLY where the value is stored in memory.
#
# ============================================
# 🔹 STEP 1: CREATE A HASH TABLE (DICTIONARY)
# ============================================

# We create a dictionary using {} braces.
# Inside, we store KEY : VALUE pairs
# KEY = what you search with
# VALUE = the data you get

student = {
    "name": "Rahul",     # key = "name", value = "Rahul"
    "age": 21,           # key = "age", value = 21
    "course": "Python"   # key = "course", value = "Python"
}

# 🧠 MENTAL MODEL:
# Think of this like a locker system:
# - "name" is the locker number
# - "Rahul" is what's inside
#
# Instead of searching ALL lockers,
# Python uses HASHING to jump directly to the correct locker.

print("Step 1 - Original Dictionary:", student)


# ============================================
# 🔹 STEP 2: ACCESSING VALUES USING KEYS
# ============================================

# We access values using keys inside square brackets []
# Python internally:
# 1. Takes the key
# 2. Converts it into a hash (a number)
# 3. Jumps to memory location
# 4. Returns value

print("Step 2 - Access Name:", student["name"])   # gets "Rahul"
print("Step 2 - Access Age:", student["age"])     # gets 21


# ============================================
# 🔹 STEP 3: ADDING NEW DATA
# ============================================

# We can add new key-value pairs easily

student["grade"] = "A"

# 🧠 WHY THIS WORKS:
# Python hashes the key "grade"
# Finds a free space
# Stores ("grade" : "A")

print("Step 3 - After Adding Grade:", student)


# ============================================
# 🔹 STEP 4: MODIFYING DATA
# ============================================

# If key already exists → value gets replaced

student["age"] = 22

# 🧠 LOGIC:
# Python hashes "age"
# Goes to SAME location
# Replaces old value (21 → 22)

print("Step 4 - After Updating Age:", student)


# ============================================
# 🔹 STEP 5: REMOVING DATA
# ============================================

# Using del keyword removes a key-value pair

del student["course"]

# 🧠 LOGIC:
# Python finds "course" via hash
# Deletes that entry

print("Step 5 - After Deleting Course:", student)


# ============================================
# 🔹 STEP 6: SAFE ACCESS USING .get()
# ============================================

# If key does NOT exist, normal access causes ERROR ❌
# Instead use .get() which is SAFE ✅

print("Step 6 - Safe Access Existing Key:", student.get("name"))

# If key doesn't exist → returns None instead of crashing
print("Step 6 - Safe Access Missing Key:", student.get("course"))


# ============================================
# 🔹 STEP 7: ITERATING (LOOPING)
# ============================================

# Loop through keys and values

print("Step 7 - Loop Through Dictionary:")

for key, value in student.items():
    # key = each key in dictionary
    # value = corresponding value
    print("Key:", key, "| Value:", value)


# ============================================
# 🔹 STEP 8: HOW HASHING WORKS (SIMPLIFIED)
# ============================================

# Python has built-in hash() function

key = "name"
hashed_value = hash(key)

# 🧠 IMPORTANT:
# This hash value is used internally to decide storage location

print("Step 8 - Hash of 'name':", hashed_value)


# ============================================
# 🔹 FINAL OUTPUT EXPLANATION
# ============================================

# The output shows:
# 1. Dictionary creation → stores key-value pairs
# 2. Access → instant lookup using hashing
# 3. Add → new key inserted
# 4. Update → existing key overwritten
# 5. Delete → key removed
# 6. Safe access → prevents crashes
# 7. Loop → traverse all data
# 8. Hash → internal magic number used for fast lookup
#
# 🧠 CORE DEEP UNDERSTANDING:
# Hash Table = KEY → HASH FUNCTION → MEMORY LOCATION → VALUE
#
# Instead of searching linearly (slow O(n)),
# it jumps directly (fast O(1)) ⚡
```
