```python
# Problem: We need to identify prime numbers in a list.
# A prime number is a number that is only divisible by 1 and itself.
# For example, 2, 3, 5, 7, 11, 13 are prime numbers.

def is_prime(num):  # We define a function called is_prime to check if a number is prime.
    if num < 2:  # Check if the number is less than 2. Prime numbers are greater than or equal to 2.
        return False  # If the number is less than 2, it’s not prime, so return False.

    # Now we will check if the number can be divided by any number other than 1 and itself.
    for i in range(2, num):  # Start a loop from 2 to num-1 (because a number cannot divide by 1 and itself).
        if num % i == 0:  # Check if num is divisible by any number from 2 to num-1.
            return False  # If it’s divisible, it’s not prime, return False.
    return True  # If no division happened, the number is prime, so return True.

def count_primes(numbers):  # We define another function called count_primes that takes a list of numbers.
    prime_numbers = []  # We create an empty list to store prime numbers.
    prime_count = 0  # Initialize a counter to count prime numbers.

    for num in numbers:  # Loop through each number in the list.
        if is_prime(num):  # Call the is_prime function to check if the current number is prime.
            prime_numbers.append(num)  # If the number is prime, add it to the prime_numbers list.
            prime_count += 1  # Increment the prime_count by 1.

    # Now we will print the results.
    print(f"Prime numbers in the list: {prime_numbers}")  # Print the list of prime numbers found.
    print(f"Total number of primes: {prime_count}")  # Print the total count of prime numbers.
    return prime_numbers, prime_count  # Return the list of primes and the count.

# Example usage of the function:
numbers_list = [10, 15, 2, 7, 19, 33, 8, 13, 50, 29]
count_primes(numbers_list)  # Call the function with a list of numbers to find primes.

# Expected output:
# Prime numbers in the list: [2, 7, 19, 13, 29]
# Total number of primes: 5
```
