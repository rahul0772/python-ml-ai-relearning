```python
# ============================================================
# HASH TABLE
# ============================================================

# ---------------------------
# 1. What is a Hash Table?
# ---------------------------

# A hash table is a special way to store data.
# It is like a magic "dictionary" where each item has a "key" and a "value".
# - Key: a name or label for the data
# - Value: the data itself
# Example:
#   Key: "age" -> Value: 10
#   Key: "name" -> Value: "Alice"

# Why do we use it?
# - Because it is super fast to find things.
# - Imagine a real dictionary: you can quickly find a word without reading the whole book.
# - Hash tables do the same in computers.

# In Python, the built-in hash table is called a "dictionary" (dict).

# ============================================================
# 2. Python Dictionary Example (Hash Table)
# ============================================================

# Create an empty hash table (dictionary)
my_hash_table = {}  # {} means empty dictionary

# Add some items to our hash table
my_hash_table["name"] = "Alice"  # key = "name", value = "Alice"
my_hash_table["age"] = 10        # key = "age", value = 10
my_hash_table["city"] = "Paris"  # key = "city", value = "Paris"

# Print the entire hash table
print("Full Hash Table:", my_hash_table)

# Accessing values using keys
print("Name:", my_hash_table["name"])  # prints "Alice"
print("Age:", my_hash_table["age"])    # prints 10

# ------------------------------------------------------------
# How it works:
# - my_hash_table["name"] looks for the key "name" in the hash table
# - Hash table finds the exact location quickly
# - Returns the value "Alice"

# ============================================================
# 3. Adding, Updating, and Removing Items
# ============================================================

# Adding a new item
my_hash_table["country"] = "France"
print("After adding country:", my_hash_table)

# Updating an existing item
my_hash_table["age"] = 11  # change age from 10 to 11
print("After updating age:", my_hash_table)

# Removing an item
del my_hash_table["city"]  # removes key "city" and its value
print("After deleting city:", my_hash_table)

# ------------------------------------------------------------
# Note:
# - Adding: my_hash_table[new_key] = new_value
# - Updating: same as adding (key already exists)
# - Deleting: del my_hash_table[key]

# ============================================================
# 4. Checking if a key exists
# ============================================================

if "name" in my_hash_table:
    print("Key 'name' exists in hash table!")

if "city" not in my_hash_table:
    print("Key 'city' does NOT exist (we deleted it earlier)")

# ============================================================
# 5. Looping Through Hash Table
# ============================================================

# Loop through keys and values
for key, value in my_hash_table.items():
    # key = "name", "age", "country"
    # value = "Alice", 11, "France"
    print("Key:", key, "| Value:", value)

# ============================================================
# 6. Example: Counting Frequency of Numbers (Classic Hash Table Problem)
# ============================================================

# Suppose we have a list of numbers
numbers = [1, 2, 2, 3, 3, 3, 4]

# We want to count how many times each number appears
# We'll use a hash table (dictionary) for this

frequency = {}  # empty hash table to store counts

for num in numbers:
    # Check if the number is already a key in the hash table
    if num in frequency:
        frequency[num] += 1  # increment count by 1
    else:
        frequency[num] = 1   # first time we see this number, count = 1

print("Frequency of numbers:", frequency)

# ------------------------------------------------------------
# Explanation of this logic:
# 1. Loop through each number in the list
# 2. If the number is already in our hash table, increase its value by 1
# 3. If it's not in the hash table, create a new key with value 1
# 4. At the end, we have counts of all numbers in O(1) time per number
# (This is why hash tables are so fast!)

# ============================================================
# 7. Accessing Frequency Example
# ============================================================

print("Number 2 appears", frequency[2], "times")
print("Number 3 appears", frequency[3], "times")

# ============================================================
# 8. Summary
# ============================================================

# - Hash table = fast dictionary of key:value pairs
# - Key = label, Value = data
# - Operations: Add, Update, Delete, Access, Loop
# - Use case: Counting, storing records, mapping keys to values
# - Python implementation: dictionary {}
```
