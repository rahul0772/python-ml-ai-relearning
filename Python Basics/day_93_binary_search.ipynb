```python
# ============================================================
# BINARY SEARCH (DIVIDE & CONQUER)

#
# 🚨 VERY IMPORTANT RULE:
# Binary Search ONLY works on SORTED data.
#
# ============================================================
# STEP 1: CREATE A SORTED LIST
# ============================================================

# We create a sorted list of numbers
# Why sorted? Because binary search depends on order
arr = [2, 5, 8, 12, 16, 23, 38, 56, 72, 91]

# Let's define the number we want to search
target = 23

# ============================================================
# STEP 2: DEFINE BINARY SEARCH FUNCTION
# ============================================================

def binary_search(arr, target):
    # --------------------------------------------------------
    # "low" = starting index of search space
    # "high" = ending index of search space
    #
    # Initially:
    # We consider the WHOLE list
    # --------------------------------------------------------
    low = 0
    high = len(arr) - 1  # last index

    # --------------------------------------------------------
    # WHY while low <= high?
    #
    # Because as long as there is SOME search space left,
    # we continue searching.
    #
    # If low > high → means search space is empty
    # --------------------------------------------------------
    while low <= high:

        # ----------------------------------------------------
        # FIND THE MIDDLE INDEX
        #
        # Formula:
        # mid = (low + high) // 2
        #
        # WHY // ?
        # Because we want integer index (no decimals)
        #
        # Example:
        # low = 0, high = 9 → mid = 4
        # ----------------------------------------------------
        mid = (low + high) // 2

        # ----------------------------------------------------
        # DEBUG PRINT (to see what's happening)
        # ----------------------------------------------------
        print(f"Checking middle index {mid}, value = {arr[mid]}")

        # ----------------------------------------------------
        # CASE 1: FOUND THE TARGET
        # ----------------------------------------------------
        if arr[mid] == target:
            # 🎯 We found the element
            return mid

        # ----------------------------------------------------
        # CASE 2: TARGET IS SMALLER → GO LEFT
        #
        # If target < middle value,
        # we IGNORE right half completely
        #
        # WHY?
        # Because list is sorted → everything on right is bigger
        # ----------------------------------------------------
        elif arr[mid] > target:
            high = mid - 1  # shrink to LEFT half

        # ----------------------------------------------------
        # CASE 3: TARGET IS BIGGER → GO RIGHT
        #
        # If target > middle value,
        # we IGNORE left half completely
        #
        # WHY?
        # Because everything on left is smaller
        # ----------------------------------------------------
        else:
            low = mid + 1  # shrink to RIGHT half

    # --------------------------------------------------------
    # If we exit loop → element NOT found
    # --------------------------------------------------------
    return -1


# ============================================================
# STEP 3: CALL FUNCTION
# ============================================================

result = binary_search(arr, target)

# ============================================================
# STEP 4: PRINT RESULT
# ============================================================

if result != -1:
    print(f"\n✅ Element {target} found at index {result}")
else:
    print(f"\n❌ Element {target} not found")


# ============================================================
# 🧠 OUTPUT EXPLANATION (READ THIS VERY CAREFULLY)
# ============================================================
#
# Suppose output was:
#
# Checking middle index 4, value = 16
# Checking middle index 7, value = 56
# Checking middle index 5, value = 23
# ✅ Element 23 found at index 5
#
# ------------------------------------------------------------
# WHAT HAPPENED STEP BY STEP:
# ------------------------------------------------------------
#
# 1️⃣ First iteration:
# low = 0, high = 9
# mid = 4 → value = 16
#
# Compare:
# 23 > 16 → go RIGHT
#
# New range:
# low = 5, high = 9
#
# ------------------------------------------------------------
#
# 2️⃣ Second iteration:
# mid = 7 → value = 56
#
# Compare:
# 23 < 56 → go LEFT
#
# New range:
# low = 5, high = 6
#
# ------------------------------------------------------------
#
# 3️⃣ Third iteration:
# mid = 5 → value = 23
#
# 🎯 FOUND!
#
# ------------------------------------------------------------
# 🧠 DEEP MENTAL MODEL:
#
# Each step CUTS THE SEARCH SPACE IN HALF.
#
# Instead of checking 10 elements:
# → we checked only 3 steps.
#
# That is why Binary Search is VERY FAST:
#
# Time Complexity:
# O(log n)
#
# Meaning:
# Even if list has 1,000,000 elements,
# you find result in ~20 steps.
#
# ------------------------------------------------------------
#
#
# Binary Search = "Divide into half again and again"
#
# ALWAYS REMEMBER:
# ✔ Data must be sorted
# ✔ Use middle
# ✔ Eliminate half each step
#
# ============================================================
```

```python
# ============================================================
# BINARY SEARCH - CONTINUATION
# ============================================================


# ============================================================
# STEP 1: CREATE A NEW SORTED LIST OF NUMBERS
# ============================================================

# This is the same approach as before, but with some extra complexity.
# We will search for multiple numbers in the same list to see how
# the binary search behaves when given multiple targets.
arr = [1, 3, 5, 7, 9, 11, 15, 18, 23, 30, 34, 40, 45, 50]

# Now we have a list with multiple elements. We're going to search
# for three different numbers.
targets = [23, 11, 100]  # Let's search for 23, 11, and 100

# ============================================================
# STEP 2: DEFINE BINARY SEARCH FUNCTION (AGAIN)
# ============================================================

# The binary search function will stay almost the same, but this time
# we will loop through multiple targets to show how Binary Search
# handles each one independently.

def binary_search(arr, target):
    low = 0
    high = len(arr) - 1

    while low <= high:
        mid = (low + high) // 2

        # Printing for clarity (this helps us see what's happening step-by-step)
        print(f"Checking middle index {mid}, value = {arr[mid]}")

        if arr[mid] == target:
            return mid  # Found the target, return index

        elif arr[mid] > target:
            high = mid - 1  # Go left

        else:
            low = mid + 1  # Go right

    # If not found, return -1 indicating failure
    return -1

# ============================================================
# STEP 3: SEARCH FOR MULTIPLE TARGETS
# ============================================================

# We'll use a loop to call binary_search for each target.
for target in targets:
    print(f"\n🔍 Searching for {target}...")
    result = binary_search(arr, target)

    # After the search is completed, we check and print the result
    if result != -1:
        print(f"✅ Element {target} found at index {result}")
    else:
        print(f"❌ Element {target} not found")


# ============================================================
# STEP 4: OUTPUT EXPLANATION
# ============================================================
#
# Here’s what happens when you run the above code:
#
# Let's assume this is the output:
#
# 🔍 Searching for 23...
# Checking middle index 6, value = 15
# Checking middle index 9, value = 30
# Checking middle index 7, value = 23
# ✅ Element 23 found at index 7
#
# 🔍 Searching for 11...
# Checking middle index 6, value = 15
# Checking middle index 4, value = 9
# Checking middle index 5, value = 11
# ✅ Element 11 found at index 5
#
# 🔍 Searching for 100...
# Checking middle index 6, value = 15
# Checking middle index 9, value = 30
# Checking middle index 11, value = 40
# Checking middle index 12, value = 45
# Checking middle index 13, value = 50
# ❌ Element 100 not found
#
# ------------------------------------------------------------
# Let's break down each search:

# 🔍 Searching for 23:
# The binary search starts by looking at the middle index (index 6, value 15).
# Since 23 is greater than 15, we move to the right half.
# We then look at the middle of the right half (index 9, value 30).
# Since 23 is less than 30, we now check the middle of the left side (index 7, value 23).
# Bingo! We found the number at index 7.

# 🔍 Searching for 11:
# First, the middle element is 15 (index 6), which is greater than 11, so we move left.
# The next middle element is 9 (index 4), which is smaller than 11, so we move right.
# We land on index 5, where the value is 11. We’ve found the target.

# 🔍 Searching for 100:
# The middle element is 15, but 100 is larger, so we move right.
# We repeat the process, checking middle elements, but none are 100.
# Eventually, we’ve checked the whole list and found that 100 isn’t present.
# Thus, we return -1 (not found).

# ============================================================
# 🧠 DEEP MENTAL MODEL:
# ============================================================
# Binary Search works by repeatedly eliminating half of the search space.
# In each iteration:
# - If the target is smaller, we eliminate the right half.
# - If the target is larger, we eliminate the left half.
#
# Let's look at time complexity again:
# - Best case: O(1) if we find the target at the middle right away.
# - Worst case: O(log n) because we keep halving the search space.

# This makes Binary Search **super efficient**, even with large datasets.
# For example, if you had a list with 1 million numbers, you'd find the result in only about 20 steps (log₂(1,000,000)).

# Binary Search also has an important property: it doesn't just work for numbers.
# It works for ANY SORTED data! So you could use it on strings, dates, or anything that can be ordered.

# ============================================================
```
