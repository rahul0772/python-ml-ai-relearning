```python
# Bubble Sort is a basic sorting algorithm that works by repeatedly swapping adjacent elements if they are in the wrong order.
# The idea is that, in each pass through the list, the largest unsorted element "bubbles" up to its correct position.

# Below is an ultra-detailed breakdown of the Bubble Sort algorithm in Python.
# I'll explain every single line of code, so you can understand the logic and how it works.

def bubble_sort(arr):
    """
    This function implements the Bubble Sort algorithm to sort the elements in the list in ascending order.
    It repeatedly compares adjacent elements and swaps them if they are in the wrong order.
    The process is repeated for all elements in the list until no more swaps are needed.
    """

    n = len(arr)  # Get the length of the list.

    # Outer loop - Iterate over the entire list.
    # The outer loop makes multiple passes through the list until the list is sorted.
    for i in range(n):

        # We use a flag 'swapped' to track if any elements were swapped in the current pass.
        # If no elements were swapped, it means the list is already sorted, and we can stop early.
        swapped = False

        # Inner loop - Go through the list from the beginning to the second-to-last element.
        # We compare each element with the next one and swap if needed.
        for j in range(0, n-i-1):  # n-i-1 ensures that after each pass, the last element is already sorted.

            # If the current element is greater than the next, we swap them.
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]  # Swap the elements.

                # Set 'swapped' to True to indicate that a swap was made.
                swapped = True

        # If no elements were swapped in this pass, the list is already sorted, so break the loop early.
        if not swapped:
            break

# Let's test our Bubble Sort function with an unsorted list of numbers.
arr = [64, 34, 25, 12, 22, 11, 90]

# Calling the bubble_sort function with the array.
bubble_sort(arr)

# Let's print the sorted array and explain the output:
print("Sorted array:", arr)

# Explanation of Output:
# We started with the unsorted list: [64, 34, 25, 12, 22, 11, 90]
# The Bubble Sort algorithm works by repeatedly swapping adjacent elements if they are in the wrong order.
# It does this for multiple passes through the list until the list is fully sorted.

# On the first pass, the largest element (90) "bubbles" up to the last position, and the list looks like this:
# [34, 25, 12, 22, 11, 64, 90]

# On the second pass, the next largest element (64) bubbles up to its correct position:
# [25, 12, 22, 11, 34, 64, 90]

# This continues, with the next largest elements bubbling to their correct positions in each pass, until the entire list is sorted.
# The final sorted list is: [11, 12, 22, 25, 34, 64, 90]

# Bubble Sort is inefficient for large lists because it makes many unnecessary comparisons.
# However, it's easy to understand and can be useful for small lists or educational purposes.

# Let's walk through the key parts of the code:

# 1. **`n = len(arr)`**:
#    - This line calculates the length of the list, which tells us how many elements we need to sort.
#    - The outer loop will run `n` times, but after each pass, the largest unsorted element is in its correct place, so we reduce the range of the inner loop by `n-i-1`.

# 2. **Outer loop (`for i in range(n):`)**:
#    - This loop runs through the list multiple times, each time bringing the largest unsorted element to its correct position.
#    - With each pass, the range of the inner loop shrinks because the largest element is already in place.

# 3. **Inner loop (`for j in range(0, n-i-1):`)**:
#    - The inner loop compares adjacent elements in the list. If the current element is greater than the next one, they are swapped.
#    - The `n-i-1` ensures that as we bubble larger elements to the end, we don't compare them again.

# 4. **Swap (`arr[j], arr[j+1] = arr[j+1], arr[j]`)**:
#    - If an element at index `j` is greater than the element at index `j+1`, the two elements are swapped.
#    - This step is crucial for "bubbling" the larger elements to the end of the list.

# 5. **Optimization with `swapped` flag**:
#    - The `swapped` flag is a performance optimization. If no elements were swapped during a pass, it means the list is already sorted, and we can exit early.
#    - This saves unnecessary passes when the list is already sorted.

# The final output is the sorted list, which should be:
# [11, 12, 22, 25, 34, 64, 90]
```
