```python
# Bubble Sort is a basic sorting algorithm that works by repeatedly swapping adjacent elements if they are in the wrong order.
# The idea is that, in each pass through the list, the largest unsorted element "bubbles" up to its correct position.

# Below is an ultra-detailed breakdown of the Bubble Sort algorithm in Python.
# I'll explain every single line of code, so you can understand the logic and how it works.

def bubble_sort(arr):
    """
    This function implements the Bubble Sort algorithm to sort the elements in the list in ascending order.
    It repeatedly compares adjacent elements and swaps them if they are in the wrong order.
    The process is repeated for all elements in the list until no more swaps are needed.
    """

    n = len(arr)  # Get the length of the list.

    # Outer loop - Iterate over the entire list.
    # The outer loop makes multiple passes through the list until the list is sorted.
    for i in range(n):

        # We use a flag 'swapped' to track if any elements were swapped in the current pass.
        # If no elements were swapped, it means the list is already sorted, and we can stop early.
        swapped = False

        # Inner loop - Go through the list from the beginning to the second-to-last element.
        # We compare each element with the next one and swap if needed.
        for j in range(0, n-i-1):  # n-i-1 ensures that after each pass, the last element is already sorted.

            # If the current element is greater than the next, we swap them.
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]  # Swap the elements.

                # Set 'swapped' to True to indicate that a swap was made.
                swapped = True

        # If no elements were swapped in this pass, the list is already sorted, so break the loop early.
        if not swapped:
            break

# Let's test our Bubble Sort function with an unsorted list of numbers.
arr = [64, 34, 25, 12, 22, 11, 90]

# Calling the bubble_sort function with the array.
bubble_sort(arr)

# Let's print the sorted array and explain the output:
print("Sorted array:", arr)

# Explanation of Output:
# We started with the unsorted list: [64, 34, 25, 12, 22, 11, 90]
# The Bubble Sort algorithm works by repeatedly swapping adjacent elements if they are in the wrong order.
# It does this for multiple passes through the list until the list is fully sorted.

# On the first pass, the largest element (90) "bubbles" up to the last position, and the list looks like this:
# [34, 25, 12, 22, 11, 64, 90]

# On the second pass, the next largest element (64) bubbles up to its correct position:
# [25, 12, 22, 11, 34, 64, 90]

# This continues, with the next largest elements bubbling to their correct positions in each pass, until the entire list is sorted.
# The final sorted list is: [11, 12, 22, 25, 34, 64, 90]

# Bubble Sort is inefficient for large lists because it makes many unnecessary comparisons.
# However, it's easy to understand and can be useful for small lists or educational purposes.

# Let's walk through the key parts of the code:

# 1. **`n = len(arr)`**:
#    - This line calculates the length of the list, which tells us how many elements we need to sort.
#    - The outer loop will run `n` times, but after each pass, the largest unsorted element is in its correct place, so we reduce the range of the inner loop by `n-i-1`.

# 2. **Outer loop (`for i in range(n):`)**:
#    - This loop runs through the list multiple times, each time bringing the largest unsorted element to its correct position.
#    - With each pass, the range of the inner loop shrinks because the largest element is already in place.

# 3. **Inner loop (`for j in range(0, n-i-1):`)**:
#    - The inner loop compares adjacent elements in the list. If the current element is greater than the next one, they are swapped.
#    - The `n-i-1` ensures that as we bubble larger elements to the end, we don't compare them again.

# 4. **Swap (`arr[j], arr[j+1] = arr[j+1], arr[j]`)**:
#    - If an element at index `j` is greater than the element at index `j+1`, the two elements are swapped.
#    - This step is crucial for "bubbling" the larger elements to the end of the list.

# 5. **Optimization with `swapped` flag**:
#    - The `swapped` flag is a performance optimization. If no elements were swapped during a pass, it means the list is already sorted, and we can exit early.
#    - This saves unnecessary passes when the list is already sorted.

# The final output is the sorted list, which should be:
# [11, 12, 22, 25, 34, 64, 90]
```

```python
# Let's dive into a more advanced and deeply detailed explanation of **Insertion Sort**,
# which is another fundamental sorting algorithm. It’s more efficient than Bubble Sort in practice,
# but still not the most optimal algorithm for large datasets. I will explain **each line of code**
# with extreme detail and show how the algorithm works step-by-step.

# Insertion Sort works similarly to how we might sort playing cards in our hands.
# We take the first card (or element) and assume it’s already sorted.
# Then, we take the next card and insert it into its correct position among the already sorted cards.
# We keep doing this until all cards are sorted.

def insertion_sort(arr):
    """
    This function implements the Insertion Sort algorithm. It sorts the input list `arr` in ascending order.
    The algorithm works by iterating over the list and "inserting" each element into its correct position
    in the already-sorted portion of the list.
    """

    # Step 1: Start iterating from the second element, because the first element is trivially sorted.
    for i in range(1, len(arr)):  # Loop starts from the second element (index 1) and goes to the end of the list.

        key = arr[i]  # 'key' is the current element that we are going to insert into the sorted portion of the list.

        # Step 2: We need to find the correct position for 'key' in the sorted part of the list.
        j = i - 1  # Start from the element just before 'key'.

        # Step 3: Shift all elements in the sorted portion that are greater than 'key' one position to the right.
        # This is done in a while loop, which runs as long as there are elements that are greater than the 'key'.
        while j >= 0 and arr[j] > key:
            arr[j + 1] = arr[j]  # Shift arr[j] one position to the right to make space for 'key'.
            j -= 1  # Move leftwards in the sorted portion.

        # Step 4: Insert the 'key' into its correct position.
        arr[j + 1] = key  # Place 'key' at the correct position.

# Now, let's test the insertion_sort function with an unsorted list.
arr = [12, 11, 13, 5, 6]  # The list we want to sort.

# Call the insertion_sort function to sort the array.
insertion_sort(arr)

# Print the sorted array to see the result.
print("Sorted array:", arr)

# Explanation of Output:
# We started with the unsorted list: [12, 11, 13, 5, 6]
# The Insertion Sort algorithm works by iterating through the list and inserting each element into its correct position
# in the already sorted portion of the list.

# Let's break down the sorting process:
# - In the first iteration (i = 1), the key is 11. We compare it with 12 (the element before it).
# - 11 is smaller than 12, so we shift 12 to the right and place 11 in the first position.
# - The list becomes: [11, 12, 13, 5, 6].

# - In the second iteration (i = 2), the key is 13. Since 13 is greater than 12 (the element before it),
#   no elements are shifted, and 13 stays in its place.
#   The list remains: [11, 12, 13, 5, 6].

# - In the third iteration (i = 3), the key is 5. We compare 5 with 13, then 12, and then 11.
#   We shift all three elements to the right to make space for 5, and 5 is placed in the first position.
#   The list becomes: [5, 11, 12, 13, 6].

# - In the final iteration (i = 4), the key is 6. We compare it with 13, then 12, and then 11.
#   We shift 13, 12, and 11 to the right, and place 6 in the correct position.
#   The final sorted list is: [5, 6, 11, 12, 13].

# The output printed will be: "Sorted array: [5, 6, 11, 12, 13]"

# Now, let's take a deep dive into the **mental model** behind this code:
# 1. **Outer loop (`for i in range(1, len(arr))`)**:
#    - The outer loop starts from the second element (index 1) because we consider the first element as sorted.
#    - As the loop progresses, the sublist arr[0:i] is always sorted. Our job is to insert arr[i] into that sorted sublist.

# 2. **Key element (`key = arr[i]`)**:
#    - For each iteration, we take the element `arr[i]` as the key. This is the element that will be inserted into its correct position in the sorted portion of the list.
#    - The idea is that we "take" the key, compare it with previous elements, and place it in the right spot.

# 3. **Inner loop (`while j >= 0 and arr[j] > key`)**:
#    - The inner loop is responsible for shifting elements to the right to make space for the key.
#    - We continue shifting elements as long as they are greater than the key and `j` is not out of bounds.
#    - For each comparison, we move the element at `arr[j]` one position to the right (`arr[j + 1] = arr[j]`).
#    - The purpose is to create space in the list so the key can be inserted in its proper position.

# 4. **Insert key into sorted portion (`arr[j + 1] = key`)**:
#    - Once the inner loop finishes (i.e., the correct position for the key is found), we insert the key at that position.
#    - The index `j + 1` is the correct position for the key because `j` is the last position where `arr[j]` was greater than the key.

# **Why Insertion Sort Works**:
# - Insertion Sort essentially builds a sorted list one element at a time.
# - Initially, the sorted portion is just the first element.
# - On each iteration, it places the next element in its correct position in the sorted list, shifting elements to the right if needed.
# - This is why it’s called “insertion” sort—it’s like inserting the key into the right place in the sorted portion of the list.

# **Time Complexity**:
# - Best case: O(n) (if the list is already sorted)
# - Average and worst case: O(n²) (because of the inner loop that shifts elements)

# Let’s look at how it behaves with a list of integers:

# The final output is the sorted list:
# [5, 6, 11, 12, 13]
```
