```python
# creating multiple instances
class Dog:
    # Constructor: Runs every time you create a new Dog
    def __init__(self, name, age):
        self.name = name
        self.age = age

    # Defines behavior all dogs can perform.
    def sit(self):
        print(self.name.title() + " is now sitting.")

#Creating multiple instances - Two separate objects, same class.
my_dog = Dog('willie', 6)
your_dog = Dog('lucy', 3)

print("My dog's name is " + my_dog.name.title() + ".")
print("My dog is " + str(my_dog.age) + " years old.")
my_dog.sit()

print("\nYour dog's name is " + your_dog.name.title() + ".")
print("Your dog is " + str(your_dog.age) + " years old.")
your_dog.sit()

# we create a dog named Willie and a dog named Lucy.
# Each dog is a separate instance with its own set of attributes, capable of the
# same set of actions
```

```python
# This example shows how classes and instances work in Python

# THE CAR CLASS
# A class is a blueprint for creating objects (cars, in this case)
class Car:
    # This docstring explains what the class represents
    '''A simple attempt to represent a car'''

    # This method runs automatically when a new Car object is created
    def __init__(self, make, model, year):
        # These are attributes that belong to each individual car
        self.make = make    # brand of the car (e.g. audi)
        self.model = model  # model of the car (e.g. a4)
        self.year = year    # year the car was made (e.g. 2016)

    # This method returns a nicely formatted description of the car
    def get_descriptive_name(self):
        # Combine the year, make, and model into one string
        long_name = str(self.year) + " " + self.make + " " + self.model
        # Return the string with proper capitalization
        return long_name.title()


# Create a new Car object using the Car class
# This creates ONE specific car
my_new_car = Car("audi", "a4", 2016)

# Call the method on the object and print the result
print(my_new_car.get_descriptive_name())
```

```python
# modifying an attributes value through a method
# Instead of accessing the attribute directly, you pass the new value to a method that handles the updating internally

class Car:
    def __init__(self, make, model, year):
        """
        This method runs automatically when a new Car object is created.
        It sets the initial values for the car.
        """
        self.make = make                  # Stores the car brand (example: Audi)
        self.model = model                # Stores the car model (example: A4)
        self.year = year                  # Stores the year of the car
        self.odometer_reading = 0          # Starts mileage at 0

    def get_descriptive_name(self):
        """
        This method returns a nicely formatted name of the car.
        """
        return f"{self.year} {self.make} {self.model}"

    def read_odometer(self):
        """
        This method prints the current mileage of the car.
        """
        print(f"This car has {self.odometer_reading} miles on it.")

    def update_odometer(self, mileage):
        """
        This method updates the odometer reading.
        We use a method instead of changing the value directly
        to protect the data and prevent mistakes.
        """
        if mileage >= self.odometer_reading:
            self.odometer_reading = mileage
        else:
            print("You cannot roll back the odometer!")

my_new_car = Car('audi', 'a4', 2016)
print(my_new_car.get_descriptive_name())

my_new_car.update_odometer(23)
my_new_car.read_odometer()
```

```python
# extend the method update_odometer() to do additional work every time the odometer reading is modified

class Car:
    def __init__(self, make, model, year):
        # Store basic information about the car
        self.make = make
        self.model = model
        self.year = year

        # Start the odometer at 0 miles
        self.odometer_reading = 0

    def update_odometer(self, mileage):
        """
        Update the car's mileage.
        Do NOT allow the mileage to go backward.
        """

        # Check if the new mileage is greater than or equal to current mileage
        if mileage >= self.odometer_reading:
            self.odometer_reading = mileage
        else:
            print("You can't roll back an odometer!")

    def read_odometer(self):
        # Display the current mileage
        print(f"This car has {self.odometer_reading} miles on it.")

my_car = Car("Audi", "A4", 2016)

my_car.update_odometer(100)   # ✅ works
my_car.read_odometer()

my_car.update_odometer(50)    # ❌ blocked
my_car.read_odometer()
```

```python
# Incrementing an Attribute’s Value Through a Method

class Car:
    def __init__(self, make, model, year):
        # Store car details
        self.make = make
        self.model = model
        self.year = year

        # Start odometer at 0
        self.odometer_reading = 0

    def get_descriptive_name(self):
        # Return a nicely formatted car name
        return f"{self.year} {self.make.title()} {self.model.title()}"

    def read_odometer(self):
        # Display current mileage
        print(f"This car has {self.odometer_reading} miles on it.")

    def update_odometer(self, mileage):
        """
        Set the odometer to a specific value.
        Prevent rolling the odometer backward.
        """
        if mileage >= self.odometer_reading:
            self.odometer_reading = mileage
        else:
            print("You can't roll back an odometer!")

    def increment_odometer(self, miles):
        """
        Add miles to the current odometer reading.
        Used when driving the car more.
        """
        if miles > 0:
            self.odometer_reading += miles
        else:
            print("Miles added must be positive.")

my_used_car = Car("subaru", "outback", 2013)

print(my_used_car.get_descriptive_name())

my_used_car.update_odometer(23500)
my_used_car.read_odometer()

my_used_car.increment_odometer(100)
my_used_car.read_odometer()
```

## Inheritance

```python
# INHERITANCE EXAMPLE
# A child class (ElectricCar) inherits from a parent class (Car)
# lets you create a new class based on an existing class, instead of starting from nothing
#A Car has wheels, a steering wheel, and an engine.
### An Electric Car is still a car, but it also has a battery and charging features.

class Car:
    """A simple attempt to represent a car."""

    def __init__(self, make, model, year):
        # Initialize attributes common to ALL cars
        self.make = make
        self.model = model
        self.year = year
        self.odometer_reading = 0  # Every new car starts at 0 miles

    def get_descriptive_name(self):
        # Return a nicely formatted name for the car
        return f"{self.year} {self.make} {self.model}".title()

    def read_odometer(self):
        # Show the car's current mileage
        print(f"This car has {self.odometer_reading} miles on it.")

    def update_odometer(self, mileage):
        # Set the odometer to a specific value (no rollback allowed)
        if mileage >= self.odometer_reading:
            self.odometer_reading = mileage
        else:
            print("You can't roll back an odometer!")

    def increment_odometer(self, miles):
        # Add miles to the current odometer reading
        self.odometer_reading += miles


class ElectricCar(Car): # By writing (Car) in parentheses, you tell Python:
#“ElectricCar is a type of Car. Give it all the features of Car automatically.”
    """
    Represent aspects of a car that are specific to electric vehicles.
    This class INHERITS everything from the Car class.
    From Car, it automatically gets:
self.make
self.model
self.year
self.odometer_reading

And all the methods:
get_descriptive_name()
read_odometer()
update_odometer()
increment_odometer()
    """

    def __init__(self, make, model, year):
        # Call the parent class's __init__ method
        # This sets make, model, year, and odometer_reading
        # super() gives the child class access to the parent class
        # super().__init__(make, model, year) runs the Car class constructor.
        # This means ElectricCar automatically sets make, model, year, and odometer_reading, so you don’t have to write it again.
        super().__init__(make, model, year)


# Create an ElectricCar object
my_tesla = ElectricCar("tesla", "model s", 2016)

# Use a method inherited from the Car class
print(my_tesla.get_descriptive_name())
```

```python
# Defining Attributes and Methods for the Child Class

# PARENT CLASS
class Car:
    """A simple attempt to represent a general car."""

    def __init__(self, make, model, year):
        # Initialize attributes common to all cars
        self.make = make
        self.model = model
        self.year = year
        self.odometer_reading = 0  # All cars start with 0 miles

    def get_descriptive_name(self):
        # Return a nicely formatted descriptive name for the car
        return f"{self.year} {self.make.title()} {self.model.title()}"

    def read_odometer(self):
        # Print the current mileage of the car
        print(f"This car has {self.odometer_reading} miles on it.")

    def update_odometer(self, mileage):
        # Set the odometer reading to a specific value
        # Reject any attempt to roll it back
        if mileage >= self.odometer_reading:
            self.odometer_reading = mileage
        else:
            print("You can't roll back an odometer!")

    def increment_odometer(self, miles):
        # Add miles to the current odometer reading
        if miles > 0:
            self.odometer_reading += miles
        else:
            print("Miles added must be positive.")


# CHILD CLASS
class ElectricCar(Car):
    """Represents aspects of a car specific to electric vehicles."""

    def __init__(self, make, model, year):
        """
        Initialize attributes of the parent class (Car) using super().
        Then initialize attributes specific to ElectricCar.
        """
        super().__init__(make, model, year)  # Initialize parent class attributes
        self.battery_size = 70  # New attribute specific to electric cars

    def describe_battery(self):
        """Print a statement describing the battery size."""
        print(f"This car has a {self.battery_size}-kWh battery.")


# CREATING AND USING AN ELECTRIC CAR OBJECT
my_tesla = ElectricCar('tesla', 'model s', 2016)

# Use a method inherited from Car
print(my_tesla.get_descriptive_name())

# Use the method specific to ElectricCar
my_tesla.describe_battery()
```

```python
# Overriding Methods from the Parent Class

# PARENT CLASS
class Car:
    """A simple attempt to represent a car."""

    def __init__(self, make, model, year):
        self.make = make
        self.model = model
        self.year = year
        self.odometer_reading = 0

    def get_descriptive_name(self):
        return f"{self.year} {self.make.title()} {self.model.title()}"

    def fill_gas_tank(self):
        """Fill the gas tank for a regular car."""
        print("The gas tank is now full.")


# CHILD CLASS
class ElectricCar(Car):
    """Represents aspects of a car specific to electric vehicles."""

    def __init__(self, make, model, year, battery_size=70):
        super().__init__(make, model, year)
        self.battery_size = battery_size

    def describe_battery(self):
        print(f"This car has a {self.battery_size}-kWh battery.")

    # METHOD OVERRIDING
    def fill_gas_tank(self):
        """
        Electric cars don't have gas tanks.
        This overrides the fill_gas_tank() method in the Car class.
        """
        print("This car doesn't need a gas tank!")


# TESTING
my_tesla = ElectricCar("tesla", "model s", 2016)

# Method from parent class (inherited) - works normally if not overridden
print(my_tesla.get_descriptive_name())  # inherited

# Method specific to ElectricCar
my_tesla.describe_battery()

# Method overridden in ElectricCar
my_tesla.fill_gas_tank()  # runs the overridden method, NOT the parent
```

```python
"""
Instances as Attributes

OVERVIEW:

This example demonstrates two key Object-Oriented Programming concepts:

1. Inheritance:
   - ElectricCar is a child class of Car.
   - It automatically gets all attributes and methods from Car (like make, model, year, odometer_reading, get_descriptive_name()).
   - We can also add child-specific attributes or methods, like battery-related functionality.

2. Composition (Instances as Attributes):
   - Instead of putting all battery details inside ElectricCar, we create a separate Battery class.
   - Each ElectricCar instance has a Battery instance stored as an attribute (self.battery).
   - This keeps the code modular, clean, and realistic: ElectricCar "has a" Battery.

3. How it works in practice:
   - Creating my_tesla = ElectricCar(...) automatically creates a Car + Battery.
   - We can call parent methods (get_descriptive_name) or Battery methods (describe_battery) through the ElectricCar instance.
   - This demonstrates combining inheritance and composition for organized, maintainable code.
"""

# PARENT CLASS
class Car:
    """A simple attempt to represent a general car."""

    def __init__(self, make, model, year):
        # Initialize attributes common to all cars
        self.make = make
        self.model = model
        self.year = year
        self.odometer_reading = 0  # All cars start with 0 miles

    def get_descriptive_name(self):
        # Return a nicely formatted descriptive name for the car
        return f"{self.year} {self.make.title()} {self.model.title()}"

    def read_odometer(self):
        # Print the current mileage of the car
        print(f"This car has {self.odometer_reading} miles on it.")


# BATTERY CLASS (SEPARATE CLASS)
class Battery:
    """A simple attempt to model a battery for an electric car."""

    def __init__(self, battery_size=70):
        """
        Initialize the battery's attributes.
        Default battery_size is 70 if not specified.
        """
        self.battery_size = battery_size

    def describe_battery(self):
        """Print a statement describing the battery size."""
        print(f"This car has a {self.battery_size}-kWh battery.")


# CHILD CLASS
class ElectricCar(Car):
    """Represents aspects of a car specific to electric vehicles."""

    def __init__(self, make, model, year):
        """
        Initialize attributes of the parent class using super().
        Then initialize attributes specific to an electric car.
        """
        super().__init__(make, model, year)
        self.battery = Battery()  # Create a Battery instance as an attribute


# CREATING AND USING AN ELECTRIC CAR OBJECT
my_tesla = ElectricCar('tesla', 'model s', 2016)

# Use a method inherited from Car
print(my_tesla.get_descriptive_name())

# Access the Battery instance and call its method
my_tesla.battery.describe_battery()
```

```python
"""
OVERVIEW:

This example demonstrates:
1. Inheritance:
   - ElectricCar inherits from Car, getting all its attributes and methods.

2. Composition (Instances as Attributes):
   - ElectricCar has a Battery instance stored as an attribute (self.battery).
   - Battery handles all battery-specific details.

3. New functionality with get_range():
   - Battery can now calculate and report the driving range based on battery size.
   - ElectricCar can use this without cluttering its own class.
"""

# PARENT CLASS
class Car:
    """A simple attempt to represent a general car."""

    def __init__(self, make, model, year):
        # Initialize attributes common to all cars
        self.make = make
        self.model = model
        self.year = year
        self.odometer_reading = 0  # All cars start with 0 miles

    def get_descriptive_name(self):
        # Return a nicely formatted descriptive name for the car
        return f"{self.year} {self.make.title()} {self.model.title()}"

    def read_odometer(self):
        # Print the current mileage of the car
        print(f"This car has {self.odometer_reading} miles on it.")


# BATTERY CLASS (SEPARATE CLASS)
class Battery:
    """A simple attempt to model a battery for an electric car."""

    def __init__(self, battery_size=70):
        """
        Initialize the battery's attributes.
        Default battery_size is 70 if not specified.
        """
        self.battery_size = battery_size

    def describe_battery(self):
        """Print a statement describing the battery size."""
        print(f"This car has a {self.battery_size}-kWh battery.")

    def get_range(self):
        """
        Print a statement about the range this battery provides.
        Simple logic: 70 kWh -> 240 miles, 85 kWh -> 270 miles.
        """
        if self.battery_size == 70:
            range_miles = 240
        elif self.battery_size == 85:
            range_miles = 270
        else:
            # For other battery sizes, estimate range roughly
            range_miles = self.battery_size * 3

        message = f"This car can go approximately {range_miles} miles on a full charge."
        print(message)


# CHILD CLASS
class ElectricCar(Car):
    """Represents aspects of a car specific to electric vehicles."""

    def __init__(self, make, model, year):
        """
        Initialize attributes of the parent class using super().
        Then initialize attributes specific to an electric car.
        """
        super().__init__(make, model, year)
        self.battery = Battery()  # Create a Battery instance as an attribute


# CREATING AND USING AN ELECTRIC CAR OBJECT
my_tesla = ElectricCar('tesla', 'model s', 2016)

# Use a method inherited from Car
print(my_tesla.get_descriptive_name())

# Access the Battery instance and call its methods
my_tesla.battery.describe_battery()
my_tesla.battery.get_range()
```

```python
'''IMPORTING CLASSES:

- Python lets you keep classes in separate files (modules) to avoid long, cluttered code.
- Use `from module import ClassName` to bring in the class.
- In Google Colab:
    1. Either define the class in the same cell (easy for learning)
    2. Or create a .py file using `%%writefile filename.py` and then import.
'''

class Car:
    """A simple attempt to represent a car."""

    def __init__(self, make, model, year):
        self.make = make
        self.model = model
        self.year = year
        self.odometer_reading = 0

    def get_descriptive_name(self):
        long_name = f"{self.year} {self.make} {self.model}"
        return long_name.title()

    def read_odometer(self):
        print(f"This car has {self.odometer_reading} miles on it.")

    def update_odometer(self, mileage):
        if mileage >= self.odometer_reading:
            self.odometer_reading = mileage
        else:
            print("You can't roll back an odometer!")

    def increment_odometer(self, miles):
        self.odometer_reading += miles

my_new_car = Car('audi', 'a4', 2016)
print(my_new_car.get_descriptive_name())
my_new_car.odometer_reading = 23
my_new_car.read_odometer()
```
