#### Removing All Instances of Specific Values from a List

```python
pets = ["dogs", "cat", "dog", "cat", "rabbit","goldfish" , "dogs"]
print(pets)

while "cat" in pets:
  pets.remove("cat")

print(pets)
```

```python
# filling a dictonary with user input

responses = {}

# set flag to indicate that polling is active
polling_active = True

while polling_active:
  # prompt for the person's name and response
  name = input('\nwhat is your name ')
  response = input("which mountain climb ")

  # store the response in the dictionary
  responses[name] = response

  # find out if anyone else is going to take the poll
  repeat = input("would you like to let another person respond? yes or no" )
  if repeat == "no":
    polling_active = False

  # polling is complete. show result
  print("\n --- Poll Results ---")
  for name, response in responses.items():
    print(name + " would like to climb " + response + ".")
```

# Functions

```python
# Define a function

def greet_user():
  """display a simple greeting"""
  print("hello ")

greet_user()
```

```python
#passing information to a function

def greet_user(username):
  '''display a simple greeting'''
  print("hello, " + username.title())

greet_user("jes")
```

```python
greet_user("rahul")
```

```python
# arguments and parameters
# username = parameter argument = rahul


# passing Argumets
## position argumets

def describe_pet(animal_type, pet_name):
  '''display information about pet'''
  print("I have a " + animal_type + ".")
  print("my " + animal_type + "'s name is " + pet_name.title())

describe_pet("hamster", "harry")
```

```python
# multiple function calls

def describe_pet(animal_type, pet_name):
  '''display information about a pet'''
  print('i have a ' + animal_type)
  print("my " + animal_type + "'s name is " + pet_name.title())

describe_pet("hamster", "harry")
describe_pet("cat", "tom")
```

```python
# orders in positional args

def describe_pet(animal_type, pet_name):
  """Display information about a pet."""
  print("I have a " + animal_type + ".")
  print("My " + animal_type + "'s name is " + pet_name.title() + ".")

describe_pet("harry", "hamster")
```

```python
# Keyword argumets :  name-value pair that you pass to a function

describe_pet(animal_type = "hamster", pet_name = "harry")
```

```python
#order not matter

describe_pet(animal_type = "hamster", pet_name = "harry")
describe_pet(pet_name='harry', animal_type='hamster')
```

```python
# default values

def describe_pet(pet_name, animal_type = "dog"):
  '''display information about a pet'''
  print("I have a " + animal_type)
  print("my " + animal_type + "'s name is " + pet_name.title())

describe_pet(pet_name = "tom")
```

```python
describe_pet("tom")
```

```python
# avoid arg error

def describe_pet(animal_type, pet_name):
  """Display information about a pet."""
  print("\nI have a " + animal_type + ".")
  print("My " + animal_type + "'s name is " + pet_name.title() + ".")

describe_pet()
```

```python
# return values

def get_formatted_name(first_name, last_name):
  '''return a full name, neat formatted'''
  full_name = first_name + ' ' + last_name
  return full_name.title()

musician = get_formatted_name("jimi", "hendrix")
print(musician)
```
