```python
# ================================================================
# LINKED LIST IN PYTHON
# ================================================================
# Today’s topic (randomly chosen): LINKED LIST (Data Structure)
#
# Think of a Linked List like a treasure hunt:
# Each box (node) contains:
#   1. Some data (value)
#   2. A clue (reference) to the next box
#
# Unlike a Python list:
#   Python list: [10, 20, 30]  → stored together in memory
#   Linked List: 10 → 20 → 30 → None (scattered but connected)
#
# WHY LINKED LIST?
# - Easy insertion/deletion (no shifting like arrays)
# - Flexible size (no fixed memory)
#
# ================================================================
# STEP 1: CREATE A NODE (the building block)
# ================================================================

class Node:
    def __init__(self, data):
        # "data" = actual value we want to store (like 10, 20, etc.)
        self.data = data

        # "next" = pointer/reference to next node
        # Initially, it points to NOTHING (None)
        self.next = None

# ================================================================
# STEP 2: CREATE LINKED LIST CLASS (manages nodes)
# ================================================================

class LinkedList:
    def __init__(self):
        # "head" is the starting point of the linked list
        # Initially empty → so head = None
        self.head = None

    # ============================================================
    # FUNCTION: INSERT AT END
    # ============================================================
    def insert_at_end(self, data):
        # Create a new node with given data
        new_node = Node(data)

        # CASE 1: If list is empty
        if self.head is None:
            # Make new node the head (first element)
            self.head = new_node
            return

        # CASE 2: Traverse till last node
        temp = self.head  # start from head

        # Move until we find last node (next = None)
        while temp.next:
            temp = temp.next

        # Now temp is last node → connect it to new node
        temp.next = new_node

    # ============================================================
    # FUNCTION: DISPLAY LIST
    # ============================================================
    def display(self):
        temp = self.head  # start from head

        # Traverse each node
        while temp:
            print(temp.data, end=" -> ")  # print current node data
            temp = temp.next  # move to next node

        print("None")  # last points to None

    # ============================================================
    # FUNCTION: SEARCH ELEMENT
    # ============================================================
    def search(self, key):
        temp = self.head  # start from head
        position = 0      # track index

        while temp:
            # If found
            if temp.data == key:
                return position

            temp = temp.next  # move forward
            position += 1     # increase index

        return -1  # not found

# ================================================================
# STEP 3: USE THE LINKED LIST
# ================================================================

# Create Linked List object
ll = LinkedList()

# Insert elements
ll.insert_at_end(10)  # First node
ll.insert_at_end(20)  # Second node
ll.insert_at_end(30)  # Third node

# Display list
print("Linked List:")
ll.display()

# Search element
result = ll.search(20)

# Print search result
if result != -1:
    print("Element 20 found at position:", result)
else:
    print("Element not found")

# ================================================================
# 🧠 OUTPUT EXPLANATION (VERY IMPORTANT FOR UNDERSTANDING)
# ================================================================
# EXPECTED OUTPUT:
#
# Linked List:
# 10 -> 20 -> 30 -> None
# Element 20 found at position: 1
#
# WHY THIS OUTPUT?
#
# 1. We inserted:
#    10 → becomes HEAD
#    20 → attached after 10
#    30 → attached after 20
#
#    So structure becomes:
#    HEAD → [10 | next] → [20 | next] → [30 | None]
#
# 2. Display function:
#    Starts at head (10)
#    Moves node by node using "next"
#    Prints each value until None
#
# 3. Search function:
#    Checks:
#      10 → not match
#      20 → match found at index 1
#
# ================================================================
# 🧠 MENTAL MODEL (SUPER IMPORTANT)
# ================================================================
# Imagine:
#
#   10 → 20 → 30 → None
#
# Each arrow is NOT position — it is a MEMORY LINK
#
# Node structure:
#   [data | next]
#
# Example:
#   [10 | address_of_20]
#   [20 | address_of_30]
#   [30 | None]
#
# ================================================================
# 🎯 WHY LINKED LIST IS POWERFUL?
# ================================================================
# - No shifting needed (unlike arrays)
# - Dynamic size
# - Efficient insertion/deletion
```

```python
# ============================================================
# 🧠 ULTRA EASY + ULTRA DETAILED LESSON: LINKED LIST IN PYTHON
# ============================================================
# Topic: DATA STRUCTURE → LINKED LIST
#
# Imagine a Linked List like a treasure hunt:
# Each box (node) contains:
#   1. Some data (like a number or name)
#   2. A pointer to the NEXT box
#
# Unlike a Python list (array), elements are NOT stored together in memory.
# Instead, each element "points" to the next one.
#
# Example Visualization:
#   [10 | next] → [20 | next] → [30 | next] → None
#
# "None" means END of the list (no more boxes).
#
# Now let's build EVERYTHING step-by-step from scratch 🚀


# ============================================================
# STEP 1: CREATE A NODE CLASS
# ============================================================

class Node:
    # This class represents ONE single node (box) in the Linked List
    def __init__(self, data):
        # "data" is the value stored inside the node (like 10, 20, etc.)
        self.data = data

        # "next" is a pointer/reference to the next node
        # Initially, we don't know the next node → so it's None
        self.next = None

        # WHY this design?
        # Because every node must store:
        #   (1) Data
        #   (2) Link to next node


# ============================================================
# STEP 2: CREATE LINKED LIST CLASS
# ============================================================

class LinkedList:
    def __init__(self):
        # "head" is the starting point of the Linked List
        # Initially, the list is empty → so head = None
        self.head = None

        # WHY "head"?
        # Because without knowing the first node,
        # we cannot access the rest of the list.


    # ========================================================
    # FUNCTION: INSERT AT END
    # ========================================================
    def insert_at_end(self, data):
        # Create a new node with given data
        new_node = Node(data)

        # CASE 1: If list is empty
        if self.head is None:
            # Then the new node becomes the first node (head)
            self.head = new_node
            return

        # CASE 2: List already has nodes
        # We need to go to the LAST node

        current = self.head  # Start from head

        # Traverse until we reach last node
        while current.next:
            # Move to next node
            current = current.next

        # Now "current" is the last node
        # So we attach new_node to it
        current.next = new_node


    # ========================================================
    # FUNCTION: DISPLAY LINKED LIST
    # ========================================================
    def display(self):
        current = self.head  # Start from head

        # Traverse until None (end of list)
        while current:
            # Print current node data
            print(current.data, end=" → ")

            # Move to next node
            current = current.next

        # Print end marker
        print("None")


    # ========================================================
    # FUNCTION: SEARCH FOR A VALUE
    # ========================================================
    def search(self, value):
        current = self.head  # Start from beginning

        position = 0  # To track index

        while current:
            # If data matches
            if current.data == value:
                return f"Value {value} found at position {position}"

            # Move forward
            current = current.next
            position += 1

        return f"Value {value} NOT found in list"


    # ========================================================
    # FUNCTION: DELETE A NODE
    # ========================================================
    def delete(self, value):
        current = self.head
        previous = None

        # Traverse list
        while current:

            # If we find the value
            if current.data == value:

                # CASE 1: Deleting HEAD
                if previous is None:
                    self.head = current.next

                # CASE 2: Deleting middle or last node
                else:
                    previous.next = current.next

                return f"Value {value} deleted"

            # Move forward
            previous = current
            current = current.next

        return f"Value {value} NOT found, cannot delete"


# ============================================================
# STEP 3: USE THE LINKED LIST
# ============================================================

# Create a Linked List object
ll = LinkedList()

# Insert values
ll.insert_at_end(10)
ll.insert_at_end(20)
ll.insert_at_end(30)
ll.insert_at_end(40)

# Display list
print("Linked List after insertion:")
ll.display()

# Search for value
print(ll.search(20))
print(ll.search(99))

# Delete a value
print(ll.delete(30))

# Display again after deletion
print("Linked List after deletion:")
ll.display()


# ============================================================
# 🧾 OUTPUT + EXPLANATION (VERY IMPORTANT 🧠)
# ============================================================

# Expected Output:
#
# Linked List after insertion:
# 10 → 20 → 30 → 40 → None
# Value 20 found at position 1
# Value 99 NOT found in list
# Value 30 deleted
# Linked List after deletion:
# 10 → 20 → 40 → None
#
#
# 🧠 EXPLANATION OF OUTPUT:
#
# 1. INSERTION:
#    We inserted 10, 20, 30, 40 one by one.
#    Each new node was attached at the END.
#
#    So structure becomes:
#    10 → 20 → 30 → 40 → None
#
#
# 2. SEARCH:
#    - Searching 20:
#        Start at head (10)
#        Move → 20 → FOUND at position 1
#
#    - Searching 99:
#        Traverse entire list
#        Not found → return message
#
#
# 3. DELETE:
#    - Deleting 30:
#        We connect 20 → 40 directly
#        So node 30 is removed from chain
#
#        New structure:
#        10 → 20 → 40 → None
#
#
#
# Linked List = Chain of nodes
# Each node knows ONLY the next node
#
# You NEVER jump directly like array indexing ❌
# You ALWAYS traverse step-by-step ✅
#
# This builds strong logic for:
# - Memory understanding
# - Pointer thinking
# - Real-world system design
```
