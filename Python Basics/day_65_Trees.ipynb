```python
# ============================================================
# PYTHON TREES
# ============================================================

# ------------------------------------------------------------
# PART 1 : WHAT IS A TREE? (Simple Explanation)
# ------------------------------------------------------------

# Imagine a real tree 🌳
#
#          10        <- ROOT (top node)
#         /  \
#        5    20
#       / \   /
#      3   7 15
#
# Each circle is called a NODE.
#
# Words to understand:
#
# ROOT
# The first node (top of tree)
#
# NODE
# A container that stores:
#   1) data (value)
#   2) links to children
#
# CHILD
# Nodes below another node
#
# PARENT
# Node above another node
#
# LEAF NODE
# Node with NO children
#
# TREE TYPE HERE:
# Binary Tree
#
# Binary means each node can have
# MAXIMUM 2 children:
#
# left child
# right child
#
# ------------------------------------------------------------
# PART 2 : CREATING A NODE CLASS
# ------------------------------------------------------------

class Node:

    # __init__ is a special function called a CONSTRUCTOR
    # It runs automatically when we create a new node.

    def __init__(self, value):

        # value is the data stored in the node
        # Example: 10 or 5 or 20

        self.value = value

        # left child pointer
        # Initially None because no child yet

        self.left = None

        # right child pointer
        # Initially None

        self.right = None


# ------------------------------------------------------------
# PART 3 : CREATE A TREE MANUALLY
# ------------------------------------------------------------

# Create root node
root = Node(10)

# Now we attach children

root.left = Node(5)   # left child of 10
root.right = Node(20) # right child of 10

root.left.left = Node(3)  # child of 5
root.left.right = Node(7)

root.right.left = Node(15)

# Tree now looks like this:
#
#          10
#         /  \
#        5    20
#       / \   /
#      3   7 15


# ------------------------------------------------------------
# PART 4 : TREE TRAVERSAL
# ------------------------------------------------------------

# Traversal means:
# "How we visit nodes one by one"

# There are 4 famous traversals:
#
# 1) INORDER
# 2) PREORDER
# 3) POSTORDER
# 4) BREADTH FIRST


# ------------------------------------------------------------
# INORDER TRAVERSAL
# ------------------------------------------------------------

# Order rule:
#
# LEFT -> ROOT -> RIGHT
#
# Example tree:
#
#          10
#         /  \
#        5    20
#       / \   /
#      3   7 15
#
# Output:
# 3 5 7 10 15 20

def inorder(node):

    # Base case
    # If node is None we stop

    if node is None:
        return

    # Step 1: visit left subtree
    inorder(node.left)

    # Step 2: print current node value
    print(node.value, end=" ")

    # Step 3: visit right subtree
    inorder(node.right)


# ------------------------------------------------------------
# PREORDER TRAVERSAL
# ------------------------------------------------------------

# Rule:
#
# ROOT -> LEFT -> RIGHT
#
# Output example:
# 10 5 3 7 20 15

def preorder(node):

    if node is None:
        return

    # visit root first
    print(node.value, end=" ")

    # go left
    preorder(node.left)

    # go right
    preorder(node.right)


# ------------------------------------------------------------
# POSTORDER TRAVERSAL
# ------------------------------------------------------------

# Rule:
#
# LEFT -> RIGHT -> ROOT
#
# Output:
# 3 7 5 15 20 10

def postorder(node):

    if node is None:
        return

    postorder(node.left)

    postorder(node.right)

    print(node.value, end=" ")


# ------------------------------------------------------------
# PART 5 : BREADTH FIRST SEARCH (LEVEL ORDER)
# ------------------------------------------------------------

# BFS visits nodes LEVEL by LEVEL
#
# Level 1
#   10
#
# Level 2
#   5   20
#
# Level 3
# 3 7 15
#
# Output:
# 10 5 20 3 7 15

# To do this we use a QUEUE

from collections import deque

def bfs(root):

    # Create queue
    queue = deque()

    # Add root to queue
    queue.append(root)

    while queue:

        # Remove first element
        node = queue.popleft()

        print(node.value, end=" ")

        # Add children to queue

        if node.left:
            queue.append(node.left)

        if node.right:
            queue.append(node.right)


# ------------------------------------------------------------
# PART 6 : RUN ALL TRAVERSALS
# ------------------------------------------------------------

print("\nInorder Traversal:")
inorder(root)

print("\n\nPreorder Traversal:")
preorder(root)

print("\n\nPostorder Traversal:")
postorder(root)

print("\n\nBreadth First Traversal:")
bfs(root)


# ------------------------------------------------------------
# PART 7 : WHY TREES ARE IMPORTANT
# ------------------------------------------------------------

# Trees are used in:
#
# 1) File systems
# 2) Databases
# 3) Search engines
# 4) AI decision trees
# 5) Compilers
# 6) Game engines
#
# Advanced trees you will learn later:
#
# Binary Search Tree
# AVL Tree
# Red Black Tree
# Heap
# Trie
# Segment Tree
```
