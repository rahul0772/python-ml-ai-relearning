```python
# ---------------------------------------------------
# LINKED LIST
# ---------------------------------------------------

# Imagine your toys in boxes in a line.
# Each box has a number and points to the next box.
# None = end of line, no more toys.

# Step 1: Make empty line
line = []  # This is our line of toy boxes
print("Step 1: Start with empty line")
print("Line:", line, "=> Nothing yet!\n")

# Step 2: Add boxes one by one (toys)
print("Step 2: Add toys 10, 20, 30 at the end")
line.append(10)
line.append(20)
line.append(30)

# Step 3: Show line function (draw boxes and arrows)
def show_line(line):
    if not line:
        print("No boxes in line! (empty)")
        return
    for i in range(len(line)):
        print(f"[{line[i]}]", end=" -> ")
    print("None")  # End of line

print("Line after adding 10, 20, 30:")
show_line(line)
print()

# Step 4: Add toy at the start
print("Step 4: Add toy 5 at the start")
line.insert(0, 5)
show_line(line)
print()

# Step 5: Add toy in the middle
print("Step 5: Add toy 15 between 10 and 20")
# Find where 10 is, insert 15 after it
index_10 = line.index(10)  # Find index of toy 10
line.insert(index_10 + 1, 15)
show_line(line)
print()

# Step 6: Remove a toy by number
print("Step 6: Remove toy 20 from the line")
line.remove(20)
show_line(line)
print()

# Step 7: Remove first toy
print("Step 7: Remove first toy (5)")
line.pop(0)  # Remove first toy
show_line(line)
print()

# Step 8: Remove last toy
print("Step 8: Remove last toy (30)")
line.pop()  # Remove last toy
show_line(line)
print()

# Step 9: Add many toys
print("Step 9: Add toys 25, 35, 45 at end")
line.append(25)
line.append(35)
line.append(45)
show_line(line)
print()

# Step 10: Insert toy after a specific toy
print("Step 10: Insert toy 28 after toy 25")
index_25 = line.index(25)
line.insert(index_25 + 1, 28)
show_line(line)
print()

# Step 11: Try removing a toy that does not exist
print("Step 11: Try removing toy 99 (not in line)")
if 99 in line:
    line.remove(99)
else:
    print("Toy 99 not found, cannot remove")
show_line(line)
print()

# Step 12: Show final line in very visual way
print("Final line of toys (step by step visualization):")
for i, toy in enumerate(line):
    arrow = " -> " if i < len(line)-1 else " -> None"
    print(f"[{toy}]{arrow}", end="")
print("\n")
```
