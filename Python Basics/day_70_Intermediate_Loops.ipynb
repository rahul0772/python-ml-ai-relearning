```python
# ================================================================
# INTERMEDIATE PYTHON
# 1) for loops
# 2) while loops
# 3) functions
# 4) lists
# 5) condition logic
# 6) how programmers THINK about logic step-by-step
# 7) breaking problems into small steps
#
# IMPORTANT IDEA:
# Programming is about LOGIC:
#
# Step 1: Understand the problem
# Step 2: Break the problem into tiny steps
# Step 3: Translate each step into Python
#
# We'll solve a small but realistic problem:
#
# PROBLEM:
# We have a list of student scores.
# We want to:
# 1) print each score
# 2) calculate the average score
# 3) find the highest score
# 4) classify each student (pass/fail)
# 5) put the logic inside FUNCTIONS
#
# ================================================================


# ------------------------------------------------
# STEP 1: CREATE DATA
# ------------------------------------------------

# A LIST stores multiple values in one variable.
# Think of a list like a row of boxes.
#
# Example visualization:
#
# index:    0   1   2   3   4
# scores = [75, 90, 60, 82, 55]
#
# "index" means POSITION in the list.
#
# scores[0] -> 75
# scores[1] -> 90
#
# ists let us process
# many values using loops.

scores = [75, 90, 60, 82, 55]


# ------------------------------------------------
# STEP 2: LOOP THROUGH DATA (FOR LOOP)
# ------------------------------------------------

# A FOR LOOP repeats an action for each item in a list.
#
# Think of it like:
#
# "For each score in the list, do something"
#
# Python syntax:
#
# for variable in list:
#     code
#
# The variable temporarily holds each item.

print("Printing each score:\n")

for score in scores:   # Python takes one value from the list at a time
                       # First iteration: score = 75
                       # Second iteration: score = 90
                       # Third iteration: score = 60
                       # Fourth iteration: score = 82
                       # Fifth iteration: score = 55

    print("Current score:", score)

    # During each loop iteration:
    #
    # loop 1 → score = 75
    # loop 2 → score = 90
    # loop 3 → score = 60
    # loop 4 → score = 82
    # loop 5 → score = 55


# ------------------------------------------------
# STEP 3: CALCULATE A TOTAL USING A LOOP
# ------------------------------------------------

# To compute an average we first need a SUM of all scores.
#
# Logic thinking:
#
# start total at 0
# add each score to total
#
# Example mental process:
#
# total = 0
#
# add 75 → total = 75
# add 90 → total = 165
# add 60 → total = 225
# add 82 → total = 307
# add 55 → total = 362

total = 0  # starting value

for score in scores:   # loop through every score

    total = total + score
    # VERY IMPORTANT LINE
    #
    # "total = total + score"
    #
    # means:
    #
    # take current total
    # add score
    # store result back into total
    #
    # Example iteration:
    #
    # total = 0
    # score = 75
    #
    # total = 0 + 75
    # total becomes 75


print("\nTotal score:", total)


# ------------------------------------------------
# STEP 4: CALCULATE AVERAGE
# ------------------------------------------------

# Average formula:
#
# average = total / number_of_items

count = len(scores)
# len() means LENGTH
#
# len(scores) counts how many items are inside the list
#
# scores = [75, 90, 60, 82, 55]
#
# count = 5

average = total / count

print("Average score:", average)


# ------------------------------------------------
# STEP 5: FIND THE HIGHEST SCORE
# ------------------------------------------------

# Logic idea:
#
# Start by assuming the first number is the highest.
#
# Then compare every other number to it.

highest = scores[0]
# scores[0] is the FIRST value in the list
#
# highest starts as 75

for score in scores:

    if score > highest:
        # CONDITION:
        #
        # "Is this score bigger than the highest so far?"

        highest = score
        # If YES:
        #
        # update highest


print("Highest score:", highest)


# ------------------------------------------------
# STEP 6: CLASSIFY PASS / FAIL
# ------------------------------------------------

# We'll create rules:
#
# score >= 70  -> PASS
# score < 70   -> FAIL

print("\nStudent results:\n")

for score in scores:

    if score >= 70:
        print(score, "PASS")

    else:
        print(score, "FAIL")


# ------------------------------------------------
# STEP 7: FUNCTIONS
# ------------------------------------------------

# A FUNCTION is a reusable block of code.
#
# Instead of writing the same logic again,
# we package it into a function.

def calculate_average(numbers):
    """
    This function calculates the average of a list.

    numbers → list of values
    """

    total = 0

    for n in numbers:
        total = total + n

    avg = total / len(numbers)

    return avg
    # return sends the result back to the caller


# calling the function

avg_result = calculate_average(scores)

print("\nAverage from function:", avg_result)


# ------------------------------------------------
# STEP 8: WHILE LOOP (LOGIC BASED LOOP)
# ------------------------------------------------

# A WHILE LOOP repeats while a condition is TRUE.
#
# Example logic:
#
# while something_is_true:
#     keep running

index = 0

print("\nPrinting scores using while loop:\n")

while index < len(scores):
    # condition check
    #
    # index starts at 0
    #
    # 0 < 5 → True → run loop
    # 1 < 5 → True → run loop
    # 2 < 5 → True → run loop
    # 3 < 5 → True → run loop
    # 4 < 5 → True → run loop
    # 5 < 5 → False → stop

    print(scores[index])

    index = index + 1
    # VERY IMPORTANT
    #
    # Without this line the loop would NEVER STOP
    #
    # index increases each loop


# ------------------------------------------------
# FINAL
# ------------------------------------------------
# 1) What data do I have?
# 2) What do I need to calculate?
# 3) What steps solve the problem?
#
# Then they convert each step into code.
#
# Logic -> Code
#
# Not the other way around.
#
# The more problems you solve,
# the stronger your logic becomes.
# ================================================================
```
