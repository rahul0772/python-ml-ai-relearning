# Dictionary


```python
#Removing key-Value pair

alien = {"color" : "green", "points" :5}
print(alien)

del alien["points"] #key-value pair is removed permanently
print(alien)
```

```python
# Dictionary of Similar Objects

favorite_language = {
    "jen" : "python",
    "sarah" : "c",
    "edward" : "ruby",
    "phil" : "python",
    }
```

```python
print("Sarah's favorite language is " +
      favorite_language['sarah'].title() +
      ".")
```

```python
# Looping through a dictionary all key value pairs

user = {
    "username" : "daves",
    "first" : "dave",
    "last" : "smith",
}

for key, value in user.items(): # returns a list of key-value pairs and key value stored in two variable key, value
  print("\nkey:" + key)
  print("Value: " + value)
```

```python
favorite_language = {
    "jen" : "python",
    "sarah" : "C",
    "edward" : "ruby",
    "phil" : "python",
}

for name, language in favorite_language.items():
  print(name.title() + "'s fav language is " + language.title() + ".")
```

```python
#looping through all the keys ina a dictionary
#when not to work with all of the values in a dictionary

favaorite_languages = {
    "jen" : "python",
    "sarah" : "C",
    "edward" : "ruby",
    "phil" : "python",
}

for name in favorite_language.keys():
  print(name.title())
```

```python
for name in favorite_language: #default without without keys()
  print(name.title())
```

```python
friends = ["phil", "sarah"]
for name in favorite_language.keys():
  print(name.title())

  if name in friends:
    print( " Hi " + name.title() +
          ", I see your favorite language is " +
           favorite_language[name].title() + ".")
```

```python
# Looping through a dictionary keys in order
#sorted() function to get a copy of the keys in order
for name in sorted(favaorite_languages.keys()):
  print(name.title() + ", thank you for taking the poll.")
```
