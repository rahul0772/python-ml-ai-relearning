# Dictionary


```python
#Removing key-Value pair

alien = {"color" : "green", "points" :5}
print(alien)

del alien["points"] #key-value pair is removed permanently
print(alien)
```

```python
# Looping through a dictionary all key value pairs

user = {
    "username" : "daves",
    "first" : "dave",
    "last" : "smith",
}

for key, value in user.items(): # returns a list of key-value pairs and key value stored in two variable key, value
  print("\nkey:" + key)
  print("Value: " + value)
```

```python
favorite_language = {
    "jen" : "python",
    "sarah" : "C",
    "edward" : "ruby",
    "phil" : "python",
}

for name, language in favorite_language.items():
  print(name.title() + "'s fav language is " + language.title() + ".")
```

```python
#looping through all the keys ina a dictionary
#when not to work with all of the values in a dictionary

favaorite_languages = {
    "jen" : "python",
    "sarah" : "C",
    "edward" : "ruby",
    "phil" : "python",
}

for name in favorite_language.keys():
  print(name.title())
```

```python
for name in favorite_language: #default without without keys()
  print(name.title())
```

```python
friends = ["phil", "sarah"]
for name in favorite_language.keys():
  print(name.title())

  if name in friends:
    print( " Hi " + name.title() +
          ", I see your favorite language is " +
           favorite_language[name].title() + ".")
```

```python
# Looping through a dictionary keys in order
#sorted() function to get a copy of the keys in order
for name in sorted(favaorite_languages.keys()):
  print(name.title() + ", thank you for taking the poll.")
```

```python
#looping through all values in a dictionary
favorite_language = {
 'jen': 'python',
 'sarah': 'c',
 'edward': 'ruby',
 'phil': 'python',
 }

print("fav language:")
for language in favorite_language.values():
  print(language.title())
```

```python
#for no repitions of the values
#set() to pull out the unique values
print("fav language:")
for language in set(favorite_language.values()):
  print(language.title())
```

### Nesting

```python
alien_0 = {"color" : "green" , "points" : 5}
alien_1 = {"color" : "yellow" , "points" : 15}
alien_2 = {"color" : "red" , "points" : 25}

aliens = [alien_0, alien_1, alien_2]

for alien in aliens:
  print(alien)
```

```python
# Make an empty list for storing alien

alien = []

# make 30 green aliens.
for alien_number in range(30): # times we want the loop to repeat
  new_alien = {"color" : "green", "points" : 5, "speed" : "slow"} # Each time the loop runs we create a new alien
  aliens.append(new_alien)  #append each new alien to the list aliens

  # first five alien
  for alien in aliens[:5]: #slice to print the first five aliens
    print(alien)
    print("..")

  # total aliens created
  print("total " + str(len(aliens)) )
```

```python
# A list in a dictionary

pizza = {
    "crust" : "thick",
    "toppings" : ["mushrooms", "extra cheese"],
}

# summarize the order
print("you ordered a " + pizza["crust"] + " -crust pizza "
      "with the following toppings:")

for topping in pizza["toppings"]:
  print("\t" + topping)
```
