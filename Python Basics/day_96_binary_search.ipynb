```python
# ============================================
# BINARY SEARCH

# --------------------------------------------
# 🧠 CORE IDEA (SUPER SIMPLE):
# --------------------------------------------
# Imagine you are searching a word in a dictionary 📖
# Do you start from page 1? ❌ NO!
# You open somewhere in the middle.
#
# That is EXACTLY what Binary Search does:
# 👉 It cuts the problem into HALF every time.
#
# This is called: "Divide and Conquer"

# --------------------------------------------
# ⚠️ VERY IMPORTANT CONDITION:
# --------------------------------------------
# Binary Search ONLY works on SORTED data!
# If data is not sorted → it WILL FAIL ❌

# Let's create a sorted list:
numbers = [2, 5, 8, 12, 16, 23, 38, 45, 56, 72]

# --------------------------------------------
# 🎯 TARGET:
# --------------------------------------------
# We want to find where number 23 is in the list.
target = 23

# --------------------------------------------
# 🧠 HOW BINARY SEARCH THINKS (MENTAL MODEL):
# --------------------------------------------
# Step 1: Look at middle element
# Step 2: Compare with target
# Step 3:
#    If equal → FOUND 🎉
#    If target < middle → search LEFT side
#    If target > middle → search RIGHT side

# --------------------------------------------
# 🧱 SETUP VARIABLES:
# --------------------------------------------

# 'low' = starting index of search range
low = 0

# 'high' = ending index of search range
high = len(numbers) - 1  # last index

# WHY len(numbers) - 1?
# Because list index starts from 0
# If list has 10 elements → last index is 9

# --------------------------------------------
# 🔍 START SEARCH LOOP:
# --------------------------------------------

# We use a WHILE loop because:
# We want to keep searching UNTIL range becomes invalid
while low <= high:

    # ----------------------------------------
    # 🧠 FIND MIDDLE INDEX
    # ----------------------------------------

    # Formula:
    mid = (low + high) // 2

    # WHY // ?
    # // means integer division → gives whole number only
    # Example:
    # (0 + 9) // 2 = 4 → perfect middle index

    # ----------------------------------------
    # 🧠 CHECK MIDDLE VALUE
    # ----------------------------------------

    print(f"Checking middle index {mid}, value = {numbers[mid]}")

    # Case 1: FOUND 🎉
    if numbers[mid] == target:
        print(f"🎉 Found {target} at index {mid}")
        break  # STOP LOOP

    # Case 2: Target is SMALLER → go LEFT
    elif target < numbers[mid]:
        print(f"{target} is smaller than {numbers[mid]} → move LEFT")

        # Move 'high' to one position LEFT of mid
        high = mid - 1

        # WHY?
        # Because mid already checked → exclude it

    # Case 3: Target is BIGGER → go RIGHT
    else:
        print(f"{target} is greater than {numbers[mid]} → move RIGHT")

        # Move 'low' to one position RIGHT of mid
        low = mid + 1

# --------------------------------------------
# 🧠 IF NOT FOUND:
# --------------------------------------------
# This executes ONLY if loop finishes WITHOUT break
else:
    print(f"❌ {target} not found in list")

# ============================================
# 📤 OUTPUT EXPLANATION (VERY IMPORTANT)
# ============================================

# Let's deeply understand what just happened:

# Initial List:
# [2, 5, 8, 12, 16, 23, 38, 45, 56, 72]

# Step 1:
# low = 0, high = 9
# mid = (0 + 9)//2 = 4 → value = 16
# 23 > 16 → go RIGHT

# Step 2:
# low = 5, high = 9
# mid = (5 + 9)//2 = 7 → value = 45
# 23 < 45 → go LEFT

# Step 3:
# low = 5, high = 6
# mid = (5 + 6)//2 = 5 → value = 23
# FOUND 🎉

# --------------------------------------------
# 🚀 WHY THIS IS POWERFUL:
# --------------------------------------------
# Linear Search: checks EVERY element → slow 🐢
# Binary Search: cuts problem in HALF → fast 🚀

# Example:
# 1000 elements:
# Linear Search → up to 1000 checks 😱
# Binary Search → ~10 checks 😎

# --------------------------------------------
# 🧠 FINAL MENTAL MODEL:
# --------------------------------------------
# Binary Search = "Keep cutting the problem in HALF until found"

# --------------------------------------------
# 🎯 BIG TAKEAWAYS:
# --------------------------------------------
# ✔ Works ONLY on sorted data
# ✔ Uses divide and conquer
# ✔ Extremely efficient
# ✔ Used in real-world systems (databases, search engines)
# ============================================
```
