```python
# ============================================================
# INTRODUCTION TO GRAPH THEORY
# + BASIC GRAPH ALGORITHMS (BFS & DFS)
# ============================================================

# Imagine a graph like a network of cities connected by roads.
# - Each city = a "NODE" (also called vertex)
# - Each road = an "EDGE" (connection between nodes)
#
# Example:
# A --- B --- C
# |     |
# D --- E
#
# We will represent this using a Python "dictionary"
# where:
# - Key = node
# - Value = list of neighbors (connected nodes)

# ------------------------------------------------------------
# STEP 1: CREATE A GRAPH USING ADJACENCY LIST
# ------------------------------------------------------------

# We choose dictionary because:
# - Fast lookup of neighbors
# - Easy to understand mapping (node -> connections)

graph = {
    'A': ['B', 'D'],      # Node A connects to B and D
    'B': ['A', 'C', 'E'], # Node B connects to A, C, E
    'C': ['B'],           # Node C connects to B
    'D': ['A', 'E'],      # Node D connects to A and E
    'E': ['B', 'D']       # Node E connects to B and D
}

# WHY THIS STRUCTURE?
# Because:
# graph['A'] gives ['B','D'] instantly
# That means: from A, we can go to B and D

print("Graph structure (Adjacency List):")
print(graph)

# ------------------------------------------------------------
# STEP 2: DEPTH FIRST SEARCH (DFS)
# ------------------------------------------------------------

# DFS = go as deep as possible first
# Think like exploring a maze:
# - Keep going forward until you hit a dead end
# - Then go back and try another path

# We use RECURSION (function calling itself)

def dfs(graph, node, visited):
    # graph = full graph
    # node = current node we are visiting
    # visited = set to track visited nodes

    # If node already visited → STOP (avoid infinite loops)
    if node in visited:
        return

    # Mark node as visited
    visited.add(node)

    # Print current node
    print(node, end=" ")

    # Explore neighbors
    for neighbor in graph[node]:
        # Recursive call → go deeper
        dfs(graph, neighbor, visited)

# Run DFS
print("\n\nDFS Traversal starting from A:")
dfs(graph, 'A', set())

# ------------------------------------------------------------
# STEP 3: BREADTH FIRST SEARCH (BFS)
# ------------------------------------------------------------

# BFS = explore level by level
# Think like:
# - First visit all immediate neighbors
# - Then neighbors of neighbors

# We use QUEUE (FIFO → First In First Out)

from collections import deque  # deque = fast queue

def bfs(graph, start):
    # visited keeps track of visited nodes
    visited = set()

    # queue stores nodes to explore
    queue = deque([start])  # start node added

    # Mark start as visited
    visited.add(start)

    while queue:
        # Remove first element (FIFO)
        node = queue.popleft()

        # Print current node
        print(node, end=" ")

        # Visit all neighbors
        for neighbor in graph[node]:
            # If not visited
            if neighbor not in visited:
                visited.add(neighbor)   # mark visited
                queue.append(neighbor) # add to queue

# Run BFS
print("\n\nBFS Traversal starting from A:")
bfs(graph, 'A')

# ------------------------------------------------------------
# STEP 4: SHORTEST PATH USING BFS
# ------------------------------------------------------------

# BFS can find shortest path in UNWEIGHTED graph

def shortest_path(graph, start, goal):
    # queue stores paths (not just nodes)
    queue = deque([[start]])  # path starts with start node

    visited = set()

    while queue:
        path = queue.popleft()  # get current path

        node = path[-1]  # last node in path

        # If goal reached
        if node == goal:
            return path

        if node not in visited:
            visited.add(node)

            for neighbor in graph[node]:
                new_path = list(path)   # copy current path
                new_path.append(neighbor)  # extend path
                queue.append(new_path)

    return None  # if no path

# Find shortest path from A to C
path = shortest_path(graph, 'A', 'C')

print("\n\nShortest path from A to C:")
print(path)

# ------------------------------------------------------------
# FINAL OUTPUT EXPLANATION (VERY IMPORTANT 🧠)
# ------------------------------------------------------------

# Graph printed:
# Shows how nodes connect. Example:
# 'A': ['B','D'] → means A connects to B and D

# DFS Output:
# A B C E D (order may vary slightly)
# Why?
# - Start at A
# - Go deep: A → B → C
# - Backtrack → B → E → D

# BFS Output:
# A B D C E
# Why?
# - Start at A
# - Visit neighbors first: B, D
# - Then their neighbors: C, E

# Shortest Path:
# ['A', 'B', 'C']
# Why?
# BFS guarantees shortest path in unweighted graph
# A → B → C is shortest (2 edges)

# 🧠 FINAL MENTAL MODEL:
# DFS = Deep exploration (stack/recursion)
# BFS = Wide exploration (queue)
# Graph = Nodes + Connections
# BFS = Best for shortest path (unweighted)

# ============================================================
# ============================================================
```
