```python
# 0/1 Knapsack Problem: Recursion + Dynamic Programming (Memoization)
def knapsack(W, weights, values, n, memo={}):
    """
    Solve the knapsack problem using recursion and memoization.

    W      : The maximum weight the knapsack can hold
    weights: List of weights of the items
    values : List of values of the items
    n      : Number of items left to consider
    memo   : A dictionary to store results of subproblems (memoization)

    Returns the maximum value that can be obtained by packing items into the knapsack
    """

    # Base case: If no items left or knapsack capacity is 0, no value can be added
    if n == 0 or W == 0:
        return 0

    # Check if this subproblem has already been solved (memoization)
    if (n, W) in memo:
        return memo[(n, W)]

    # If the weight of the current item is greater than the capacity, we cannot include it
    if weights[n-1] > W:
        result = knapsack(W, weights, values, n-1, memo)
    else:
        # Recursive case 1: Include the current item
        include = values[n-1] + knapsack(W - weights[n-1], weights, values, n-1, memo)

        # Recursive case 2: Exclude the current item
        exclude = knapsack(W, weights, values, n-1, memo)

        # Take the maximum of including or excluding the item
        result = max(include, exclude)

    # Memoize the result to avoid recomputing
    memo[(n, W)] = result
    return result

# Test the knapsack function with items, their weights, values, and knapsack capacity
weights = [1, 3, 4, 5]     # Weight of each item
values = [1, 4, 5, 7]      # Value of each item
W = 7                      # Maximum weight capacity of the knapsack
n = len(weights)           # Number of items

# Call the knapsack function and print the result
print("Maximum value that can be obtained:", knapsack(W, weights, values, n))
```
