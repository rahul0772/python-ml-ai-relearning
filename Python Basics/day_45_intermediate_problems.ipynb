```python
# ================================================================
# PROBLEM:
# Find the SECOND LARGEST number in a list (without using sort())
# ================================================================

# ------------------------------------------------
# STEP 1: Create a list of numbers
# ------------------------------------------------

# A list is like a box that stores many values.
# Think of it like a row of lockers:
# [10, 5, 20, 8, 15]
#  index: 0   1   2   3   4

numbers = [10, 5, 20, 8, 15]

# ------------------------------------------------
# STEP 2: Check if the list has at least 2 numbers
# ------------------------------------------------

# Why?
# Because to find SECOND largest,
# we need at least TWO numbers.
# If list has only 1 number â†’ impossible.

if len(numbers) < 2:
    # len(numbers) means:
    # "How many items are inside the list?"
    print("List must have at least 2 numbers")
else:

    # ------------------------------------------------
    # STEP 3: Create two variables to store answers
    # ------------------------------------------------

    # largest will store biggest number found so far
    # second_largest will store second biggest number

    # We start both as VERY SMALL number.
    # float('-inf') means NEGATIVE INFINITY
    # It is smaller than any number.
    # So any number in our list will be bigger than this.

    largest = float('-inf')
    second_largest = float('-inf')

    # ------------------------------------------------
    # STEP 4: Loop through every number in the list
    # ------------------------------------------------

    # This means:
    # "Go through each number one by one"

    for num in numbers:

        # Let's print to understand what's happening
        print("Checking number:", num)

        # --------------------------------------------
        # CASE 1: If current number is BIGGER than largest
        # --------------------------------------------

        if num > largest:

            # If we found a NEW biggest number,
            # then old largest becomes second_largest

            second_largest = largest

            # And update largest
            largest = num

            print("  New largest found!")
            print("  Largest is now:", largest)
            print("  Second largest is now:", second_largest)

        # --------------------------------------------
        # CASE 2: If number is NOT biggest
        # but still bigger than second_largest
        # --------------------------------------------

        elif num > second_largest and num != largest:

            # Update only second_largest
            second_largest = num

            print("  Updated second largest to:", second_largest)

        # --------------------------------------------
        # OTHERWISE
        # --------------------------------------------

        else:
            print("  No change needed.")

    # ------------------------------------------------
    # STEP 5: Print final result
    # ------------------------------------------------

    print("\nFinal Answer:")
    print("Largest number:", largest)
    print("Second largest number:", second_largest)
```
