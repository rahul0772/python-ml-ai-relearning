```python
# **Bubble Sort Algorithm**.
# Bubble Sort is a simple sorting algorithm that repeatedly steps through the list, compares adjacent elements, and swaps them if they are in the wrong order. The pass through the list is repeated until the list is sorted.

# I will provide a detailed breakdown of each line of code, explaining everything step-by-step, with easy-to-understand explanations, so that it becomes crystal clear how the logic works.

# Bubble Sort Implementation with extreme ultra-detailed explanations inside the code:

def bubble_sort(arr):
    """
    The function takes a list 'arr' and sorts it using the Bubble Sort algorithm.
    This algorithm works by repeatedly stepping through the list, comparing adjacent
    elements, and swapping them if they are in the wrong order. The process repeats
    until the list is sorted.
    """

    # Step 1: Start by getting the length of the list to determine how many passes we need
    n = len(arr)  # We store the length of the array in a variable 'n'

    # For example: if arr = [5, 3, 8, 4], len(arr) will give us 4.
    # This means we will need to perform up to 4 passes over the array to ensure it's sorted.

    # Step 2: Outer loop to manage the number of passes. In Bubble Sort, each pass ensures
    # that the largest element 'bubbles' to the correct position.
    for i in range(n):  # Outer loop, we will iterate n times, once for each pass.
        swapped = False  # This variable tracks whether any elements were swapped during the pass.

        # Step 3: Inner loop to go through the unsorted part of the list in each pass
        for j in range(0, n - i - 1):  # Inner loop compares adjacent elements
            # We are comparing arr[j] with arr[j+1], if arr[j] > arr[j+1], they are swapped.
            if arr[j] > arr[j + 1]:  # Compare two adjacent elements
                arr[j], arr[j + 1] = arr[j + 1], arr[j]  # Swap if they are in the wrong order

                # After this swap, arr[j] is now smaller than arr[j + 1] which means this
                # pair is sorted. The larger element moves towards the right end.

                swapped = True  # If a swap occurred, set the swapped flag to True.

        # Step 4: If no elements were swapped during this pass, the list is already sorted.
        if not swapped:  # If no elements were swapped in the inner loop, break out early.
            break  # The list is sorted, we can stop the algorithm here.

    # Final sorted array will be in 'arr'
    return arr  # Return the sorted array after all passes

# Step 5: Now let's test our function with an example input list.
test_list = [64, 34, 25, 12, 22, 11, 90]  # Initial unsorted list
print("Unsorted List:", test_list)  # Print the unsorted list for reference
sorted_list = bubble_sort(test_list)  # Call the bubble_sort function on the unsorted list
print("Sorted List:", sorted_list)  # Print the sorted list

# Explanation of Output:
# For example, the input list is [64, 34, 25, 12, 22, 11, 90].
# In the first pass, the largest number (90) will bubble to the right-most position.
# After the second pass, the second-largest number (64) will be in its correct position.
# This process continues until the array is fully sorted.
```
