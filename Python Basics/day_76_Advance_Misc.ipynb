```python
# ---------------- PROBLEM 6: Lambda Function ----------------
# Lambda = small anonymous function, like a mini calculator
# Example: we want a function to add two numbers
# Normal function: def add(x, y): return x+y
# Lambda makes it shorter

add = lambda x, y: x + y
# Explanation:
# lambda -> tells Python "this is a small function"
# x, y -> input numbers
# x + y -> what we return (sum)
# No need to write def, return, etc.

print("Problem 6: Lambda Function")
print("Add 5 + 3 =", add(5,3))  # Example: 5 + 3 = 8

# ---------------- PROBLEM 7: Map and Filter ----------------
# Map = applies a function to every item in a list
# Filter = selects items that meet a condition

numbers = [1,2,3,4,5,6]

# Map example: square every number
squared_numbers = list(map(lambda x: x**2, numbers))
# Explanation:
# map(function, list) -> apply function to every item
# lambda x: x**2 -> square each x
# list() -> convert result to a list

# Filter example: select only even numbers
even_numbers = list(filter(lambda x: x%2==0, numbers))
# Explanation:
# filter(function, list) -> keep only items where function returns True
# lambda x: x%2==0 -> True if x is divisible by 2
# list() -> convert to list

print("\nProblem 7: Map & Filter")
print("Squared numbers:", squared_numbers)
print("Even numbers:", even_numbers)

# ---------------- PROBLEM 8: Generator Function ----------------
# Generator = function that gives numbers one by one, saves memory
# Example: Fibonacci generator

def fibonacci_gen(n):
    a, b = 0, 1  # start numbers
    count = 0
    while count < n:
        yield a  # "yield" gives value but keeps function alive for next call
        a, b = b, a + b  # update numbers
        count += 1

print("\nProblem 8: Fibonacci Generator")
for num in fibonacci_gen(10):  # print first 10 numbers
    print(num, end=" ")  # numbers printed on same line

# ---------------- PROBLEM 9: Decorator Function ----------------
# Decorator = function that adds extra features to another function
# Example: we want to print "Start" and "End" around any function

def my_decorator(func):
    def wrapper():
        print("Start of function")  # extra feature
        func()  # call original function
        print("End of function")    # extra feature
    return wrapper

@my_decorator  # attach decorator to say_hello function
def say_hello():
    print("Hello, World!")

print("\n\nProblem 9: Decorator Function")
say_hello()  # automatically wrapped with Start/End prints

# ---------------- PROBLEM 10: Error Handling ----------------
# Try/Except = prevent program from crashing when error occurs
# Example: dividing by zero

print("\nProblem 10: Error Handling")
try:
    x = int(input("Enter a number to divide 10 by: "))
    result = 10 / x
    print("Result =", result)
except ZeroDivisionError:
    print("Oops! Cannot divide by zero!")  # handle division by zero
except ValueError:
    print("Invalid input! Please enter a number.")  # handle non-numbers
```
