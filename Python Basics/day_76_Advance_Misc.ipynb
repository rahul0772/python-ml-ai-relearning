```python
# ============================================================
# 🔴 PROBLEM 1: FIBONACCI (DYNAMIC PROGRAMMING)
# ============================================================

# Fibonacci means:
# 0, 1, 1, 2, 3, 5, 8, 13, ...
# Rule:
# Next number = previous + previous previous

def fibonacci(n):
    # n = how many numbers we want

    # Step 1: Handle small cases
    if n <= 0:
        return []  # no numbers
    if n == 1:
        return [0]  # only first number

    # Step 2: Create a list to store results
    fib = [0, 1]
    # We already know first two values

    # Step 3: Loop to calculate next values
    for i in range(2, n):
        # i = current position

        # fib[i-1] = last number
        # fib[i-2] = second last number
        next_value = fib[i-1] + fib[i-2]

        # Add new value to list
        fib.append(next_value)

    return fib


print("Fibonacci:", fibonacci(10))


# ============================================================
# 🔴 PROBLEM 2: PRIME NUMBER CHECK (OPTIMIZED)
# ============================================================

# Prime number = only divisible by 1 and itself
# Example: 7 → only 1 and 7

def is_prime(num):
    # Step 1: Handle small numbers
    if num <= 1:
        return False  # 0 and 1 are NOT prime

    # Step 2: Check from 2 to square root of number
    # Why sqrt? Because factors repeat after that
    for i in range(2, int(num ** 0.5) + 1):
        # num % i == 0 means divisible
        if num % i == 0:
            return False  # found a divisor

    return True  # no divisor found


print("Is 29 prime?", is_prime(29))


# ============================================================
# 🔴 PROBLEM 3: PALINDROME CHECK
# ============================================================

# Palindrome = same forward and backward
# Example: "madam", "racecar"

def is_palindrome(text):
    # Step 1: Convert to lowercase (ignore case)
    text = text.lower()

    # Step 2: Reverse the string
    reversed_text = text[::-1]
    # [::-1] means reverse

    # Step 3: Compare original and reversed
    return text == reversed_text


print("Is 'madam' palindrome?", is_palindrome("madam"))


# ============================================================
# 🔴 PROBLEM 4: FLATTEN A NESTED LIST
# ============================================================

# Example:
# Input: [1, [2, 3], [4, [5, 6]]]
# Output: [1, 2, 3, 4, 5, 6]

def flatten(lst):
    # Create empty result list
    result = []

    for item in lst:
        # If item is a list → go deeper (recursion)
        if isinstance(item, list):
            # Call function again
            result.extend(flatten(item))
        else:
            # If normal number → add directly
            result.append(item)

    return result


nested = [1, [2, 3], [4, [5, 6]]]
print("Flatten:", flatten(nested))


# ============================================================
# 🔴 PROBLEM 5: COUNT FREQUENCY OF ELEMENTS
# ============================================================

# Example:
# Input: [1,1,2,3,3,3]
# Output: {1:2, 2:1, 3:3}

def count_frequency(lst):
    # Create empty dictionary
    freq = {}

    for item in lst:
        # If item already exists → increase count
        if item in freq:
            freq[item] += 1
        else:
            # First time seeing this item
            freq[item] = 1

    return freq


print("Frequency:", count_frequency([1,1,2,3,3,3]))


# 1. List:
#    A collection of items → [1, 2, 3]

# 2. Loop:
#    Repeat something many times
#    for i in range(5):

# 3. Function:
#    A reusable block of code
#    def name():

# 4. Recursion:
#    Function calling itself

# 5. Dictionary:
#    Key → Value pairs
#    {"apple": 2}

# 6. Condition:
#    if something is true → do this

# 7. Modulus (%):
#    Gives remainder
#    10 % 3 = 1

# 8. Slicing:
#    text[::-1] → reverse string
```
