```python
# Selection Sort is a simple algorithm that sorts an array (or list) by repeatedly finding the smallest (or largest) element
# from the unsorted portion of the list and swapping it with the leftmost unsorted element. This process is repeated
# until the entire list is sorted. It's a very intuitive algorithm, but it is not the most efficient for large datasets.

# Here’s a fully detailed Python implementation of Selection Sort with ultra-detailed explanations inside each line of code.

# Define the selection_sort function which takes a list as an argument
def selection_sort(arr):
    # The 'arr' is the list we want to sort
    # We are using the Selection Sort algorithm to sort 'arr'

    # Step 1: Loop over the entire list to sort it
    for i in range(len(arr)):
        # 'i' is the current index in the list (we will find the minimum value after this index)
        # Think of 'i' as representing the position where the sorted portion ends
        # Initially, there is no sorted portion, so we start at index 0 and go all the way to the end

        # Step 2: Assume that the current index 'i' holds the smallest value
        min_index = i
        # 'min_index' is the index of the smallest element we find during each iteration
        # We start by assuming that the smallest value is at 'i'

        # Step 3: Loop through the unsorted portion of the list (starting from 'i+1' to end)
        for j in range(i + 1, len(arr)):
            # 'j' is the index where we are currently checking for a smaller value than 'min_index'
            # Start from 'i + 1' because everything before 'i' is already sorted

            # Step 4: If we find a smaller element, update 'min_index'
            if arr[j] < arr[min_index]:
                min_index = j
                # Now, 'min_index' points to the smallest element found so far

        # Step 5: After finding the smallest element, swap it with the element at index 'i'
        # We only swap if the minimum value is not already at 'i', to avoid unnecessary swaps
        if min_index != i:
            # Swap the element at 'i' with the element at 'min_index'
            arr[i], arr[min_index] = arr[min_index], arr[i]
            # This swaps arr[i] with arr[min_index], placing the smallest element found
            # in the correct sorted position for this iteration.

    # After completing the loop, the list 'arr' is sorted.
    return arr


# Example to test the function:
# We define an unsorted list that we will sort using the Selection Sort algorithm.
unsorted_list = [64, 25, 12, 22, 11]

# Call the selection_sort function and print the sorted result
sorted_list = selection_sort(unsorted_list)
print(sorted_list)

# Now, let's break down the output:

# Output explanation:
# The list [64, 25, 12, 22, 11] is unsorted. The Selection Sort algorithm will work as follows:
# 1. Start with the first element (64). We find the smallest element in the rest of the list (11) and swap them.
#   - The list becomes [11, 25, 12, 22, 64].
# 2. Now, starting from index 1 (25), the smallest element in the remaining unsorted portion is 12, so we swap them.
#   - The list becomes [11, 12, 25, 22, 64].
# 3. Starting from index 2 (25), the smallest element in the remaining unsorted portion is 22, so we swap them.
#   - The list becomes [11, 12, 22, 25, 64].
# 4. Now, starting from index 3 (25), no smaller element is found, so no swap happens.
#   - The list remains [11, 12, 22, 25, 64].
# 5. Finally, starting from index 4 (64), there are no more elements left, and the list is fully sorted.

# The final sorted list is [11, 12, 22, 25, 64].
# We successfully sorted the list using the Selection Sort algorithm.
```
