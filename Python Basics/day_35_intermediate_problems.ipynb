```python
# ============================================================
# PYTHON LEARNING NOTEBOOK
# LEVEL: INTERMEDIATE
# ============================================================

# ------------------------------------------------------------
# PROBLEM 1: COUNT HOW MANY TIMES A NUMBER APPEARS IN A LIST
# ------------------------------------------------------------

# A "list" is like a box that can store many numbers together.
# Example: [1, 2, 3] is a list with 3 numbers.

numbers = [1, 2, 3, 2, 4, 2, 5]
# numbers list explanation:
# Position 0 → 1
# Position 1 → 2
# Position 2 → 3
# Position 3 → 2
# Position 4 → 4
# Position 5 → 2
# Position 6 → 5

target = 2
# target means: the number we want to count

count = 0
# count starts at 0
# We use count to remember how many times we found the target number

# "for" loop means:
# "Go through each number in the list ONE BY ONE"
for num in numbers:
    # num becomes each number in the list
    # First time → num = 1
    # Second time → num = 2
    # Third time → num = 3
    # and so on...

    if num == target:
        # This checks:
        # "Is the current number equal to the target number?"

        count = count + 1
        # If YES, increase count by 1
        # Example:
        # count was 0 → now becomes 1
        # then 2 → then 3

print("Problem 1 Answer:", count)
# This prints how many times 2 appeared in the list


# ------------------------------------------------------------
# PROBLEM 2: FIND THE BIGGEST NUMBER IN A LIST
# ------------------------------------------------------------

numbers = [10, 45, 2, 99, 23, 67]

# We assume the first number is the biggest
biggest = numbers[0]
# numbers[0] means the first number in the list
# biggest = 10 at the start

for num in numbers:
    # Go through each number one by one

    if num > biggest:
        # Check if current number is bigger than what we think is biggest

        biggest = num
        # If YES, update biggest

print("Problem 2 Answer:", biggest)
# This prints the largest number


# ------------------------------------------------------------
# PROBLEM 3: REVERSE A WORD
# ------------------------------------------------------------

word = "python"
# word is a string (text)

reversed_word = ""
# This is empty at first
# We will slowly build the reversed word

# Loop through each letter in the word
for letter in word:
    # First letter → 'p'
    # Second → 'y'
    # Third → 't'
    # etc...

    reversed_word = letter + reversed_word
    # This puts the letter IN FRONT of the old word
    # Example:
    # letter = 'p' → reversed_word = 'p'
    # letter = 'y' → 'y' + 'p' → 'yp'
    # letter = 't' → 't' + 'yp' → 'typ'

print("Problem 3 Answer:", reversed_word)
# This prints the reversed word


# ------------------------------------------------------------
# PROBLEM 4: CHECK IF A NUMBER IS PRIME
# ------------------------------------------------------------

# A PRIME number:
# 1️⃣ Greater than 1
# 2️⃣ Divisible ONLY by 1 and itself
# Examples:
# 2, 3, 5, 7 → Prime
# 4 (2×2), 6 (2×3) → NOT prime

number = 17

is_prime = True
# We first assume the number IS prime

if number <= 1:
    # Numbers 0 and 1 are NOT prime
    is_prime = False

# Check from 2 up to number-1
for i in range(2, number):
    # range(2, number) means:
    # 2, 3, 4, ..., number-1

    if number % i == 0:
        # % means remainder
        # If remainder is 0 → divisible

        is_prime = False
        break
        # break stops the loop immediately

print("Problem 4 Answer:", is_prime)
# True → prime
# False → not prime


# ------------------------------------------------------------
# PROBLEM 5: COUNT LETTERS USING A DICTIONARY
# ------------------------------------------------------------

# A dictionary is like:
# key → value
# Example:
# {"a": 2, "b": 1}

text = "banana"

letter_count = {}
# Empty dictionary

for letter in text:
    # Go through each letter:
    # 'b', 'a', 'n', 'a', 'n', 'a'

    if letter in letter_count:
        # Check if letter already exists in dictionary

        letter_count[letter] = letter_count[letter] + 1
        # Increase the count by 1

    else:
        # If letter NOT in dictionary yet

        letter_count[letter] = 1
        # Add it with count 1

print("Problem 5 Answer:", letter_count)

# ============================================================
# END OF NOTEBOOK
# ============================================================
```
