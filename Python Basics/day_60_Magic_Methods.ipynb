```python

# --------------------------
# Advanced OOP: Magic Methods
# --------------------------
# Magic methods let us define how objects behave with operators
class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    # String representation
    def __str__(self):
        return f"Vector({self.x}, {self.y})"  # how object prints

    # Addition of two vectors
    def __add__(self, other):
        return Vector(self.x + other.x, self.y + other.y)

v1 = Vector(2, 3)
v2 = Vector(4, 5)
v3 = v1 + v2  # calls v1.__add__(v2)
print("v3 =", v3)  # prints Vector(6, 8)

# --------------------------
# Class Methods & Static Methods
# --------------------------
class Circle:
    pi = 3.14159  # class variable (shared by all objects)

    def __init__(self, radius):
        self.radius = radius  # instance variable

    @classmethod  # works with the class, not instance
    def from_diameter(cls, diameter):
        return cls(diameter / 2)  # create Circle from diameter

    @staticmethod  # no access to class or instance
    def area_formula(radius):
        return Circle.pi * radius**2

c1 = Circle.from_diameter(10)
print("\nCircle radius from diameter:", c1.radius)
print("Area using static method:", Circle.area_formula(c1.radius))

# --------------------------
# Context Managers: with + __enter__/__exit__
# --------------------------
# Useful to handle resources automatically
class MyFile:
    def __init__(self, filename, mode):
        self.filename = filename
        self.mode = mode

    def __enter__(self):
        self.file = open(self.filename, self.mode)
        return self.file  # object to use inside 'with'

    def __exit__(self, exc_type, exc_val, exc_tb):
        self.file.close()  # automatically closes file
        if exc_type:
            print(f"An exception occurred: {exc_val}")
        return True  # suppress exception if any

with MyFile("advanced.txt", "w") as f:
    f.write("Hello using custom context manager!\n")

# --------------------------
# Custom Iterators
# --------------------------
class ReverseRange:
    def __init__(self, n):
        self.n = n
        self.current = n

    def __iter__(self):
        return self

    def __next__(self):
        if self.current <= 0:
            raise StopIteration  # stops iteration
        self.current -= 1
        return self.current

print("\nReverse range iterator:")
for i in ReverseRange(5):
    print(i)  # prints 4,3,2,1,0

# --------------------------
# Advanced Decorators with parameters
# --------------------------
def repeat(n_times):  # decorator factory
    def decorator(func):
        def wrapper(*args, **kwargs):
            for _ in range(n_times):
                func(*args, **kwargs)
        return wrapper
    return decorator

@repeat(3)
def say_hi():
    print("Hi!")

print("\nDecorator with parameters:")
say_hi()  # prints "Hi!" 3 times

# --------------------------
# Property Decorators
# --------------------------
class Temperature:
    def __init__(self, celsius):
        self._celsius = celsius

    @property  # getter
    def celsius(self):
        return self._celsius

    @celsius.setter  # setter
    def celsius(self, value):
        if value < -273.15:  # physically impossible temperature
            raise ValueError("Temperature below -273.15C is impossible!")
        self._celsius = value

    @property
    def fahrenheit(self):
        return (self._celsius * 9/5) + 32

t = Temperature(25)
print("\nTemperature in Celsius:", t.celsius)
print("Temperature in Fahrenheit:", t.fahrenheit)
t.celsius = 30  # update
print("Updated Celsius:", t.celsius)

# --------------------------
# Nested Functions & Closures
# --------------------------
def outer_function(msg):
    def inner_function():
        print("Message from outer:", msg)
    return inner_function  # returns inner function without running

closure_func = outer_function("Hello Closure!")
closure_func()  # runs inner_function, remembers 'msg' from outer

# --------------------------
# Custom Exceptions
# --------------------------
class MyCustomError(Exception):
    pass

try:
    raise MyCustomError("Something went wrong!")
except MyCustomError as e:
    print("\nCaught custom exception:", e)
```
