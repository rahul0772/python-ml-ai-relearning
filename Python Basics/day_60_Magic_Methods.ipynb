```python
# --------------------------
# 1️⃣ Magic Methods (__init__, __str__, __add__)
# --------------------------
# Magic methods let us control how our objects behave with Python operators.
# For example: print(), +, -, etc.

class Vector:
    def __init__(self, x, y):
        # __init__ is the constructor. It runs when you create a new object.
        # Here we store x and y coordinates in the object
        self.x = x
        self.y = y

    def __str__(self):
        # __str__ controls how the object prints with print()
        return f"Vector({self.x}, {self.y})"

    def __add__(self, other):
        # __add__ allows using + operator between two Vector objects
        return Vector(self.x + other.x, self.y + other.y)

# Example usage:
v1 = Vector(2, 3)
v2 = Vector(4, 5)
v3 = v1 + v2  # calls v1.__add__(v2), result is a new Vector(6,8)
print("v3 =", v3)  # prints nicely thanks to __str__

# --------------------------
# 2️⃣ Class Methods & Static Methods
# --------------------------
# Class methods work with the class itself, not specific objects
# Static methods are just functions inside a class (no self or cls)

class Circle:
    pi = 3.14159  # class variable, same for all circles

    def __init__(self, radius):
        # instance variable, each circle can have its own radius
        self.radius = radius

    @classmethod
    def from_diameter(cls, diameter):
        # cls refers to the class itself
        # lets us create Circle using diameter instead of radius
        return cls(diameter / 2)

    @staticmethod
    def area_formula(radius):
        # static method: just a function inside class
        return Circle.pi * radius**2

# Example usage:
c1 = Circle.from_diameter(10)  # radius becomes 5
print("\nCircle radius from diameter:", c1.radius)
print("Area using static method:", Circle.area_formula(c1.radius))

# --------------------------
# 3️⃣ Context Managers (__enter__ / __exit__)
# --------------------------
# 'with' keyword ensures resources are properly opened & closed (like files)

class MyFile:
    def __init__(self, filename, mode):
        self.filename = filename
        self.mode = mode

    def __enter__(self):
        # runs when 'with' starts
        self.file = open(self.filename, self.mode)
        return self.file  # what 'as f' will reference

    def __exit__(self, exc_type, exc_val, exc_tb):
        # runs when 'with' ends, even if an error occurs
        self.file.close()  # always close file
        if exc_type:
            # if an exception occurred, print it
            print(f"An exception occurred: {exc_val}")
        return True  # True suppresses the exception

# Example usage:
with MyFile("advanced.txt", "w") as f:
    f.write("Hello using custom context manager!\n")

# --------------------------
# 4️⃣ Custom Iterators (__iter__, __next__)
# --------------------------
# Lets you make objects that can be looped with 'for'

class ReverseRange:
    def __init__(self, n):
        self.n = n
        self.current = n

    def __iter__(self):
        # __iter__ returns the iterator object itself
        return self

    def __next__(self):
        # __next__ gives the next item each time loop runs
        if self.current <= 0:
            raise StopIteration  # stops the loop
        self.current -= 1
        return self.current

# Example usage:
print("\nReverse range iterator:")
for i in ReverseRange(5):  # prints 4,3,2,1,0
    print(i)

# --------------------------
# 5️⃣ Decorators with Parameters
# --------------------------
# Decorators modify behavior of functions
# Here we create a decorator that repeats a function N times

def repeat(n_times):  # outer function receives the number of repeats
    def decorator(func):
        # decorator wraps the original function
        def wrapper(*args, **kwargs):
            for _ in range(n_times):
                func(*args, **kwargs)  # call the original function
        return wrapper
    return decorator

@repeat(3)  # decorate say_hi to repeat 3 times
def say_hi():
    print("Hi!")

print("\nDecorator with parameters:")
say_hi()  # prints "Hi!" 3 times

# --------------------------
# 6️⃣ Property Decorators (@property, setter)
# --------------------------
# Lets us control access to attributes with getter/setter

class Temperature:
    def __init__(self, celsius):
        self._celsius = celsius  # use _ to indicate "private" variable

    @property
    def celsius(self):
        # getter: lets us access t.celsius instead of t._celsius
        return self._celsius

    @celsius.setter
    def celsius(self, value):
        # setter: lets us validate before updating
        if value < -273.15:
            raise ValueError("Temperature below -273.15C is impossible!")
        self._celsius = value

    @property
    def fahrenheit(self):
        # calculated property: no setter, read-only
        return (self._celsius * 9/5) + 32

# Example usage:
t = Temperature(25)
print("\nTemperature in Celsius:", t.celsius)
print("Temperature in Fahrenheit:", t.fahrenheit)
t.celsius = 30
print("Updated Celsius:", t.celsius)

# --------------------------
# 7️⃣ Nested Functions & Closures
# --------------------------
# A function inside another function. Inner remembers outer variables

def outer_function(msg):
    def inner_function():
        print("Message from outer:", msg)
    return inner_function  # return inner function without running it

# Example usage:
closure_func = outer_function("Hello Closure!")
closure_func()  # prints message and remembers 'msg'

# --------------------------
# 8️⃣ Custom Exceptions
# --------------------------
# Lets you create your own errors

class MyCustomError(Exception):
    # inherit from Exception
    pass

try:
    raise MyCustomError("Something went wrong!")  # manually raise error
except MyCustomError as e:
    print("\nCaught custom exception:", e)

# ------------------------------------------------------
```
