```python


# 1. For Loop:
# The "for" loop is used to iterate over a sequence (like a list, range, or string).
# It executes a block of code a specific number of times.

# Example 1: Using a for loop to print numbers from 1 to 5:
for i in range(1, 6):  # range(1, 6) generates numbers from 1 to 5
    print(i)  # This will print each number in the range one by one
# Output:
# 1
# 2
# 3
# 4
# 5

# Here's how it works:
# - `range(1, 6)` creates a sequence of numbers from 1 to 5.
# - The `for` loop goes through each number in the sequence and assigns it to the variable `i`.
# - Then, it prints `i` on each iteration.

# 2. While Loop:
# The "while" loop continues executing as long as the condition is true.

# Example 2: Using a while loop to print numbers from 1 to 5:
i = 1  # Start with i = 1
while i <= 5:  # As long as i is less than or equal to 5
    print(i)  # Print the current value of i
    i += 1  # Increment i by 1 (this is important to avoid an infinite loop)
# Output:
# 1
# 2
# 3
# 4
# 5

# Here's how the while loop works:
# - The condition `i <= 5` is checked before each loop iteration.
# - If the condition is true, the code inside the loop runs.
# - After each iteration, `i` is incremented by 1 (`i += 1`), making sure the loop doesn't run forever.

# -------------------------------------
# Why It Works:
# Loops are a powerful tool for automating repetitive tasks.
# - The `for` loop is ideal when you know the number of iterations in advance (like iterating through a list).
# - The `while` loop is used when you don’t know how many iterations you'll need, but you have a condition to check.
```
