```python
# ✅ HUFFMAN CODING

import heapq
from collections import Counter

# 1️⃣ Tree Node Class
class HuffmanNode:
    def __init__(self, char=None, freq=0, left=None, right=None):
        self.char = char      # Leaf nodes store characters
        self.freq = freq      # Frequency/probability weight
        self.left = left      # Left child (represents '0')
        self.right = right    # Right child (represents '1')

    # heapq requires < comparison. Tie-break with id() to avoid errors on equal freq
    def __lt__(self, other):
        return self.freq < other.freq

# 2️⃣ Build Huffman Tree
def build_huffman_tree(text):
    if not text:
        return None

    # Count character frequencies
    freq = Counter(text)

    # Initialize min-heap with leaf nodes
    heap = [HuffmanNode(char=c, freq=f) for c, f in freq.items()]
    heapq.heapify(heap)

    # Combine lowest-frequency nodes until one tree remains
    while len(heap) > 1:
        left = heapq.heappop(heap)
        right = heapq.heappop(heap)
        merged = HuffmanNode(freq=left.freq + right.freq, left=left, right=right)
        heapq.heappush(heap, merged)

    return heap[0]

# 3️⃣ Generate Binary Codes via Tree Traversal
def generate_codes(node, current_code="", codes=None):
    if codes is None:
        codes = {}
    if node is None:
        return codes

    # Leaf node: store the accumulated path as its code
    if node.char is not None:
        codes[node.char] = current_code or "0"  # '0' fallback for single-char text
        return codes

    # Traverse left (add '0') and right (add '1')
    generate_codes(node.left, current_code + "0", codes)
    generate_codes(node.right, current_code + "1", codes)
    return codes

# 4️⃣ Encode & Decode Functions
def huffman_encode(text, codes):
    return "".join(codes[char] for char in text)

def huffman_decode(encoded_text, root):
    if root is None:
        return ""
    decoded = []
    current = root
    for bit in encoded_text:
        current = current.left if bit == '0' else current.right
        if current.char is not None:
            decoded.append(current.char)
            current = root  # Reset to root for next character
    return "".join(decoded)

# ========================
# 🚀 EXAMPLE USAGE & METRICS
# ========================
if __name__ == "__main__":
    text = "this is an example for huffman encoding"
    print(f"📝 Original Text: '{text}'")
    print(f"📏 Original Size: {len(text) * 8} bits (8-bit ASCII)\n")

    # Build tree & generate codes
    root = build_huffman_tree(text)
    codes = generate_codes(root)

    print("🔑 Huffman Code Table (sorted by code length):")
    for char, code in sorted(codes.items(), key=lambda x: (len(x[1]), x[0])):
        char_display = "'SPACE'" if char == ' ' else f"'{char}'"
        print(f"  {char_display}: {code}")

    # Encode & Decode
    encoded = huffman_encode(text, codes)
    decoded = huffman_decode(encoded, root)

    print(f"\n📦 Encoded Bitstream: {encoded}")
    print(f"📏 Encoded Size: {len(encoded)} bits")
    print(f"🔄 Decoded Text: '{decoded}'")
    print(f"✅ Lossless Verification: {text == decoded}")

    # Compression stats
    original_bits = len(text) * 8
    compressed_bits = len(encoded)
    ratio = original_bits / compressed_bits
    print(f"\n📊 Compression Ratio: {ratio:.2f}x")
    print(f"💾 Space Saved: {original_bits - compressed_bits} bits ({(1 - 1/ratio)*100:.1f}%)")
```
