```python
# Bubble Sort Algorithm

# Bubble Sort is a simple sorting algorithm that works by repeatedly stepping through the list to be sorted,
# comparing each pair of adjacent items and swapping them if they are in the wrong order.
# The process continues until no more swaps are needed, meaning the list is sorted.

# Here's a deeper dive into the Bubble Sort algorithm:

def bubble_sort(arr):
    # Step 1: Determine the length of the array, 'n'.
    # This will help us determine how many iterations of the list are needed for sorting.
    # In each iteration, one less comparison is needed because the largest elements bubble up to the end.
    n = len(arr)  # The variable 'n' holds the total number of elements in the list.

    # Step 2: Outer loop controls how many passes we make over the list.
    # We will loop n times to ensure the entire list is sorted.
    for i in range(n):
        # Step 3: Set a flag to track if any swaps happen during this pass.
        # The 'swapped' flag tells us if the list is already sorted.
        swapped = False  # We assume that the list is sorted at the start of each pass.

        # Step 4: Inner loop is where we do the actual comparisons and swaps.
        # The inner loop compares adjacent elements and swaps them if needed.
        # The range of the inner loop reduces by one each time because the last elements are already sorted.
        for j in range(0, n-i-1):  # The '-i' part prevents re-checking the sorted elements at the end of the list.
            # Step 5: Compare adjacent elements in the list, arr[j] and arr[j+1].
            # If the left element (arr[j]) is greater than the right element (arr[j+1]),
            # then the two elements are swapped to ensure they are in the correct order.
            if arr[j] > arr[j+1]:
                # Swap the two elements: if arr[j] is greater, move arr[j+1] to the left.
                arr[j], arr[j+1] = arr[j+1], arr[j]  # Swapping elements to sort them in ascending order.
                swapped = True  # We have swapped elements, so the list might still need more sorting.

        # Step 6: After the inner loop completes, check if we made any swaps during this pass.
        # If no swaps were made, the list is already sorted, and we can break out of the outer loop early.
        if not swapped:
            break  # If no swaps occurred, the list is already sorted, and there's no need to continue.

# Step 7: Testing the bubble_sort function with an example list.

# We define a test list with unsorted numbers.
test_list = [64, 34, 25, 12, 22, 11, 90]

# Step 8: Call the bubble_sort function with the test list.
# This will modify the list in place and sort it in ascending order.
bubble_sort(test_list)

# Step 9: Print the sorted list after calling the bubble_sort function.
print("Sorted list is:", test_list)

# EXPLANATION OF THE OUTPUT:

# Let's break down the output, which is:
# Sorted list is: [11, 12, 22, 25, 34, 64, 90]

# The bubble_sort function works by going through the list multiple times and comparing adjacent elements.
# Here's what happens step by step for our example input, [64, 34, 25, 12, 22, 11, 90]:

# 1st Pass (Outer Loop Iteration 1):
# - The algorithm starts with the first element and compares it with the next element.
# - 64 vs 34 → 64 is greater than 34, so we swap them.
#   List: [34, 64, 25, 12, 22, 11, 90]
# - 64 vs 25 → 64 is greater than 25, so we swap them.
#   List: [34, 25, 64, 12, 22, 11, 90]
# - 64 vs 12 → 64 is greater than 12, so we swap them.
#   List: [34, 25, 12, 64, 22, 11, 90]
# - 64 vs 22 → 64 is greater than 22, so we swap them.
#   List: [34, 25, 12, 22, 64, 11, 90]
# - 64 vs 11 → 64 is greater than 11, so we swap them.
#   List: [34, 25, 12, 22, 11, 64, 90]
# - 64 vs 90 → 64 is smaller than 90, so no swap.
#   List: [34, 25, 12, 22, 11, 64, 90]
# - The largest number (64) is now at the correct position.
# - After the first pass, the list looks like: [34, 25, 12, 22, 11, 64, 90].

# 2nd Pass (Outer Loop Iteration 2):
# - 34 vs 25 → 34 is greater than 25, so we swap them.
#   List: [25, 34, 12, 22, 11, 64, 90]
# - 34 vs 12 → 34 is greater than 12, so we swap them.
#   List: [25, 12, 34, 22, 11, 64, 90]
# - 34 vs 22 → 34 is greater than 22, so we swap them.
#   List: [25, 12, 22, 34, 11, 64, 90]
# - 34 vs 11 → 34 is greater than 11, so we swap them.
#   List: [25, 12, 22, 11, 34, 64, 90]
# - 34 vs 64 → 34 is smaller than 64, so no swap.
#   List: [25, 12, 22, 11, 34, 64, 90]
# - 34 vs 90 → 34 is smaller than 90, so no swap.
#   List: [25, 12, 22, 11, 34, 64, 90]
# - After the second pass, the second largest number (34) is in place.
# - The list after the second pass: [25, 12, 22, 11, 34, 64, 90].

# 3rd Pass (Outer Loop Iteration 3):
# - 25 vs 12 → 25 is greater than 12, so we swap them.
#   List: [12, 25, 22, 11, 34, 64, 90]
# - 25 vs 22 → 25 is greater than 22, so we swap them.
#   List: [12, 22, 25, 11, 34, 64, 90]
# - 25 vs 11 → 25 is greater than 11, so we swap them.
#   List: [12, 22, 11, 25, 34, 64, 90]
# - 25 vs 34 → 25 is smaller than 34, so no swap.
#   List: [12, 22, 11, 25, 34, 64, 90]
# - 25 vs 64 → 25 is smaller than 64, so no swap.
#   List: [12, 22, 11, 25, 34, 64, 90]
# - 25 vs 90 → 25 is smaller than 90, so no swap.
#   List: [12, 22, 11, 25, 34, 64, 90]
# - After the third pass, the third largest number (25) is in place.
# - The list after the third pass: [12, 22, 11, 25, 34, 64, 90].

# This continues until the entire list is sorted, and by the end, the list is fully sorted:
# Sorted list is: [11, 12, 22, 25, 34, 64, 90].

# We can stop early if we make a pass without any swaps (this is why the 'swapped' flag is important).
# If no swaps occur in a pass, the list is already sorted, and the function terminates early.

# Final Output:
# Sorted list is: [11, 12, 22, 25, 34, 64, 90]
```
