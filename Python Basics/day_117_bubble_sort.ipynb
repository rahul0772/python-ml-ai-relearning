```python
# ===========================================================
# Topic: Bubble Sort (Basic Sorting Algorithm)
# ===========================================================

# Bubble Sort is a simple sorting algorithm that repeatedly steps through
# the list, compares adjacent elements and swaps them if they are in the
# wrong order. This process is repeated until the list is sorted.

# We will create a list of numbers, sort it using Bubble Sort,
# and explain everything line by line in extreme detail.

# Step 1: Create a sample list of numbers that we want to sort
numbers = [64, 34, 25, 12, 22, 11, 90]
# Explanation:
# We are initializing a Python list named 'numbers'.
# A list is a data structure that can hold multiple items in order.
# This list contains integers in random order.
# Our goal: sort them from smallest to largest using Bubble Sort.

# Step 2: Get the total number of elements in the list
n = len(numbers)
# Explanation:
# 'len(numbers)' returns the total number of elements in the list.
# In our case, len([64, 34, 25, 12, 22, 11, 90]) = 7
# We store this number in variable 'n' to control our loops later.
# This is important because we need to know how many elements to compare.

# Step 3: Start the Bubble Sort algorithm
for i in range(n):
    # Explanation:
    # 'for i in range(n)' means we will repeat the process n times (0 to n-1)
    # Each pass through the list moves the next largest element to its correct position.
    # i represents the current pass number (iteration).
    # Bubble Sort requires multiple passes to ensure all elements are in order.

    # Step 4: Inner loop for comparing adjacent elements
    for j in range(0, n-i-1):
        # Explanation:
        # 'range(0, n-i-1)' ensures we don't compare already sorted elements
        # Each pass 'i' places the largest unsorted element at the end.
        # So, after first pass, last element is sorted, after second pass, last 2 are sorted, etc.
        # j is the index of the current element being compared.

        # Step 5: Compare the current element with the next element
        if numbers[j] > numbers[j+1]:
            # Explanation:
            # We check if the current element is bigger than the next element.
            # If it is, they are in the wrong order, so we need to swap them.
            # This ensures that the larger elements "bubble up" to the end.

            # Step 6: Swap the elements if they are in wrong order
            numbers[j], numbers[j+1] = numbers[j+1], numbers[j]
            # Explanation:
            # This is a Pythonic way to swap two values without a temporary variable.
            # numbers[j], numbers[j+1] = numbers[j+1], numbers[j] swaps the values.
            # Example: if numbers[j]=64 and numbers[j+1]=34, after swap:
            # numbers[j]=34, numbers[j+1]=64
            # This moves the larger element one step closer to its correct position.

# Step 7: Print the sorted list
print("Sorted list is:", numbers)
# Explanation:
# After all passes are done, the 'numbers' list is now sorted in ascending order.
# print() function outputs the final sorted list.
# This line allows us to see the result of our Bubble Sort algorithm.

# Output explanation:
# The largest elements "bubble up" to the end of the list in each pass.
# After first pass: [34, 25, 12, 22, 11, 64, 90]
# After second pass: [25, 12, 22, 11, 34, 64, 90]
# After all passes: [11, 12, 22, 25, 34, 64, 90]
# The output shows the fully sorted list.
# This demonstrates the core idea of Bubble Sort: repeated comparisons and swaps
# to gradually move elements to their correct positions.
```
