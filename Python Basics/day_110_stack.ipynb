```python
# ---------------------------------------------------------
# ULTRA DETAILED PYTHON LESSON: DATA STRUCTURES -> STACK
# Topic: Stack (LIFO: Last In First Out)
# We will implement a simple stack using a Python list and
# understand each step deeply with super detailed explanations
# ---------------------------------------------------------

# Step 1: Initialize an empty stack
stack = []  # We use a Python list to simulate a stack
# Explanation:
# A stack is a collection where the last element added is the first to be removed (LIFO).
# Using a list is simple because Python lists already support adding/removing from the end efficiently.

# Step 2: Push elements onto the stack
stack.append(10)  # Add 10 to the stack
# Explanation:
# append() is a Python list method that adds an element at the end.
# Since the end is the "top" of our stack, append acts as the "push" operation in stack terminology.

stack.append(20)  # Add 20 to the stack
# Now the stack looks like [10, 20]
# 10 was pushed first, then 20. The top is 20.

stack.append(30)  # Add 30 to the stack
# Stack is now [10, 20, 30]
# Top of the stack is 30. In LIFO, this will be the first element to be removed.

# Step 3: Peek at the top element (without removing it)
top_element = stack[-1]  # Access the last element using negative indexing
# Explanation:
# stack[-1] gives the last item in the list, which is the "top" of our stack.
# Peek is important when we want to see the top value but not remove it.
# top_element will hold 30 in this case.

# Step 4: Pop elements from the stack
popped_element1 = stack.pop()  # Remove and return the top element (LIFO)
# Explanation:
# pop() removes the last element of the list (the top of the stack)
# It returns the value so we can store it or use it.
# popped_element1 will be 30
# Now the stack is [10, 20]

popped_element2 = stack.pop()  # Remove the next top element
# popped_element2 will be 20
# Stack after this pop is [10]

# Step 5: Check if stack is empty
is_empty = len(stack) == 0
# Explanation:
# len(stack) returns the number of elements in the stack
# If it is 0, the stack is empty.
# Currently, stack has [10], so is_empty will be False

# Step 6: Print all outputs with explanations
print("Initial stack after pushes: [10, 20, 30]")
print("Top element after pushes (peek):", top_element)
print("First popped element:", popped_element1)
print("Second popped element:", popped_element2)
print("Stack now after pops:", stack)
print("Is stack empty?", is_empty)

# ---------------------------------------------------------
# OUTPUT EXPLANATION
# 1. "Initial stack after pushes" shows the order elements were added. LIFO means last added (30) is at top.
# 2. "Top element after pushes" shows peek operation result, which is 30 because it was added last.
# 3. "First popped element" is 30. LIFO: last in, first out.
# 4. "Second popped element" is 20, the next top element.
# 5. "Stack now after pops" shows remaining elements: [10].
# 6. "Is stack empty?" returns False because one element (10) remains.
# This connects all operations directly to LIFO stack behavior and internal Python list mechanics.
# ---------------------------------------------------------
```
