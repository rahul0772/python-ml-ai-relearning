```python
# ==========================
# Dynamic Programming (Longest Increasing Subsequence - LIS)
# ==========================

# Importing lru_cache to enable memoization (caching results of recursive calls)
from functools import lru_cache

# Define the sequence we want to find the Longest Increasing Subsequence (LIS) for
sequence = [10, 22, 9, 33, 21, 50, 41, 60]

# Use memoization to avoid recalculating LIS for the same subproblems
@lru_cache(maxsize=None)  # maxsize=None allows unlimited caching
def lis_ending_at(index):
    """
    Recursive function to find the length of the Longest Increasing Subsequence (LIS)
    ending at a given index.

    index: int - the current position in the sequence we're calculating LIS for
    Returns: int - length of LIS ending at this index
    """
    # Base case: LIS of a single element is always 1 (the element itself)
    max_length = 1

    # Check all previous indices to see if we can extend the LIS
    for prev_index in range(index):
        # Only consider previous elements smaller than the current one
        if sequence[prev_index] < sequence[index]:
            # Recursively find LIS ending at prev_index and extend by 1
            possible_length = lis_ending_at(prev_index) + 1
            # Update max_length if extending the LIS gives a longer sequence
            max_length = max(max_length, possible_length)

    # Return the maximum length of LIS ending at this index
    return max_length

# Compute LIS for all indices and take the maximum
lis_length = max(lis_ending_at(i) for i in range(len(sequence)))

# Print the length of the Longest Increasing Subsequence
print("Length of Longest Increasing Subsequence:", lis_length)

# ==========================
# Explanation of Output:
# ==========================
# The sequence analyzed: [10, 22, 9, 33, 21, 50, 41, 60]
# The LIS can be: [10, 22, 33, 50, 60] which has length 5
# Here's why:
# - Start with 10, extend to 22 (10 < 22)
# - 22 extends to 33 (22 < 33)
# - 33 extends to 50 (33 < 50)
# - 50 extends to 60 (50 < 60)
# Memoization ensures that repeated calculations (like LIS ending at the same index) are not recomputed,
# which drastically improves performance for larger sequences.
```
