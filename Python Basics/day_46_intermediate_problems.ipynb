```python
# Importing the random library to make random decisions (like fighting a monster)
import random

# A function that starts the game and introduces the player to the world
def intro():
    # The intro gives a message to the player, explaining the beginning of the game.
    print("Welcome to the Adventure Game!")  # Prints a welcome message.
    print("You find yourself standing in a dark room.")  # Describes the setting.
    print("There are two doors: one to the left and one to the right.")  # Gives options for the player.
    print("Choose which door to enter.")  # Instructs the player to make a choice.

# Function that handles what happens when the player chooses the left door
def left_door():
    # Prints what happens when the player opens the left door.
    print("You open the left door and find a treasure chest!")
    print("Do you want to open the chest or leave?")  # Gives the player another choice.

    # The input function asks the player to type 'open' or 'leave'
    choice = input("Type 'open' to open the chest, or 'leave' to leave: ").lower()  # Gets the player's choice in lowercase (to avoid case sensitivity).

    # Checking the player's choice:
    if choice == 'open':  # If the player chooses to open the chest
        print("You open the chest and find a golden coin. Congratulations! You win!")  # Player wins!
        return True  # The game ends with the player winning.
    elif choice == 'leave':  # If the player chooses to leave
        print("You leave the chest, and the door slams shut behind you. You are trapped forever!")  # Player loses!
        return False  # The game ends with the player losing.
    else:  # If the player doesn't type 'open' or 'leave'
        print("Invalid choice. You are stuck in the room forever!")  # Invalid choice message.
        return False  # The game ends with the player losing.

# Function that handles what happens when the player chooses the right door
def right_door():
    # Prints what happens when the player opens the right door.
    print("You open the right door and find a scary monster!")
    print("Do you want to fight the monster or run away?")  # Player is given two choices.

    # The input function asks the player to type 'fight' or 'run'
    choice = input("Type 'fight' to fight, or 'run' to run away: ").lower()  # The player's choice is taken and converted to lowercase.

    if choice == 'fight':  # If the player chooses to fight the monster
        outcome = random.choice(['win', 'lose'])  # Randomly pick between 'win' or 'lose'
        if outcome == 'win':  # If the outcome is a win
            print("You defeat the monster and find a hidden treasure. Congratulations! You win!")  # Player wins!
            return True  # The game ends with the player winning.
        else:  # If the outcome is a loss
            print("The monster defeats you. Game Over!")  # Player loses!
            return False  # The game ends with the player losing.
    elif choice == 'run':  # If the player chooses to run away
        print("You run away safely but you get lost in the dark forest. Game Over!")  # Player loses!
        return False  # The game ends with the player losing.
    else:  # If the player types anything other than 'fight' or 'run'
        print("Invalid choice. You are stuck in the room forever!")  # Invalid choice message.
        return False  # The game ends with the player losing.

# This is the main function that starts the game and controls the flow of the game.
def play_game():
    intro()  # Call the intro function to print the starting message.

    while True:  # This starts an infinite loop so that the game keeps asking for choices until it ends.
        print("\nWhich door will you choose? (left or right)")  # Ask the player which door they want to go through.
        choice = input("Type 'left' for the left door, or 'right' for the right door: ").lower()  # Take the player's choice.

        if choice == 'left':  # If the player chooses the left door
            if left_door():  # Call the left_door function, and if the player wins, break the loop and end the game.
                break
        elif choice == 'right':  # If the player chooses the right door
            if right_door():  # Call the right_door function, and if the player wins, break the loop and end the game.
                break
        else:  # If the player types something other than 'left' or 'right'
            print("Invalid choice. Please choose 'left' or 'right'.")  # Let the player know the choice was invalid.

# Call the play_game function to start the game when the code is run.
play_game()
```
