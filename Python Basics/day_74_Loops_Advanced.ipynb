```python
# ==============================================
# ADVANCED LOOPS IN PYTHON
# ==============================================

# -------------------------------
# 1. LOOPING WITH MULTIPLE VARIABLES
# -------------------------------

# Sometimes you need two variables in a loop.
# Example: print pairs of numbers

print("Example 1: Looping with two variables using range")
for i in range(1, 4):  # outer number 1,2,3
    for j in range(4, 7):  # inner number 4,5,6
        print("Pair:", i, j)  # prints each combination

# -------------------------------
# 2. ENUMERATE() -> Get index and value in a loop
# -------------------------------

fruits = ["apple", "banana", "cherry"]

print("\nExample 2: Using enumerate() to get index and item")
for index, fruit in enumerate(fruits):  # index = position, fruit = item
    print("Index", index, "->", fruit)

# -------------------------------
# 3. ZIP() -> Looping two lists together
# -------------------------------

names = ["Alice", "Bob", "Charlie"]
ages = [10, 11, 12]

print("\nExample 3: Using zip() to loop two lists together")
for name, age in zip(names, ages):  # pairs first element with first element
    print(name, "is", age, "years old")

# -------------------------------
# 4. LIST COMPREHENSIONS (Advanced loop shortcut)
# -------------------------------

# Normal loop to create a new list
squares = []
for i in range(1, 6):
    squares.append(i**2)  # i**2 = i squared
print("\nSquares using normal loop:", squares)

# Same thing with list comprehension (shorter)
squares2 = [i**2 for i in range(1, 6)]
print("Squares using list comprehension:", squares2)

# You can even add conditions in list comprehension
even_squares = [i**2 for i in range(1, 11) if i % 2 == 0]  # only even numbers
print("Squares of even numbers:", even_squares)

# -------------------------------
# 5. ADVANCED NESTED LOOPS FOR PATTERNS
# -------------------------------

print("\nExample 4: Number pyramid pattern")
rows = 5
for i in range(1, rows+1):  # row 1 to 5
    for j in range(1, i+1):  # numbers in each row
        print(j, end=" ")
    print()  # move to next row

print("\nExample 5: Inverted pyramid pattern")
for i in range(rows, 0, -1):  # rows from 5 down to 1
    for j in range(1, i+1):
        print("*", end=" ")
    print()  # move to next line

# -------------------------------
# 6. LOOPS + CONDITIONS FOR PROBLEMS
# -------------------------------

# Example: Print all prime numbers up to 20
print("\nExample 6: Prime numbers up to 20")
for num in range(2, 21):
    is_prime = True  # assume number is prime
    for i in range(2, num):  # check divisibility
        if num % i == 0:  # divisible by i
            is_prime = False  # not prime
            break  # no need to check further
    if is_prime:
        print(num, "is prime")

# -------------------------------
# 7. LOOPS + FUNCTIONS
# -------------------------------

print("\nExample 7: Function to print Fibonacci sequence using loops")

def fibonacci(n):
    a, b = 0, 1  # first two numbers
    sequence = []
    for _ in range(n):  # repeat n times
        sequence.append(a)
        a, b = b, a + b  # update numbers
    return sequence

print("First 10 Fibonacci numbers:", fibonacci(10))

# -------------------------------
# 8. MORE LOOP PRACTICE IDEAS
# -------------------------------
# - Create multiplication tables
# - Sum of numbers with conditions
# - Pattern problems (triangles, pyramids)
# - Nested loops with multiple lists
# - Using break/continue/else in complex loops
```
