```python

# ----------------------------------------------------------
# PROBLEM 1: COUNT TOTAL LETTERS USING A LOOP
# ----------------------------------------------------------
# We want to count letters manually (like a human)

word = "ELEPHANT"      # this word has letters
count = 0             # start counting from zero

for letter in word:   # loop through each letter
    count = count + 1 # add 1 for every letter we see

print("Total letters:", count)


# ----------------------------------------------------------
# PROBLEM 2: CHECK IF A NUMBER EXISTS IN A LIST
# ----------------------------------------------------------
# This uses a "flag" (YES / NO thinking)

numbers = [2, 4, 6, 8, 10]
found = False          # we assume number is NOT found

for n in numbers:
    if n == 6:         # check each number
        found = True  # we found it!

if found == True:
    print("Number is found ðŸ˜Š")
else:
    print("Number not found ðŸ˜¢")


# ----------------------------------------------------------
# PROBLEM 3: SUM OF NUMBERS UNTIL LOOP STOPS
# ----------------------------------------------------------
# Robot keeps adding until total becomes 10 or more

total = 0
number = 1

while total < 10:
    total = total + number
    number = number + 1

print("Final total:", total)


# ----------------------------------------------------------
# PROBLEM 4: MULTIPLICATION TABLES (1 TO 3)
# ----------------------------------------------------------
# Loop inside loop (advanced but easy)

for table in range(1, 4):        # table number
    print("Table of", table)

    for i in range(1, 6):        # numbers 1 to 5
        print(table, "x", i, "=", table * i)

    print()  # empty line


# ----------------------------------------------------------
# PROBLEM 5: COUNT HOW MANY NUMBERS ARE BIGGER THAN 5
# ----------------------------------------------------------

nums = [1, 3, 5, 7, 9, 2, 8]
count_big = 0

for n in nums:
    if n > 5:
        count_big = count_big + 1

print("Numbers bigger than 5:", count_big)


# ----------------------------------------------------------
# PROBLEM 6: BUILD A WORD LETTER BY LETTER
# ----------------------------------------------------------
# This shows how loops build things slowly

letters = ["C", "A", "T"]
word = ""

for letter in letters:
    word = word + letter   # add letter to word

print("Word made:", word)


# ----------------------------------------------------------
# PROBLEM 7: FIND SMALLEST NUMBER (ADVANCED THINKING)
# ----------------------------------------------------------

numbers = [9, 3, 7, 1, 8]

smallest = numbers[0]   # assume first is smallest

for n in numbers:
    if n < smallest:
        smallest = n

print("Smallest number:", smallest)


# ----------------------------------------------------------
# PROBLEM 8: STAR PYRAMID (ADVANCED PATTERN)
# ----------------------------------------------------------
# Each row prints more stars

for row in range(1, 6):
    print("*" * row)


# ----------------------------------------------------------
# PROBLEM 9: STOP LOOP WHEN WRONG NUMBER COMES
# ----------------------------------------------------------

numbers = [2, 4, 6, 9, 8]

for n in numbers:
    if n % 2 != 0:   # if number is odd
        print("Odd number found:", n)
        break        # stop loop
    print("Even number:", n)


# ----------------------------------------------------------
# PROBLEM 10: ROBOT WALKING STEPS
# ----------------------------------------------------------
# Robot walks until step 5

step = 1

while True:
    print("Robot step:", step)

    if step == 5:
        print("Robot stops ðŸ›‘")
        break

    step = step + 1


# ==========================================================
# ðŸŽ‰ YOU ARE DOING REAL PROGRAMMING ðŸŽ‰
# ==========================================================
```
