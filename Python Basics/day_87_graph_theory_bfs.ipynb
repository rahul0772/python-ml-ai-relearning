```python

# ============================================================
# GRAPH THEORY - BREADTH FIRST SEARCH (BFS)
# ============================================================
#
# Imagine a graph like a map of cities connected by roads.
# Each city = a NODE (also called vertex)
# Each road = an EDGE (connection between nodes)
#
# Example:
# A --- B --- D
# |     |
# C     E
#
# Our goal:
# 👉 Start from one node (like A)
# 👉 Visit all connected nodes step by step
#
# BFS = Breadth First Search
# 👉 It explores LEVEL BY LEVEL (like ripples in water)
#
# Think like:
# Level 0 → A
# Level 1 → B, C
# Level 2 → D, E
#
# We use a QUEUE (FIFO: First In First Out)
# 👉 Like a line of people waiting
#
# ============================================================
# STEP 1: CREATE GRAPH USING DICTIONARY
# ============================================================

# We represent graph using dictionary
# Key = node
# Value = list of connected neighbors

graph = {
    'A': ['B', 'C'],     # A connects to B and C
    'B': ['A', 'D', 'E'],# B connects to A, D, E
    'C': ['A'],          # C connects to A
    'D': ['B'],          # D connects to B
    'E': ['B']           # E connects to B
}

# Why dictionary?
# 👉 Fast lookup
# 👉 Easy to represent adjacency list
# 👉 Very common in graph problems

# ============================================================
# STEP 2: DEFINE BFS FUNCTION
# ============================================================

def bfs(graph, start_node):
    # graph → our data structure
    # start_node → where we begin traversal

    # visited set → to avoid visiting same node again
    visited = set()
    # Why set?
    # 👉 Fast lookup (O(1))
    # 👉 Avoid duplicates easily

    # queue → stores nodes to visit
    queue = []

    # Step 1: add starting node
    queue.append(start_node)
    visited.add(start_node)

    # Explanation:
    # 👉 We push start node into queue
    # 👉 Also mark it visited immediately

    # Store traversal result
    traversal_order = []

    # ========================================================
    # MAIN BFS LOOP
    # ========================================================

    while queue:
        # Take first element from queue
        current = queue.pop(0)

        # Why pop(0)?
        # 👉 Because queue works FIFO (First In First Out)
        # 👉 Remove oldest element first

        # Add to traversal result
        traversal_order.append(current)

        # Now explore neighbors of current node
        for neighbor in graph[current]:

            # If neighbor NOT visited yet
            if neighbor not in visited:

                # Mark as visited
                visited.add(neighbor)

                # Add to queue for future processing
                queue.append(neighbor)

                # Why add to queue?
                # 👉 So we process it later in correct order
                # 👉 Ensures level-by-level traversal

    # Return final BFS traversal order
    return traversal_order

# ============================================================
# STEP 3: RUN BFS
# ============================================================

start = 'A'  # starting node

result = bfs(graph, start)

# ============================================================
# STEP 4: PRINT OUTPUT
# ============================================================

print("BFS Traversal Order:", result)

# ============================================================
# OUTPUT EXPLANATION (VERY IMPORTANT 🧠)
# ============================================================
#
# Expected Output:
# BFS Traversal Order: ['A', 'B', 'C', 'D', 'E']
#
# WHY THIS OUTPUT?
#
# Step-by-step:
#
# Start at A
# Queue = [A]
#
# 1️⃣ Visit A
# → Add neighbors B, C
# Queue = [B, C]
#
# 2️⃣ Visit B
# → Add neighbors D, E (A already visited)
# Queue = [C, D, E]
#
# 3️⃣ Visit C
# → Only A (already visited)
# Queue = [D, E]
#
# 4️⃣ Visit D
# → Only B (already visited)
# Queue = [E]
#
# 5️⃣ Visit E
# → Only B (already visited)
# Queue = []
#
# DONE ✅
#
# FINAL ORDER:
# A → B → C → D → E
#
# ============================================================
# 🧠 SUPER IMPORTANT INSIGHT:
# ============================================================
#
# BFS ALWAYS:
# ✔ Visits nearest nodes first
# ✔ Uses QUEUE
# ✔ Works level by level
#
# ============================================================
# 🚀 REAL LIFE USE CASES:
# ============================================================
#
# ✔ Shortest path in unweighted graphs
# ✔ Finding minimum moves in games
# ✔ Social networks (friends of friends)
# ✔ Web crawling
#
# BFS = "Breadth First = Level First"
# QUEUE = "First Come First Serve"
#
# ============================================================
```

```python
# ALT
# -------------------------------
# STEP 1: CREATE GRAPH
# -------------------------------

graph = {
    # We are creating a GRAPH using a Python DICTIONARY.
    # A dictionary stores data in KEY : VALUE format.
    # Here:
    # KEY = Node (like A, B, C)
    # VALUE = List of connected nodes (neighbors)

    'A': ['B', 'C'],
    # Node A is connected to B and C
    # Example: If you are at city A, you can go to B or C

    'B': ['A', 'D', 'E'],
    # Node B connects to A, D, E
    # Example: From B, you can go to A, D, or E

    'C': ['A'],
    # Node C connects only to A

    'D': ['B'],
    # Node D connects only to B

    'E': ['B']
    # Node E connects only to B
}

# ----------------------------------------------------------
# VISUAL REPRESENTATION:
#
#        A
#       / \
#      B   C
#     / \
#    D   E
#
# ----------------------------------------------------------


# -------------------------------
# STEP 2: DEFINE BFS FUNCTION
# -------------------------------

def bfs(graph, start_node):
    # "def" means we are DEFINING a function
    # bfs = function name
    # graph = input graph
    # start_node = where we start traversal

    visited = set()
    # Create an EMPTY SET to store visited nodes
    # Why set?
    # 👉 Fast lookup (very fast checking)
    # 👉 Prevent visiting same node again

    queue = []
    # Create EMPTY LIST to use as QUEUE
    # Queue = First In First Out (FIFO)
    # Example:
    # Add: A → B → C
    # Remove: A first

    queue.append(start_node)
    # Add starting node into queue
    # Example:
    # If start_node = 'A'
    # queue = ['A']

    visited.add(start_node)
    # Mark A as visited
    # visited = {'A'}

    traversal_order = []
    # This will store final BFS result
    # Example: ['A', 'B', 'C', ...]


    # -------------------------------
    # MAIN LOOP
    # -------------------------------

    while queue:
        # "while queue" means:
        # Run loop until queue becomes EMPTY
        # If queue = ['A'], loop runs
        # If queue = [], loop stops

        current = queue.pop(0)
        # Remove FIRST element from queue
        # pop(0) removes index 0 (front of queue)

        # Example:
        # queue = ['A', 'B', 'C']
        # current = 'A'
        # queue becomes ['B', 'C']

        traversal_order.append(current)
        # Add current node into result list

        # Example:
        # traversal_order = ['A']


        for neighbor in graph[current]:
            # Loop through neighbors of current node

            # Example:
            # current = 'A'
            # graph['A'] = ['B', 'C']
            # neighbor = 'B', then 'C'


            if neighbor not in visited:
                # Check if neighbor is NOT visited

                # Example:
                # visited = {'A'}
                # neighbor = 'B' → not visited → TRUE

                visited.add(neighbor)
                # Mark neighbor as visited

                # Example:
                # visited = {'A', 'B'}

                queue.append(neighbor)
                # Add neighbor to queue

                # Example:
                # queue = ['B']


    return traversal_order
    # Return final BFS order


# -------------------------------
# STEP 3: RUN FUNCTION
# -------------------------------

start = 'A'
# We choose starting node as 'A'

result = bfs(graph, start)
# Call BFS function
# graph → our graph
# start → 'A'

# Function runs step by step internally


# -------------------------------
# STEP 4: PRINT RESULT
# -------------------------------

print("BFS Traversal Order:", result)
# Print final traversal


# ============================================================
# 🧠 OUTPUT EXPLANATION
# ============================================================
#
# Output:
# BFS Traversal Order: ['A', 'B', 'C', 'D', 'E']
#
# FULL STEP-BY-STEP EXECUTION:
#
# INITIAL:
# queue = ['A']
# visited = {'A'}
#
# ---------------------------------
# STEP 1:
# current = 'A'
# traversal = ['A']
# neighbors = ['B', 'C']
# queue = ['B', 'C']
# visited = {'A', 'B', 'C'}
#
# ---------------------------------
# STEP 2:
# current = 'B'
# traversal = ['A', 'B']
# neighbors = ['A', 'D', 'E']
# A already visited
# queue = ['C', 'D', 'E']
# visited = {'A', 'B', 'C', 'D', 'E'}
#
# ---------------------------------
# STEP 3:
# current = 'C'
# traversal = ['A', 'B', 'C']
# neighbors = ['A'] (already visited)
# queue = ['D', 'E']
#
# ---------------------------------
# STEP 4:
# current = 'D'
# traversal = ['A', 'B', 'C', 'D']
# neighbors = ['B'] (already visited)
# queue = ['E']
#
# ---------------------------------
# STEP 5:
# current = 'E'
# traversal = ['A', 'B', 'C', 'D', 'E']
# neighbors = ['B'] (already visited)
# queue = []
#
# LOOP ENDS ✅
#
# FINAL RESULT:
# ['A', 'B', 'C', 'D', 'E']
#
# ============================================================
#
# BFS = LEVEL BY LEVEL
# QUEUE = FIRST IN FIRST OUT
#
# ============================================================
```

```python
start = 'D'

result = bfs(graph, start)


print("BFS Traversal Order:", result)
```
