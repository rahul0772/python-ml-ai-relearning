```python
# ===============================================================
# INTRODUCTION TO GRAPH THEORY
# ===============================================================

# ---------------------------------------------------------------
# PART 1: WHAT IS A GRAPH?
# ---------------------------------------------------------------

# A GRAPH is a way to represent things and their connections.
#
# Example:
# Imagine cities and roads.
#
# City A ---- road ---- City B
# City B ---- road ---- City C
#
# Cities = NODES (also called VERTICES)
# Roads = EDGES (connections between nodes)
#
# In graph theory:
# Node = a point
# Edge = a connection between two nodes
#
# Example graph:
#
#    A
#   / \
#  B   C
#   \
#    D
#
# Nodes: A, B, C, D
# Edges:
# A-B
# A-C
# B-D


# ---------------------------------------------------------------
# PART 2: HOW DO WE STORE A GRAPH IN PYTHON?
# ---------------------------------------------------------------

# One very common way is called an "ADJACENCY LIST".

# Adjacency list means:
# Each node keeps a list of its neighbors (connected nodes)

# Example:
# A -> [B, C]
# B -> [A, D]
# C -> [A]
# D -> [B]

# This means:
# A connects to B and C
# B connects to A and D


# ---------------------------------------------------------------
# CREATE AN EMPTY GRAPH
# ---------------------------------------------------------------

# We will use a Python DICTIONARY.
# A dictionary stores key-value pairs.

# Example dictionary:
# {
#   "A": ["B","C"],
#   "B": ["A","D"]
# }

graph = {}  # This creates an empty dictionary to store the graph


# ---------------------------------------------------------------
# PART 3: ADDING A NODE
# ---------------------------------------------------------------

def add_node(graph, node):
    """
    Function to add a node into the graph
    """

    # First we check if node already exists in graph
    # "node not in graph" means the node key does not exist

    if node not in graph:

        # If node does not exist
        # we create a new key in dictionary

        # graph[node] means create key named node

        # We assign it an empty list []
        # because it has no neighbors yet

        graph[node] = []

        # Print message so we see what happened
        print("Node added:", node)

    else:
        # If node already exists
        print("Node already exists:", node)


# ---------------------------------------------------------------
# PART 4: ADDING AN EDGE
# ---------------------------------------------------------------

def add_edge(graph, node1, node2):
    """
    This function connects two nodes together
    """

    # First check if node1 exists
    if node1 not in graph:
        add_node(graph, node1)

    # Check if node2 exists
    if node2 not in graph:
        add_node(graph, node2)

    # Now we connect the nodes

    # graph[node1] gives list of neighbors of node1

    # append() means add item to list

    graph[node1].append(node2)

    # For an undirected graph we also add reverse connection

    graph[node2].append(node1)

    print("Edge added between", node1, "and", node2)


# ---------------------------------------------------------------
# PART 5: PRINT GRAPH
# ---------------------------------------------------------------

def print_graph(graph):

    # Loop through every node in graph

    # graph.items() gives (node, neighbors)

    for node, neighbors in graph.items():

        # Print node and its neighbors

        print(node, "->", neighbors)


# ---------------------------------------------------------------
# CREATE A SAMPLE GRAPH
# ---------------------------------------------------------------

add_edge(graph, "A", "B")
add_edge(graph, "A", "C")
add_edge(graph, "B", "D")
add_edge(graph, "C", "E")

print("\nGraph structure:")
print_graph(graph)


# ---------------------------------------------------------------
# PART 6: BREADTH FIRST SEARCH (BFS)
# ---------------------------------------------------------------

# BFS explores nodes layer by layer.

# Example:
#
# Start from A
#
# Level 0: A
# Level 1: B C
# Level 2: D E

from collections import deque

def bfs(graph, start):

    # visited set keeps track of visited nodes
    visited = set()

    # Queue structure
    # BFS uses a queue (First In First Out)

    queue = deque()

    # Start node goes in queue
    queue.append(start)

    # Mark start visited
    visited.add(start)

    print("\nBFS traversal starting from:", start)

    while queue:

        # popleft removes first element of queue
        node = queue.popleft()

        print(node, end=" ")

        # visit neighbors
        for neighbor in graph[node]:

            # If neighbor not visited
            if neighbor not in visited:

                visited.add(neighbor)

                queue.append(neighbor)


bfs(graph, "A")


# ---------------------------------------------------------------
# PART 7: DEPTH FIRST SEARCH (DFS)
# ---------------------------------------------------------------

# DFS explores deep first.

# Example path:
# A -> B -> D -> back -> C -> E

def dfs(graph, start):

    visited = set()

    stack = []

    stack.append(start)

    print("\nDFS traversal starting from:", start)

    while stack:

        node = stack.pop()

        if node not in visited:

            print(node, end=" ")

            visited.add(node)

            # Add neighbors to stack
            for neighbor in graph[node]:

                if neighbor not in visited:
                    stack.append(neighbor)


dfs(graph, "A")


# ---------------------------------------------------------------
# PART 8: KEY CONCEPT SUMMARY
# ---------------------------------------------------------------

print("\n\nIMPORTANT CONCEPTS:")
print("1. Graph = nodes + edges")
print("2. Node = object or point")
print("3. Edge = connection between nodes")
print("4. Adjacency List = dictionary storing neighbors")
print("5. BFS = explores level by level")
print("6. DFS = explores deep paths first")
```

```python
# ================================================================
# INTRODUCTION TO GRAPH THEORY USING PYTHON
# ================================================================


# ------------------------------------------------
# PART 1 : WHAT IS A GRAPH ?
# ------------------------------------------------

# A GRAPH is a structure used in mathematics and computer science.
# It helps represent relationships between things.

# Example real world graph:
# Cities connected by roads.

# Example:
# Mumbai ---- Pune
# Pune   ---- Delhi
# Delhi  ---- Bangalore

# In graph theory:
# Cities = NODES (also called vertices)
# Roads  = EDGES (connections)

# Graph = collection of nodes + edges


# ------------------------------------------------
# PART 2 : HOW DO WE STORE A GRAPH IN PYTHON ?
# ------------------------------------------------

# There are multiple ways:
# 1. Adjacency Matrix
# 2. Adjacency List

# We will use ADJACENCY LIST because:
# ✔ easier to understand
# ✔ efficient
# ✔ used commonly

# Example graph:

# A -- B
# |    |
# C -- D

# Adjacency List representation:

# A -> B, C
# B -> A, D
# C -> A, D
# D -> B, C

# Python dictionary is perfect for this!

# dictionary structure:
# {
#   node : [list_of_connected_nodes]
# }


# ------------------------------------------------
# PART 3 : CREATE GRAPH CLASS
# ------------------------------------------------

# Class = blueprint to create objects
# Think of it like a "machine design"

class Graph:

    # ------------------------------------------------
    # CONSTRUCTOR METHOD
    # ------------------------------------------------
    # __init__ runs automatically when object is created
    def __init__(self):

        # self means "this object itself"

        # adjacency_list will store graph connections
        # It is a dictionary

        # Example:
        # {
        #   'A': ['B','C'],
        #   'B': ['A']
        # }

        self.adjacency_list = {}


    # ------------------------------------------------
    # ADD NODE FUNCTION
    # ------------------------------------------------

    def add_node(self, node):

        # Check if node already exists
        # "not in" checks membership in dictionary

        if node not in self.adjacency_list:

            # Create empty list for connections
            self.adjacency_list[node] = []

        # Example after adding A:
        # { 'A': [] }


    # ------------------------------------------------
    # ADD EDGE FUNCTION
    # ------------------------------------------------

    def add_edge(self, node1, node2):

        # An edge connects two nodes

        # Step 1 : ensure nodes exist
        self.add_node(node1)
        self.add_node(node2)

        # Step 2 : connect node1 -> node2
        self.adjacency_list[node1].append(node2)

        # Step 3 : connect node2 -> node1
        # This makes graph UNDIRECTED

        self.adjacency_list[node2].append(node1)



    # ------------------------------------------------
    # PRINT GRAPH
    # ------------------------------------------------

    def print_graph(self):

        print("\nGRAPH STRUCTURE (Adjacency List):")

        # .items() lets us loop through dictionary
        # node = key
        # neighbors = value list

        for node, neighbors in self.adjacency_list.items():

            print(node, "->", neighbors)



    # ------------------------------------------------
    # DEPTH FIRST SEARCH (DFS)
    # ------------------------------------------------

    def dfs(self, start):

        print("\nDEPTH FIRST SEARCH STARTING FROM:", start)

        # visited set prevents infinite loops
        visited = set()

        # stack helps DFS go deep
        stack = [start]

        while stack:

            # pop removes last item
            node = stack.pop()

            if node not in visited:

                print("Visited:", node)

                visited.add(node)

                # Add neighbors to stack
                for neighbor in self.adjacency_list[node]:

                    if neighbor not in visited:
                        stack.append(neighbor)



    # ------------------------------------------------
    # BREADTH FIRST SEARCH (BFS)
    # ------------------------------------------------

    def bfs(self, start):

        print("\nBREADTH FIRST SEARCH STARTING FROM:", start)

        visited = set()

        # queue structure
        queue = [start]

        visited.add(start)

        while queue:

            # pop first element
            node = queue.pop(0)

            print("Visited:", node)

            for neighbor in self.adjacency_list[node]:

                if neighbor not in visited:

                    visited.add(neighbor)

                    queue.append(neighbor)



    # ------------------------------------------------
    # SHORTEST PATH (UNWEIGHTED GRAPH)
    # ------------------------------------------------

    def shortest_path(self, start, goal):

        print("\nFINDING SHORTEST PATH FROM", start, "TO", goal)

        visited = set()

        queue = [[start]]  # list of paths

        while queue:

            path = queue.pop(0)

            node = path[-1]

            if node == goal:
                print("Shortest Path Found:", path)
                return path

            if node not in visited:

                visited.add(node)

                for neighbor in self.adjacency_list[node]:

                    new_path = list(path)

                    new_path.append(neighbor)

                    queue.append(new_path)

        print("No path found")
        return None



# ------------------------------------------------
# PART 4 : CREATE GRAPH OBJECT
# ------------------------------------------------

# Object = real instance of class
g = Graph()



# ------------------------------------------------
# PART 5 : ADD EDGES
# ------------------------------------------------

g.add_edge("A","B")
g.add_edge("A","C")
g.add_edge("B","D")
g.add_edge("C","D")
g.add_edge("D","E")



# ------------------------------------------------
# PART 6 : PRINT GRAPH
# ------------------------------------------------

g.print_graph()



# ------------------------------------------------
# PART 7 : RUN SEARCH ALGORITHMS
# ------------------------------------------------

g.dfs("A")

g.bfs("A")



# ------------------------------------------------
# PART 8 : SHORTEST PATH
# ------------------------------------------------

g.shortest_path("A","E")



# ================================================================
# FINAL CONCEPT SUMMARY
# ================================================================

# Graph = Nodes + Edges

# Node = point (A, B, C)
# Edge = connection (A-B)

# Adjacency List = dictionary storing connections

# DFS = go deep first
# BFS = explore layer by layer

# Shortest path in unweighted graph = BFS
```
