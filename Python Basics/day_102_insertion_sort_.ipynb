```python
# =========================================================
# INSERTION SORT IN PYTHON
# =========================================================

# Insertion Sort is a simple sorting algorithm that builds the final sorted list one element at a time.
# Think of how you might sort playing cards in your hand: you take one card at a time and insert it into the correct position.

# Let's start by creating an unsorted list of numbers
numbers = [8, 3, 5, 2, 9]
# Here, 'numbers' is a Python list containing integers.
# Each number represents a "card" in our hand that we want to sort from smallest to largest.
# Initial list: [8, 3, 5, 2, 9] (unsorted)

# Now we will implement insertion sort step by step
for i in range(1, len(numbers)):
    # 'i' is the index of the element we are currently trying to insert into the sorted part of the list.
    # We start from 1 because the element at index 0 is trivially "sorted" (a single element is always sorted).

    key = numbers[i]
    # 'key' stores the value of the current element that we want to place at the correct position.
    # Example: When i=1, key=3. We want to find where 3 should go in the sorted portion [8].

    j = i - 1
    # 'j' is the index of the last element in the "sorted portion" of the list.
    # We will use 'j' to compare elements in the sorted part and find the correct position for 'key'.

    # Now we move elements that are greater than 'key' one position ahead to make space for 'key'
    while j >= 0 and numbers[j] > key:
        # Condition explanation:
        # 1. j >= 0 ensures we do not go out of the list bounds.
        # 2. numbers[j] > key means we shift elements to the right if they are bigger than our 'key'.
        numbers[j + 1] = numbers[j]
        # We move the element at index 'j' one position ahead to make space for 'key'.
        # Example: If numbers = [8,3,5,2,9] and key=3, we move 8 to index 1.
        j -= 1
        # We decrease 'j' to check the previous element in the sorted portion.

    # Insert the 'key' at the correct position found
    numbers[j + 1] = key
    # After shifting larger elements, we place 'key' in the correct position.
    # Example: numbers = [3, 8, 5, 2, 9] after first pass.

    # Print the list after each insertion to see how the list is gradually sorted
    print(f"After inserting element at index {i} (value={key}): {numbers}")

# Final sorted list
print("\nFinal sorted list:", numbers)
# Explanation of final output:
# The list has been sorted in ascending order using insertion sort.
# Step by step, each element was taken ('key') and placed into the correct position among the already sorted elements.
# This mimics the human way of sorting cards: pick one card at a time and place it correctly.
```
