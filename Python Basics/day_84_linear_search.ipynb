```python
# Linear Search

# Linear Search is a basic algorithm to search for an element in a list by checking each element one by one.

def linear_search(arr, target):
    # The function linear_search takes in two parameters:
    # arr (a list of elements) and target (the element we want to search for).

    # We will iterate through each element in the list 'arr' using a for loop.
    for index in range(len(arr)):  # Iterate through every index of the list 'arr'
        # For each element, we check if it is equal to the target element.
        if arr[index] == target:  # If the element matches the target value
            return index  # We found the target! Return the index of the target in the list.

    # If we finish the loop and the target was not found, return -1 to indicate not found.
    return -1  # If target is not found, return -1

# Example list to search in
arr = [5, 3, 8, 6, 7, 2]

# The target element we want to find in the list
target = 6

# Calling the linear_search function with the list 'arr' and target value 'target'
result = linear_search(arr, target)

# Printing the result of the linear search function
print("Index of target:", result)

# Explanation of Output:
# The linear_search function looks through each element in the list to find if it matches the target.
# It starts with index 0 and checks if arr[0] (5) is equal to the target (6).
# It goes through each index, comparing the value at that index with the target (6).
# When it reaches index 3 (arr[3] = 6), it finds the target and returns the index 3.
# Therefore, the output will be:
# "Index of target: 3"
# This means the target 6 is found at index 3 in the list.
```
