```python
# ============================================
# BINARY SEARCH 🚀
# ============================================
# Today we randomly picked topic: BINARY SEARCH (from search algorithms)
#
# 🧠 CORE IDEA (VERY IMPORTANT — BURN THIS INTO YOUR BRAIN):
# Instead of checking EVERY element one by one (like linear search),
# Binary Search works by repeatedly dividing the search space into HALF.
#
# ⚡ REQUIREMENT:
# The list MUST be SORTED (very very important condition).
#
# 🔥 WHY?
# Because we decide to go LEFT or RIGHT based on comparison.
# If unsorted → decision becomes meaningless.
#
# Think like guessing a number between 1–100:
# You don't start from 1 → you go to middle (50), then adjust.
# That’s EXACTLY what binary search does.

# --------------------------------------------
# STEP 1: DEFINE THE FUNCTION
# --------------------------------------------
def binary_search(arr, target):
    # arr → the sorted list in which we search
    # target → the value we want to find

    # ----------------------------------------
    # STEP 2: DEFINE SEARCH BOUNDARIES
    # ----------------------------------------
    left = 0
    # left pointer starts at index 0 (beginning of list)

    right = len(arr) - 1
    # right pointer starts at last index of list

    # ----------------------------------------
    # STEP 3: LOOP UNTIL SEARCH SPACE EXISTS
    # ----------------------------------------
    while left <= right:
        # This means: "as long as there is a valid range to search"
        # If left crosses right → search space is gone → element not found

        # ------------------------------------
        # STEP 4: FIND MIDDLE INDEX
        # ------------------------------------
        mid = (left + right) // 2
        # WHY // ?
        # Because we want integer index (no decimals allowed in list indexing)

        # Example:
        # left = 0, right = 9 → mid = (0+9)//2 = 4

        # ------------------------------------
        # STEP 5: CHECK IF MID IS TARGET
        # ------------------------------------
        if arr[mid] == target:
            # If middle element equals target → SUCCESS 🎯
            return mid  # return index where found

        # ------------------------------------
        # STEP 6: DECIDE WHICH SIDE TO SEARCH
        # ------------------------------------
        elif arr[mid] < target:
            # If middle element is SMALLER than target
            # → target must be on RIGHT side

            left = mid + 1
            # Move left pointer forward
            # WHY mid + 1?
            # Because mid is already checked → skip it

        else:
            # If middle element is GREATER than target
            # → target must be on LEFT side

            right = mid - 1
            # Move right pointer backward
            # Skip mid again because already checked

    # ----------------------------------------
    # STEP 7: IF NOT FOUND
    # ----------------------------------------
    return -1
    # Convention: return -1 means "not found"

# ============================================
# 🔍 LET'S TEST OUR BRAIN + CODE
# ============================================

# Sorted list (MANDATORY for binary search)
numbers = [2, 5, 8, 12, 16, 23, 38, 56, 72, 91]

# Target we want to find
target_value = 23

# Call the function
result = binary_search(numbers, target_value)

# Print result
print("Index of target:", result)

# ============================================
# 🧠 OUTPUT EXPLANATION (VERY IMPORTANT)
# ============================================
# Let's TRACE mentally step-by-step how algorithm worked:
#
# Initial:
# left = 0, right = 9
#
# Step 1:
# mid = (0+9)//2 = 4 → numbers[4] = 16
# 16 < 23 → go RIGHT → left = 5
#
# Step 2:
# mid = (5+9)//2 = 7 → numbers[7] = 56
# 56 > 23 → go LEFT → right = 6
#
# Step 3:
# mid = (5+6)//2 = 5 → numbers[5] = 23
# 🎯 FOUND!
#
# So output = 5
#
# --------------------------------------------
# 🔥 FINAL MENTAL MODEL (LOCK THIS IN):
# --------------------------------------------
# Each step cuts problem into HALF:
#
# 10 elements → 5 → 2 → 1 → DONE
#
# This is why Binary Search is SUPER FAST ⚡
#
# Time Complexity:
# O(log n)  ← VERY IMPORTANT (logarithmic time)
#
# --------------------------------------------
# 💡 KEY MEMORY TRIGGERS:
# --------------------------------------------
# Sorted array? → Use Binary Search
# Need speed? → Binary Search
# Divide into halves? → Binary Search
#
# --------------------------------------------
# 🧠 MASTER LEVEL UNDERSTANDING:
# --------------------------------------------
# Binary Search is NOT just code.
# It is a WAY OF THINKING:
#
# "Eliminate half possibilities instantly."
#
# This concept appears everywhere:
# - Searching
# - Optimization problems
# - Decision making
```
