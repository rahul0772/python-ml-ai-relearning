#Organizing a List
### Sorting a List Permanently with the sort() Method

```python
fruits = ['orange', 'apple', 'pear', 'watermelon']
fruits.sort()
print(fruits)
```

```python
fruits = ['orange', 'apple', 'pear', 'watermelon']
fruits.sort(reverse=True)
print(fruits)
```

### Sorting a List Temporarily with the sorted() Function

```python
fruits = ['orange', 'apple', 'pear', 'watermelon']
print("\nfruits in sorted order:")
print(sorted(fruits))
print(fruits)
print("\nfruits in reverse order:")
print(sorted(fruits, reverse=True))
```

```python
#slicing
fruits = [ "plum","watermelon", "raspberry", "orange", "grapes", "litchi", "pear", "guava", "pineapple", "cherry", "fig", "peach"]
print(fruits[0:3])
print(fruits[3:8])
print(fruits[:2])
print(fruits[2:])
print(fruits[:-2]) #last two values
print(fruits[-2:])
```

```python
#Loopin through slice
fruits = [ "plum","watermelon", "raspberry", "orange", "grapes", "litchi", "pear", "guava", "pineapple", "cherry", "fig", "peach"]
print("iteration in a list item:")
for fruit in fruits[3]:
  print(fruit.title())

print("\nfirst three fruits:")
for fruit in fruits[:3]:
  print(fruit.title())
```
