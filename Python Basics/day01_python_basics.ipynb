```python
#Removing an Item Using the pop() Method
Fruits = ['orange', 'apple', 'pear', 'watermelon']
Fruits[0] = "dragonfruit"
print(Fruits)

popped_Fruits = Fruits.pop()
print(popped_Fruits)
print(Fruits)

print("The last fruit I ate was  " + popped_Fruits.title() + ".")
```

```python
#Popping Items from any Position in a List

Fruits = ['orange', 'apple', 'pear', 'watermelon']
Fruits[0] = "dragonfruit"
print(Fruits)

first_Fruits = Fruits.pop(2)
print(first_Fruits)
print("The first fruit I ate today " + first_Fruits.title() + "." )
```

```python
# Removing an Item by Value

Fruits = ['orange', 'apple', 'pear', 'watermelon']
print(Fruits)

Fruits.remove("pear")
print(Fruits)
```

#Organizing a List
### Sorting a List Permanently with the sort() Method

```python
fruits = ['orange', 'apple', 'pear', 'watermelon']
fruits.sort()
print(fruits)
```

```python
fruits = ['orange', 'apple', 'pear', 'watermelon']
fruits.sort(reverse=True)
print(fruits)
```

### Sorting a List Temporarily with the sorted() Function

```python
fruits = ['orange', 'apple', 'pear', 'watermelon']
print("\nfruits in sorted order:")
print(sorted(fruits))
print(fruits)
print("\nfruits in reverse order:")
print(sorted(fruits, reverse=True))
```
