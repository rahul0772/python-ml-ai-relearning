```python
# -----------------------------
# INTERMEDIATE PYTHON LOOPS
# -----------------------------

# -----------------------------
# PROBLEM 1: PRINT NUMBERS FROM 1 TO 10
# -----------------------------
# Goal: We want Python to print numbers from 1 to 10.
# Concept: Loops repeat actions. We use "for loops" to do something many times.

# 'for' loop syntax:
# for variable in iterable:
#     # do something

# Here, 'variable' is a name we give for the item in the loop
# 'iterable' is something we can go through step by step (like a list of numbers)

# Step 1: Using range() to create numbers 1 to 10
# range(start, stop, step) generates numbers starting from 'start', up to 'stop-1', with a step size 'step'.
# Example: range(1, 11, 1) → 1, 2, 3, ..., 10
# start = 1, stop = 11 (stop is always 1 more than the last number you want), step = 1 (go 1 number at a time)

for number in range(1, 11, 1):  # 'number' will take each value from 1 to 10 one by one
    print(number)               # print the current value of 'number'

# Explanation:
# 1st iteration: number = 1 → prints 1
# 2nd iteration: number = 2 → prints 2
# ...
# 10th iteration: number = 10 → prints 10
# Loop ends after the last number (10) is printed

# -----------------------------
# PROBLEM 2: PRINT EVEN NUMBERS BETWEEN 1 AND 20
# -----------------------------
# We only want numbers divisible by 2
# Concept: Use if condition inside loops

for number in range(1, 21):    # number goes from 1 to 20
    if number % 2 == 0:        # % is modulo, gives remainder when number is divided by 2
        # If remainder is 0 → number is divisible by 2 → even number
        print(number)          # print even number

# Explanation:
# 1 % 2 = 1 → not even → skip
# 2 % 2 = 0 → even → print 2
# 3 % 2 = 1 → not even → skip
# 4 % 2 = 0 → even → print 4
# ... continues until 20

# -----------------------------
# PROBLEM 3: NESTED LOOPS (MULTIPLICATION TABLE)
# -----------------------------
# Goal: Print multiplication table from 1 to 3
# Concept: Nested loops → a loop inside another loop
# Outer loop: controls the first number
# Inner loop: controls the second number

for i in range(1, 4):             # i goes 1 to 3 (outer loop)
    for j in range(1, 4):         # j goes 1 to 3 (inner loop)
        result = i * j            # multiply outer and inner numbers
        print(f"{i} x {j} = {result}")  # formatted print
    print("------")               # separates tables for clarity

# Explanation:
# i=1 → inner loop j=1,2,3 → prints 1x1=1, 1x2=2, 1x3=3
# i=2 → inner loop j=1,2,3 → prints 2x1=2, 2x2=4, 2x3=6
# i=3 → inner loop j=1,2,3 → prints 3x1=3, 3x2=6, 3x3=9
# '------' just separates the tables visually

# -----------------------------
# PROBLEM 4: USING BREAK AND CONTINUE
# -----------------------------
# Goal: Learn how to stop a loop early or skip certain steps
# 'break' → stops the entire loop
# 'continue' → skips current step and moves to next iteration

# Example: print numbers 1 to 10, but stop if number is 6
for number in range(1, 11):
    if number == 6:
        print("We reached 6, stopping the loop!")
        break  # stops loop completely
    print(number)

# Example: print numbers 1 to 10, but skip 5
for number in range(1, 11):
    if number == 5:
        print("Skipping number 5!")
        continue  # skips this iteration
    print(number)

# -----------------------------
# PROBLEM 5: FOR LOOP WITH ELSE
# -----------------------------
# Goal: Learn about 'else' with loops
# 'else' in loops runs only if loop was NOT stopped by break
# Concept: Useful when searching for something

# Example: check if 7 exists in numbers 1 to 10
for number in range(1, 11):
    if number == 7:
        print("Found 7!")
        break  # stops loop
else:
    print("7 was not found")  # only runs if no break occurred

# -----------------------------
# SUMMARY OF INTERMEDIATE LOOPS
# -----------------------------
# 1. for loops are for repeating actions over a sequence
# 2. range(start, stop, step) generates numbers
# 3. Nested loops help with multiple levels of repetition
# 4. if statements inside loops can filter or control actions
# 5. break stops the loop, continue skips current iteration
# 6. else after loop runs only if loop did not encounter break
```
