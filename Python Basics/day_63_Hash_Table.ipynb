```python
# =========================================================
# HASH TABLES IN PYTHON
# =========================================================

# --------------------------
# STEP 1: WHAT IS A HASH TABLE?
# --------------------------
# Imagine you have a toy box with labels.
# Each label tells you what toy is inside a small box.
# You don't have to search through all toys. You just look at the label and find it immediately.
# That is what a hash table does: it uses a "key" (the label) to find the "value" (the toy) super fast.

# In Python, the easiest hash table is called a "dictionary"
# A dictionary is written using curly braces {} with key:value pairs inside.

# --------------------------
# STEP 2: CREATE AN EMPTY HASH TABLE
# --------------------------
my_hash_table = {}  # This creates an empty hash table
# {} means "nothing inside yet"
# Think: empty toy box with no toys

# --------------------------
# STEP 3: ADD ITEMS TO THE HASH TABLE
# --------------------------
# To add items, we give a key and a value
my_hash_table["apple"] = 10  # Key = "apple", Value = 10
# Explanation:
# - Key "apple" is like a label
# - Value 10 is like "10 apples"
# - The hash table stores this pair
my_hash_table["banana"] = 5
my_hash_table["orange"] = 7
# Now the hash table has 3 items:
# {'apple': 10, 'banana': 5, 'orange': 7}
# Think of it as a shelf with labeled boxes:
# apple -> 10, banana -> 5, orange -> 7

# --------------------------
# STEP 4: SEE WHAT IS INSIDE
# --------------------------
print("Current Hash Table:", my_hash_table)
# Output: {'apple': 10, 'banana': 5, 'orange': 7}
# This prints all the key-value pairs inside the hash table

# --------------------------
# STEP 5: LOOK UP A VALUE
# --------------------------
# If we want to know how many apples we have:
apples = my_hash_table["apple"]
# Explanation:
# - my_hash_table["apple"] looks at the key "apple"
# - It finds the value 10 and gives it to us
print("Number of apples:", apples)
# Output: Number of apples: 10
# Super fast lookup, no searching like a list

# --------------------------
# STEP 6: CHECK IF A KEY EXISTS
# --------------------------
# We can check if a key is in the hash table using "in"
if "banana" in my_hash_table:
    print("Banana is in the hash table!")
# If the key does not exist:
if "grape" not in my_hash_table:
    print("Grape is NOT in the hash table!")
# Explanation:
# - "in" checks if the key is present
# - "not in" checks if the key is absent
# - Very useful to avoid errors

# --------------------------
# STEP 7: UPDATE A VALUE
# --------------------------
# Suppose we ate 2 bananas:
my_hash_table["banana"] = my_hash_table["banana"] - 2
# Step by step:
# - my_hash_table["banana"] is currently 5
# - 5 - 2 = 3
# - We put 3 back into the hash table for key "banana"
print("Updated hash table:", my_hash_table)
# Output: {'apple': 10, 'banana': 3, 'orange': 7}

# --------------------------
# STEP 8: DELETE A KEY
# --------------------------
# Remove orange completely
del my_hash_table["orange"]
print("After deleting orange:", my_hash_table)
# Output: {'apple': 10, 'banana': 3}
# Explanation:
# - del removes the key and its value
# - Now "orange" is gone

# --------------------------
# STEP 9: LOOP THROUGH HASH TABLE
# --------------------------
# We can see all key-value pairs using a loop
for key, value in my_hash_table.items():
    # Explanation:
    # - key = each key in the hash table
    # - value = value corresponding to that key
    print("Key:", key, "| Value:", value)
# Output:
# Key: apple | Value: 10
# Key: banana | Value: 3

# --------------------------
# STEP 10: REAL-LIFE EXAMPLE - COUNTING NUMBERS
# --------------------------
numbers = [1, 2, 3, 2, 1, 2, 3, 4, 5, 1]
# We want to count how many times each number appears

count_table = {}  # empty hash table for counting

for num in numbers:
    if num in count_table:
        count_table[num] += 1  # already exists? add 1
    else:
        count_table[num] = 1   # first time? set count = 1

print("Count of each number:", count_table)
# Output: {1: 3, 2: 3, 3: 2, 4: 1, 5: 1}
# Explanation:
# - key = the number
# - value = how many times it appears
# - Super fast counting using hash table

# --------------------------
# STEP 11: IMPORTANT CONCEPTS
# --------------------------
# 1. Key = label (must be unique)
# 2. Value = data you want to store
# 3. Hash table = magical storage with super fast lookup
# 4. Operations:
#    - Add: table[key] = value
#    - Lookup: table[key]
#    - Update: table[key] = new_value
#    - Delete: del table[key]
# 5. Great for counting, grouping, and organizing data quickly
# 6. Python uses a "hash function" behind the scenes to find the key fast

# =========================================================# =========================================================
```
