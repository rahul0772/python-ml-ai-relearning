```python
# ==========================================
#  ADVANCED PYTHON LOOPS 🔥
# ==========================================

# ==========================================
# 1. DEEP NESTED LOOPS (3 LEVELS)
# ==========================================

# We create combinations of 3 numbers
for i in range(2):  # first loop → 0,1
    for j in range(2):  # second loop → 0,1
        for k in range(2):  # third loop → 0,1
            print("i:", i, "j:", j, "k:", k)

# 👉 Total runs = 2 × 2 × 2 = 8
# This is called O(n^3) (very slow if n is big)

print("\n------------------\n")


# ==========================================
# 2. LOOP OPTIMIZATION (IMPORTANT FOR EXPERTS)
# ==========================================

# BAD WAY (slow)
for i in range(5):
    for j in range(5):
        print(i, j)

# 👉 runs 25 times

print("\nOptimized version:\n")

# BETTER WAY: reduce work using logic
for i in range(5):
    for j in range(i, 5):  # starts from i instead of 0
        print(i, j)

# 👉 fewer operations (faster)

print("\n------------------\n")


# ==========================================
# 3. BREAKING MULTIPLE LOOPS (TRICKY)
# ==========================================

# Python cannot directly break outer loops easily
# So we use a FLAG variable

found = False  # flag to stop outer loop

for i in range(5):
    for j in range(5):
        if i == 2 and j == 3:
            print("Found at:", i, j)
            found = True  # mark as found
            break  # break inner loop

    if found:
        break  # break outer loop

print("\n------------------\n")


# ==========================================
# 4. LOOP + ELSE + BREAK (ADVANCED LOGIC)
# ==========================================

# Check if a number exists in list

numbers = [1, 3, 5, 7]

target = 4

for num in numbers:
    if num == target:
        print("Found!")
        break
else:
    print("Not found!")

# 👉 else runs ONLY if break never happens

print("\n------------------\n")


# ==========================================
# 5. ADVANCED LIST COMPREHENSION (DOUBLE LOOP)
# ==========================================

# create pairs (i, j)

pairs = [(i, j) for i in range(3) for j in range(3)]

print("Pairs:", pairs)

# 👉 same as nested loops but in ONE LINE

print("\n------------------\n")


# ==========================================
# 6. MATRIX (2D LIST) LOOPING
# ==========================================

matrix = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

# Access each element
for row in matrix:  # row = [1,2,3] etc.
    for value in row:  # each number inside row
        print(value, end=" ")
    print()  # new line after each row

print("\n------------------\n")


# ==========================================
# 7. TRANSPOSE MATRIX (ADVANCED THINKING)
# ==========================================

# swap rows and columns

transpose = []

for i in range(len(matrix)):  # column index
    new_row = []
    for j in range(len(matrix)):  # row index
        new_row.append(matrix[j][i])  # swap indices
    transpose.append(new_row)

print("Transpose:", transpose)

print("\n------------------\n")


# ==========================================
# 8. GENERATOR FUNCTION (VERY ADVANCED)
# ==========================================

# generator function uses "yield"

def my_generator(n):
    i = 0
    while i < n:
        yield i  # returns value one by one
        i += 1

# use generator
for value in my_generator(5):
    print("Generated:", value)

# 👉 memory efficient (does not store full list)

print("\n------------------\n")


# ==========================================
# 9. LAZY EVALUATION (PRO CONCEPT)
# ==========================================

# This generator does NOT calculate immediately
gen = (i*i for i in range(5))

print("Generator created, not executed yet")

for val in gen:
    print("Now computing:", val)

# 👉 computation happens ONLY when needed

print("\n------------------\n")


# ==========================================
# 10. ITERTOOLS (PROFESSIONAL TOOL)
# ==========================================

import itertools

# combinations
comb = list(itertools.combinations([1, 2, 3], 2))
print("Combinations:", comb)

# permutations
perm = list(itertools.permutations([1, 2, 3], 2))
print("Permutations:", perm)

# 👉 used in real-world coding + interviews

print("\n------------------\n")


# ==========================================
# 11. INFINITE LOOP (DANGEROUS BUT IMPORTANT)
# ==========================================

count = 0

while True:  # runs forever
    print("Count:", count)
    count += 1

    if count == 3:
        break  # stop manually

# 👉 always control infinite loops

print("\n------------------\n")


# ==========================================
# 12. TIME COMPLEXITY THINKING (VERY IMPORTANT)
# ==========================================

# O(n)
for i in range(5):
    print(i)

# O(n^2)
for i in range(5):
    for j in range(5):
        print(i, j)

# 👉 More loops = slower program

print("\n------------------\n")


# ==========================================
# 13. REAL PROBLEM: PATTERN LOGIC
# ==========================================

# Print triangle pattern

n = 5

for i in range(1, n+1):
    for j in range(i):
        print("*", end=" ")
    print()

# 👉 builds logic for interviews + thinking

print("\n------------------\n")


# ==========================================
# 🧠 FINAL EXPERT LESSON
# ==========================================

# If you understand this:
# ✔ nested loops deeply
# ✔ optimization
# ✔ generators
# ✔ itertools
# ✔ matrix logic
# ✔ break + else logic
```
