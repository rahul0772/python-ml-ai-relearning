```python
# Insertion Sort Algorithm
# The Insertion Sort algorithm is one of the simplest sorting algorithms. It sorts the list by picking one element at a time from the unsorted portion of the list
# and placing it in the correct position in the sorted portion of the list.

# Imagine you are sorting a hand of cards, and for each card, you pick it up and place it in the correct position
# among the cards that you have already sorted. This is exactly how Insertion Sort works.

# Let's break it down with extremely detailed comments at every step!

def insertion_sort(arr):
    # Loop through the list starting from the second element (index 1).
    # We don't need to worry about the first element, since a list with one element is already sorted.
    for i in range(1, len(arr)):  # i starts from 1 because we assume the first element is "sorted".

        # Store the current element (arr[i]) in a variable 'key'.
        # This is the element we are going to compare with the sorted part of the list.
        key = arr[i]  # "key" is the element that needs to be inserted into the sorted portion of the list.

        # Set 'j' to one position to the left of 'i', so we can compare 'key' with elements in the sorted portion of the list.
        j = i - 1  # j points to the last element of the sorted portion.

        # The while loop will check if the element in the sorted portion of the list (arr[j]) is larger than the key.
        # If it is larger, we will move that element one position to the right to make space for the key.
        while j >= 0 and arr[j] > key:  # As long as we are within bounds (j >= 0) and arr[j] is greater than the key...

            # Shift arr[j] to the right to make room for the key.
            arr[j + 1] = arr[j]  # Move arr[j] one position to the right.

            # Decrease 'j' by 1 to check the next element in the sorted portion.
            j -= 1  # We keep moving left in the sorted portion to find the correct place for the key.

        # After the while loop, 'j' is pointing to the position where the key should be placed.
        # We insert the key into its correct position in the sorted portion of the list.
        arr[j + 1] = key  # Insert the key after the last smaller element (or at the start if key is smaller than all elements).

    # Finally, after all the passes, we return the sorted list.
    return arr  # Return the sorted list after all insertions are done.

# Let's test the insertion_sort function with an example list.
arr = [12, 11, 13, 5, 6]  # This is our example list that needs to be sorted.

# We will call the insertion_sort function on this list and print the result.
sorted_arr = insertion_sort(arr)

# Now we print the sorted array and explain the output.
print("Sorted Array:", sorted_arr)
```

```python
# Output of the code will be:
# Sorted Array: [5, 6, 11, 12, 13]

# Explanation:
# The Insertion Sort algorithm sorts the list one element at a time by inserting each element into the correct position in the sorted portion of the list.
# Let's walk through the steps with the given example:

# Original array: [12, 11, 13, 5, 6]

# **Step 1: i = 1 (Compare 11 with 12)**
# - The first element (12) is considered sorted. Now, we need to insert the second element (11).
# - We compare 11 with 12, and since 11 is smaller than 12, we shift 12 to the right.
# - Now, we place 11 at the start, giving us: [11, 12, 13, 5, 6]

# **Step 2: i = 2 (Compare 13 with 12)**
# - Now we take 13 and compare it with 12. Since 13 is greater than 12, we leave 13 where it is.
# - No shifting is needed: [11, 12, 13, 5, 6]

# **Step 3: i = 3 (Compare 5 with 13, 12, 11)**
# - Next, we take 5 and compare it with 13, 12, and 11. Since 5 is smaller than all of them, we shift 13, 12, and 11 to the right.
# - We place 5 in the first position: [5, 11, 12, 13, 6]

# **Step 4: i = 4 (Compare 6 with 13, 12, 11, 5)**
# - Finally, we take 6 and compare it with 13, 12, and 11. We shift these elements to the right, and place 6 after 5: [5, 6, 11, 12, 13]

# The final sorted list is: [5, 6, 11, 12, 13].

# Insertion Sort works well for small lists or lists that are already nearly sorted. However, for larger lists, it becomes inefficient because it
# requires shifting elements to make room for each new element.

# The time complexity of Insertion Sort is O(n^2) in the worst case (when the list is in reverse order)
# because it has to move each element for every insertion. However, its best-case time complexity is O(n) when the list is already sorted.
```
