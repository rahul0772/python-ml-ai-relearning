```python
# Define a function to find the longest increasing subsequence
def longest_increasing_subsequence(nums):
    # If the list is empty, we return 0 because there's no subsequence
    # Example: If nums = [], return 0 because no numbers exist
    if not nums:
        return 0

    # Create a list called 'dp' to store the length of the longest increasing subsequence
    # that ends at each position. Initially, every number can at least form a subsequence of length 1
    dp = [1] * len(nums)  # Initialize all values to 1
    # Example: If nums = [10, 9, 2, 5], then dp = [1, 1, 1, 1]
    # Each element can at least be its own subsequence.

    # Now we loop through the list starting from the second element (index 1)
    for i in range(1, len(nums)):  # i starts from 1 because we compare with previous elements
        # We now check for every element nums[i], if it can form a longer subsequence by adding itself
        # after any previous number nums[j] (where j < i).
        # Example: For nums[i] = 5 (at index 3), we compare with nums[j] = [10, 9, 2]

        for j in range(i):  # Inner loop to compare nums[i] with all elements before it
            # If nums[i] is greater than nums[j], then nums[i] can extend the subsequence at nums[j]
            # This means nums[i] can be added to the subsequence that ends at nums[j]
            if nums[i] > nums[j]:
                # Update dp[i] to the maximum subsequence length that can be formed by adding nums[i]
                # to the subsequence ending at nums[j]. We take the largest possible subsequence length.
                dp[i] = max(dp[i], dp[j] + 1)
                # Example: If nums[i] = 5, nums[j] = 2, we can form the subsequence [2, 5]
                # So, dp[3] = dp[2] + 1 = 2, meaning subsequence length is now 2.

    # After the loops, 'dp' will contain the longest subsequence length for each position.
    # Example: For nums = [10, 9, 2, 5, 3, 7, 101, 18], dp will be [1, 1, 1, 2, 2, 3, 4, 4]

    # The longest subsequence length will be the maximum value in the dp list
    return max(dp)  # Get the maximum subsequence length from the dp array

# Test the function with a sample input
nums = [10, 9, 2, 5, 3, 7, 101, 18]
result = longest_increasing_subsequence(nums)

# Print the result, which is the length of the longest increasing subsequence
print(f"Length of Longest Increasing Subsequence: {result}")
```
