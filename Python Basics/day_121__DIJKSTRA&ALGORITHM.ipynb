```python
# =========================
# DIJKSTRA ALGORITHM
# =========================

"""
WHAT IS DIJKSTRA ALGORITHM?
---------------------------
Dijkstra's Algorithm is used to find the SHORTEST PATH
from one source node to all other nodes in a weighted graph.

IMPORTANT:
- Works only for NON-NEGATIVE edge weights.
- Can be used in:
    * GPS Navigation
    * Network Routing
    * Maps
    * Shortest path problems

TIME COMPLEXITY:
----------------
Using Priority Queue (Min Heap):
    O((V + E) log V)

Where:
    V = Number of vertices
    E = Number of edges

CORE IDEA:
-----------
1. Start from source node.
2. Keep track of minimum known distance to every node.
3. Always pick the node with the smallest distance.
4. Relax its neighbors:
       if new_distance < old_distance:
            update it

DATA STRUCTURES USED:
---------------------
1. Graph (Adjacency List)
2. Priority Queue / Min Heap

-------------------------------------------------------
EXAMPLE GRAPH
-------------------------------------------------------

        4
    A ------- B
    |         |
  1 |         | 2
    |         |
    C ------- D
        1

Adjacency List Representation:
{
    'A': [('B',4), ('C',1)],
    'B': [('A',4), ('D',2)],
    'C': [('A',1), ('D',1)],
    'D': [('B',2), ('C',1)]
}

Goal:
Find shortest distance from source node 'A'
"""

# heapq provides priority queue implementation
import heapq

# ---------------------------------------------------
# GRAPH REPRESENTATION (Adjacency List)
# Each node stores:
#   (neighbor, weight)
# ---------------------------------------------------

graph = {
    'A': [('B', 4), ('C', 1)],
    'B': [('A', 4), ('D', 2), ('E', 5)],
    'C': [('A', 1), ('D', 1)],
    'D': [('B', 2), ('C', 1), ('E', 3)],
    'E': [('B', 5), ('D', 3)]
}

# ---------------------------------------------------
# DIJKSTRA FUNCTION
# ---------------------------------------------------

def dijkstra(graph, start):

    # -----------------------------------------------
    # Step 1:
    # Initialize distances dictionary
    # Set all distances to infinity initially
    # -----------------------------------------------

    distances = {node: float('inf') for node in graph}

    # Distance from source to itself is 0
    distances[start] = 0

    # -----------------------------------------------
    # Priority Queue (Min Heap)
    # Stores:
    #   (distance, node)
    # -----------------------------------------------

    priority_queue = [(0, start)]

    # -----------------------------------------------
    # Previous dictionary
    # Used to reconstruct shortest path
    # -----------------------------------------------

    previous = {node: None for node in graph}

    # ------------------------------------------------
    # MAIN LOOP
    # ------------------------------------------------

    while priority_queue:

        # Get node with smallest distance
        current_distance, current_node = heapq.heappop(priority_queue)

        # ------------------------------------------------
        # Skip outdated entries
        # (Optimization)
        # ------------------------------------------------

        if current_distance > distances[current_node]:
            continue

        # ------------------------------------------------
        # Explore neighbors
        # ------------------------------------------------

        for neighbor, weight in graph[current_node]:

            # Calculate new possible distance
            distance = current_distance + weight

            # ------------------------------------------------
            # RELAXATION STEP
            # If shorter path found -> update
            # ------------------------------------------------

            if distance < distances[neighbor]:

                distances[neighbor] = distance

                previous[neighbor] = current_node

                # Push updated distance into heap
                heapq.heappush(priority_queue, (distance, neighbor))

                print(f"Updated: {neighbor}")
                print(f"New Distance: {distance}")
                print(f"Previous Node: {current_node}")
                print("-" * 40)

    return distances, previous

# ---------------------------------------------------
# FUNCTION TO RECONSTRUCT SHORTEST PATH
# ---------------------------------------------------

def shortest_path(previous, start, target):

    path = []

    current = target

    # Backtrack using previous dictionary
    while current is not None:
        path.append(current)
        current = previous[current]

    # Reverse path because we backtracked
    path.reverse()

    # Check if valid path exists
    if path[0] == start:
        return path
    else:
        return []

# ---------------------------------------------------
# RUN DIJKSTRA
# ---------------------------------------------------

source = 'A'

distances, previous = dijkstra(graph, source)

# ---------------------------------------------------
# PRINT FINAL SHORTEST DISTANCES
# ---------------------------------------------------

print("\nFINAL SHORTEST DISTANCES")
print("=" * 40)

for node, distance in distances.items():
    print(f"{source} --> {node} = {distance}")

# ---------------------------------------------------
# PRINT SHORTEST PATHS
# ---------------------------------------------------

print("\nSHORTEST PATHS")
print("=" * 40)

for target in graph:

    path = shortest_path(previous, source, target)

    print(f"Path to {target}: {' -> '.join(path)}")

"""
-------------------------------------------------------
STEP-BY-STEP DRY RUN
-------------------------------------------------------

Source = A

INITIAL:
---------
Distances:
A = 0
B = inf
C = inf
D = inf
E = inf

Priority Queue:
[(0, 'A')]

-------------------------------------------------------
STEP 1:
Pop A

Neighbors:
B with weight 4
C with weight 1

Update:
B = 4
C = 1

Queue:
[(1,C), (4,B)]

-------------------------------------------------------
STEP 2:
Pop C

Neighbors:
A already smaller
D = 1 + 1 = 2

Update:
D = 2

Queue:
[(2,D), (4,B)]

-------------------------------------------------------
STEP 3:
Pop D

Neighbors:
B = min(4, 2+2=4)
E = 2+3 = 5

Update:
E = 5

Queue:
[(4,B), (5,E)]

-------------------------------------------------------
FINAL ANSWER:
--------------
A -> A = 0
A -> B = 4
A -> C = 1
A -> D = 2
A -> E = 5

-------------------------------------------------------
WHY PRIORITY QUEUE?
-------------------------------------------------------

Without heap:
    O(V^2)

With heap:
    O((V + E) log V)

Heap always gives smallest distance node quickly.

-------------------------------------------------------
IMPORTANT INTERVIEW QUESTIONS
-------------------------------------------------------

1. Why Dijkstra fails for negative weights?
   Because greedy choice becomes invalid.

2. Difference between BFS and Dijkstra?
   BFS:
      - Unweighted graph
   Dijkstra:
      - Weighted graph

3. Difference between Dijkstra and Bellman-Ford?
   Dijkstra:
      Faster
      No negative weights

   Bellman-Ford:
      Handles negative weights
      Slower

-------------------------------------------------------
REAL WORLD APPLICATIONS
-------------------------------------------------------

1. Google Maps
2. Network Routing
3. Flight Systems
4. GPS Navigation
5. Robot Path Planning

-------------------------------------------------------
END
-------------------------------------------------------
"""
```
