```python
# ================================================================
# BUBBLE SORT (Sorting Algorithm)
# ================================================================
# We are going to learn a VERY IMPORTANT concept in programming:
#
# Imagine:
# You have numbers: [5, 3, 8, 1]
# You want:         [1, 3, 5, 8]
#
# Bubble Sort is one of the SIMPLEST ways to do this.
# It works by repeatedly comparing adjacent (neighbor) elements
# and swapping them if they are in the wrong order.
#
# Think like this:
# Big numbers "bubble up" to the top (end of the list),
# just like bubbles rise in water. 🫧
#
# ================================================================
# STEP 1: DEFINE THE LIST (OUR DATA)
# ================================================================

numbers = [5, 3, 8, 1, 2]
# We create a list (collection of values)
# Think of this like a row of boxes:
# [5] [3] [8] [1] [2]
#
# Our goal:
# Rearrange these numbers in ascending order (small → big)

# ================================================================
# STEP 2: GET LENGTH OF LIST
# ================================================================

n = len(numbers)
# len(numbers) gives total count of elements in list
# Here: n = 5
#
# WHY we need this?
# Because we must know how many times to loop through the list
# to compare elements

# ================================================================
# STEP 3: OUTER LOOP (NUMBER OF PASSES)
# ================================================================

for i in range(n):
    # This loop controls how many times we go over the list
    # range(n) means: 0 → 4 (5 times)
    #
    # WHY multiple passes?
    # Because in one pass, only ONE big element reaches correct position
    #
    # Think:
    # Pass 1 → biggest number goes to end
    # Pass 2 → second biggest goes near end
    # and so on...

    # ============================================================
    # STEP 4: INNER LOOP (COMPARING NEIGHBORS)
    # ============================================================

    for j in range(0, n - i - 1):
        # j is used to compare elements
        #
        # Why (n - i - 1)?
        # Because after each pass, last elements are already sorted
        #
        # Example:
        # After 1st pass → last element is correct
        # So no need to check it again

        # ========================================================
        # STEP 5: COMPARE ADJACENT ELEMENTS
        # ========================================================

        if numbers[j] > numbers[j + 1]:
            # We compare current element with next one
            #
            # Example:
            # numbers[j] = 5
            # numbers[j+1] = 3
            #
            # If 5 > 3 → WRONG ORDER → we SWAP

            # ====================================================
            # STEP 6: SWAP ELEMENTS
            # ====================================================

            numbers[j], numbers[j + 1] = numbers[j + 1], numbers[j]
            # This is Python's EASY swap trick
            #
            # Before swap:
            # [5, 3]
            #
            # After swap:
            # [3, 5]
            #
            # WHY swap?
            # To move smaller number left and bigger right
            #
            # This is the CORE LOGIC of sorting!

    # ============================================================
    # PRINT AFTER EACH PASS (TO SEE LEARNING STEP-BY-STEP)
    # ============================================================

    print(f"After pass {i+1}: {numbers}")
    # This helps us SEE how list changes step-by-step
    #
    # Cognitive trick:
    # Always visualize algorithm → makes understanding 10x easier

# ================================================================
# FINAL RESULT
# ================================================================

print("\nFinal Sorted List:", numbers)
# After all passes → list is fully sorted

# ================================================================
#
# Example Output:
#
# After pass 1: [3, 5, 1, 2, 8]
# After pass 2: [3, 1, 2, 5, 8]
# After pass 3: [1, 2, 3, 5, 8]
# After pass 4: [1, 2, 3, 5, 8]
# After pass 5: [1, 2, 3, 5, 8]
#
# Final Sorted List: [1, 2, 3, 5, 8]

# PASS 1:
# Biggest number (8) moves to end
#
# PASS 2:
# Next biggest (5) moves near end
#
# PASS 3:
# Next (3) goes correct place
#
# PASS 4 & 5:
# Already sorted → no changes
#
# Bubble Sort works like:
# Repeatedly pushing the biggest unsorted value to its correct position
#
# Each pass = one "layer" of correctness added
#
# ================================================================
# BIG IDEA TO REMEMBER:
#
# 1. Compare neighbors
# 2. Swap if wrong
# 3. Repeat passes
```
