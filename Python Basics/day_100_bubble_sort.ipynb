```python
# ============================================================
# BUBBLE SORT (Sorting Algorithm)
# ============================================================
#
# GOAL:
# We will take a list of numbers and sort them step-by-step.
#
# 🧠 CORE IDEA (VERY SIMPLE):
# Imagine bubbles in water:
# - Bigger bubbles rise to the top
# - Similarly, bigger numbers "bubble up" to the end of the list
#
# We repeatedly compare adjacent (next-to-each-other) elements
# and swap them if they are in the wrong order.
#
# ------------------------------------------------------------
# 🚀 LET'S START CODING + LEARNING EVERYTHING LINE BY LINE
# ------------------------------------------------------------

# Step 1: Create a list (array) of numbers that we want to sort
numbers = [5, 3, 8, 4, 2]

# 🧠 Explanation:
# - This is our INPUT data
# - Think of this as a messy line of numbers
# - Our goal is to transform it into sorted order
#
# Current state:
# [5, 3, 8, 4, 2]


# Step 2: Get the length of the list
n = len(numbers)

# 🧠 Explanation:
# - len(numbers) counts how many elements are in the list
# - Here n = 5
#
# WHY we need this?
# Because sorting requires us to iterate (loop) over elements
# multiple times, and we need to know how many elements exist


# Step 3: Outer loop → controls how many passes we make
for i in range(n):

    # 🧠 Explanation:
    # - We repeat the process multiple times
    # - Why? Because one pass only partially sorts the list
    #
    # Think of it like cleaning a messy room:
    # - First pass: move big things
    # - Second pass: adjust smaller things
    # - Eventually everything becomes clean (sorted)
    #
    # range(n) → gives values 0,1,2,3,4
    #
    # Each iteration = ONE FULL PASS through the list


    # Step 4: Inner loop → compares adjacent elements
    for j in range(0, n - i - 1):

        # 🧠 Explanation:
        # This is VERY IMPORTANT LOGIC:
        #
        # Why (n - i - 1)?
        #
        # Because:
        # - After each pass, the largest element is already at the end
        # - So we don't need to check it again
        #
        # Example:
        # After 1st pass → largest number is fixed at last position
        # After 2nd pass → second largest is fixed
        #
        # So we reduce the range each time → efficiency


        # Step 5: Compare two adjacent elements
        if numbers[j] > numbers[j + 1]:

            # 🧠 Explanation:
            # We compare:
            # numbers[j]     → current element
            # numbers[j+1]   → next element
            #
            # If left > right → WRONG ORDER → we swap


            # Step 6: Swap the elements
            numbers[j], numbers[j + 1] = numbers[j + 1], numbers[j]

            # 🧠 Explanation:
            # This is Python's shortcut swap!
            #
            # Before:
            # numbers[j] = 5
            # numbers[j+1] = 3
            #
            # After swap:
            # numbers[j] = 3
            # numbers[j+1] = 5
            #
            # WHY swap?
            # Because we want smaller number on left, bigger on right
            #
            # This is the CORE ACTION of Bubble Sort:
            # COMPARE → DECIDE → SWAP


    # Step 7: Print list after each pass
    print(f"After pass {i+1}: {numbers}")

    # 🧠 Explanation:
    # This helps us VISUALIZE learning step-by-step
    #
    # Each pass moves one largest element to correct position


# Step 8: Final sorted result
print("\nFinal Sorted List:", numbers)

# ============================================================
# 🧠🔥 OUTPUT + ULTRA DEEP EXPLANATION
# ============================================================
#
# Expected Output:
#
# After pass 1: [3, 5, 4, 2, 8]
# After pass 2: [3, 4, 2, 5, 8]
# After pass 3: [3, 2, 4, 5, 8]
# After pass 4: [2, 3, 4, 5, 8]
# After pass 5: [2, 3, 4, 5, 8]
#
# Final Sorted List: [2, 3, 4, 5, 8]
#
# ------------------------------------------------------------
# 🧠 STEP-BY-STEP MENTAL MODEL BREAKDOWN:
#
# PASS 1:
# [5,3,8,4,2]
# Compare 5 & 3 → swap → [3,5,8,4,2]
# Compare 5 & 8 → OK
# Compare 8 & 4 → swap → [3,5,4,8,2]
# Compare 8 & 2 → swap → [3,5,4,2,8]
# 👉 Biggest (8) moved to end
#
# PASS 2:
# [3,5,4,2,8]
# Compare 5 & 4 → swap
# Compare 5 & 2 → swap
# 👉 5 moves closer to correct place
#
# PASS 3:
# [3,4,2,5,8]
# Compare 4 & 2 → swap
#
# PASS 4:
# [3,2,4,5,8]
# Compare 3 & 2 → swap
#
# PASS 5:
# Already sorted → no changes
#
# ------------------------------------------------------------
# 🧠 FINAL DEEP UNDERSTANDING:
#
# Bubble Sort works by:
# 1. Repeatedly stepping through list
# 2. Comparing adjacent elements
# 3. Swapping if needed
# 4. Largest elements "bubble up" to the end
#
# TIME COMPLEXITY:
# - Worst case: O(n^2) → very slow for big data
```
