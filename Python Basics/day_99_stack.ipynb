```python
# STACK (LIFO: Last In First Out)
# ===============================================================

# 🧠 CORE IDEA (VERY IMPORTANT MENTAL MODEL):
# A STACK is like a stack of plates 🍽️
# - You ADD (push) plates on top
# - You REMOVE (pop) plates from the top
# - The LAST plate you added is the FIRST one you remove
# This is called LIFO = Last In, First Out

# ===============================================================
# Step 1: Create a Stack class
# ===============================================================

class Stack:
    # We define a class called "Stack"
    # A class is like a blueprint to create objects (like creating your own custom data structure)

    def __init__(self):
        # __init__ is a special function that runs automatically when we create an object
        # It initializes (sets up) our stack

        self.items = []
        # VERY IMPORTANT:
        # We are using a Python LIST to store stack elements
        # Why list?
        # Because Python lists already support:
        # - append() -> add element at end (perfect for push)
        # - pop() -> remove last element (perfect for pop)

        # So internally, our stack is just a list
        # But we will control how it's used (only like a stack)

    # ===========================================================
    # Step 2: PUSH operation (add element to stack)
    # ===========================================================
    def push(self, item):
        # 'push' means: add element on TOP of the stack

        self.items.append(item)
        # append() adds element to the END of list
        # END of list = TOP of stack

        # Example:
        # If stack = [1, 2]
        # push(3) → stack becomes [1, 2, 3]

        print(f"Pushed {item} onto stack")
        # Print to show what is happening step-by-step

    # ===========================================================
    # Step 3: POP operation (remove top element)
    # ===========================================================
    def pop(self):
        # 'pop' means: remove the TOP element

        if self.is_empty():
            # Before removing, we check:
            # Is the stack empty?

            print("Stack is empty, cannot pop")
            return None
            # If empty, we cannot remove anything

        return self.items.pop()
        # pop() removes LAST element of list
        # LAST = TOP of stack

        # Example:
        # [1, 2, 3] → pop() → returns 3 → stack becomes [1, 2]

    # ===========================================================
    # Step 4: PEEK operation (see top element without removing)
    # ===========================================================
    def peek(self):
        # 'peek' means: look at TOP element without removing

        if self.is_empty():
            print("Stack is empty, nothing to peek")
            return None

        return self.items[-1]
        # [-1] means LAST element of list
        # That is the TOP of stack

        # Example:
        # [1, 2, 3] → peek() → returns 3 (but stack stays same)

    # ===========================================================
    # Step 5: Check if stack is empty
    # ===========================================================
    def is_empty(self):
        # Returns True if stack has no elements

        return len(self.items) == 0
        # len(self.items) gives number of elements
        # If 0 → empty

    # ===========================================================
    # Step 6: Get size of stack
    # ===========================================================
    def size(self):
        return len(self.items)
        # Returns how many elements are inside stack

# ===============================================================
# Step 7: Create a Stack object
# ===============================================================

my_stack = Stack()
# We create an object of Stack class
# This is our actual working stack

# ===============================================================
# Step 8: Perform operations step by step
# ===============================================================

# Push elements
my_stack.push(10)
my_stack.push(20)
my_stack.push(30)

# Current stack mentally:
# Bottom → [10, 20, 30] ← Top

# Peek top element
top_element = my_stack.peek()
print(f"Top element is: {top_element}")

# Pop elements
removed_element = my_stack.pop()
print(f"Popped element: {removed_element}")

# Stack now:
# [10, 20]

# Check size
print(f"Stack size: {my_stack.size()}")

# Pop all elements
my_stack.pop()
my_stack.pop()

# Try popping from empty stack
my_stack.pop()

# ===============================================================
# Step 9: Output Explanation (VERY IMPORTANT UNDERSTANDING)
# ===============================================================

# 🧠 Let's break down what happened logically:

# 1) push(10), push(20), push(30)
#    Stack became: [10, 20, 30]
#    30 is at TOP (last added)

# 2) peek()
#    Returned 30 (top element) WITHOUT removing it

# 3) pop()
#    Removed 30 → stack became [10, 20]

# 4) size()
#    Returned 2 (because two elements left)

# 5) pop(), pop()
#    Removed 20 → then 10 → stack becomes empty []

# 6) pop() on empty stack
#    Shows message: cannot pop

# 🧠 DEEP LOGIC IMPRINT:
# Stack always works ONLY from one side (TOP)
# - Add → TOP
# - Remove → TOP
# - View → TOP

# This is why it's called LIFO:
# Last In → First Out
```
