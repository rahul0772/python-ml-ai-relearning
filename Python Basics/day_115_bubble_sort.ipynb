```python
# ===========================
# Bubble Sort is a very simple sorting algorithm.
# It works by repeatedly swapping adjacent elements if they are in the wrong order.
# Think of it like bubbles in water: the largest numbers "bubble up" to the top of the list.

# Step 1: Let's create an example list of numbers to sort
numbers = [64, 34, 25, 12, 22, 11, 90]

# Explanation:
# Here we define a Python list called 'numbers'
# Lists are like containers that can hold multiple items
# In this case, our list holds integers: 64, 34, 25, 12, 22, 11, 90
# Our goal is to rearrange these numbers from smallest to largest using Bubble Sort

# Step 2: Define the Bubble Sort algorithm
def bubble_sort(arr):
    # 'arr' is the list that we want to sort
    n = len(arr)  # Find the length of the list
    # Explanation: len(arr) gives us the total number of items in the list.
    # We need to know this because Bubble Sort will repeatedly compare elements up to the end of the list

    # Outer loop: We will do multiple passes through the list
    for i in range(n):
        # Explanation:
        # 'i' represents the current pass number
        # In Bubble Sort, we need to go through the list multiple times to ensure it's fully sorted
        # Each pass will push the next largest element to its correct position

        # Inner loop: Compare adjacent elements and swap if needed
        for j in range(0, n-i-1):
            # Explanation:
            # 'j' is the index of the current element we are comparing
            # We stop at n-i-1 because the last i elements are already sorted after each pass
            # This reduces unnecessary comparisons and makes the algorithm slightly faster

            # Compare current element with the next element
            if arr[j] > arr[j+1]:
                # Explanation:
                # If the current element is greater than the next element, we need to swap them
                # This is the "bubble" effect: larger elements move up towards the end of the list

                # Swap the elements
                arr[j], arr[j+1] = arr[j+1], arr[j]
                # Explanation:
                # Python allows a very elegant way to swap values without a temporary variable
                # arr[j], arr[j+1] = arr[j+1], arr[j] swaps the values of arr[j] and arr[j+1]
                # After this step, the bigger number moves one step closer to its correct position

# Step 3: Call the bubble_sort function on our numbers list
bubble_sort(numbers)

# Step 4: Print the sorted list
print("Sorted list is:", numbers)
# Explanation of output:
# After calling bubble_sort(numbers), our original list is now sorted
# The print statement outputs: [11, 12, 22, 25, 34, 64, 90]
# Step-by-step logic of Bubble Sort for this list:
# 1st pass: 64 moves to the end → [34, 25, 12, 22, 11, 64, 90]
# 2nd pass: 34 moves closer to the end → [25, 12, 22, 11, 34, 64, 90]
# 3rd pass: 25 moves closer → [12, 22, 11, 25, 34, 64, 90]z

# 4th pass: 12 moves to the front → [12, 11, 22, 25, 34, 64, 90]
# 5th pass: 11 moves to the front → [11, 12, 22, 25, 34, 64, 90]
# Final list is fully sorted in ascending order
# Cognitive imprint logic:
# Each comparison teaches the mind about order relationships
# Each swap reinforces the concept of moving items into the right place
# Visualizing each pass helps build a mental model of iterative improvement
```
