```python
# ============================================================
# Prim's Algorithm
# ============================================================
#
# What is Prim's Algorithm?
# --------------------------
# Prim's Algorithm is used to find the:
#   ➜ Minimum Spanning Tree (MST)
#
# Minimum Spanning Tree means:
#   - Connect all vertices
#   - Use minimum total edge weight
#   - No cycles
#
# Time Complexity:
#   Using Priority Queue (Heap):
#       O(E log V)
#
# Where:
#   V = Number of vertices
#   E = Number of edges
#
# ------------------------------------------------------------
# Example Graph:
#
#        2
#   A ------- B
#   | \       |
# 6 |  \3     | 5
#   |   \     |
#   C ------- D
#        4
#
# MST Output:
#   A - B : 2
#   A - D : 3
#   D - C : 4
#
# Total Cost = 9
# ------------------------------------------------------------

import heapq

# ------------------------------------------------------------
# Step 1: Create Graph using Adjacency List
# ------------------------------------------------------------
#
# Graph Format:
# graph[node] = [(weight, neighbor), ...]
#
# Undirected Graph:
# Add both directions
# ------------------------------------------------------------

graph = {
    'A': [(2, 'B'), (6, 'C'), (3, 'D')],
    'B': [(2, 'A'), (5, 'D')],
    'C': [(6, 'A'), (4, 'D')],
    'D': [(3, 'A'), (5, 'B'), (4, 'C')]
}

# ------------------------------------------------------------
# Step 2: Prim's Algorithm Function
# ------------------------------------------------------------

def prim_mst(graph, start):

    # Min Heap / Priority Queue
    # Stores edges as:
    # (weight, current_node, parent)
    min_heap = [(0, start, None)]

    # To track visited nodes
    visited = set()

    # Store MST edges
    mst = []

    # Total minimum cost
    total_cost = 0

    # --------------------------------------------------------
    # Process until heap becomes empty
    # --------------------------------------------------------
    while min_heap:

        # Get edge with minimum weight
        weight, current, parent = heapq.heappop(min_heap)

        # Skip if node already visited
        if current in visited:
            continue

        # Mark node as visited
        visited.add(current)

        # Add edge to MST
        # Ignore first node because parent is None
        if parent is not None:
            mst.append((parent, current, weight))
            total_cost += weight

        # ----------------------------------------------------
        # Add neighboring edges into heap
        # ----------------------------------------------------
        for edge_weight, neighbor in graph[current]:

            # Only add unvisited neighbors
            if neighbor not in visited:
                heapq.heappush(
                    min_heap,
                    (edge_weight, neighbor, current)
                )

    return mst, total_cost


# ------------------------------------------------------------
# Step 3: Run Algorithm
# ------------------------------------------------------------

mst, cost = prim_mst(graph, 'A')

# ------------------------------------------------------------
# Step 4: Print Results
# ------------------------------------------------------------

print("Minimum Spanning Tree Edges:\n")

for u, v, w in mst:
    print(f"{u} -- {v}  weight = {w}")

print("\nTotal Minimum Cost =", cost)
```
