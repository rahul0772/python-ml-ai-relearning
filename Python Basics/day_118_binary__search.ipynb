```python
# ============================================================
#                BINARY SEARCH IN PYTHON
# ============================================================
#
# Binary Search works only on SORTED lists.
#
# It is much faster than Linear Search because:
# Instead of checking every element one by one,
# it repeatedly divides the list into HALF.
#
# Time Complexity:
# Binary Search  -> O(log n)
# Linear Search  -> O(n)
#
# ============================================================


def binary_search(arr, target):
    """
    Function to perform Binary Search

    Parameters:
    arr    -> Sorted list
    target -> Element to search

    Returns:
    Index of target if found
    -1 if not found
    """

    # Starting index
    start = 0

    # Ending index
    end = len(arr) - 1

    # Loop continues until start becomes greater than end
    while start <= end:

        # Find middle index
        # Safer formula to avoid overflow
        mid = start + (end - start) // 2

        # ----------------------------------------------------
        # CASE 1: Target Found
        # ----------------------------------------------------
        if arr[mid] == target:
            return mid

        # ----------------------------------------------------
        # CASE 2: Target is greater than middle element
        # Search in RIGHT half
        # ----------------------------------------------------
        elif arr[mid] < target:
            start = mid + 1

        # ----------------------------------------------------
        # CASE 3: Target is smaller than middle element
        # Search in LEFT half
        # ----------------------------------------------------
        else:
            end = mid - 1

    # Element not found
    return -1


# ============================================================
#                    MAIN PROGRAM
# ============================================================

# Sorted list
arr = [2, 5, 8, 12, 16, 23, 38, 56, 72]

print("Sorted List:")
print(arr)

# Take input from user
target = int(input("\nEnter number to search: "))

# Call binary search function
result = binary_search(arr, target)

# ============================================================
#                    OUTPUT RESULT
# ============================================================

if result != -1:
    print(f"\nElement {target} found at index {result}")
else:
    print(f"\nElement {target} not found")


# ============================================================
#                    DRY RUN EXAMPLE
# ============================================================
#
# Suppose:
# arr = [2, 5, 8, 12, 16, 23, 38, 56, 72]
# target = 23
#
# Step 1:
# start = 0
# end = 8
# mid = 4
# arr[mid] = 16
#
# 23 > 16
# Search RIGHT side
#
# Step 2:
# start = 5
# end = 8
# mid = 6
# arr[mid] = 38
#
# 23 < 38
# Search LEFT side
#
# Step 3:
# start = 5
# end = 5
# mid = 5
# arr[mid] = 23
#
# FOUND
#
# ============================================================


# ============================================================
#              RECURSIVE BINARY SEARCH (EXTRA)
# ============================================================

def recursive_binary_search(arr, start, end, target):

    # Base condition
    if start > end:
        return -1

    # Find middle
    mid = start + (end - start) // 2

    # Element found
    if arr[mid] == target:
        return mid

    # Search right half
    elif arr[mid] < target:
        return recursive_binary_search(arr, mid + 1, end, target)

    # Search left half
    else:
        return recursive_binary_search(arr, start, mid - 1, target)


# Example usage of recursive binary search
print("\n--- Recursive Binary Search Example ---")

target2 = 38

result2 = recursive_binary_search(arr, 0, len(arr) - 1, target2)

if result2 != -1:
    print(f"Element {target2} found at index {result2}")
else:
    print(f"Element {target2} not found")
```
