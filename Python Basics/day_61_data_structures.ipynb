```python
# STACK, QUEUE, LINKED LIST, BINARY TREE, HASH TABLE
# Every line is explained with examples, logic, and definitions
# =====================================================

# -----------------------------------------------------
# 1. STACK (LIFO - Last In First Out)
# -----------------------------------------------------
# Imagine a stack of books:
# You can only put a book on top or take the top book.
# The last book you put on top is the first one you take off.
# This is called LIFO (Last In First Out)

print("=== STACK EXAMPLE ===")

# Step 1: Create an empty stack
stack = []  # [] means an empty list. This list will act as our stack
# Example: stack = []  means there are no books yet

# Step 2: Push items (add books on top of the stack)
stack.append(10)  # Add 10 to the stack, stack = [10]
stack.append(20)  # Add 20 on top, stack = [10, 20]
stack.append(30)  # Add 30 on top, stack = [10, 20, 30]
# Now the top of stack is 30, bottom is 10

print("Stack after pushing 10,20,30:", stack)
# Output will show: [10, 20, 30]

# Step 3: Pop items (remove the top book)
top_item = stack.pop()  # Removes 30, stack becomes [10, 20]
print("Popped item (top of stack removed):", top_item)
print("Stack after popping:", stack)

# Step 4: Peek at the top item without removing
top_item = stack[-1]  # -1 means last item in the list
print("Peek top item (without removing):", top_item)
# Top item is 20 now

# -----------------------------------------------------
# 2. QUEUE (FIFO - First In First Out)
# -----------------------------------------------------
# Imagine a line of people waiting for ice cream:
# First person to get in line is the first to get ice cream.
# This is called FIFO (First In First Out)

print("\n=== QUEUE EXAMPLE ===")

from collections import deque  # deque is like a line where you can add/remove people easily

# Step 1: Create an empty queue
queue = deque()  # deque() creates an empty line
# Example: queue = [] means no one is in line

# Step 2: Enqueue (add people to the end of the line)
queue.append(10)  # Add 10, queue = [10]
queue.append(20)  # Add 20, queue = [10, 20]
queue.append(30)  # Add 30, queue = [10, 20, 30]
print("Queue after enqueue 10,20,30:", list(queue))

# Step 3: Dequeue (remove person from front of line)
first_item = queue.popleft()  # Remove 10 (first in line), queue = [20, 30]
print("Dequeued item (removed from front):", first_item)
print("Queue after dequeue:", list(queue))

# -----------------------------------------------------
# 3. LINKED LIST (CHAIN OF NODES)
# -----------------------------------------------------
# Imagine a treasure chain where each link has a value and a pointer to the next link.
# Example: 10 -> 20 -> 30 -> None
# Each link is called a "Node" (value + next pointer)

print("\n=== LINKED LIST EXAMPLE ===")

# Step 1: Create a Node class
class Node:
    def __init__(self, data):
        self.data = data  # data stores the value of the node
        self.next = None  # next stores the reference to the next node
        # Example: Node(10) -> data=10, next=None

# Step 2: Create Linked List class
class LinkedList:
    def __init__(self):
        self.head = None  # head is the start of the linked list
        # Example: Empty list -> head=None

    # Add node at the end
    def append(self, data):
        new_node = Node(data)  # Create a new node with data
        if self.head is None:  # If list is empty
            self.head = new_node  # Make new node the head
            return
        last = self.head  # Start from head
        while last.next:  # Traverse till the last node
            last = last.next
        last.next = new_node  # Link last node to new node

    # Print the linked list
    def print_list(self):
        current = self.head  # Start from head
        while current:  # While there is a node
            print(current.data, end=" -> ")  # Print value
            current = current.next  # Move to next node
        print("None")  # End of list

# Step 3: Create linked list and add items
llist = LinkedList()
llist.append(10)  # Linked list: 10 -> None
llist.append(20)  # Linked list: 10 -> 20 -> None
llist.append(30)  # Linked list: 10 -> 20 -> 30 -> None
print("Linked List contents:")
llist.print_list()  # Output: 10 -> 20 -> 30 -> None

# -----------------------------------------------------
# 4. BINARY TREE (HIERARCHY OF NODES)
# -----------------------------------------------------
# Imagine a family tree:
# Each person (node) can have at most 2 children: left and right
# Example:
#        10
#       /  \
#      5    20

print("\n=== BINARY TREE EXAMPLE ===")

# Step 1: Create TreeNode class
class TreeNode:
    def __init__(self, value):
        self.value = value  # value of the node
        self.left = None    # left child
        self.right = None   # right child
        # Example: TreeNode(10) -> value=10, left=None, right=None

# Step 2: Create root and children
root = TreeNode(10)       # Root node with value 10
root.left = TreeNode(5)   # Left child of 10 is 5
root.right = TreeNode(20) # Right child of 10 is 20

# Step 3: Pre-order traversal (Root -> Left -> Right)
def preorder(node):
    if node:  # If node exists
        print(node.value, end=" ")  # Print root value
        preorder(node.left)   # Traverse left subtree
        preorder(node.right)  # Traverse right subtree

print("Pre-order traversal of tree:")
preorder(root)  # Output: 10 5 20

# -----------------------------------------------------
# 5. HASH TABLE (DICTIONARY - KEY-VALUE STORE)
# -----------------------------------------------------
# Imagine a phone book: you look for a name (key) to find the number (value)
# Example: {"Alice": 1234, "Bob": 5678}

print("\n=== HASH TABLE / DICTIONARY EXAMPLE ===")

# Step 1: Create dictionary
fruits = {"apple": 3, "banana": 5}  # apple=3, banana=5
print("Initial dictionary:", fruits)

# Step 2: Add item
fruits["orange"] = 7  # Add key="orange", value=7
print("After adding orange:", fruits)

# Step 3: Access value by key
print("Banana count:", fruits["banana"])  # Output: 5

# Step 4: Remove item
del fruits["apple"]  # Remove key "apple"
print("After removing apple:", fruits)

# =====================================================
# =====================================================
print("\nAll done!")
print("- Stack (LIFO)")
print("- Queue (FIFO)")
print("- Linked List (Nodes with pointers)")
print("- Binary Tree (Parent with children)")
print("- Hash Table / Dictionary (Key-Value store)")
```
