(1) WHAT IS A HASH TABLE?
------------------------------------------------------------
A hash table is a data structure that stores data as:

        KEY  ‚Üí  VALUE

Example:
        RollNumber ‚Üí StudentName
        101        ‚Üí "Alex"

Meaning:
- KEY   = what you search with
- VALUE = what you get

Goal of a hash table:
üëâ Find data VERY FAST


(2) WHY NORMAL LISTS ARE SLOW
------------------------------------------------------------
Example list:
        index:  0   1   2   3   4
        data : [10, 20, 30, 40, 50]

To find 50:
- check index 0 ‚ùå
- check index 1 ‚ùå
- check index 2 ‚ùå
- check index 3 ‚ùå
- check index 4 ‚úÖ

This is SLOW for big data.


(3) BASIC IDEA OF HASH TABLE
------------------------------------------------------------
Instead of searching everywhere,
we DIRECTLY go to the correct location.

We do this using:
üëâ HASH FUNCTION


(4) WHAT IS A HASH FUNCTION?
------------------------------------------------------------
A hash function:
- Takes a KEY
- Converts it into a NUMBER (index)

Example:
        key = 23
        hash(key) = 23 % 5 = 3

So data goes into:
        index 3


(5) BUCKETS (IMPORTANT)
------------------------------------------------------------
A hash table is just a list of BUCKETS.

Example: size = 5

Index (bucket number):
        0   1   2   3   4

Buckets:
        []  []  []  []  []

Each bucket can store data.


(6) INSERTING DATA (STEP BY STEP)
------------------------------------------------------------
Insert:
        key = 23
        value = "Apple"

Step 1: Apply hash function
        23 % 5 = 3

Step 2: Go to bucket 3

Step 3: Store data
        bucket 3 ‚Üí [(23, "Apple")]

Now table looks like:
        0 ‚Üí []
        1 ‚Üí []
        2 ‚Üí []
        3 ‚Üí [(23, "Apple")]
        4 ‚Üí []


(7) INSERTING MORE DATA
------------------------------------------------------------
Insert:
        key = 18
        value = "Banana"

Step 1:
        18 % 5 = 3

Step 2:
        Goes to SAME bucket (3)

This is called:
üëâ COLLISION


(8) WHAT IS COLLISION?
------------------------------------------------------------
Collision happens when:
        two keys give SAME index

Example:
        23 % 5 = 3
        18 % 5 = 3

Both want bucket 3.


(9) HOW COLLISION IS HANDLED (CHAINING)
------------------------------------------------------------
We store MULTIPLE values in same bucket.

Bucket 3:
        [(23, "Apple"), (18, "Banana")]

This method is called:
üëâ CHAINING


(10) SEARCHING DATA
------------------------------------------------------------
To find key = 18

Step 1:
        18 % 5 = 3

Step 2:
        Go directly to bucket 3

Step 3:
        Search inside bucket:
            (23, "Apple") ‚ùå
            (18, "Banana") ‚úÖ

Value found FAST.


(11) WHY HASH TABLE IS FAST
------------------------------------------------------------
- We DO NOT search whole table
- We jump directly to one bucket
- Usually only 1‚Äì2 comparisons

Time Complexity (average):
        O(1)  ‚Üí VERY FAST


(12) REAL LIFE EXAMPLE
------------------------------------------------------------
Think of a hash table like:
        Library with numbered shelves

- Book ID ‚Üí Shelf number
- Hash function ‚Üí Librarian
- Collision ‚Üí Multiple books on same shelf
- Chaining ‚Üí Books arranged on that shelf


(13) IMPORTANT POINTS TO REMEMBER
------------------------------------------------------------
1) Hash table stores KEY ‚Üí VALUE
2) Hash function decides location
3) Buckets store data
4) Collision is normal
5) Chaining solves collision
6) Average search is O(1)

```python
# ============================================================
#                  HASH TABLE
# ============================================================
# RULE FOR YOU:
# Read TOP ‚Üí BOTTOM
# ============================================================


# ------------------------------------------------------------
# (1) WHAT IS A HASH TABLE?
# ------------------------------------------------------------
# A hash table stores data in this form:
#
#        KEY  ‚Üí  VALUE
#
# Example:
#        "apple"  ‚Üí  10
#        "banana" ‚Üí  20
#
# KEY   = something you search with
# VALUE = data you want
#
# A hash table uses a HASH FUNCTION to decide
# WHERE to store the key-value pair
# ------------------------------------------------------------


# ------------------------------------------------------------
# (2) WHY DO WE NEED A HASH FUNCTION?
# ------------------------------------------------------------
# Computers cannot directly store strings like "apple" as index
#
# This is INVALID:
#     table["apple"] ‚ùå
#
# So we convert the key into a NUMBER
#
# Example:
#     "apple" ‚Üí 834738475
#
# Then we reduce it into a SMALL number
#
# Example:
#     834738475 % 5 = 0
#
# So "apple" goes into bucket 0
# ------------------------------------------------------------


# ------------------------------------------------------------
# (3) CREATING THE HASH TABLE USING A CLASS
# ------------------------------------------------------------
# A class lets us group:
#  - data (the table)
#  - behavior (insert, get, display)
# ------------------------------------------------------------

class HashTable:

    # --------------------------------------------------------
    # (4) CONSTRUCTOR FUNCTION
    # --------------------------------------------------------
    # This runs automatically when we create the hash table
    #
    # Example:
    #     ht = HashTable(5)
    #
    # size = 5
    # --------------------------------------------------------
    def __init__(self, size):

        # (4.1) Save the size
        # This tells us how many buckets we have
        # You store how many buckets the table has
        self.size = size

        # (4.2) Create empty buckets
        #
        # This line creates a list of empty lists
        #
        # If size = 5, result is:
        # range(5) ‚Üí 0, 1, 2, 3, 4

        # [] for _ in range(size) means: Create an empty list [] and Do it 5 times
        # result:
        # [
        #   [],  # bucket 0
        #   [],  # bucket 1
        #   [],  # bucket 2
        #   [],  # bucket 3
        #   []   # bucket 4
        # ]
        # Each inner list is called a bucket.
        self.table = [[] for _ in range(size)]


    # --------------------------------------------------------
    # (5) HASH FUNCTION
    # --------------------------------------------------------
    # INPUT  ‚Üí key (example: "apple")
    # OUTPUT ‚Üí index number (example: 0, 1, 2...)
    # Convert a key (like "apple") into a bucket index (0‚Äì4)
    # --------------------------------------------------------
    def hash_function(self, key):

        # (5.1) Python's built-in hash() function
        # Converts a value into a large integer: Very large, Looks random and Can change between program runs
        # Example:
        #     hash("apple") = 7428394729384 (random big number)
        big_number = hash(key)

        # (5.2) Use modulo (%) to keep index in range or limit range
        # we force the number into a valid bucket index
        # If size = 5:
        #     big_number % 5 ‚Üí 0 to 4 bucket only
        index = big_number % self.size

        # (5.3) Return the index
        # we know which bucket to use.
        return index


    # --------------------------------------------------------
    # (6) INSERT FUNCTION (ADD DATA)
    # --------------------------------------------------------
    # Store key ‚Üí value in the hash table
    # Example:
    #     insert("apple", 10)
    # --------------------------------------------------------
    def insert(self, key, value):

        # (6.1) Find bucket index using hash function
        # hash("apple") % 5 ‚Üí 0
        index = self.hash_function(key)

        # (6.2) Go to that bucket
        # If index = 2, then:, bucket = self.table[2]
        bucket = self.table[index]

        # (6.3) CHECK FOR COLLISION
        #
        # Collision = two keys go to same bucket
        #
        # Example bucket:
        # apple ‚Üí bucket 2
        # banana ‚Üí bucket 2
        #     [ ["apple",10], ["banana",20] ]
        for pair in bucket:

            # (6.4) pair looks like: ["apple", 10]
            # pair[0] = key
            # pair[1] = value

            # (6.5) If key already exists ‚Üí update value
            if pair[0] == key:
                pair[1] = value
                return  # stop function

            # insert("apple", 50)
            # Update: ["apple", 10] ‚Üí ["apple", 50]

        # (6.6) If key NOT found, add new pair
        bucket.append([key, value])


    # --------------------------------------------------------
    # (7) GET FUNCTION (READ DATA)
    # --------------------------------------------------------
    # Find the value for a given key
    # Example:
    #     get("apple") ‚Üí 10
    # --------------------------------------------------------
    def get(self, key):

        # (7.1) Find index for key
        index = self.hash_function(key)

        # (7.2) Go to correct bucket
        bucket = self.table[index]

        # (7.3) Search inside bucket
        for pair in bucket:

            # (7.4) If key matches, return value
            if pair[0] == key:
                return pair[1]

        # (7.5) Key not found
        return "Key not found"


    # --------------------------------------------------------
    # (8) DISPLAY FUNCTION (VISUALIZE TABLE)
    # --------------------------------------------------------
    # Show all buckets and their contents
    def display(self):

        # (8.1) Loop through each bucket
        for i in range(self.size):

            # (8.2) Print bucket number and contents
            print("Bucket", i, "‚Üí", self.table[i])



# ============================================================
# (9) USING THE HASH TABLE ‚Äî STEP BY STEP
# ============================================================

# (9.1) Create hash table with 5 buckets
ht = HashTable(5)

# Current table:
# [
#   [], [], [], [], []
# ]

# (9.2) Insert values
ht.insert("apple", 10)
# Example:
# hash("apple") % 5 ‚Üí 0
# Bucket 0: [["apple", 10]]

ht.insert("banana", 20)
# Example:
# hash("banana") % 5 ‚Üí 2
# Bucket 2: [["banana", 20]]

ht.insert("orange", 30)
# Example:
# hash("orange") % 5 ‚Üí 2
# Collision happens!
# Bucket 2: [["banana",20], ["orange",30]]

# (9.3) Read values
print("apple ‚Üí", ht.get("apple"))
print("banana ‚Üí", ht.get("banana"))
print("grape ‚Üí", ht.get("grape"))  # not present

# (9.4) Display full hash table
ht.display()
```

```python
# # We make 5 empty boxes
boxes = [[], [], [], [], []]

# This is our magic rule
def magic_box(name):
    # count letters in name
    return len(name) % 5

# PUT a toy
def put(name, toy):
    box = magic_box(name)
    boxes[box].append([name, toy])

# FIND a toy
def find(name):
    box = magic_box(name)
    for item in boxes[box]:
        if item[0] == name:
            return item[1]
    return "Not found"

# LET'S PLAY
put("car", "üöó")
put("ball", "‚öΩ")
put("doll", "üß∏")

print(find("car"))
print(find("ball"))
print(boxes)
```

```python
# ============================================================
# üß† DATA STRUCTURE: HASH TABLE
# ============================================================
# ============================================================


# ------------------------------------------------------------
# üß© WHAT IS A HASH TABLE?
# ------------------------------------------------------------
# A Hash Table is like a set of boxes.
# Each box can store information.
#
# Example:
# Name  -> Age
# "Tom" -> 10
# "Bob" -> 12
#
# We use a MAGIC FUNCTION (hash function)
# to decide WHICH box to put the data in.
# ------------------------------------------------------------


# ------------------------------------------------------------
# üèóÔ∏è STEP 1: CREATE A CLASS
# ------------------------------------------------------------
# A class is like a blueprint or a recipe.
# It tells Python how to make a Hash Table.
# ------------------------------------------------------------

class HashTable:

    # --------------------------------------------------------
    # üß± STEP 2: CONSTRUCTOR (__init__)
    # --------------------------------------------------------
    # This function runs AUTOMATICALLY
    # when we create a HashTable.
    #
    # "size" means how many boxes we want.
    # --------------------------------------------------------
    def __init__(self, size):

        # Save the size (number of boxes)
        self.size = size

        # Create empty boxes (lists inside a list)
        # Each box will store key-value pairs
        #
        # Example:
        # [
        #   [],  <- box 0
        #   [],  <- box 1
        #   [],  <- box 2
        # ]
        self.table = [[] for _ in range(size)]


    # --------------------------------------------------------
    # üî¢ STEP 3: HASH FUNCTION
    # --------------------------------------------------------
    # This function turns a KEY into a NUMBER.
    #
    # Example:
    # "cat" -> 3
    # "dog" -> 1
    #
    # This number tells us WHICH box to use.
    # --------------------------------------------------------
    def hash_function(self, key):

        # Start with number 0
        hash_value = 0

        # Go through EACH letter in the key
        for char in key:

            # ord(char) gives a number for each letter
            # Example: 'a' -> 97, 'b' -> 98
            hash_value += ord(char)

        # Use modulo (%) to stay inside table size
        # This keeps the number SMALL
        return hash_value % self.size


    # --------------------------------------------------------
    # ‚ûï STEP 4: INSERT (ADD DATA)
    # --------------------------------------------------------
    # This function adds a key-value pair
    # Example: "apple" -> 5
    # --------------------------------------------------------
    def insert(self, key, value):

        # Find the box number using hash function
        index = self.hash_function(key)

        # Go through items already in that box
        for pair in self.table[index]:

            # If key already exists, update value
            if pair[0] == key:
                pair[1] = value
                return

        # If key does not exist, add new pair
        self.table[index].append([key, value])


    # --------------------------------------------------------
    # üîç STEP 5: SEARCH (FIND DATA)
    # --------------------------------------------------------
    # This function finds a value using a key
    # --------------------------------------------------------
    def search(self, key):

        # Find the box number
        index = self.hash_function(key)

        # Look inside the box
        for pair in self.table[index]:

            # If key matches, return the value
            if pair[0] == key:
                return pair[1]

        # If key not found
        return "Key not found"


    # --------------------------------------------------------
    # ‚ùå STEP 6: DELETE (REMOVE DATA)
    # --------------------------------------------------------
    def delete(self, key):

        # Find the box number
        index = self.hash_function(key)

        # Go through box items
        for i, pair in enumerate(self.table[index]):

            # If key found, remove it
            if pair[0] == key:
                self.table[index].pop(i)
                return

        print("Key not found")


    # --------------------------------------------------------
    # üì¶ STEP 7: DISPLAY HASH TABLE
    # --------------------------------------------------------
    # This shows ALL boxes and data
    # --------------------------------------------------------
    def display(self):

        print("\nüì¶ HASH TABLE CONTENTS:")

        # Go through each box
        for i, box in enumerate(self.table):
            print(f"Box {i}:", box)


# ============================================================
# üéÆ LET'S USE THE HASH TABLE
# ============================================================

# Create a Hash Table with 5 boxes
hash_table = HashTable(5)

# Insert data
hash_table.insert("apple", 10)
hash_table.insert("banana", 20)
hash_table.insert("grape", 30)
hash_table.insert("orange", 40)

# Display table
hash_table.display()

# Search for values
print("\nüîç Searching:")
print("apple ->", hash_table.search("apple"))
print("banana ->", hash_table.search("banana"))
print("mango ->", hash_table.search("mango"))

# Delete a value
print("\n‚ùå Deleting 'banana'")
hash_table.delete("banana")

# Display again
hash_table.display()
```
