```python
# ============================================================
# ðŸ§  GRAPH THEORY
# ============================================================
# RULE:
# READ COMMENTS FIRST.
# CODE IS SECOND.
# COMMENTS ARE THE LESSON.
# ============================================================


# ------------------------------------------------------------
# STEP 1: WHAT IS A GRAPH IN CODE?
# ------------------------------------------------------------
# We use a DICTIONARY.
#
# Think of it like:
# "Who is connected to who?"
#
# Example:
# A is connected to B and C
# B is connected to A and D
# ------------------------------------------------------------

graph = {
    'A': ['B', 'C'],   # A has roads to B and C
    'B': ['A', 'D'],   # B has roads to A and D
    'C': ['A'],        # C has road to A
    'D': ['B']         # D has road to B
}

# ------------------------------------------------------------
# STEP 2: BFS â€” BREADTH FIRST SEARCH
# ------------------------------------------------------------
# IDEA:
# Visit NEAR places first.
#
# TOOL:
# A QUEUE (a line)
#
# Example queue:
# People stand in line.
# First person enters â†’ first person leaves.
# ------------------------------------------------------------

from collections import deque  # deque = special list for queue

def bfs(graph, start):
    # visited = places we already went to
    visited = []

    # queue = places we will visit next
    queue = deque()

    # Step 1: start position goes into queue
    queue.append(start)

    # Step 2: mark start as visited
    visited.append(start)

    print("BFS order:")

    # Step 3: repeat until queue is empty
    while queue:
        # Take first element from queue
        current = queue.popleft()

        # Show where we are
        print(current, end=" ")

        # Look at neighbors
        for neighbor in graph[current]:
            # If not visited yet
            if neighbor not in visited:
                visited.append(neighbor)
                queue.append(neighbor)

    print("\n")

# Run BFS
bfs(graph, 'A')


# ------------------------------------------------------------
# STEP 3: DFS â€” DEPTH FIRST SEARCH
# ------------------------------------------------------------
# IDEA:
# Go DEEP first.
#
# Like:
# Keep going forward until you can't.
# ------------------------------------------------------------

def dfs(graph, start):
    visited = []

    print("DFS order:")

    def go(node):
        visited.append(node)
        print(node, end=" ")

        for neighbor in graph[node]:
            if neighbor not in visited:
                go(neighbor)

    go(start)
    print("\n")

# Run DFS
dfs(graph, 'A')


# ------------------------------------------------------------
# STEP 4: SHORTEST PATH (VERY SIMPLE)
# ------------------------------------------------------------
# We will just calculate distances manually using logic.
# ------------------------------------------------------------

weighted_graph = {
    'A': {'B': 1, 'C': 4},
    'B': {'D': 2},
    'C': {'D': 1},
    'D': {}
}

# Distance table
distance = {
    'A': 0,
    'B': 999,
    'C': 999,
    'D': 999
}

# From A
distance['B'] = distance['A'] + 1
distance['C'] = distance['A'] + 4

# From B
distance['D'] = distance['B'] + 2

# Compare with C path
if distance['C'] + 1 < distance['D']:
    distance['D'] = distance['C'] + 1

print("Shortest distance from A:")
for node in distance:
    print(node, "=", distance[node])
```

```python
def dfs(graph, start_node):
  visited = set()

  print("DFS Traversal Order:")

  def dfs_helper(node):
    visited.add(node)
    print(node, end=" ")

    for neighbor in graph[node]:
      if neighbor not in visited:
        dfs_helper(neighbor)

  dfs_helper(start_node)
  print("\n")

dfs(graph, "A")
```

```python
weighted_graph = {
    "A" : {"B" : 1, "C" : 4},
    "B" : {"A" : 1, "D" : 2},
    "C" : {"A" : 4, "D" : 1},
    "D" : {"B" : 2, "C" : 1, "E" : 3},
    "E" : {"D" : 3}
}

import heapq

def dijkstra(graph, start):
  distances = {}

  for node in graph:
    distances[node] = float("inf")

  distances[start] = 0

  priority_queue = []

  heapq.heappush(priority_queue, (0, start))

  while priority_queue:

    current_distance, current_node = heapq.heappop(priority_queue)

    if current_distance > distances[current_node]:
      continue

    for neighbor, weight in graph[current_node].items():
      distance = current_distance + weight

      if distance < distances[neighbor]:
        distances[neighbor] = distance
        heapq.heappush(priority_queue, (distance, neighbor))

  return distances

shortest_paths = dijkstra(weighted_graph, "A")

print("Shortest Distance from Node A:")
for node in shortest_paths:
  print("A ->", node, "=", shortest_paths[node])
```
