```python
# =====================================================================
# üå≥ DATA STRUCTURE: TREE (BINARY SEARCH TREE - BST)
# =====================================================================
# READ THIS LIKE A STORY üìñ
# NO RUSH ‚è≥
# EVERYTHING IS EXPLAINED FROM ZERO
# =====================================================================


# =====================================================================
# üß† PART 1: WHAT IS A DATA STRUCTURE?
# =====================================================================
# A data structure is a way to STORE data in a computer.
#
# Example:
# - A box üì¶ stores toys
# - A cupboard stores clothes
# - A data structure stores numbers, words, etc.
#
# TREE is one type of data structure.


# =====================================================================
# üå≥ PART 2: WHAT IS A TREE?
# =====================================================================
# A TREE looks like an upside-down tree üå≥
#
# Real tree:
#     root at bottom
#
# Computer tree:
#     ROOT at top
#
# Example TREE:
#
#            10        ‚Üê ROOT (top)
#           /  \
#          5    15
#         / \
#        3   7
#
# IMPORTANT WORDS:
# - ROOT: first/top node
# - NODE: each box with a number
# - EDGE: line connecting nodes
# - LEAF: node with no children
#
# Node 10 is ROOT
# Node 3 is a LEAF


# =====================================================================
# üå≥ PART 3: WHAT IS A BINARY TREE?
# =====================================================================
# BINARY means TWO ‚úåÔ∏è
#
# So:
# - Each node can have MAXIMUM 2 children
# - LEFT child
# - RIGHT child
#
# NOT MORE THAN 2 ‚ùå


# =====================================================================
# üå≥ PART 4: WHAT IS A BINARY SEARCH TREE (BST)?
# =====================================================================
# A Binary Search Tree follows ONE GOLDEN RULE üèÜ
#
# RULE:
# - Smaller numbers go LEFT ‚¨ÖÔ∏è
# - Bigger numbers go RIGHT ‚û°Ô∏è
#
# Example:
#
# Insert numbers: 10, 5, 15
#
# Step 1: 10 ‚Üí ROOT
#
# Step 2: 5
# 5 < 10 ‚Üí go LEFT
#
# Step 3: 15
# 15 > 10 ‚Üí go RIGHT
#
# Result:
#
#            10
#           /  \
#          5    15


# =====================================================================
# üß© PART 5: WHAT IS A NODE?
# =====================================================================
# A NODE is like a small box üì¶
#
# Each node has:
# 1. A VALUE (number)
# 2. LEFT child (another node or nothing)
# 3. RIGHT child (another node or nothing)
#
# Example:
# Node value = 10
# left = 5
# right = 15


# =====================================================================
# üßë‚Äçüíª PART 6: NODE CLASS (CREATING A NODE)
# =====================================================================
class Node:

    # This function runs when we create a new node
    def __init__(self, value):

        # Store the number inside the node
        self.value = value

        # Left child is empty at start
        self.left = None

        # Right child is empty at start
        self.right = None


# =====================================================================
# üå≥ PART 7: CREATING THE TREE
# =====================================================================
class BinarySearchTree:

    # This runs when we create the tree
    def __init__(self):

        # Tree is empty first
        self.root = None


    # =================================================================
    # ‚ûï PART 8: INSERTING A NUMBER INTO TREE
    # =================================================================
    def insert(self, value):

        # STEP 1: If tree is empty
        if self.root is None:

            # Make this value the ROOT
            self.root = Node(value)
            return

        # STEP 2: Tree is not empty
        current = self.root

        # Loop until number is placed
        while True:

            # CASE 1: value is SMALLER
            if value < current.value:

                # If LEFT is empty
                if current.left is None:
                    current.left = Node(value)
                    return
                else:
                    # Move LEFT
                    current = current.left

            # CASE 2: value is BIGGER or EQUAL
            else:

                # If RIGHT is empty
                if current.right is None:
                    current.right = Node(value)
                    return
                else:
                    # Move RIGHT
                    current = current.right


    # =================================================================
    # üîÅ PART 9: TREE TRAVERSAL (IN-ORDER)
    # =================================================================
    # Traversal means VISITING nodes
    #
    # IN-ORDER RULE:
    # LEFT ‚Üí ROOT ‚Üí RIGHT
    #
    # This prints numbers in SORTED ORDER
    def inorder(self, node):

        # If node exists
        if node:

            # Visit LEFT child
            self.inorder(node.left)

            # Print current node
            print(node.value, end=" ")

            # Visit RIGHT child
            self.inorder(node.right)


# =====================================================================
# üß™ PART 10: REAL EXAMPLE WITH NUMBERS
# =====================================================================

tree = BinarySearchTree()

# Insert numbers ONE BY ONE
tree.insert(10)   # ROOT
tree.insert(5)    # goes LEFT of 10
tree.insert(15)   # goes RIGHT of 10
tree.insert(3)    # goes LEFT of 5
tree.insert(7)    # goes RIGHT of 5
tree.insert(12)   # goes LEFT of 15
tree.insert(18)   # goes RIGHT of 15


# Final Tree:
#
#               10
#              /  \
#             5    15
#            / \   / \
#           3   7 12  18


# =====================================================================
# üñ®Ô∏è PART 11: PRINT TREE
# =====================================================================
print("\nTree values in sorted order:")
tree.inorder(tree.root)


# =====================================================================
# üéØ PART 12: WHY TREE IS IMPORTANT?
# =====================================================================
# - FAST searching
# - FAST inserting
# - Used in:
#   - Databases
#   - Games
#   - File systems
#   - Google search
#
# YOU JUST LEARNED A REAL COMPUTER SCIENCE CONCEPT üéâ
# =====================================================================
```

```python
# ============================================================
# üå≥ DATA STRUCTURE: TREE (BINARY TREE)
# ============================================================
# We will explain EVERYTHING in EXTREME DETAIL
# Assume you are 10 years old üë∂
# Read comments slowly and carefully
# ============================================================


# ------------------------------------------------------------
# STEP 1: WHAT IS A NODE?
# ------------------------------------------------------------
# A Tree is made of small pieces called NODES
# Each NODE has:
# 1. A value (data)
# 2. A left child (another node)
# 3. A right child (another node)
#
# Example:
#        10
#       /  \
#      5    15
#
# 10 is a NODE
# 5 and 15 are children of 10


class Node:
    # This function runs when we CREATE a new node
    def __init__(self, value):
        # Store the value inside the node
        self.value = value

        # Left child starts as NOTHING (None)
        self.left = None

        # Right child starts as NOTHING (None)
        self.right = None


# ------------------------------------------------------------
# STEP 2: WHAT IS A TREE?
# ------------------------------------------------------------
# A TREE has:
# - A ROOT (top node)
# - Many child nodes
#
# We will make a Binary Search Tree (BST)
# RULE:
# - Smaller numbers go LEFT
# - Bigger numbers go RIGHT


class BinarySearchTree:

    # This function runs when we create the tree
    def __init__(self):
        # At start, tree is EMPTY
        self.root = None


    # --------------------------------------------------------
    # STEP 3: INSERTING DATA INTO THE TREE
    # --------------------------------------------------------
    def insert(self, value):

        # If tree is empty, make first node the ROOT
        if self.root is None:
            self.root = Node(value)
            return

        # If tree is NOT empty, we find correct position
        current = self.root

        # Loop until we place the value
        while True:

            # If value is smaller, go LEFT
            if value < current.value:

                # If left is empty, insert here
                if current.left is None:
                    current.left = Node(value)
                    return
                else:
                    # Move left
                    current = current.left

            # If value is bigger, go RIGHT
            else:

                # If right is empty, insert here
                if current.right is None:
                    current.right = Node(value)
                    return
                else:
                    # Move right
                    current = current.right


    # --------------------------------------------------------
    # STEP 4: PRINT TREE (IN-ORDER TRAVERSAL)
    # --------------------------------------------------------
    # In-order means:
    # LEFT ‚Üí ROOT ‚Üí RIGHT
    # This prints numbers in SORTED order
    def inorder(self, node):

        # If node exists
        if node:

            # Go left first
            self.inorder(node.left)

            # Print current node value
            print(node.value, end=" ")

            # Go right
            self.inorder(node.right)


# ------------------------------------------------------------
# STEP 5: USING THE TREE
# ------------------------------------------------------------

# Create the tree
tree = BinarySearchTree()

# Insert numbers
tree.insert(10)
tree.insert(5)
tree.insert(15)
tree.insert(3)
tree.insert(7)
tree.insert(12)
tree.insert(18)

# Print the tree
print("Tree values in sorted order:")
tree.inorder(tree.root)


# ------------------------------------------------------------
# STEP 6: FINAL EXPLANATION
# ------------------------------------------------------------
# The tree now looks like this:
#
#            10
#           /  \
#          5    15
#         / \   / \
#        3   7 12  18
#
# In-order traversal prints:
# 3 5 7 10 12 15 18
#
# WHY?
# Because:
# LEFT ‚Üí ROOT ‚Üí RIGHT
# ------------------------------------------------------------
```
