# What is Selection Sort?

Selection Sort is a simple sorting algorithm. It works by repeatedly finding the smallest (or largest, depending on the order you want) element from the unsorted part of the list and swapping it with the first unsorted element. This process is repeated until the whole list is sorted.

# How Selection Sort Works:

Start with the whole list unsorted.

Look through the entire list and find the smallest element.

Swap that smallest element with the first element of the list.

Now, consider the part of the list starting from the second element (ignore the first one since it’s sorted).

Repeat this process for the remaining unsorted part of the list.

Continue until all elements are sorted.

```python
# Let's write the Selection Sort algorithm with detailed explanations

# This is a function that implements the Selection Sort algorithm
# selection_sort(arr): This is the main function where all the sorting happens.
# It takes a list arr as input and sorts it using the Selection Sort algorithm.
def selection_sort(arr):
    # Length of the array (this will tell us how many elements we have)
    # This n line finds out how many elements are in the list.
    # We need this to know how many times we have to loop through the list.
    n = len(arr)

    # Outer loop: We loop through each element of the list one by one
    # The outer loop starts at index 0 and goes through every element of the list.
    # Each time we run this loop, we are sorting one element in place.
    for i in range(n):

        # Let's assume the first unsorted element is the smallest
        # We assume that the element at index i is the smallest.
        # We are going to check if there’s any smaller element in the rest of the list.
        min_index = i

        # Inner loop: We look at the rest of the list to find the smallest element
        # This loop looks at all the elements that come after the current element i (because we already think i is sorted).
        # It compares them to find the smallest element.
        for j in range(i + 1, n):
            # If the element at index 'j' is smaller than the element at 'min_index'
            # If we find a number smaller than the current smallest number, we update min_index to the new smaller element's index.
            if arr[j] < arr[min_index]:
                # If we find a smaller element, update 'min_index'
                min_index = j

        # After finding the smallest element, we swap it with the element at index 'i'
        # This makes sure the first 'i+1' elements are sorted.
        # After finding the smallest element in the unsorted part of the list, we swap it with the first unsorted element (arr[i]).
        # This is done with the line:
        arr[i], arr[min_index] = arr[min_index], arr[i]

        # This is just to visualize the process while sorting (not required for the algorithm)
        print(f"Step {i + 1}: {arr}")

    # After sorting, return the sorted array
    return arr


# Example array to test our selection sort
arr = [64, 25, 12, 22, 11]
print("Original Array:", arr)

# Calling the selection_sort function to sort the array
sorted_array = selection_sort(arr)

# Print the final sorted array
print("Sorted Array:", sorted_array)
```
