```python
# A stack works based on the rule "Last In, First Out" (LIFO),
# which means the last item you add is the first one you take out.

class Stack:
  # This line defines the Stack class. A class is like a blueprint that we can use to create objects. Weâ€™ll use this class to create a stack object.
    # This is the constructor. It runs as soon as we create a stack object.
    def __init__(self):
        # We use a list to store the items in the stack.
        # Initially, the stack is empty, so we start with an empty list.
        # The constructor (__init__) is automatically called when we create a new Stack object.
        # Inside the constructor, self.stack = [] initializes an empty list that will store the items in the stack.
        self.stack = []

    # The push method adds a new item to the stack.
    def push(self, item):
      # The push method adds an item to the stack.
      # We use self.stack.append(item) to add the item to the end of the list.
      # The print statement shows the item that was added to the stack.
        # The append() method adds the item to the end of the list.
        # In a stack, the last item added is the one we will remove first.
        self.stack.append(item)
        # This prints a message so we can see the item we added to the stack.
        print(f"Item {item} pushed to stack.")

    # The pop method removes the topmost item from the stack.
    # The pop method removes the top item from the stack.
      # Before popping, we check if the stack is empty using self.is_empty().
      # If the stack is not empty, we pop the item from the stack using self.stack.pop().
      # We then print the item that was popped and return it.
      # If the stack is empty, we print a warning message and return None.
    def pop(self):
        # First, we check if the stack is empty using the is_empty() method.
        if not self.is_empty():
            # If the stack isn't empty, we pop the last item from the stack.
            # The pop() method removes and gives back the last item of the list.
            item = self.stack.pop()
            # This prints a message showing which item was removed.
            print(f"Item {item} popped from stack.")
            # The popped item is returned so we can use it later.
            return item
        else:
            # If the stack is empty, we can't pop anything.
            print("Stack is empty. Cannot pop.")
            return None  # If the stack is empty, we return None.

    # The peek method allows us to look at the top item of the stack without removing it.
    # The peek method lets us see the topmost item without removing it.
    # We use self.stack[-1] to get the last item in the list (which is the top of the stack).
    # If the stack is empty, we print a message and return None.
    def peek(self):
        # We first check if the stack is empty.
        if not self.is_empty():
            # If the stack is not empty, we return the last item without removing it.
            return self.stack[-1]
        else:
            # If the stack is empty, there's nothing to peek at.
            print("Stack is empty. Nothing to peek.")
            return None  # If the stack is empty, we return None.

    # The is_empty method checks if the stack is empty.
    # If the length of the stack (using len(self.stack)) is 0, it means the stack is empty, so we return True. Otherwise, we return False.
    def is_empty(self):
        # If the length of the stack is 0, it means the stack is empty.
        return len(self.stack) == 0

    # The size method tells us how many items are in the stack.
    def size(self):
        # We return the length of the stack (number of items).
        return len(self.stack)

    # The print_stack method shows the current state of the stack.
    # If the stack is not empty, it prints the list of items in the stack.
    # If the stack is empty, it prints "Stack is empty."
    def print_stack(self):
        # We check if the stack is empty before printing it.
        if not self.is_empty():
            # If the stack is not empty, print the list of items.
            print("Current stack:", self.stack)
        else:
            # If the stack is empty, print that the stack is empty.
            print("Stack is empty.")

# Now, let's test our stack class and see how it works!
# Create a new stack object. This will call the __init__ method to create an empty stack.
my_stack = Stack()

# Let's check the initial state of the stack. It should be empty.
my_stack.print_stack()

# Push some items to the stack.
my_stack.push(10)  # Pushes the number 10 onto the stack.
my_stack.push(20)  # Pushes the number 20 onto the stack.
my_stack.push(30)  # Pushes the number 30 onto the stack.

# Print the current state of the stack to see what items are inside.
my_stack.print_stack()

# Use peek() to see the topmost item in the stack without removing it.
top_item = my_stack.peek()
# Print the item we peeked at. This will be 30, because it's the most recently added item.
print(f"Top item is: {top_item}")

# Pop an item from the stack. This removes the topmost item, which is 30.
my_stack.pop()

# Print the state of the stack after popping an item. It should show 10 and 20 now.
my_stack.print_stack()

# Pop another item. This removes the next topmost item, which is 20.
my_stack.pop()

# Print the state of the stack again. Now it should show only 10.
my_stack.print_stack()

# Check if the stack is empty.
# If it's empty, it should return True, otherwise False.
print(f"Is the stack empty? {my_stack.is_empty()}")

# Try to pop from the stack when it's empty. This will print a warning message.
my_stack.pop()

# Finally, print the stack one last time. It should be empty now.
my_stack.print_stack()
```

```python
class Stack:
  def __init__(self):
    self.stack = []

  def push(self, item):
    self.stack.append(item)
    print(f'Item {item} pushed to stack')

  def pop(self):
    if not self.is_empty():
      item = self.stack.pop()
      print(f"Item {item} popped from stack")
      return item
    else:
      print("Stack is empty. cannot pop")
      return None

  def peek(self):
    if not self.is_empty():
      return self.stack[-1]
    else:
      print("stack is empty. nothing to peek")

  def is_empty(self):
    return len(self.stack) == 0

  def size(self):
    return len(self.stack)

  def print_stack(self):
    if not self.is_empty():
      print("current stack: ", self.stack)
    else:
      print("stack is empty")


my_stack = Stack()

my_stack.push(10)
my_stack.push(20)
my_stack.push(40)

my_stack.print_stack()

top_item = my_stack.peek()
print(f"Top item is: {top_item}")

my_stack.pop()

my_stack.print_stack()

my_stack.pop()

my_stack.print_stack()

print(f"IS the stack empty? {my_stack.is_empty()}")

my_stack.pop()

my_stack.print_stack()
```
