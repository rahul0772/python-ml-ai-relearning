```python
# ============================================================
# LINKED LIST
# ============================================================

# ------------------------------------------------------------
# WHAT IS A LINKED LIST?
# ------------------------------------------------------------
COMMENT = ''' A Linked List is a data structure used to store a collection of elements (values), but unlike normal lists or arrays, the elements are not stored in contiguous memory locations.
Instead, each element (called a node) points to the next element in the list.
Think of a Linked List as a series of train cars, where each car holds a value and points to the next car in the train.
If the train stops at the last car, the last car's pointer will be None, indicating there's no more cars in the train.'''
# A Linked List is a way to store many values (numbers, words, etc.)
# BUT unlike a normal list, the values are NOT stored next to each other
#
# Instead, each value lives inside a small box called a NODE ðŸ“¦
# And each node knows where the NEXT node is
# Value: The actual data (e.g., a number, string).
# Pointer (or Link): A reference to the next node in the list.
# Example:
# [10] -> [20] -> [30] -> None
#The first node holds 10 and points to the second node.
# The second node holds 20 and points to the third node.
# The third node holds 30 and points to None (end of list).
# "None" means: âŒ nothing comes after this
# ------------------------------------------------------------


# ------------------------------------------------------------
# STEP 1: CREATE A NODE
# ------------------------------------------------------------
# The Node class defines the basic structure of a node. A node has two key properties:
# Data: The value you want to store (e.g., 10, 20, 30).
# Next: A reference (pointer) to the next node in the list.
# Initially, when we create a node, this is set to None (because there is no next node yet).

# A Node is ONE small box in the linked list
# Think of it like ONE train coach ðŸš†
#
# Every coach has:
# 1ï¸âƒ£ A value (data)
# 2ï¸âƒ£ A hook that connects to the next coach (next)

class Node:
    # --------------------------------------------------------
    # __init__ is a SPECIAL function called a CONSTRUCTOR
    # Python AUTOMATICALLY runs it when we create a Node
    # --------------------------------------------------------
    def __init__(self, data):
        # ----------------------------------------------------
        # "data" is the value we want to store (e.g., 10, 20, 30)
        # A data field to store the value.
        # "self.data" means:
        # ðŸ‘‰ store this value INSIDE the node
        # ----------------------------------------------------
        self.data = data

        # ----------------------------------------------------
        # "self.next" will store the ADDRESS of the next node
        # The 'next' pointer will initially point to None because the node doesn't have a next node yet
        # At the beginning, we DON'T have a next node
        # So we put None (means nothing)
        # A next field that points to None at the start (until we connect it to another node).
        # ----------------------------------------------------
        self.next = None


# ------------------------------------------------------------
# STEP 2: CREATE LINKED LIST CLASS
# ------------------------------------------------------------
# This LinkedList class controls the ENTIRE linked list
# It knows where the first node (head) is and provides operations to manage the list, like adding or deleting nodes.

class LinkedList:
    def __init__(self):
        # ----------------------------------------------------
        # "head" means the FIRST node of the linked list
         # The linked list starts empty, so the 'head' is set to None
        #  The head is the first node in the list. If the list is empty, head is None.
        # When the list is EMPTY:
        # there is no first node
        # so head = None
        # ----------------------------------------------------
        self.head = None


    # --------------------------------------------------------
    # ADD NODE AT END (APPEND)
    # The append method adds a new node with the given data to the end of the list. If the list is empty, it sets the new node as the head of the list.
    # --------------------------------------------------------
    def append(self, data):
        # ----------------------------------------------------
        # This function adds a NEW node at the END of the list
        # ----------------------------------------------------

        # Create a new Node object
        # This calls Node.__init__(data)
        new_node = Node(data)

        # ----------------------------------------------------
       # Case 1: If the list is empty (head is None), the new node becomes the first node
        # ----------------------------------------------------
        if self.head is None:
            # No nodes exist yet
            # So the new node becomes the FIRST node
            self.head = new_node

            print(f"Added {data} as the first node.")
            return  # stop the function here

        # ----------------------------------------------------
       # Case 2: If the list is not empty, traverse to the last node
        # ----------------------------------------------------

        # Start from the first node
        current = self.head

        # ----------------------------------------------------
        # Move through the list
        # Keep going until current.next is None
        # That means: we reached the LAST node
        # ----------------------------------------------------
        while current.next is not None:
            current = current.next

        # ----------------------------------------------------
        # Now we are at the LAST node
        # Attach the new node at the end
        # ----------------------------------------------------
        current.next = new_node

        print(f"Added {data} to the end of the list.")

        ### Create a new node with the provided data.
        # If the list is empty (head is None), set the new node as the first node.
        # Otherwise, traverse through the list to find the last node and append the new node there.


    # --------------------------------------------------------
    # PRINT LINKED LIST
    # The print_list method prints all the elements of the linked list, from the head to the end (indicated by None).
    # If the list is empty (head is None), print a message indicating the list is empty.
    # Otherwise, start from the head node and traverse through each node, printing its data followed by ->.
    # When you reach the end (current is None), print None to indicate the end of the list.
    # --------------------------------------------------------
    def print_list(self):
        # ----------------------------------------------------
        # This function shows all values in the linked list
        # ----------------------------------------------------

        # If the list is empty
        if self.head is None:
            print("Linked List is empty.")
            return

        # Start from the first node
        current = self.head

        print("Linked List:")

        # ----------------------------------------------------
        # Go through each node one by one
        # ----------------------------------------------------
        while current is not None:
            # Print the data of current node
            print(current.data, end=" -> ")

            # Move to the next node
            current = current.next

        # When current becomes None, we stop
        print("None")  # shows end of list


    # --------------------------------------------------------
    # DELETE FIRST NODE
    # The delete_first method removes the first node in the list (the node pointed to by the head).
    # If the list is empty (head is None), print a message and do nothing.
    # Otherwise, print the value of the node being deleted (the current head).
    # Move the head to the next node in the list (self.head = self.head.next).
    # --------------------------------------------------------
    def delete_first(self):
        # ----------------------------------------------------
        # This removes the FIRST node (head)
        # ----------------------------------------------------

        # If list is empty
        if self.head is None:
            print("List is empty. Nothing to delete.")
            return

        # Print what we are deleting
        print(f"Deleted {self.head.data} from the start.")

        # ----------------------------------------------------
        # Move head to the NEXT node
        #
        # Old head is forgotten (Python deletes it automatically)
        # ----------------------------------------------------
        self.head = self.head.next


    # --------------------------------------------------------
    # SEARCH FOR A VALUE
    # The search method checks if a specific value exists in the list and prints the position of the first occurrence.
    # Start from the head node.
    # Traverse through the list, checking each node's data.
    # If the data matches the key you're looking for, print the position and stop.
    # If the list is finished and the key isn't found, print that the value wasn't found.
    # # --------------------------------------------------------
    def search(self, key):
        # ----------------------------------------------------
        # This function looks for a value inside the list
        # ----------------------------------------------------

        # Start from the first node
        current = self.head

        # Position tells us where we are (0, 1, 2...)
        position = 0

        # ----------------------------------------------------
        # Go through all nodes
        # ----------------------------------------------------
        while current is not None:
            # If data matches what we are searching
            if current.data == key:
                print(f"Found {key} at position {position}.")
                return

            # Move to next node
            current = current.next
            position += 1

        # If we finish loop and didn't find it
        print(f"{key} not found in the list.")


# ============================================================
# STEP 3: TEST THE LINKED LIST
# After defining the LinkedList and Node classes, you can create an instance of the LinkedList class and use the methods to manipulate and view the list.
# ============================================================

# Create a LinkedList object
my_list = LinkedList()

# Print empty list
my_list.print_list()

# ------------------------------------------------------------
# Add elements to the list
# ------------------------------------------------------------
my_list.append(10)
my_list.append(20)
my_list.append(30)

# ------------------------------------------------------------
# Print list again
# ------------------------------------------------------------
my_list.print_list()

# ------------------------------------------------------------
# Search values
# ------------------------------------------------------------
my_list.search(20)  # exists
my_list.search(40)  # does not exist

# ------------------------------------------------------------
# Delete first node
# ------------------------------------------------------------
my_list.delete_first()

# ------------------------------------------------------------
# Print list after deletion
# ------------------------------------------------------------
my_list.print_list()

a = ''' We create an empty linked list.
Add 10, 20, and 30 to the list using append.
Print the list to check the elements.
Search for a value (20 and 40).
Delete the first node (10).
Print the list again after deleting the first node.'''
```
