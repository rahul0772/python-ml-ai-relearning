```python
# Insertion Sort Algorithm

# This defines a function insertion_sort that accepts a list arr as input. We will sort this list using the insertion sort algorithm.
def insertion_sort(arr):
    # Step 1: Print the initial state of the array/sorting process started
    print("Starting Insertion Sort...")
    print(f"Initial array: {arr}\n")  # Print the array we are going to sort

    # Step 2: Loop through the array starting from index 1 (second element)
    # Because the first element is trivially considered sorted.
    # We start iterating through the array from index 1. This is because we assume that the element at index 0 is already sorted.
    # The loop will go from i = 1 to i = len(arr) - 1.
    for i in range(1, len(arr)):  # Start at index 1 (second element)

        # Step 3: Pick the key element
        # Store the current element to be inserted in the sorted part of the array
        # key stores the element at index i that we will insert into the correct position within the sorted portion of the array.
        # j is initialized to i - 1 to compare the key with the previous elements.

        key = arr[i]  # `key` is the element we're going to find the right position for
        j = i - 1     # `j` starts just before `i`, to compare with previous elements

        # We print the step number (STEP i) to indicate which step of the sorting process we're at.
        # We also print the current key (the element we are sorting).
        print(f"STEP {i}:")
        print(f"Key selected: {key} (from index {i})")  # Print the key element

        # Step 4: Shift elements in the sorted part of the array to the right, if they're greater than `key`
        # This while loop checks whether any element in the sorted portion of the array (left side) is greater than the key.
        # If it is, the element will be shifted to the right.
        # j >= 0 ensures we don't go out of bounds, and arr[j] > key checks if the current element needs to be shifted.
        while j >= 0 and arr[j] > key:  # While `j` is valid and arr[j] is greater than the `key`

            # Step 5: Print comparison and shifting of elements to the right
            # We print which element is being compared with key and is being shifted to the right.
            print(f"  Compare: {arr[j]} > {key} → shift {arr[j]} right")  # Print which element is greater than key
            arr[j + 1] = arr[j]  # Shift the element at `arr[j]` to `arr[j+1]`
            # If the element at arr[j] is greater than key, we shift it to the right by one position.
            # The element at index j moves to arr[j + 1], creating space for the key.

            # We also print the current state of the array after the shift.
            print(f"  Array after shift: {arr}")  # Print array after shifting an element to the right

            # j is decremented to check the next previous element in the array. We continue shifting until we find an element smaller than key.
            j -= 1  # Move `j` one position left to compare the next element

        # Step 6: Place the key element at its correct position
        # Once we find the correct position for the key (i.e., when arr[j] is no longer greater than key), we place the key in its correct sorted position at arr[j + 1].
        arr[j + 1] = key  # Insert the `key` into the correct position (after the last element that was smaller)

        # After placing the key in its sorted position, we print the insertion message, including the position where the key was inserted and the array's state.
        print(f"Insert key {key} at index {j + 1}")  # Print where the key has been inserted
        print(f"Array after insertion: {arr}\n")  # Print the array after insertion of the key

    # Step 7: Print the final sorted array
    # Once the loop finishes, it means the entire array is sorted.
    # We print "Insertion Sort completed." to indicate the end of the sorting process.
    # Finally, the sorted array is returned.
    print("Insertion Sort completed.")
    return arr  # Return the sorted array
arr = [5,67,22,3, 54, 4, 21,56,11,1]
sorted_arr = insertion_sort(arr)
print("Final Sorted Array: ", sorted_arr)
```

```python
def insertion_sort(arr):
  print("starting Insertion sort...")

  for i in range(1, len(arr)):
    key = arr[i]
    j= i-1

    print(f"STEP {i}:")
    print(f"Key selected: {key} from index {i}")

    while j >= 0 and arr[j] > key:
      print(f" Compare: arr[j] > {key} -> shift {arr[j]} right")
      arr[j+1] = arr[j]
      print(f" Array afters shift: {arr}")
      j-=1

    arr[j+1] = key
    print(f"Insert key {key} at index {j+1}")
    print(f"Array after insertion: {arr}\n")

  print("insertion sort completed")
  return arr

arr = [5,3,5,67,22,54,21,56,11]
sorted_arr = insertion_sort(arr)
print("Final Sorted Array: ", sorted_arr)
```
