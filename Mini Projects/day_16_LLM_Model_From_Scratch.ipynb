```python
# Step 1: Sample sentences to help our model "learn"
sentences = [
    "The cat is on the mat.",
    "The dog is in the house.",
    "The bird is in the tree.",
    "The cat is playing with the ball."
]

# Step 2: Breaking sentences into words
words = []
for sentence in sentences:
    words += sentence.split()  # split each sentence into words and add to our list

# Step 3: Print all the words
print("All words:", words)

# sentences = [...]: This is a list that holds some example sentences. The model is going to learn from these.
# sentence.split(): This takes each sentence and splits it into individual words. For example, the sentence "The cat is on the mat." will become ["The", "cat", "is", "on", "the", "mat."].
# words += ...: We add each word to the list words. This helps us collect all the words we want the model to learn from.
```

```python
# Step 4: Count the frequency of each word
word_count = {}
for word in words:
    if word in word_count:
        word_count[word] += 1
    else:
        word_count[word] = 1

# Step 5: Print the word frequencies
print("Word Frequencies:", word_count)

# word_count = {}: This creates an empty dictionary. A dictionary is like a box that stores things in pairs (the word and how many times it appears).
# if word in word_count:: This checks if the word is already in the dictionary.
# word_count[word] += 1: If the word is already in the dictionary, we increase its count by 1.
# else:: If the word isnâ€™t in the dictionary yet, we add it with a count of 1.
# The result is a dictionary that shows how many times each word appeared in the sentences.
```

```python
# Step 6: Guess the next word based on previous words
def guess_next_word(current_word):
    possible_next_words = []
    for i in range(len(sentences)):
        sentence_words = sentences[i].split()
        for j in range(len(sentence_words) - 1):
            if sentence_words[j] == current_word:
                possible_next_words.append(sentence_words[j + 1])

    # Choose the word that appears most often
    if possible_next_words:
        return max(possible_next_words, key=possible_next_words.count)
    else:
        return "I don't know!"

# Example: Guess what comes after "The"
print(guess_next_word("The"))
print(guess_next_word("mat"))
print(guess_next_word("is"))
print(guess_next_word("cat"))

# def guess_next_word(current_word):: This defines a function where you give it a word, and it tries to guess what comes after that word in the sentences.
# possible_next_words.append(...): For each sentence, if it finds the current_word, it remembers the word that comes after it (the next word).
# max(..., key=...): This finds the word that appears the most after the given word and returns it.
```
