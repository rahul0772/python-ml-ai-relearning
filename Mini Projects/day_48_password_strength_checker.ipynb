```python
# ==============================================================
# MINI PASSWORD STRENGTH CHECKER (Intermediate Level)
# ==============================================================


# --------------------------------------------------------------
# STEP 1: Ask the user to enter a password
# --------------------------------------------------------------

# input() means:
# "Hey user, type something and press Enter"

password = input("Enter your password: ")

# Now whatever the user types is stored inside the variable "password"
# Example:
# If user types: Hello123!
# Then:
# password = "Hello123!"



# --------------------------------------------------------------
# STEP 2: Create variables to track rules
# --------------------------------------------------------------

# We will use these variables to check conditions.

has_upper = False        # Will become True if we find uppercase letter
has_lower = False        # Will become True if we find lowercase letter
has_digit = False        # Will become True if we find number
has_special = False      # Will become True if we find special character
score = 0                # This will count password strength points



# --------------------------------------------------------------
# STEP 3: Loop through each character in the password
# --------------------------------------------------------------

# A string is just text.
# Example: "Hello"
# It is actually:
# H e l l o
# 0 1 2 3 4 (indexes)

# A for loop goes one character at a time.

for char in password:

    # Example:
    # If password = "Hello123!"
    # First loop: char = 'H'
    # Second loop: char = 'e'
    # Third loop: char = 'l'
    # etc...


    # ----------------------------------------------------------
    # Check if character is uppercase letter
    # ----------------------------------------------------------

    # isupper() checks:
    # Is this character A-Z ?

    if char.isupper():
        has_upper = True


    # ----------------------------------------------------------
    # Check if character is lowercase letter
    # ----------------------------------------------------------

    # islower() checks:
    # Is this character a-z ?

    elif char.islower():
        has_lower = True


    # ----------------------------------------------------------
    # Check if character is a digit (number)
    # ----------------------------------------------------------

    # isdigit() checks:
    # Is this character 0-9 ?

    elif char.isdigit():
        has_digit = True


    # ----------------------------------------------------------
    # If it is none of the above, it's special character
    # ----------------------------------------------------------

    # Example special characters:
    # ! @ # $ % ^ & * ( ) _ + etc.

    else:
        has_special = True



# --------------------------------------------------------------
# STEP 4: Calculate Score
# --------------------------------------------------------------

# Now we give points based on rules.

# Rule 1: Length check
# len(password) means:
# How many characters are in the password?

if len(password) >= 8:
    score += 1   # same as score = score + 1


# Rule 2: Has uppercase
if has_upper:
    score += 1


# Rule 3: Has lowercase
if has_lower:
    score += 1


# Rule 4: Has digit
if has_digit:
    score += 1


# Rule 5: Has special character
if has_special:
    score += 1



# --------------------------------------------------------------
# STEP 5: Show Detailed Results
# --------------------------------------------------------------

print("\n--------- PASSWORD ANALYSIS ---------")

print("Length of password:", len(password))
print("Contains Uppercase:", has_upper)
print("Contains Lowercase:", has_lower)
print("Contains Number:", has_digit)
print("Contains Special Character:", has_special)
print("Total Score:", score)



# --------------------------------------------------------------
# STEP 6: Final Strength Decision
# --------------------------------------------------------------

# We now decide strength based on score.
# Maximum score possible = 5

if score <= 2:
    print("Strength: WEAK ❌")
elif score == 3 or score == 4:
    print("Strength: MEDIUM ⚠️")
else:
    print("Strength: STRONG ✅")
```
