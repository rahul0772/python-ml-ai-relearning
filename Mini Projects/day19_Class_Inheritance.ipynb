```python
# Step 1: Create a Base Class called "Book"
class Book:
    # Constructor to initialize common properties for all books
    # The __init__ method is a special function that gets called when we create a new object (a book in this case).
    # It initializes the properties of the book, like title, author, and year.
    def __init__(self, title, author, year):
        self.title = title  # Book title (e.g., 'Harry Potter')
        self.author = author  # Author name (e.g., 'J.K. Rowling')
        self.year = year  # Year of publication (e.g., 1997)

    # Method to display book details
    # This method will print the basic details of the bookâ€”title, author, and year. It is common for all books.
    def display_details(self):
        print(f"Title: {self.title}")
        print(f"Author: {self.author}")
        print(f"Year of Publication: {self.year}")

    # Method to display/print the type of book
    # This method will print the type of book. In this case, we simply print "This is a general book."
    def book_type(self):
        print(f"This is a general book.")

# Step 2: Create a Subclass called "FictionBook"
# The FictionBook class is a child class of the Book class.
# This means it inherits the properties and methods of the Book class but can also have its own unique properties and methods.
class FictionBook(Book):
    # Constructor to initialize properties specific to fiction books
    def __init__(self, title, author, year, genre):
      # The FictionBook class is a child class of the Book class.
      # This means it inherits the properties and methods of the Book class but can also have its own unique properties and methods.
        # Use the parent class constructor to initialize common properties
        super().__init__(title, author, year)
        self.genre = genre  # Genre of fiction book (e.g., Fantasy, Mystery)

    # Method to display the genre of the book (specific to FictionBook)
    # This method is specific to fiction books. It prints the genre (like Fantasy or Mystery).
    def display_genre(self):
        print(f"Genre: {self.genre}")

    # Override the method to describe the type of book (for fiction)
    # We override the book_type() method here to specify that this is a fiction book.
    # This is an example of method overriding, where we provide a new implementation of a method that was already defined in the parent class.
    def book_type(self):
        print(f"This is a Fiction Book.")

# Step 3: Create another Subclass called "NonFictionBook"
class NonFictionBook(Book):
    # Constructor to initialize properties specific to non-fiction books
    def __init__(self, title, author, year, subject):
        # Use the parent class constructor to initialize common properties
        super().__init__(title, author, year)
        self.subject = subject  # Subject of non-fiction book (e.g., History, Science)

    # Method to display the subject of the book (specific to NonFictionBook)
    def display_subject(self):
        print(f"Subject: {self.subject}")

    # Override the method to describe the type of book (for non-fiction)
    def book_type(self):
        print(f"This is a Non-Fiction Book.")

# Step 4: Create Objects of FictionBook and NonFictionBook classes
# Fiction Book Example
fiction_book = FictionBook("Harry Potter", "J.K. Rowling", 1997, "Fantasy")

# Non-Fiction Book Example
non_fiction_book = NonFictionBook("Sapiens", "Yuval Noah Harari", 2011, "History")


# We create a FictionBook object called fiction_book with title "Harry Potter", author "J.K. Rowling", year 1997, and genre "Fantasy".
# We create a NonFictionBook object called non_fiction_book with title "Sapiens", author "Yuval Noah Harari", year 2011, and subject "History".
# For both books, we call various methods (display_details(), book_type(), display_genre(), display_subject()) to see how the properties and methods work.

# Step 5: Call Methods on the Objects
print("Fiction Book Details:")
fiction_book.display_details()  # Call method from Book class
fiction_book.display_genre()  # Call method from FictionBook class
fiction_book.book_type()  # Call overridden method from FictionBook class

print("\nNon-Fiction Book Details:")
non_fiction_book.display_details()  # Call method from Book class
non_fiction_book.display_subject()  # Call method from NonFictionBook class
non_fiction_book.book_type()  # Call overridden method from NonFictionBook class
```
