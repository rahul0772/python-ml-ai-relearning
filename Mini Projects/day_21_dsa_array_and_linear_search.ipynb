```python
# Let's start by writing everything in ONE cell of Google Colab.

# Define the Linear Search function
def linear_search(arr, target):
# The function linear_search(arr, target) takes two inputs:
# arr: This is the array (list) where we will search for the target number.
# target: This is the number we are looking for in the array.

    """
    This function performs a linear search in an array to find the target element.
    It will check each element one by one and return the index if the element is found.
    If not, it will return -1 to indicate the target is not in the array.
    """

    # Step 1: Loop through each element of the array (from index 0 to the last index)
    for index in range(len(arr)):  # 'len(arr)' gives the total number of elements in the array
    # The for index in range(len(arr)): part tells the program to go through each element of the array one by one.
    # len(arr) tells us how many elements there are in the array.
    # index will start at 0 and go all the way up to the last index (e.g., for an array of length 5, it will check index 0, 1, 2, 3, and 4).


    # range(len(arr)) creates a sequence of numbers from 0 to len(arr)-1 (basically the indices of the array).
    # For example, if arr is [10, 20, 30], len(arr) would be 3, so range(3) would generate the numbers 0, 1, 2.
    # for index in range(len(arr)): loops through each of these numbers (which represent the indices of the array).
    # For each iteration, index takes the value of the current index (0, 1, 2, etc.) and you can use that to access the elements in arr like this: arr[index].
      # ----------
          # Example:
          # arr = [10, 20, 30]
          # for index in range(len(arr)):
          # print(f"Index {index} has value {arr[index]}")


          # Output:
          # Index 0 has value 10
          # Index 1 has value 20
          # Index 2 has value 30
          # -----------------


        # Step 2: Compare each element with the target value
        if arr[index] == target:
          # The line if arr[index] == target: compares the number at the current position (arr[index]) with the target number.
          # If it finds the target, it immediately returns the index of that element.

            # If we find the element, return its index
            return index  # This means we found the target, and we're returning where it is.

    # Step 3: If we finished checking all elements and didn't find the target, return -1
    # If the loop finishes without finding the target, the function returns -1 to indicate the target is not in the array.
    return -1  # This means the element is not in the array.

# Now, let's try it with an example array and target.
# Array (List) with numbers:
arr = [15, 30, 50, 70, 90]

# Let's say we want to find the number 50 in this array.
target = 50

# Now, let's call the linear_search function with our array and target.
result = linear_search(arr, target)

# Finally, let's print out the result to see what we got.
# If we find the number, the index (position) where it is located will be shown.
# If it's not there, it will print -1.

if result != -1:
    print(f"Element {target} found at index {result}.")
else:
    print(f"Element {target} not found in the array.")
```

```python
# Let's start by writing everything in ONE cell of Google Colab.

# Define the Linear Search function
def linear_search(arr, target):
    """
    This function performs a linear search in an array to find the target element.
    It will check each element one by one and return the index if the element is found.
    If not, it will return -1 to indicate the target is not in the array.
    """

    # Step 1: Loop through each element of the array (from index 0 to the last index)
    for index in range(len(arr)):  # 'len(arr)' gives the total number of elements in the array

        # Step 2: Print what we're checking right now
        print(f"Checking element at index {index}: {arr[index]}")

        # Step 3: Compare each element with the target value
        if arr[index] == target:
            # If we find the element, return its index
            print(f"Found the target {target} at index {index}!\n")  # Print when we find the target
            return index  # This means we found the target, and we're returning where it is.

        # Print a message if the current element is not the target
        else:
            print(f"Target {target} is NOT at index {index}. Moving to next element.\n")

    # Step 4: If we finished checking all elements and didn't find the target, return -1
    print(f"Target {target} is NOT in the array.\n")
    return -1  # This means the element is not in the array.

# Now, let's try it with an example array and target.
# Array (List) with numbers:
arr = [15, 30, 50, 70, 90]

# Let's say we want to find the number 50 in this array.
target = 50

# Now, let's call the linear_search function with our array and target.
result = linear_search(arr, target)

# Finally, let's print out the result to see what we got.
# If we find the number, the index (position) where it is located will be shown.
# If it's not there, it will print -1.

if result != -1:
    print(f"Element {target} found at index {result}.")
else:
    print(f"Element {target} not found in the array.")
```
