```python

# -------------------------------------------------
# 1Ô∏è‚É£ CLASS (BLUEPRINT)
# -------------------------------------------------
# A class is like a recipe.
# This recipe tells how an ice cream should be made.

class IceCream:

    # -------------------------------------------------
    # 2Ô∏è‚É£ CONSTRUCTOR (__init__)
    # -------------------------------------------------
    # This runs automatically when we create an object
    # It gives starting values to the ice cream

    def __init__(self, name, price):
        # ATTRIBUTES (variables inside class)
        self.name = name      # name of ice cream
        self.price = price    # price of ice cream

    # -------------------------------------------------
    # 3Ô∏è‚É£ METHOD (FUNCTION INSIDE CLASS)
    # -------------------------------------------------
    def serve(self):
        print("Serving ice cream üç¶")


# -------------------------------------------------
# 4Ô∏è‚É£ OBJECT (REAL ICE CREAM)
# -------------------------------------------------
# Object is the real thing made from class

simple_icecream = IceCream("Plain Ice Cream", 20)

print("Ice Cream Name:", simple_icecream.name)
print("Ice Cream Price:", simple_icecream.price)
simple_icecream.serve()


# =================================================
# 5Ô∏è‚É£ INHERITANCE
# =================================================
# Inheritance means:
# One class can use another class's features
#
# IceCream ‚Üí Parent
# ChocolateIceCream ‚Üí Child

class ChocolateIceCream(IceCream):

    def __init__(self, name, price, cocoa_level):
        # super() calls parent class constructor
        super().__init__(name, price)
        self.cocoa_level = cocoa_level

    # Same method name, different behavior
    def serve(self):
        print(self.name, "is served with chocolate syrup üç´")


# Create chocolate ice cream object
choco = ChocolateIceCream("Chocolate Ice Cream", 40, "High")

print("\nName:", choco.name)
print("Price:", choco.price)
print("Cocoa Level:", choco.cocoa_level)
choco.serve()


# =================================================
# 6Ô∏è‚É£ ANOTHER CHILD CLASS
# =================================================

class MangoIceCream(IceCream):

    def __init__(self, name, price, season):
        super().__init__(name, price)
        self.season = season

    def serve(self):
        print(self.name, "is served cold and fresh ü•≠")


# Create mango ice cream object
mango = MangoIceCream("Mango Ice Cream", 35, "Summer")

print("\nName:", mango.name)
print("Price:", mango.price)
print("Best Season:", mango.season)
mango.serve()


# =================================================
# 7Ô∏è‚É£ POLYMORPHISM
# =================================================
# Polymorphism means "many shapes" or "many forms".
# In programming, it means that one function or method name can work in many different ways depending on the type of object (or class) that is calling it.
# # Big word but EASY meaning:
#
# Same function name ‚Üí serve()
# Different result for each ice cream

icecreams = [simple_icecream, choco, mango]
# Each of these objects is created from a different class, but they all share the same method name serve()

# Same method name (serve), but each object (plain, chocolate, mango) performs it in its own way.
print("\n--- Ice Cream Shop Serving ---")
for ice in icecreams:
    ice.serve()

# The for loop is going through each ice cream in the list.
# Each ice cream is an object from a different class:
# simple_icecream is from the IceCream class
# choco is from the ChocolateIceCream class
# mango is from the MangoIceCream class
# But they all have the same method name serve().
# When we call ice.serve(), polymorphism happens because each ice cream (each object) does something different with that serve() method, depending on its class.
```
