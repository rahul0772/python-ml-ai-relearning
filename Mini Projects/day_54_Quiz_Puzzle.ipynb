```python
import random  # We will use random to shuffle questions

# ============================================================
# QUIZ GAME PROJECT
# ============================================================
# This is a fun quiz game where players answer multiple-choice questions.
# It will help you learn:
# - User input
# - Randomization
# - Functions
# - Loops
# - Conditionals
# - Tracking score
# ============================================================

# ------------------------------------------------------------
# Step 1: Create a list of questions and answers
# ------------------------------------------------------------
# Each question has a question, a list of choices, and the correct answer.
questions = [
    {
        "question": "What is the capital of France?",
        "choices": ["Berlin", "Madrid", "Paris", "Lisbon"],
        "answer": "Paris"
    },
    {
        "question": "Which planet is known as the Red Planet?",
        "choices": ["Earth", "Mars", "Jupiter", "Venus"],
        "answer": "Mars"
    },
    {
        "question": "What is 2 + 2?",
        "choices": ["3", "4", "5", "6"],
        "answer": "4"
    },
    {
        "question": "Which programming language is this quiz written in?",
        "choices": ["Python", "Java", "C++", "Ruby"],
        "answer": "Python"
    },
    {
        "question": "Who is known as the father of computers?",
        "choices": ["Charles Babbage", "Albert Einstein", "Isaac Newton", "Nikola Tesla"],
        "answer": "Charles Babbage"
    }
]

# ------------------------------------------------------------
# Step 2: Function to ask a question
# ------------------------------------------------------------
def ask_question(q):
    """
    This function will display the question and choices,
    then check if the answer given by the user is correct.
    """

    print(q["question"])  # Display the question
    # Print all choices (options)
    for idx, choice in enumerate(q["choices"], start=1):
        print(f"{idx}. {choice}")

    # Ask the user to select the answer (as a number)
    try:
        answer = int(input("Choose the correct answer (1-4): "))
        # Check if the answer is within the valid range
        if answer < 1 or answer > 4:
            print("Invalid choice. Please choose between 1 and 4.")
            return False
    except ValueError:
        print("Invalid input. Please enter a number between 1 and 4.")
        return False

    # Map the choice number to the actual choice text
    selected_answer = q["choices"][answer - 1]

    # Check if the answer is correct
    if selected_answer == q["answer"]:
        print("Correct! Well done.\n")
        return True
    else:
        print(f"Wrong! The correct answer is: {q['answer']}\n")
        return False

# ------------------------------------------------------------
# Step 3: Main Game Function
# ------------------------------------------------------------
def start_game():
    """
    This function will run the game, asking all questions,
    keeping track of the score, and shuffling the questions.
    """
    score = 0  # Start with 0 score

    # Shuffle the questions so they appear in random order
    random.shuffle(questions)

    # Ask each question in the shuffled list
    for question in questions:
        if ask_question(question):
            score += 1  # Increase score if the answer is correct

    # Show the final score
    print(f"\nGame Over! Your final score is: {score}/{len(questions)}")
    if score == len(questions):
        print("Perfect score! You're a quiz master!")
    elif score > len(questions) // 2:
        print("Great job! You did well.")
    else:
        print("Better luck next time. Keep practicing!")

# ------------------------------------------------------------
# Step 4: Run the Game
# ------------------------------------------------------------
# We call start_game to start the quiz when the program is run.
start_game()
```
