```python
# Simple Chatbot in Python
print("Hi, I'm your chatbot! Type 'quit' to end.")
while True:
    user_input = input("You: ")
    if user_input.lower() == "quit":
        print("Goodbye!")
        break
    else:
        print("Chatbot: You said:", user_input)
```
