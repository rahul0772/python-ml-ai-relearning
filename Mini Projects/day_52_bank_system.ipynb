```python
# ============================================================
# BANK SYSTEM
# ============================================================



# ============================================================
# STEP 1: VARIABLES (STORING DATA)
# ============================================================

# A VARIABLE is like a labeled box ðŸ“¦
# Example:
# Imagine you have a box and you write "money" on it.
# Whatever you put inside becomes stored there.

correct_password = "1234"

# correct_password â†’ This is the NAME of the box.
# "=" â†’ means "store this inside"
# "1234" â†’ This is TEXT (because of quotes "")
#
# VERY IMPORTANT:
# If something is inside quotes "like this"
# It becomes a STRING (text).
#
# Even though 1234 looks like a number,
# because it has quotes, Python treats it as TEXT.
#
# Example:
# "1234"  â†’ text
# 1234    â†’ number


balance = 1000.0

# balance â†’ another box ðŸ“¦
# 1000.0 â†’ This is a FLOAT number.
#
# FLOAT means number with decimal.
# Examples:
# 5.0
# 10.25
# 999.99
#
# We use float for money because:
# Money can have decimal like:
# 1000.50
# 99.99


# ============================================================
# STEP 2: FUNCTION FOR PASSWORD CHECK
# ============================================================

def check_password():

    # def means: "define a function"
    # A FUNCTION is like a machine ðŸ¤–
    # You press button â†’ machine runs code
    #
    # check_password is name of machine.

    for attempt in range(3):

        # for â†’ repeat something
        # range(3) means:
        # 0, 1, 2  (3 times total)
        #
        # So this loop runs 3 times.
        #
        # That means:
        # User gets 3 attempts to enter password.

        entered_password = input("Enter your password: ")

        # input() waits for user to type something.
        # Whatever user types becomes STRING.
        #
        # Example:
        # If user types 1234
        # Python sees it as "1234"
        #
        # It is ALWAYS text.

        if entered_password == correct_password:

            # == means compare
            #
            # Example:
            # 5 == 5 â†’ True
            # 5 == 3 â†’ False
            #
            # So here we check:
            # Is entered_password SAME as correct_password?

            print("Access Granted âœ…")

            return True

            # return means:
            # Send value back to where function was called.
            #
            # Also IMPORTANT:
            # return STOPS the function immediately.
            #
            # So if password correct,
            # it exits the function.

        else:
            print("Wrong password âŒ")

    # If loop finishes (all 3 attempts wrong)
    print("Too many wrong attempts!")

    return False

    # This means:
    # Password failed.


# ============================================================
# STEP 3: MAIN BANK SYSTEM
# ============================================================

def bank_system():

    global balance

    # global means:
    # We want to use the balance variable
    # that we created outside this function.
    #
    # Without global, Python would think
    # balance inside function is NEW variable.
    #
    # So global connects to the original balance.

    while True:

        # while True means:
        # Keep running forever.
        #
        # This is called an infinite loop.
        #
        # It only stops when we use:
        # break

        print("\n----- BANK MENU -----")
        print("1. Check Balance")
        print("2. Deposit Money")
        print("3. Withdraw Money")
        print("4. Exit")

        choice = input("Choose option (1-4): ")

        # User types 1 or 2 or 3 or 4
        # It becomes STRING.
        #
        # Example:
        # If user types 1
        # It becomes "1" (text)


        # ================= CHECK BALANCE =================

        if choice == "1":

            # If user typed "1"

            print("Your balance is:", balance)

            # It prints current balance value.


        # ================= DEPOSIT =================

        elif choice == "2":

            amount = float(input("Enter amount to deposit: "))

            # input gives STRING.
            # float() converts text â†’ decimal number.
            #
            # Example:
            # User types: 500
            # input gives: "500"
            # float("500") becomes: 500.0

            if amount > 0:

                # We check if amount is positive.
                #
                # Example:
                # 500 > 0 â†’ True
                # -10 > 0 â†’ False

                balance = balance + amount

                # This means:
                # Take old balance
                # Add amount
                # Store result back
                #
                # Example:
                # balance = 1000
                # amount = 200
                # New balance = 1200

                print("Money deposited successfully!")
                print("New balance:", balance)

            else:
                print("Deposit must be positive!")



        # ================= WITHDRAW =================

        elif choice == "3":

            amount = float(input("Enter amount to withdraw: "))

            if amount <= 0:

                # If user enters 0 or negative
                print("Invalid amount!")

            elif amount > balance:

                # Example:
                # balance = 1000
                # user tries 2000
                # Not allowed

                print("Not enough balance!")

            else:

                balance = balance - amount

                # Example:
                # balance = 1000
                # withdraw 200
                # new balance = 800

                print("Withdrawal successful!")
                print("Remaining balance:", balance)



        # ================= EXIT =================

        elif choice == "4":

            print("Thank you for using the bank system!")
            break

            # break stops the while loop.
            # So program exits menu.


        else:
            print("Invalid option!")

            # If user types anything else
            # like 9 or abc
            # It shows error message.



# ============================================================
# STEP 4: STARTING THE PROGRAM
# ============================================================

if check_password():

    # This means:
    # Call the check_password function.
    #
    # If it returns True
    # Then run bank_system()

    bank_system()

else:

    print("Access denied. Program ending.")

    # If password function returned False,
    # program ends here.


# ============================================================
# COMPLETE PROGRAM FLOW (SUPER SIMPLE SUMMARY)
# ============================================================

# 1. Ask password (3 chances)
# 2. If correct â†’ open bank menu
# 3. User can:
#       - Check balance
#       - Deposit money
#       - Withdraw money
#       - Exit
# 4. Loop keeps running until Exit
# 5. If password wrong 3 times â†’ program stops
```
