```python
# Tic-Tac-Toe game.
# It lets two players (X and O) play the game.
# The board is a 3x3 grid, and players will take turns to place their symbols ('X' or 'O').

# Create a function to print the board
def print_board(board):
    """
    This function prints the Tic-Tac-Toe board.
    We use a for loop to print each row of the board.
    """
    for row in board:  # Loop through each row of the board
        print("|".join(row))  # Join the row items (cells) with '|' and print the row

# Example:
# If the board looks like this:
# [['X', 'O', ' '],
#  [' ', 'X', ' '],
#  ['O', ' ', 'O']]
# It will print:
# X|O|
#  |X|
# O| |O

# Create a function to check if a player has won
def check_win(board, player):
    """
    This function checks if the player has won.
    It checks the 3 rows, 3 columns, and the 2 diagonals.
    If any of them are filled with the player's symbol, the player wins.
    """
    # Check each row (horizontal check)
    for row in board:
        if row[0] == row[1] == row[2] == player:  # If all 3 cells in a row are the player's symbol
            return True  # The player has won in this row

    # Check each column (vertical check)
    for col in range(3):  # Loop through columns (0, 1, 2)
        if board[0][col] == board[1][col] == board[2][col] == player:  # Check if all cells in the column are the player's symbol
            return True  # The player has won in this column

    # Check diagonals (2 diagonals: top-left to bottom-right, and top-right to bottom-left)
    if board[0][0] == board[1][1] == board[2][2] == player:  # Check top-left to bottom-right diagonal
        return True  # Player wins along this diagonal

    if board[0][2] == board[1][1] == board[2][0] == player:  # Check top-right to bottom-left diagonal
        return True  # Player wins along this diagonal

    return False  # If no win is found

# Example:
# If Player 'X' has this board:
# [['X', 'X', 'X'],
#  [' ', 'O', ' '],
#  [' ', ' ', 'O']]
# The function will return True because Player X has filled the first row.

# Create a function to check if the board is full (draw condition)
def is_full(board):
    """
    This function checks if the board is full.
    If there are no empty spaces left, it is a draw.
    """
    for row in board:
        if ' ' in row:  # If there's an empty space in any row
            return False  # The board is not full, so return False
    return True  # The board is full, return True

# Example:
# If the board is [['X', 'O', 'X'], ['O', 'X', 'O'], ['X', 'O', 'X']]
# There are no empty spaces, so the function will return True (board is full).

# Function to get a player's move
def player_move(board, player):
    """
    This function asks the player where they want to place their symbol ('X' or 'O').
    It checks if the position is valid (if the cell is empty).
    """
    while True:  # Keep asking for input until the move is valid
        try:
            row = int(input(f"Player {player}, enter the row (0-2): "))  # Ask the player for a row (0, 1, or 2)
            col = int(input(f"Player {player}, enter the column (0-2): "))  # Ask the player for a column (0, 1, or 2)

            # Check if the chosen cell is empty
            if board[row][col] == ' ':  # If the cell is empty
                board[row][col] = player  # Place the player's symbol in the cell
                break  # Stop asking and move on to the next step
            else:
                print("That spot is already taken! Try again.")  # Tell the player to choose another spot
        except (ValueError, IndexError):
            print("Invalid input! Please enter numbers between 0 and 2.")  # Handle invalid input

# Example:
# If Player 'X' chooses row=0, column=0, the board will be updated with 'X' in that position.
# If the cell is already occupied (e.g., 'X' or 'O' is already there), the function will ask the player again.

# Main function to play the game
def play_game():
    """
    This function controls the game flow.
    It alternates between Player 'X' and Player 'O', checks for a winner or draw, and updates the board.
    """
    # Create an empty 3x3 board filled with spaces (' ') for the start of the game.
    board = [[' ' for _ in range(3)] for _ in range(3)]  # This creates a 3x3 grid of empty spaces

    current_player = 'X'  # Player 'X' starts the game

    # Loop the game until there is a winner or the board is full (draw)
    while True:
        print_board(board)  # Print the current board

        player_move(board, current_player)  # Let the current player make their move

        # Check if the current player has won
        if check_win(board, current_player):
            print_board(board)  # Print the board after the winning move
            print(f"Player {current_player} wins!")  # Announce the winner
            break  # Stop the game

        # Check if the board is full (draw condition)
        if is_full(board):
            print_board(board)  # Print the board after the last move
            print("It's a draw!")  # Announce a draw
            break  # Stop the game

        # Switch to the next player (X -> O, O -> X)
        current_player = 'O' if current_player == 'X' else 'X'

# Start the game
play_game()
```
